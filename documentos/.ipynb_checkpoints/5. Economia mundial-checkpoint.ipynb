## Comercio internacional, ventaja comparativa y proteccionismo

En los últimos 35 años, las transacciones internacionales se
han vuelto cada vez más importantes
para la economía estadounidense. En
1970, las importaciones representaban sólo alrededor de 7 por ciento del
producto interno bruto (PIB) de Estados Unidos. La participación es ahora
de alrededor de 16 por ciento. En
2004, Estados Unidos importó bienes y
servicios con valor de más de 147,000
millones de dólares cada mes.
La “internacionalización” o “globalización” de la economía estadounidense se ha presentado en el sector
público y el privado, en los mercados
de insumos y productos, y en las empresas mercantiles y las familias. Los
productos extranjeros, que en alguna época fueron poco comunes, se encuentran en todas partes en la
actualidad, desde los utensilios con los que comemos hasta los automóviles que conducimos. En 1970,
los automóviles fabricados en el extranjero sólo representaban un pequeño porcentaje de todos los automóviles en Estados Unidos. En esa época, era difícil encontrar mecánicos que supieran reparar autos
extranjeros y era difícil obtener refacciones. Hoy en día, las calles están llenas de automóviles Toyota y
Nissan de Japón, Volvo de Suecia y BMW de Alemania, y cualquier taller de servicio que no pueda reparar automóviles fabricados en el extranjero seguramente no tendrá mucha clientela. La mitad de
todos los autos y 80 por ciento de todos aparatos electrónicos de consumo (computadoras, televisores,
reproductores de discos compactos, etcétera) que los consumidores estadounidenses compran se producen en otros países.
Al mismo tiempo, Estados Unidos exporta miles de millones de dólares en productos agrícolas,
aviones y maquinaria industrial. Además, Estados Unidos exporta e importa grandes cantidades de
servicios. Cuando un estudiante pakistaní se inscribe en una universidad o institución de educación
superior estadounidense, o una coreana enferma busca atención médica en un hospital de Estados
Unidos, o un keniano contrata un abogado en Miami para que le ayude en la compra de un bien inmueble, o un turista de Indonesia come en un restaurante en la ciudad de Nueva York, Estados Unidos
exporta un servicio. De manera similar, cuando un estudiante de Estados Unidos estudia un año en Escocia, o un turista estadounidense se hospeda en un hotel en Singapur o recibe un masaje en un balneario de Jamaica, Estados Unidos importa un servicio.
Además de que los bienes y servicios (productos) fluyen con facilidad entre las fronteras, ocurre lo
mismo con los insumos: el capital y la mano de obra. Por supuesto, es muy sencillo comprar activos financieros en el extranjero. Millones de estadounidenses tienen acciones de empresas extranjeras o han
invertido en bonos emitidos por otros países. Al mismo tiempo, millones de personas invierten dinero
en los mercados de acciones y bonos de Estados Unidos.
Un nuevo fenómeno, el outsourcing, también está cambiando el carácter del mercado global de
mano de obra. Ahora es sencillo y muy común que una llamada al servicio a clientes de una empresa
de software realizada por un usuario del producto en Bend, Oregon, se envíe a Bangalore, India, donde
un joven y ambicioso indio ofrece asistencia al cliente a través de Internet. En esencia, Internet ha hecho posible que la mano de obra fluya sin contratiempos entre fronteras internacionales.
La inextricable conexión de la economía estadounidense con las economías del resto del mundo
ha tenido un profundo impacto en la disciplina de la economía y constituye la base de uno de sus más
importantes conceptos:
Todas las economías, independientemente de su tamaño, dependen hasta cierto punto de otras
economías y resultan afectadas por los sucesos que ocurren fuera de sus fronteras.
Para familiarizarlo más con la economía internacional, este capítulo analiza la economía del comercio internacional. Primero, se describirá la tendencia reciente en Estados Unidos a importar más de lo
que exporta. En seguida, se explorará la lógica básica del comercio. ¿Por qué Estados Unidos o cualquier
otro país debe participar en actividades de comercio internacional? Por último, se abordará el controvertido asunto del proteccionismo. ¿Debe un país proporcionar protección a ciertas industrias en la forma
de cuotas de importación o aranceles, que son impuestos que gravan las importaciones? ¿Debe un país
ofrecer subsidios para ayudar a su industria nacional a competir en los mercados internacionales?
SUPERÁVIT Y DÉFICIT COMERCIAL
Hasta la década de 1970, Estados Unidos exportaba en general más de lo que importaba. Cuando un país
exporta más de lo que importa, tiene un superávit comercial. Cuando un país importa más de lo que exporta, tiene un déficit comercial. La tabla 20.1 muestra que antes de 1976 Estados Unidos registraba un
superávit comercial. Esto cambió en 1976, y desde ese año Estados Unidos ha registrado déficit comercial.
El déficit alcanzó su nivel máximo de 145,200 millones de dólares en 1987, disminuyó a 27,500 millones de
dólares en 1991 y después aumentó de forma espectacular a más de 600,000 millones de dólares en 2004.
El gran déficit comercial de mediados y finales de la década de 1980 desencadenó una controversia política que continúa en la actualidad. La competencia extranjera golpeó con fuerza los mercados
de Estados Unidos. Los bienes extranjeros menos caros (entre ellos, el acero, los textiles y los automóviles) empezaron a poner en desventaja a los fabricantes estadounidenses que se vieron obligados a cerrar y miles de empleos se perdieron en industrias importantes. Ciudades como Pittsburgh, Youngstown y Detroit tuvieron problemas graves de desempleo.
La reacción natural fue exigir protección para las industrias estadounidenses. Muchas personas
querían que el Presidente y el Congreso gravaran impuestos y restricciones a las importaciones a fin de
que los productos extranjeros estuvieran menos disponibles y fueran más caros para proteger así los
empleos en Estados Unidos. Este argumento no era nuevo. Durante cientos de años, las industrias han
pedido protección a los gobiernos, y las sociedades han debatido las ventajas y desventajas del comercio libre y abierto. En el último siglo y medio, el principal argumento contra el proteccionismo ha sido
la teoría de la ventaja comparativa, que se explicó en el capítulo 2.
LA BASE ECONÓMICA DEL COMERCIO:
VENTAJA COMPARATIVA
Quizá el debate más famoso sobre el tema del libre comercio tuvo lugar en el Parlamento británico en
los primeros años del siglo XIX. En esa época, la aristocracia rural —los terratenientes— controlaban el
Parlamento. Durante varios años, las importaciones y exportaciones de granos habían estado sujetas a
un conjunto de aranceles, subsidios y restricciones llamados de manera colectiva Leyes del maíz. Destinadas a desalentar las importaciones de granos y estimular las exportaciones, el propósito de las Leyes del maíz era mantener el precio de los alimentos en un nivel alto. Por supuesto, los ingresos de los
terratenientes dependían de los precios que cobraban por lo que se producía en sus tierras. Por supuesto, las Leyes del maíz funcionaban en beneficio de quienes detentaban el poder.
Con la Revolución Industrial, surgió una clase de capitalistas industriales acaudalados. El sector
industrial tenía que pagar a los trabajadores por lo menos lo suficiente para subsistir, y el salario de
subsistencia dependía en gran medida del precio de los alimentos. Los aranceles sobre las importaciones de granos y los subsidios a las exportaciones que mantenían los precios de los granos y alimentos
en un nivel alto, aumentaban los salarios que los capitalistas tenían que pagar y reducían sus utilidades.
La encarnizada batalla política duró varios años. Sin embargo, a medida que transcurría el tiempo, el
poder de los terratenientes en la Cámara de los Lores se redujo de manera significativa. Cuando el conflicto terminó en 1848, las Leyes del maíz fueron revocadas.
A favor de la revocación se hallaba David Ricardo, empresario, economista, miembro del Parlamento y uno de los padres de la economía moderna. La obra principal de Ricardo, Principles of Political Economy and Taxation, se publicó en 1817, dos años antes de que ingresara al Parlamento. La teoría
de la ventaja comparativa de Ricardo, que empleó para argumentar en contra de las Leyes del maíz,
postulaba que el comercio permite a los países especializarse en la producción de los bienes que elaboran mejor.
La especialización y el libre comercio benefician a todos socios comerciales (los salarios reales aumentan), incluso a aquellos que podrían ser productores menos eficientes en términos absolutos.
Este argumento básico sigue constituyendo el fondo de los debates sobre el libre comercio hasta el
presente. Los Presidentes Reagan y Bush lo invocaron en numerosas ocasiones cuando se enfrentaron
al Congreso para defender varias leyes proteccionistas.
Especialización y comercio: el caso de dos personas La manera más sencilla de entender la teoría de la ventaja comparativa es examinar una sociedad simple de dos personas. Suponga
que Bill y Colleen, perdidos en una isla desierta en el capítulo 2, tienen sólo dos tareas que realizar cada semana: recolectar comida y cortar madera para construir una casa. Si Colleen pudiera cortar más
madera que Bill en un día y Bill pudiera recolectar más bayas y frutas, es evidente que la especialización los beneficiaría a ambos.
Pero suponga que Bill es lento y torpe y que Colleen es mejor tanto para cortar madera como para recolectar comida. El punto de Ricardo es que de todos modos les conviene especializarse. Pueden
producir más en total si se especializan que si comparten el trabajo por igual. (Quizá sea útil repasar la
explicación de la ventaja comparativa en el capítulo 2 antes de continuar).
VENTAJA ABSOLUTA FRENTE A VENTAJA
COMPARATIVA
Un país disfruta de una ventaja absoluta sobre otro en la elaboración de un producto si utiliza menos
recursos que el otro país para producirlo. Suponga que el país A y el país B producen trigo, pero el clima de A es más adecuado para el trigo y su mano de obra es más productiva. El país A producirá más
trigo por hectárea que el país B y usará menos mano de obra para cultivarlo y llevarlo al mercado. El
país A disfruta de una ventaja absoluta sobre el país B en la producción de trigo.
Un país disfruta de una ventaja comparativa en la producción de un bien si ese bien puede producirse a menor costo en términos de otros bienes. Suponga que los países C y D producen trigo y maíz
y C disfruta de una ventaja absoluta en la producción de los dos; es decir, el clima de C es mejor que el
de D y se necesitan menos recursos de C para producir una cantidad determinada tanto de trigo como de
maíz. Ahora C y D pueden elegir cada uno entre usar la tierra para cultivar trigo o maíz. Para producir más
trigo, cada país debe transferir tierra de la producción de maíz; para producir más maíz, cada país debe transferir tierra de la producción de trigo. El costo del trigo en cada país puede medirse en un
bushel de maíz, y el costo del maíz puede medirse en un bushel de trigo.
Suponga que en el país C, un bushel de trigo tiene un costo de oportunidad de dos bushels de
maíz. Es decir, para producir un bushel adicional de trigo, C tiene que sacrificar 2 bushels de maíz.
Asimismo, producir un bushel de trigo en el país D requiere el sacrificio de sólo 1 bushel de maíz. A pesar de que C tiene una ventaja absoluta en la producción de ambos productos, D disfruta de una ventaja
comparativa en la producción de trigo porque el costo de oportunidad de producir trigo es menor en
D. En estas circunstancias, afirma Ricardo, D puede beneficiarse del comercio si se especializa en la
producción de trigo.
Ganancias de la ventaja absoluta mutua Para ilustrar la lógica de Ricardo con mayor
detalle, suponga que Australia y Nueva Zelanda tienen cada uno una cantidad fija de tierra y no tienen
relaciones comerciales con el resto del mundo. Hay sólo dos bienes: trigo, para producir pan, y algodón, para producir ropa. Este tipo de mundo de dos países y dos productos no existe, pero sus operaciones pueden generalizarse a muchos países y muchos productos.
Para continuar, es necesario establecer varios supuestos sobre las preferencias de la gente que vive en
Nueva Zelanda y la que vive en Australia. Si los ciudadanos de los dos países andan desnudos, no hay necesidad de producir algodón; toda la tierra puede usarse para producir trigo. Sin embargo, suponga que
los habitantes de ambos países tienen preferencias similares con respecto a la comida y la ropa: las poblaciones de los dos países usan algodón y trigo y las preferencias de comida y ropa son tales que en
ambos países se consumen cantidades iguales de trigo y algodón.
Por último, suponga que cada país tiene sólo 100 hectáreas de tierra para cultivar, cuyo rendimiento se presenta en la tabla 20.2. Nueva Zelanda puede producir tres veces el trigo que Australia cultiva en una hectárea de tierra, y Australia puede producir tres veces el algodón que Nueva Zelanda
cultiva en el mismo espacio. Nueva Zelanda tiene una ventaja absoluta en la producción de trigo y
Australia tiene una ventaja absoluta en la producción de algodón. En casos como éste, se dice que ambos países tienen ventaja absoluta mutua.
Si no hay comercio y cada país divide su tierra para obtener unidades iguales de producción de algodón y trigo, cada país produce 150 bushels de trigo y 150 pacas de algodón. Nueva Zelanda usa 75
hectáreas para el cultivo del algodón, pero sólo 25 hectáreas para el de trigo, en tanto que Australia hace lo contrario (tabla 20.3).
La misma información se puede organizar en forma gráfica como fronteras de posibilidades de
producción de cada país. En la figura 20.1, que presenta las posiciones de los dos países antes del trueque, cada país está limitado por sus propios recursos y productividad. Si Australia usa toda su tierra
para cultivar algodón, produciría 600 pacas de algodón (100 hectáreas  6 pacas/hectárea) y nada de
trigo; si usa toda su tierra para cultivar trigo, produciría 200 bushels de trigo (100 hectáreas  2
bushels/hectárea) y nada de algodón. Lo contrario aplica a Nueva Zelanda. Recuerde, por el capítulo 2,
que la frontera de posibilidades de producción de un país representa todas las combinaciones de bienes que pueden producirse, dados los recursos del país y el estado de la tecnología. Cada país debe elegir un punto a lo largo de su propia curva de posibilidades de producción.
Debido a que ambos países tienen una ventaja absoluta en la elaboración de un producto, la especialización y el comercio los beneficiarán a ambos. Australia debe producir algodón, Nueva Zelanda
debe producir trigo. Transferir toda la tierra a la producción de trigo en Nueva Zelanda rinde 600
bushels; transferir toda la tierra a la producción de algodón en Australia rinde 600 pacas. Un acuerdo
para intercambiar 300 bushels de trigo por 300 pacas de algodón duplicaría tanto el consumo de trigo como el de algodón en los dos países. (Recuerde que antes del trueque, los dos países producían
150 bushels de trigo y 150 pacas de algodón. Después del trueque, cada país tendrá 300 bushels de trigo y 300 pacas de algodón que consumir. Las cifras finales de producción y trueque se presentan en la
tabla 20.4 y la figura 20.2).
El comercio permite a ambos países dejar atrás las restricciones de recursos y productividad que
anteriormente los limitaban.
Las ventajas de la especialización y el comercio parecen obvias cuando un país es tecnológicamente superior en la producción de un bien y otro país es tecnológicamente superior en la producción de
otro bien. Sin embargo, estudiemos el caso en el que un país tiene ventaja absoluta en la producción de
ambos productos.
Ganancias de la ventaja comparativa La tabla 20.5 contiene cifras diferentes de rendimiento de la tierra de Nueva Zelanda y Australia. Ahora Nueva Zelanda tiene una ventaja absoluta
considerable en la producción tanto de algodón como de trigo, ya que una hectárea de tierra produce
seis veces la cantidad de trigo y dos veces la cantidad de algodón que una hectárea en Australia. Ricardo diría que la especialización y el comercio siguen siendo mutuamente beneficiosos.
Una vez más, las preferencias implican el consumo de unidades iguales de algodón y trigo en ambos países. Sin comercio, Nueva Zelanda dividiría de manera uniforme sus 100 hectáreas, ó 50/50 entre los dos cultivos. El resultado serían 300 pacas de algodón y 300 bushels de trigo. Australia dividiría
su tierra 75/25. La tabla 20.6 muestra que la producción final en Australia sería de 75 pacas de algodón
y 75 bushels de trigo. (Recuerde que se ha supuesto que en cada país la gente consume cantidades iguales de algodón y trigo). De nuevo, antes de que tenga lugar el trueque, cada país se ve limitado por su
propia curva de posibilidades de producción nacional.
Imagine que nos encontramos en una reunión de representantes comerciales de los dos países.
Como asesor especial, se le pide a David Ricardo que demuestre que el comercio beneficia a ambos
países. Ricardo divide su demostración en tres etapas, que usted puede seguir en la tabla 20.7.
En la etapa 1, Australia transfiere toda su tierra a la producción de algodón. Tendrá 300 pacas de
algodón y nada de trigo. Nueva Zelanda no puede especializarse por completo en la producción de trigo porque necesita 300 pacas de algodón y no podrá conseguir suficiente algodón de Australia. La razón es que se ha supuesto que cada país quiere consumir cantidades iguales de algodón y trigo.
En la etapa 2, Nueva Zelanda retira 25 hectáreas del cultivo de algodón para dedicarlas al trigo.
Ahora Nueva Zelanda tiene 25 hectáreas de algodón que producen 150 pacas y 75 hectáreas de trigo que
producen 450 bushels.
Por último, los dos países hacen un trueque. Suponemos que Nueva Zelanda envía 100 bushels de
trigo a Australia a cambio de 200 pacas de algodón. Después del trueque, Nueva Zelanda tiene 350 pacas
de algodón y 350 bushels de trigo; Australia tiene 100 pacas de algodón y 100 bushels de trigo. Ambos
países se encuentran en mejor situación que antes del trueque (tabla 20.6) y los dos han avanzado más
allá de sus propias fronteras de posibilidades de producción.
¿Por qué funciona el plan de Ricardo? Para entender por qué funciona el plan de Ricardo, volvamos a la definición de ventaja comparativa.
El costo real de producir algodón es el trigo que debe sacrificarse para producirlo. Cuando se piensa
en el costo de esta manera, es menos costoso producir algodón en Australia que producirlo en Nueva Zelanda,
a pesar de que una hectárea de tierra produce más algodón en Nueva Zelanda. Considere el “costo” de 3
pacas de algodón en los dos países. En términos del costo de oportunidad, 3 pacas de algodón en Nueva
Zelanda cuestan 3 bushels de trigo; en Australia, 3 pacas de algodón cuestan sólo 1 bushel de trigo.
Como en tierras australianas se producen 3 pacas por hectárea, para obtener 3 pacas, un australiano
debe transferir 1 hectárea de tierra de producción de trigo a algodón. Debido a que una hectárea de
tierra produce un bushel de trigo, la pérdida de una hectárea para dedicarla al cultivo de algodón implica la pérdida de un bushel de trigo. Australia tiene una ventaja comparativa en producción de algodón
porque su costo de oportunidad, en términos de trigo, es menor que el de Nueva Zelanda. Esto se ilustra en la figura 20.3.
A la inversa, Nueva Zelanda tiene una ventaja comparativa en producción de trigo. Una unidad de
trigo en Nueva Zelanda cuesta una unidad de algodón; una unidad de trigo en Australia cuesta tres
unidades de algodón.
Cuando los países se especializan en producir bienes en los cuales tienen una ventaja comparativa, maximizan su producción combinada y asignan sus recursos con mayor eficiencia.
TÉRMINOS DE COMERCIO
Ricardo podría sugerir varias opciones abiertas a los socios comerciales. La que acabamos de examinar
beneficia a los dos socios; en términos porcentuales, Australia sale ganando un poco más. Otras transacciones podrían haber sido más ventajosas para Nueva Zelanda.
La razón a la que un país puede cambiar productos nacionales por productos importados se llama términos de comercio. Los términos de comercio determinan cómo se distribuyen las ganancias
del intercambio entre los socios comerciales. En el caso que acabamos de considerar, los términos de comercio convenidos fueron 1 bushel de trigo por 2 pacas de algodón. Dichos términos de comercio benefician a Nueva Zelanda, que puede conseguir 2 pacas de algodón por cada bushel de trigo. Si transfiriera su
propia tierra de trigo a algodón, sólo obtendría 1 paca de algodón. Los mismos términos de comercio benefician a Australia, que puede obtener 1 bushel de trigo por 2 pacas de algodón. Una transferencia directa de su propia tierra la obligaría a sacrificar 3 pacas de algodón por 1 bushel de trigo.
Si los términos de comercio cambiaran a 3 pacas de algodón por cada bushel de trigo, sólo Nueva
Zelanda se beneficiaría. Con esos términos de comercio, todas las ganancias del trueque corresponderían
a Nueva Zelanda. Estos términos no benefician en absoluto a Australia porque el costo de oportunidad
de producir trigo a escala nacional es exactamente el mismo que el costo del trueque: un bushel de trigo
cuesta 3 pacas de algodón. Si los términos de comercio fueran en sentido contrario (1 paca de algodón
por cada bushel de trigo), sólo Australia se beneficiaría. Nueva Zelanda no gana nada porque ya puede
sustituir el trigo por algodón a esa razón. Sin embargo, para obtener un bushel de trigo a escala nacional, Australia debe sacrificar 3 pacas de algodón y los términos de comercio de uno a uno harían que
el trigo resultara mucho menos costoso para Australia.
Ambas partes deben ganar algo para que el trueque se lleve a cabo. En este caso, se puede ver que
tanto Australia como Nueva Zelanda ganan cuando los términos de comercio se establecen entre 1:1 y
3:1, algodón a trigo.
TIPOS DE CAMBIO
Los ejemplos que hemos presentado hasta el momento han demostrado que el comercio puede producir ganancias para ambas partes. Aún no se ha analizado cómo se realiza de manera efectiva el comercio.
Cuando el comercio es libre —libre de las trabas que plantean las barreras instituidas por el gobierno—, las pautas y flujos comerciales son el resultado de las decisiones independientes de miles de importadores y exportadores y de millones de familias y empresas privadas.
Las familias privadas deciden si comprar autos Toyota o Chevrolet, y las empresas privadas deciden si
comprar herramientas mecánicas fabricadas en Estados Unidos o en Taiwán, y acero bruto producido
en Alemania o en Pittsburgh.
Para que un ciudadano de un país pueda comprar un producto hecho en otro país o vendido por
alguien de otro país, tiene que efectuarse un intercambio de divisas. Considere a Shane, que compra un
Toyota a un distribuidor de Boston. Paga en dólares, pero los trabajadores japoneses que fabricaron el
auto reciben sus salarios en yenes. En algún punto entre el comprador y el productor del auto, tendrá
que realizarse un intercambio de moneda. El distribuidor regional probablemente acepte el pago en
dólares que luego convierte a yenes antes de enviar el producto de la venta a Japón.
Para comprar un bien producido en otro país, es necesario, en efecto, comprar moneda extranjera. El precio del Toyota de Shane en dólares depende tanto del precio del auto expresado en yenes como del precio en dólares del yen. Tal vez usted conozca muy bien las peculiaridades del cambio de moneda si alguna vez ha viajado a otro país.
En octubre de 2005, la libra esterlina valía 1.75 dólares. Ahora suponga que se encuentra en Londres, cenando. En el menú aparece una botella de buen vino que vale 15 libras. ¿Cómo puede decidir si
desea comprarla? Usted sabe qué puede comprar con dólares en Estados Unidos, por lo que tiene que
convertir el precio a dólares. Cada libra le costará 1.75 dólares, de modo que 15 libras le costarán $1.75
 15 = $26.25.
El atractivo de los bienes extranjeros para los compradores estadounidenses y de los bienes estadounidenses para los compradores extranjeros depende en parte del tipo de cambio, la razón a la cual
se intercambian dos divisas. Si el precio de las libras bajara a 1.20 dólares, esa misma botella de vino
costaría 18 dólares.
Para entender las pautas de comercio que resultan de los actos de cientos de miles de compradores y vendedores independientes —familias y empresas—, es preciso conocer algo sobre los factores
que determinan los tipos de cambio. La determinación de los tipos de cambio es muy complicada. Sin
embargo, aquí se pueden demostrar dos cosas:
Primero, por cada par de países, hay una gama de tipos de cambio que pueden conducir automáticamente a que ambos países realicen las ganancias de especialización y ventaja comparativa. Segundo, dentro de esa gama, el tipo de cambio determinará qué país gana más del intercambio comercial. En resumen, los tipos de cambio determinan los términos de comercio.
Comercio y tipos de cambio en un mundo de dos países y dos bienes Considere primero un modelo sencillo de dos países y dos bienes. Suponga que tanto Estados Unidos como
Brasil producen sólo dos bienes: madera sin procesar y acero laminado. La tabla 20.8 presenta los precios actuales de ambos bienes según los ven los compradores nacionales. En Brasil, la madera tiene un
precio de 3 reales (R) por pie, y el acero tiene un precio de 4 R por metro. En Estados Unidos, la madera cuesta $1 por pie y el acero cuesta $2 por metro.
Suponga que los compradores estadounidenses y brasileños tienen la opción de comprar en su
país o importar para satisfacer sus necesidades. Las opciones que elijan dependerán del tipo de cambio. Por el momento, haremos caso omiso de los costos de transporte entre los dos países y supondremos que los productos brasileños y estadounidenses son de igual calidad.
Comencemos con la suposición de que el tipo de cambio es $1 = 1 R. Desde el punto de vista de
los compradores estadounidenses, ni el acero brasileño ni la madera brasileña son competitivos a este
tipo de cambio. Un dólar compra un pie de madera en Estados Unidos, pero si se convierte en un real,
sólo comprará una tercera parte de un pie. El precio de la madera brasileña para un estadounidense es
de $3 porque se requieren $3 para comprar los 3 R necesarios. Asimismo, $2 compran un metro de
acero laminado en Estados Unidos, pero los mismos $2 sólo compran medio metro de acero brasileño.
El precio del acero brasileño para un estadounidense es de $4, el doble del precio del acero producido
nacionalmente.
A este tipo de cambio, sin embargo, los brasileños consideran que el acero y la madera producidos
en Estados Unidos son menos caros que los que se producen en Brasil. La madera nacional —de Brasil— cuesta 3 R, pero con 3 R se compran $3, que compran el triple de madera en Estados Unidos. Del
mismo modo, el acero cuesta 4 R en Brasil, pero con 4 R se compran $4, que compran el doble de acero fabricado en Estados Unidos. Al tipo de cambio de $1 = 1 R, Brasil importará acero y madera y Estados Unidos no importará nada.
Sin embargo, ahora suponga que el tipo de cambio es de 1 R = $0.25. Esto significa que 1 dólar
compra 4 R. A este tipo de cambio, los brasileños compran madera y acero en su país y los estadounidenses importan ambos bienes. A este tipo de cambio, los estadounidenses deben pagan un dólar por
un pie de madera estadounidense, pero la misma cantidad de madera puede adquirirse en Brasil por el
equivalente de $0.75. (Porque 1 R cuesta $0.25, 3 R pueden comprarse con $0.75). Asimismo, el acero
que cuesta $2 por metro en Estados Unidos, a un estadounidense le cuesta la mitad en Brasil, por
lo que con $2 compra 8 R, que compran 2 metros de acero brasileño. Al mismo tiempo, a los brasileños no les interesa importar, porque ambos bienes son más baratos cuando se compran a un productor brasileño. En este caso, Estados Unidos importa ambos productos y Brasil no importa nada.
Hasta el momento, se puede ver que a los tipos de cambio de $1 = 1 R y $1 = 4 R, el comercio fluye sólo en una dirección. Ahora probemos con el tipo de cambio de $1 = 2 R, o 1 R = $0.50. Primero, los brasileños comprarán madera en Estados Unidos. La madera brasileña cuesta 3 R por pie, pero 3 R compran
$1.50, suficiente para comprar 1.5 pies de madera estadounidense. Los compradores en Estados Unidos
considerarán que la madera brasileña es muy cara, pero Brasil importará madera de Estados Unidos. A este mismo tipo de cambio, sin embargo, tanto los compradores brasileños como los estadounidenses serán
indiferentes entre el acero brasileño y estadounidense. Para los compradores estadounidenses, el acero
producido en su país cuesta $2. Como con $2 se compran 4 R, un metro de acero brasileño importado
también cuesta $2. Los compradores brasileños también considerarán que el acero cuesta 4 R, ya sea de
producción nacional o importado. Por tanto, es probable que no haya comercio en acero.
¿Qué sucede si el tipo de cambio varía y con $1 se pueden comprar 2.1 R? Aunque la madera estadounidense sigue siendo más barata tanto para los brasileños como para los estadounidenses, el acero
brasileño comienza a resultar atractivo para los compradores estadounidenses. El acero producido en
Estados Unidos cuesta $2 por metro, pero con $2 se compran 4.2 R, que compran más de un metro de
acero en Brasil. Cuando $1 compra más de 2 R, el comercio empieza a fluir en ambas direcciones: Brasil importa madera y Estados Unidos importa acero.
Si examina detenidamente la tabla 20.9, verá que el comercio fluye en ambas direcciones siempre
que el tipo de cambio se establezca entre $1 = 2 R y $1 = 3 R. Dicho a la inversa, el comercio fluye en
ambas direcciones si el precio de un real se ubica entre $0.33 y $0.50.
Tipos de cambio y ventaja comparativa Si el mercado de divisas mueve el tipo de cambio a algún punto entre 2 y 3 R por dólar, los países automáticamente se ajustarán y se realizará una
ventaja comparativa. A estos tipos de cambio, los compradores estadounidenses empezarán a comprar
todo el acero en Brasil. La industria del acero estadounidense se encontrará en problemas. Las fábricas
cerrarán y los trabajadores estadounidenses comenzarán a cabildear para conseguir protección arancelaria contra el acero brasileño. Al mismo tiempo, la industria maderera estadounidense prosperará,
impulsada por la fuerte demanda de exportaciones de Brasil. El sector productor de madera se expandirá. Y la producción de madera atraerá recursos, incluidos el capital y la mano de obra.
Lo contrario ocurre en Brasil. La industria maderera de Brasil sufre pérdidas cuando no puede
hacer exportaciones y los brasileños prefieren comprar la madera estadounidense importada, que es
más barata. En Brasil, las empresas madereras recurren al gobierno y solicitan protección contra la madera barata de Estados Unidos. Sin embargo, los productores de acero en Brasil están encantados. No
sólo están suministrando el 100 por ciento del acero demandado a nivel nacional, sino que también están vendiendo a los compradores estadounidenses. La industria siderúrgica se expande, y la maderera
se contrae. Los recursos, incluida la mano de obra, fluyen hacia el acero.
Con este panorama de expansión y contracción en mente, examinemos de nuevo la definición de
ventaja comparativa. Si suponemos que los precios reflejan el uso de los recursos y éstos pueden transferirse de un sector a otro, podremos calcular el costo de oportunidad del acero y la madera en ambos
países. En Estados Unidos, la producción de un metro de acero laminado consume el doble de recursos que la producción de un pie de madera. Si suponemos que los recursos pueden transferirse, el costo de oportunidad de un metro de acero será de 2 pies de madera (véase la tabla 20.8). En Brasil, un
metro de acero usa recursos que cuestan 4 R, mientras que una unidad de madera cuesta 3 R. Producir
un metro de acero significa el sacrificio de sólo cuatro tercios (o uno y un tercio) de pies de madera.
Como el costo de oportunidad de un metro de acero (en términos de la madera) es inferior en Brasil,
decimos que Brasil tiene una ventaja comparativa en la producción de acero.
A la inversa, considere el costo de oportunidad de la madera en los dos países. Aumentar la producción de madera en Estados Unidos requiere el sacrificio de medio metro de acero por cada pie de madera: producir un metro de acero usa recursos con valor de $2, mientras que producir un pie de madera
requiere recursos que sólo valen $1. No obstante, cada pie de producción de madera en Brasil requiere
el sacrificio de tres cuartos de un metro de acero. Debido a que el costo de oportunidad de la madera es
inferior en Estados Unidos, este país tiene una ventaja comparativa en la producción de madera.
Si los tipos de cambio se ubican en los rangos indicados, el libre mercado llevará a cada país a
trasladar recursos a los sectores en los que disfruta de una ventaja comparativa. Sólo los productos en los que un país tiene ventaja comparativa serán competitivos en los mercados mundiales.
LOS ORÍGENES DE LA VENTAJA COMPARATIVA
La especialización y el comercio pueden beneficiar a todos los socios comerciales, incluso aquellos que
podrían ser productores ineficientes en términos absolutos. Si los mercados son competitivos, y si los
mercados de divisas están vinculados al intercambio de bienes y servicios, los países se especializarán
en elaborar productos en los que tienen ventaja comparativa.
Hasta el momento, no se ha dicho nada sobre los orígenes de la ventaja comparativa. ¿Qué determina si un país tiene ventaja comparativa en la industria manufacturera pesada o en la agricultura?
¿Qué explica los flujos de comercio reales que se observan en todo el mundo? Varias teorías y trabajos
empíricos sobre el comercio internacional han proporcionado algunas respuestas. La mayoría de los economistas consideran que las dotaciones de factores (la cantidad y calidad de mano de obra, tierra y recursos naturales) son las causas principales de ventaja comparativa. Las dotaciones de factores parecen
explican una parte considerable de las pautas de comercio mundial reales.
EL TEOREMA DE HECKSCHER-OHLIN
Eli Heckscher y Bertil Ohlin, dos economistas suecos que escribieron en la primera mitad del siglo XX,
ampliaron y elaboraron la teoría de la ventaja comparativa de Ricardo. El teorema de Heckscher-Ohlin
relaciona la teoría de la ventaja comparativa con la dotación de factores. Supone que los bienes pueden
producirse usando diferentes proporciones de insumos y que los insumos son móviles entre los sectores de cada economía, pero que los factores no son móviles entre las economías. Según este teorema:
Un país tiene ventaja comparativa en la producción de un bien si dicho país está relativamente
bien dotado de los insumos que se usan intensivamente en la producción de ese bien.
Esta idea es sencilla. Es probable que un país que tiene mucha tierra fértil tenga una ventaja comparativa en agricultura. Un país que tiene una gran cantidad de capital acumulado probablemente tendrá una ventaja comparativa en la fabricación de maquinaria pesada. Un país con mucho capital humano de seguro tendrá una ventaja comparativa en bienes muy técnicos.
OTRAS EXPLICACIONES DE LOS FLUJOS
DE COMERCIO OBSERVADOS
La ventaja comparativa no es la única razón por la que los países sostienen relaciones comerciales. Ésta no explica por qué muchos países importan y exportan el mismo tipo de bienes. Por ejemplo, Estados Unidos exporta e importa automóviles.
Al igual que las industrias dentro de un país diferencian sus productos para captar un mercado
interno, así también diferencian sus productos para complacer a la amplia variedad de gustos que existe en el mundo. Por ejemplo, la industria automotora japonesa comenzó a producir autos pequeños,
que hacían uso eficiente del combustible, mucho antes de que los fabricantes de automóviles estadounidenses lo hicieran. Al hacerlo, adquirieron pericia y experiencia en crear productos que atraían a
muchos seguidores devotos y una considerable lealtad a la marca. Los BMW, fabricados sólo en Alemania, y los Volvo, fabricados sólo en Suecia, también tienen sus campeones en muchos países. Así como
la diferenciación de los productos es una respuesta natural a las preferencias diversas dentro de una
economía, también es una respuesta natural a las preferencias diversas en distintas economías.
Esta idea no se contrapone a la teoría de la ventaja comparativa. Si los japoneses adquirieron técnicas y conocimientos que les dieron una ventaja en la producción de automóviles que hacen uso eficiente del combustible, puede pensarse que ese conocimiento representa un tipo muy específico de capital que en la actualidad no está disponible para otros productores. La compañía Volvo invirtió en
una forma de capital intangible que se llama crédito mercantil. Dicho crédito mercantil, que proviene
de haber establecido la reputación de seguridad y calidad a través de los años, es una causa de la ventaja comparativa que hace que los Volvo se sigan vendiendo en el mercado internacional. Algunos economistas distinguen entre las ganancias de ventajas comparativas adquiridas y las de las ventajas comparativas naturales.
Otra explicación del comercio internacional es que puede haber varias economías de escala en la
producción para un mercado mundial que no podrían conseguirse cuando se produce para un mercado nacional más limitado. Sin embargo, los hechos indican que las economías de escala se agotan a un
tamaño relativamente pequeño en la mayoría de las industrias, por lo que parece poco probable que
constituyan una explicación válida de las pautas de comercio mundial.
BARRERAS COMERCIALES: ARANCELES,
SUBSIDIOS A LAS EXPORTACIONES Y CUOTAS
Las barreras comerciales —también llamadas obstáculos para el comercio— adoptan muchas formas; las
tres más comunes son: aranceles, subsidios a las exportaciones y cuotas. Todas son formas de protección de algún sector de la economía contra la competencia externa.
Un arancel es un impuesto sobre las importaciones. El arancel promedio sobre las importaciones
que llegan a Estados Unidos es de menos de 5 por ciento. Ciertos artículos protegidos tienen aranceles
mucho más altos. Por ejemplo, en 2005, los aranceles sobre el calzado de goma variaron entre 20 y 48
por ciento, y sobre el atún enlatado fue de 35 por ciento.
Los subsidios a las exportaciones (pagos del gobierno efectuados a las empresas nacionales para
estimular las exportaciones) también pueden actuar como barrera comercial. Una de las disposiciones
de las Leyes del maíz que dieron pie a las reflexiones de Ricardo fue un subsidio a la exportación que
automáticamente pagaba el gobierno británico a los agricultores cuando el precio de los granos caía
por debajo de un nivel especificado. El subsidio servía para mantener los precios internos en un nivel
alto, pero inundaba el mercado mundial con grano subsidiado barato. Los agricultores de otros países
que no estaban subsidiados se quedaban fuera del mercado internacional por los precios artificialmente bajos.
Los subsidios a la agricultura siguen siendo parte del panorama del comercio internacional en la
actualidad. Muchos países siguen apaciguando a sus agricultores mediante fuertes subsidios a la exportación de productos agrícolas. El poder político del cabildeo agrícola en muchos países ha tenido
un efecto importante en las negociaciones recientes de comercio internacional dirigidas a reducir las
barreras comerciales.
El dumping se relaciona muy estrechamente con los subsidios. El dumping tiene lugar cuando una
empresa o una industria venden productos en el mercado mundial a precios inferiores al costo de producción. Se ha acusado de esta práctica a varias industrias japonesas, incluidas las de automóviles, aparatos electrónicos de consumo y chips de silicio para computadora. En general, una compañía practica el dumping cuando quiere dominar un mercado mundial. Después de que los precios reducidos de
los productos con que se ha inundado el mercado logran eliminar a toda la competencia, la empresa
que realizó el dumping puede explotar su posición si aumenta el precio de su producto. Una empresa estadounidense que intenta monopolizar un mercado interno infringe la Ley Antimonopolios Sherman
de 1890, que prohíbe la fijación de precios depredadores.
Las leyes arancelarias vigentes en Estados Unidos contienen varias disposiciones que intentan
contrarrestar los efectos del dumping. La Ley de Comercio de 1974 contiene una cláusula que permite
dar protección a una industria si ha resultado “lesionada” por la competencia extranjera. Con base en
esa legislación, los proyectos de ley de comercio más recientes, incluida la Ley de Comercio Integral de
1988, contienen cláusulas que permiten al Presidente imponer sanciones comerciales cuando las investigaciones ponen al descubierto prácticas de dumping de compañías extranjeras u otros países.
Una cuota es un límite que se impone a la cantidad de importaciones. Las cuotas pueden ser obligatorias o voluntarias y pueden legislarse o negociarse con otros gobiernos. La cuota voluntaria mejor
conocida, o “restricción voluntaria”, se negoció con el gobierno japonés en 1981. Japón accedió a reducir sus exportaciones de automóviles a Estados Unidos en 7.7 por ciento del nivel de 1980 de 1.82 millones de unidades a 1.68 millones de unidades. En la actualidad, muchas cuotas limitan el comercio
mundial. Tal vez el caso reciente más famoso sea la cuota de textiles impuesta por la Unión Europea a
las importaciones de textiles de China en agosto de 2005. En virtud de que China rebasó las cuotas que
se habían acordado a principios de año, la UE bloqueó la entrada de los textiles producidos en China,
y más de 100 millones de prendas se apilaron en los puertos europeos.
Políticas comerciales de Estados Unidos y el GATT Estados Unidos ha sido una nación de aranceles altos, donde el promedio de los aranceles ha sido de más de 50 por ciento en buena
parte de su historia. Los más altos estuvieron en vigor durante la Gran Depresión después del arancel
de Smoot-Hawley, que elevó la tasa promedio de los aranceles a 60 por ciento en 1930. El arancel de
Smoot-Hawley desencadenó una guerra internacional de comercio cuando los socios comerciales de
Estados Unidos establecieron en represalia sus propios aranceles. Muchos economistas opinan que el
desplome del comercio que siguió fue una de las causas de la depresión mundial de la década de 1930.1
En 1947, Estados Unidos, con otros 22 países, se comprometió a reducir las barreras al comercio.
También estableció una organización para promover la liberalización del comercio exterior. El Acuerdo General sobre Aranceles y Comercio (GATT), que en un principio se consideró un arreglo provisional, sigue funcionando en la actualidad y ha sido muy eficaz. La más reciente ronda de charlas sobre
comercio mundial auspiciada por el GATT, la Ronda de Uruguay, empezó en Uruguay en 1986. Fue
suscrita por 116 países el 15 de diciembre de 1993, y aprobada formalmente por el Congreso estadounidense tras un gran debate después de la elección de 1994. El acta definitiva de la Ronda Uruguay de
negociaciones es el acuerdo de comercio multilateral más completo y complejo de la historia.
Cada presidente que ha ocupado el cargo desde que se suscribió la primera ronda de este acuerdo
general ha argumentado a favor de las políticas de libre comercio; sin embargo, cada uno ha hecho uso
de sus facultades para proteger un sector económico u otro. Eisenhower y Kennedy restringieron las
importaciones estadounidenses de textiles japoneses; Johnson restringió las importaciones de carne;
Nixon restringió las importaciones de acero e hizo más estrictas las restricciones sobre los textiles; Carter protegió el acero, textiles y calzado; Reagan restringió las importaciones de azúcar y automóviles. A
principios de 2002, el Presidente George W. Bush impuso un arancel de 30 por ciento al acero que se
importa a Estados Unidos. La industria siderúrgica estadounidense empleaba a 175,000 trabajadores en
2002, pero casi 20,000 empleos se perdieron en los cuatro años anteriores debido a los bajos precios del
acero importado. El arancel que Bush impuso al acero fue recibido con enojo por los demás países productores de acero, muchos de los cuales amenazaron con una guerra comercial. Aunque el arancel del
acero se retiró para la mayoría de los países, seguía en vigor para las importaciones de Japón, Rusia y
Brasil a tasas de hasta 50 por ciento en 2005.
No obstante, la tendencia en Estados Unidos ha sido dejar atrás los aranceles y cuotas y avanzar
hacia un comercio más libre. La Ley de Tratados Recíprocos de Comercio de 1934 autorizó al Presidente a negociar tratados comerciales en representación de Estados Unidos. Como parte de las negociaciones comerciales, el Presidente puede conferir la categoría de nación más favorecida a sus socios comerciales en lo individual. Las importaciones de los países que gozan de la categoría de nación más
favorecida se gravan con aranceles a las tasas más bajas negociadas. Además, en los últimos años, varias
rondas de negociaciones exitosas sobre disminución de aranceles han reducido las barreras comerciales a los niveles más bajos de la historia.
Integración económica La integración económica ocurre cuando dos o más naciones se unen
para formar una zona de libre comercio. En 1991, la Comunidad Europea (CE, o el Mercado Común)
empezaron a formar la zona de libre comercio más grande del mundo. El proceso de integración económica se inició en diciembre de ese año, cuando los doce miembros originales (Reino Unido, Bélgica,
Francia, Alemania, Italia, Países Bajos, Luxemburgo, Dinamarca, Grecia, Irlanda, España y Portugal)
firmaron el Tratado de Maastricht. El tratado exigía el fin de los controles fronterizos, una moneda común, la eliminación de todos los aranceles y la coordinación de los asuntos monetarios e incluso políticos. La Unión Europea (UE), como ahora se llama la CE, tiene 25 miembros y 4 solicitantes (consulte
la lista en el resumen del capítulo, p. 413). El 1 de enero de 1993, se eliminaron todos los aranceles y barreras comerciales entre los países miembros. Los controles fronterizos se cerraron a principios de
1995. Los ciudadanos pueden viajar ahora entre los países miembros sin pasaporte.
Estados Unidos no forma parte de la UE. Sin embargo, en 1988, Estados Unidos (durante la administración del Presidente Reagan) y Canadá (durante la administración del Primer Ministro Mulroney) suscribieron el Tratado de Libre Comercio entre Estados Unidos y Canadá, que eliminó todas las
barreras al comercio, incluidos los aranceles y cuotas, entre los dos países en 1998.
En los últimos días de la administración de George H.W. Bush, Estados Unidos, México y Canadá
suscribieron el Tratado de Libre Comercio de América del Norte (TLCAN), mediante el cual los tres países se comprometieron a establecer una zona de libre comercio en toda América del Norte. La zona de
libre comercio de América del Norte comprende 360 millones de personas y una producción total de
más de 7 billones de dólares, mayor que la producción de la UE. El tratado elimina todos los aranceles
a lo largo de un periodo de 10 a 15 años y retira las restricciones sobre la mayoría de las inversiones.
Durante la campaña presidencial de 1992, el TLCAN fue tema de acalorados debates. Tanto Bill
Clinton como George Bush apoyaron el tratado. Los sindicatos de trabajadores industriales que podrían resultar afectados por las crecientes importaciones de México (como las de la industria automotora) se opusieron al tratado, mientras que las industrias cuyas exportaciones a México podrían aumentar como resultado del tratado (por ejemplo, la industria de las herramientas mecánicas) lo
apoyaron. Otro motivo de preocupación fue que las empresas mexicanas no estaban sujetas a las mismas reglamentaciones ambientales que las compañías estadounidenses, por lo que estas últimas podrían mudarse a México por esta razón.
El Congreso de Estados Unidos ratificó el TLCAN a finales de 1993 y entró en vigor el primer día
de 1994. El Departamento de Comercio estadounidense ha calculado que como resultado del TLCAN,
el comercio entre Estados Unidos y México aumentó casi 16,000 millones de dólares en 1994. Además,
las exportaciones de Estados Unidos a México superaron a las importaciones de México durante 1994.
Sin embargo, en 1995, el tratado cayó bajo la sombra del espectacular colapso del valor del peso. Las
exportaciones de Estados Unidos a México cayeron en picada y Estados Unidos pasó de tener un superávit comercial a un enorme déficit comercial con México. Aparte de unos cuantos aranceles, sin embargo, todos los compromisos establecidos en el TLCAN se implementaron por completo en 2003, y
un informe sobre los primeros ocho años del tratado, firmado por los tres países, declaró que el pacto
era un éxito. El informe concluye: “Ocho años de crecimiento del comercio, mayor empleo e inversión
y una mejor oportunidad para los ciudadanos de los tres países han demostrado que el TLCAN funciona y seguirá funcionando”.
¿LIBRE COMERCIO O PROTECCIONISMO?
Uno de los grandes debates económicos de todos los tiempos gira en torno de la controversia entre libre
comercio o proteccionismo. A continuación, se resumen brevemente los argumentos a favor de cada uno.
ARGUMENTOS A FAVOR DEL LIBRE COMERCIO
En cierto sentido, la teoría de la ventaja comparativa es un argumento a favor del libre comercio. El comercio ofrece beneficios potenciales a todas las naciones. Un bien no se importa a menos que su precio neto para los compradores sea inferior al precio neto de la alternativa producida internamente.
Cuando para los brasileños del ejemplo anterior la madera estadounidense resultaba menos cara que
la propia, la compraban; sin embargo, seguían pagando el mismo precio por el acero nacional. Los estadounidenses compraban acero brasileño menos caro, pero seguían comprando madera nacional al
mismo precio bajo. En estas condiciones, tanto los estadounidenses como los brasileños acababan pagando menos y consumiendo más.
Al mismo tiempo, los recursos (incluida la mano de obra) se trasladan de la producción de acero
a la producción de madera en Estados Unidos. En Brasil, los recursos (incluida la mano de obra) se
trasladan de la producción de madera a la de acero. Los recursos de ambos países se usan de manera
más eficiente. Los aranceles, los subsidios a las exportaciones y las cuotas, que interfieren con la libre
circulación de bienes y servicios en todo el mundo, reducen o eliminan las ganancias de la ventaja
comparativa.
Se pueden usar curvas de oferta y demanda para ilustrar esto. Suponga que la figura 20.4 muestra
la oferta y demanda nacional de textiles. En la ausencia de comercio, el mercado alcanza el equilibrio
en un precio de $4.20. En equilibrio, se producen y consumen 450 millones de metros de telas.
Suponga ahora que los textiles se venden al precio mundial de $2. Éste es el precio en dólares que
los estadounidenses deben pagar por textiles de otros países. Si suponemos que existe una cantidad ilimitada de textiles disponibles a $2 y que no hay diferencia en calidad entre los textiles nacionales y
extranjeros, ningún productor nacional podrá cobrar más de $2. En la ausencia de barreras comerciales, el precio mundial establece el precio en Estados Unidos. A medida que el precio baja en Estados
Unidos de $4.20 a $2.00, la cantidad demandada por los consumidores aumenta de 450 millones de
metros a 700 millones de metros, pero la cantidad ofrecida por los productores nacionales disminuye
de 450 millones de metros a 200 millones de metros. La diferencia, 500 millones de metros, es la cantidad de textiles importados.
El argumento a favor del libre comercio es que cada país debe especializarse en producir los bienes y servicios en los que disfruta de una ventaja comparativa. Si los productores extranjeros pueden
producir textiles a un precio mucho más bajo que los productores nacionales, tendrán una ventaja
comparativa. A medida que el precio mundial de los textiles disminuye a $2, la cantidad nacional (Estados Unidos) ofrecida se reduce y los recursos se transfieren a otros sectores. Estos otros sectores, que
pueden ser industrias de exportación o nacionales, no aparecen en la figura 20.4a. Es evidente que la
asignación de recursos es más eficiente al precio de $2. ¿Por qué debe usar Estados Unidos recursos nacionales para producir lo que los productores extranjeros pueden producir a un menor costo? Los recursos estadounidenses deben usarse en la producción de las cosas que este país produce mejor.
Ahora considere lo que ocurre con el precio nacional de los textiles cuando se impone una barrera comercial. La figura 20.4b muestra el efecto de un arancel de $1 por metro impuesto a los textiles
importados. El arancel aumenta el precio nacional de los textiles a $2 + $1 = $3. El resultado es que
parte de las ganancias del comercio se pierden. Primero, los consumidores se ven obligados a pagar un
precio más alto por el mismo bien; la cantidad de textiles demandada se reduce de 700 millones de metros en el libre comercio a 600 millones de metros, porque algunos consumidores no están dispuestos
a pagar el precio más alto.
Al mismo tiempo, el precio más alto de los textiles atrae a la producción de textiles a algunos productores nacionales marginales que no podían obtener utilidades al precio de $2. (Recuerde que los
productores nacionales no pagan el arancel). A medida que el precio aumenta a $3, la cantidad ofrecida por los productores aumenta de 200 a 300 millones de metros. El resultado es una disminución de
las importaciones de 500 a 300 millones de metros.
Por último, la imposición del arancel significa que el gobierno recauda ingresos iguales al área
sombreada en la figura 20.4b. Esta área sombreada es igual a la tasa del arancel por unidad ($1) multiplicada por el número de unidades importadas desde de la entrada en vigor del arancel (300 millones
de metros). En consecuencia, los ingresos generados por el arancel ascienden a $300 millones.
¿Cuál es el resultado final del arancel? Los productores nacionales que recibían ingresos de sólo $2
por unidad antes de que se impusiera el arancel reciben ahora un precio más alto y ganan más utilidades. Sin embargo, estas utilidades más altas se logran a costa de una pérdida de eficiencia.
Las barreras comerciales impiden que una nación coseche los beneficios de la especialización,
ejercen presión para que se adopten técnicas de producción relativamente ineficientes y obligan a
los consumidores a pagar precios más altos por los productos protegidos de lo que pagarían en
circunstancias diferentes.
ARGUMENTOS A FAVOR DEL PROTECCIONISMO
Se pueden presentar también argumentos a favor de los aranceles y cuotas. En el transcurso de la historia de Estados Unidos, se han expuesto argumentos proteccionistas tantas veces por tantas industrias
ante tantos comités del Congreso que tal parece que todas las peticiones de protección comparten los
mismos temas. A continuación se describen las que se oyen con mayor frecuencia.
El proteccionismo salva empleos El principal argumento a favor del proteccionismo es
que la competencia extranjera cuesta empleos a los estadounidenses. Cuando éstos compran Toyotas,
los automóviles fabricados en Estados Unidos no se venden. Como consecuencia, hay despidos en la
industria automotora nacional. Cuando los estadounidenses compran acero japonés o alemán, los trabajadores del acero en Pittsburgh pierden sus empleos. Cuando los estadounidenses compran zapatos
o textiles de Corea o Taiwán, los obreros fabriles de Maine y Massachusetts, así como los del sur de California y Georgia, pierden sus empleos.
Es verdad que cuando compramos bienes de productores extranjeros, los productores nacionales
sufren. Sin embargo, no hay razón para creer que los trabajadores despedidos en los sectores en contracción no serán reempleados finalmente en otros sectores en expansión. La competencia extranjera
en textiles, por ejemplo, ha significado la pérdida de empleos estadounidenses en esa industria. Miles
de trabajadores de la industria textil en Nueva Inglaterra perdieron sus empleos cuando las fábricas de
textiles cerraron durante los últimos 40 años. No obstante, con la expansión de las industrias de alta
tecnología, el índice de desempleo en Massachusetts se redujo a uno de los más bajos del país a mediados de los años ochenta, y New Hampshire, Vermont y Maine también experimentaron un repunte
significativo. En la década de 1990, Nueva Inglaterra sufrió otra baja económica pronunciada, debido
en parte a las empresas manufactureras de hardware de alta tecnología que se mudaron al extranjero.
Pero a finales de los años noventa, su economía estaba floreciendo de nuevo, esta vez apoyada en lo que
se llamó la “Nueva Revolución Industrial”: el surgimiento de las empresas basadas en Internet.
El ajuste dista mucho de no tener costos. El conocimiento de que alguna otra industria, quizá en
alguna otra parte del país, pueda estar expandiéndose ofrece poco consuelo a las personas cuyas competencias se vuelven obsoletas o cuyas pensiones se pierden cuando su compañía cierra de pronto una
planta o se declara en quiebra. Los problemas sociales y personales que provoca el desempleo específico de una industria, las competencias obsoletas y la quiebra, como resultado de la competencia extranjera, son considerables.
Estos problemas pueden solucionarse de dos maneras. Podemos prohibir las importaciones y sacrificar las ganancias del libre comercio, si estamos dispuestos a pagar precios muy altos para salvar los
empleos nacionales en industrias que pueden producir de manera más eficiente en el extranjero, o podemos apoyar a las víctimas del libre comercio de una manera constructiva, ayudándolas a volver a capacitarse en trabajos que ofrezcan futuro. En algunos casos, los programas para trasladar personal a regiones en expansión pueden ser lo más conveniente. Algunos programas enfrentan directamente la
transición sin privarse de las ganancias del comercio.
Algunos países llevan a cabo prácticas de comercio injustas Los intentos de empresas estadounidenses por monopolizar una industria son ilegales de conformidad con lo previsto en
las leyes Sherman y Clayton. Si una compañía fuerte decide establecer precios inferiores al costo para
expulsar a la competencia del mercado, sería procesada con toda energía por la División Antimonopolios del Departamento de Justicia. Sin embargo, se argumenta, si no permitimos que una empresa estadounidense fije precios predatorios o monopolice una industria o mercado, ¿podemos hacernos a un
lado y dejar que una empresa alemana o japonesa lo haga en el nombre del libre comercio? Se trata de
un argumento legítimo y que ha ganado muchos partidarios en los últimos años. ¿Cómo debemos responder cuando una empresa internacional grande o un país se comporta de manera estratégica en
contra de una empresa o industria nacional? El libre comercio puede ser la mejor solución cuando todo el mundo se sujeta a las reglas, pero a veces es necesario defenderse.
La mano de obra barata hace que la competencia sea injusta Digamos que un
país particular adquirió su “ventaja comparativa” en textiles porque sus trabajadores reciben salarios
muy bajos. ¿Cómo pueden las compañías de textiles estadounidenses competir con empresas que pagan salarios que son inferiores a una cuarta parte de lo que se paga en las compañías estadounidenses?
Primero, recuerde que los salarios en una economía competitiva reflejan productividad: una proporción elevada de producción por unidades de mano de obra. Los trabajadores en Estados Unidos ganan salarios más altos porque son más productivos. Estados Unidos tiene más capital por trabajador,
es decir, el trabajador típico trabaja con mejor maquinaria y equipo, y sus trabajadores están mejor capacitados. Segundo, el comercio fluye no de acuerdo con la ventaja absoluta sino de acuerdo con la ventaja comparativa: todos los países se benefician, incluso si un país produce todo con mayor eficiencia.
El proteccionismo salvaguarda la seguridad nacional Aparte de salvar empleos,
ciertos sectores de la economía pueden solicitar protección por otras razones. Se ha argumentado durante años, y con cierto éxito, que la industria del acero es vital para la defensa nacional. En caso de una
guerra, Estados Unidos no querría depender de otros países para conseguir un producto tan vital como el acero. Incluso si reconocemos la ventaja comparativa de otro país, es posible que convenga más
proteger nuestros propios recursos.
Ninguna industria ha pedido protección jamás sin invocar el argumento de la defensa nacional.
En el testimonio a favor de la industria de las tijeras y las cizallas se argumentó que: “en el caso de una
emergencia nacional e interrupción de las importaciones, Estados Unidos se quedaría sin una fuente
de tijeras y cizallas, herramientas básicas para muchas industrias y oficios, esencial para nuestra defensa nacional”. La cuestión no reside en el mérito del argumento, sino en la seriedad con que puede tomarse si todas las industrias lo utilizaran.
El proteccionismo desalienta la dependencia La aseveración de que los países, en particular los pequeños o en vías de desarrollo, pueden llegar a depender demasiado de uno o más socios
comerciales para conseguir muchos artículos se relaciona de manera estrecha con el argumento de la
defensa nacional. Si un país pequeño llega a depender de una potencia para conseguir alimentos, energía o alguna materia prima importante en que la nación grande tenga una ventaja comparativa, puede
ser difícil que la nación pequeña siga siendo neutral en el terreno político. Algunos críticos del libre comercio argumentan que los países grandes, como Estados Unidos, Rusia y China, comercian de forma
deliberada con países pequeños para crear este tipo de dependencias.
Por lo tanto, ¿deben los países pequeños independientes evitar de manera conciente las relaciones
comerciales que podrían provocar dependencia política? Este objetivo puede implicar el desarrollo de
industrias nacionales en áreas donde un país tenga una desventaja comparativa. Hacerlo significaría
proteger a esa industria de la competencia internacional.
El proteccionismo salvaguarda las industrias incipientes Las industrias jóvenes
en un país determinado pueden tener dificultades para competir con las industrias establecidas en
otros países. En un mundo dinámico, una industria incipiente protegida podría madurar y convertirse
en una industria mundial fuerte debido a una ventaja comparativa adquirida, pero real. Si los competidores venden más barato que dicha industria y la expulsan de los mercados mundiales al principio de
su vida, esa ventaja comparativa podría nunca desarrollarse.
Sin embargo, los esfuerzos para proteger las industrias incipientes pueden resultar contraproducentes. En julio de 1991, el gobierno de Estados Unidos impuso un arancel de 62.67 por ciento a las
importaciones de pantallas de cristal líquido de matriz activa (también conocidas como “pantallas planas”, que se utilizan sobre todo para computadoras portátiles) de Japón. El Departamento de Comercio y la Comisión de Comercio Internacional acordaron que los productores japoneses estaban vendiendo sus pantallas en el mercado estadounidense a un precio inferior al costo y que este dumping
amenazaba la supervivencia de los productores nacionales de pantallas para computadoras portátiles.
El arancel tenía el propósito de proteger la industria incipiente norteamericana hasta que pudiera
competir frente a frente con la japonesa.
Por desgracia para los productores estadounidenses de computadoras portátiles y para los consumidores que las compraron, el arancel tuvo un efecto no deliberado (pero predecible) en la industria. Debido a que en general se reconocía que las pantallas estadounidenses para computadoras portátiles eran
de menor calidad que las pantallas japonesas, la imposición del arancel dejó tres opciones a los fabricantes estadounidenses de computadoras: (1) podían usar las pantallas que vendían los productores estadounidenses y observar cómo disminuían las ventas de su producto final ante la competencia de calidad
superior del exterior; (2) podían pagar el arancel de las pantallas de mayor calidad y observar cómo las
ventas de su producto final disminuían ante la competencia del precio más bajo del exterior; o (3) podían
hacer lo que era más rentable para ellos: mudar sus instalaciones de producción al extranjero y evitar el
arancel por completo. Lo último fue precisamente lo que tanto Apple como IBM hicieron. A final de
cuentas, no sólo la industria de las computadoras portátiles y sus consumidores resultaron afectados por
la imposición del arancel (debido a los costos mayores de producción y a los precios más altos de las computadoras portátiles), sino que la industria estadounidense de pantallas también se vio afectada (debido
a la pérdida de compradores de su producto) por una política diseñada específicamente a ayudarla.
UN CONSENSO ECONÓMICO
Ya sabe algo acerca de cómo encaja el comercio internacional en la estructura de la economía.
El debate entre los partidarios del libre comercio y el proteccionismo es crucial para nuestro estudio de la economía internacional. Por un lado está la teoría de la ventaja comparativa, formalizada por
David Ricardo en la primera parte del siglo XIX. Según este punto de vista, todos los países se benefician de la especialización y el comercio. Las ganancias del comercio son reales, y pueden ser cuantiosas; el libre comercio internacional aumenta los ingresos y mejora el nivel de vida.
Por el otro lado, se encuentran los proteccionistas, que señalan la pérdida de empleos y propugnan por la protección de los trabajadores contra la competencia externa. Aunque ésta puede ocasionar
pérdida de empleos en sectores específicos, no es probable que cause una pérdida neta de empleos en
una economía y los sectores en expansión absorberán con el tiempo a los trabajadores.
El comercio exterior y el empleo pleno pueden buscarse de manera simultánea. Aunque los economistas no están de acuerdo en muchas cosas, la gran mayoría favorece el libre comercio.
El capítulo 22 retoma este debate sobre el libre comercio.

### Resumen

1. Todas las economías, independientemente de su tamaño, dependen hasta cierto punto de otras y resultan afectadas por sucesos fuera de sus fronteras.
SUPERÁVIT Y DÉFICIT COMERCIAL
2. Hasta la década de 1970, Estados Unidos exportaba en general
más de lo que importaba: tenía un superávit comercial. A mediados de los años setenta, Estados Unidos empezó a importar más
mercancías de las que exportaba: un déficit comercial.
LA BASE ECONÓMICA DEL COMERCIO:
VENTAJA COMPARATIVA
3. La teoría de la ventaja comparativa, que se remonta a David Ricardo en el siglo XIX, sostiene que la especialización y el libre comercio benefician a todos los socios comerciales, incluso a
aquellos que pueden ser productores menos eficientes en términos absolutos.
4. Un país disfruta de una ventaja absoluta sobre otro en la producción de un producto si utiliza menos recursos para producir
ese bien que el otro país. Un país disfruta de una ventaja comparativa en la producción de un bien si éste puede producirse a
menor costo en términos de otros bienes.
5. El comercio permite a los países dejar atrás sus restricciones anteriores de recursos y productividad. Cuando los países se especializan en producir los bienes en los que tienen una ventaja
comparativa, maximizan su producción combinada y asignan
los recursos con mayor eficiencia.
6. Cuando el comercio es libre, las pautas y flujos comerciales son
resultado de las decisiones independientes de miles de importadores y exportadores y millones de familias y empresas privadas.
7. El atractivo relativo de los bienes extranjeros para los compradores estadounidenses y de los bienes estadounidenses para los
compradores extranjeros depende en parte del tipo de cambio, la
razón a la cual se intercambian dos divisas.
8. Por cada par de países, hay una gama de tipos de cambio que
conducen automáticamente a que ambos países hagan realidad
las ganancias de la especialización y la ventaja comparativa.
Dentro de esa gama, el tipo de cambio determinará qué país gana más del intercambio comercial. Esto nos lleva a concluir que
los tipos de cambio determinan los términos del comercio.
9. Si los tipos de cambio se ubican en el rango correcto, es decir,
dentro de un rango que facilite el flujo de bienes entre las naciones, el libre mercado obligará a cada país a trasladar recursos a
los sectores en los que disfruta de una ventaja comparativa.
Sólo los productos en los que un país tiene ventaja comparativa
serán competitivos en los mercados mundiales.
LOS ORÍGENES DE LA VENTAJA COMPARATIVA
10. El teorema de Heckscher-Ohlin se basa en la relativa dotación
de factores para explicar la ventaja comparativa y los flujos de
comercio. Según el teorema, un país tiene ventaja comparativa
en la producción de un bien si dicho país está relativamente
dotado de los insumos que se usan de manera intensiva en la
producción de ese bien.
11. Una lista relativamente corta de insumos (recursos naturales,
capital en conocimientos, capital físico, tierra y mano de obra
calificada y no calificada) explica una parte sorprendentemente
grande de los patrones del comercio mundial. Sin embargo, la
versión sencilla de la teoría de la ventaja comparativa no puede
explicar por qué muchos países importan y exportan los mismos productos.
12. Algunas teorías sostienen que la ventaja comparativa puede adquirirse. Así como las industrias dentro de un país diferencian
sus productos para captar un mercado interno, así también
diferencian sus productos para complacer a una amplia variedad de gustos en todo el mundo. Esta teoría concuerda con la
teoría de la ventaja comparativa.
BARRERAS COMERCIALES: ARANCELES,
SUBSIDIOS A LAS EXPORTACIONES Y CUOTAS
13. Las barreras comerciales adoptan muchas formas; las tres más
comunes son: aranceles, subsidios a las exportaciones y cuotas.
Todas son formas de protección de algún sector de la economía
contra la competencia externa.
14. Aunque Estados Unidos ha sido históricamente un país de aranceles altos, la tendencia general ahora es dejar atrás los aranceles
y cuotas. El Acuerdo General sobre Aranceles y Comercio (GATT),
suscrito por Estados Unidos y otros 22 países en 1947, sigue en
vigor en la actualidad; su propósito es reducir las barreras para
el comercio mundial y mantenerlas en un nivel bajo. También
son importantes el Tratado de Libre Comercio entre Estados Unidos y Canadá, firmado en 1988, y el Tratado de Libre Comercio
de América del Norte, firmado por Estados Unidos, México y
Canadá en los últimos días de la administración del Presidente
George H.W. Bush, que entró en vigor en 1994.
15. La Unión Europea (UE) es un bloque de libre comercio compuesto por 25 naciones, que se mencionan a continuación.
Otras cuatro han solicitado su inclusión: Bulgaria, Croacia, Rumania y Turquía. Muchos economistas creen que las ventajas
del libre comercio dentro de un bloque, la Alemania reunificada
y la capacidad de funcionar bien como grupo convertirán a la
UE en el actor más poderoso en el mercado internacional durante las próximas décadas.
¿LIBRE COMERCIO O PROTECCIONISMO?
16. En cierto sentido, la teoría de la ventaja comparativa es un argumento a favor del libre comercio. Las barreras comerciales impiden que una nación coseche los beneficios de la especialización,
ejercen presión para que se adopten técnicas de producción relativamente ineficientes y obligan a los consumidores a pagar
precios más altos por los productos protegidos de lo que pagarían en circunstancias diferentes.
17. Los argumentos a favor del proteccionismo se basan en diversas
propuestas, una de las cuales es que la competencia extranjera
provoca la pérdida de empleos nacionales, pero no hay razón
para creer que los trabajadores despedidos en los sectores en
contracción no se reemplearán en otros sectores en expansión.
Sin embargo, este proceso de ajuste dista mucho de no tener
costos.
18. Otros argumentos a favor de la protección sostienen que la mano de obra barata de otros países hace que la competencia sea
injusta; que algunos países tienen prácticas de comercio injustas
y que el proteccionismo salvaguarda la seguridad nacional, desalienta la dependencia y protege a las industrias incipientes. A
pesar de estos argumentos, la mayoría de los economistas favorecen el libre comercio.

## Macroeconomía de la economía abierta: balanza de pagos y tipo de cambio

Las economías del mundo se
han vuelto cada vez más interdependientes en las últimas cuatro
décadas. Ninguna opera en el vacío
y los acontecimientos económicos
de un país tienen repercusiones importantes en las economías de otras
naciones.
El comercio internacional es
una parte considerable de la economía mundial contemporánea. Las
importaciones estadounidenses suman alrededor de 15 por ciento del
producto interno bruto (PIB) de
Estados Unidos y todos los días se
mueven miles de millones de dólares en los mercados internacionales
de capital. En el capítulo 20 exploramos las principales razones de que haya comercio internacional. Los países comercian unos con
otros para obtener bienes y servicios que no producen o porque otras naciones los generan a
menor costo. En el diagrama de flujo circular de la figura 5.1 vimos las conexiones entre la economía nacional y el resto del mundo. Diversos países abastecen de bienes y servicios a la Unión
Americana, la cual también los aporta al resto del mundo.
Desde el punto de vista de la macroeconomía, la principal diferencia entre una transacción
internacional y una nacional está en el tipo de cambio:
Cuando personas de distintos países compran y venden de manera recíproca, también tiene que realizarse un intercambio de divisas.
Los exportadores de café brasileño no pueden gastar sus dólares estadounidenses en Brasil:
necesitan su moneda nacional. Una compañía estadounidense exportadora de trigo no puede
comprar con reales brasileños un tractor a una firma norteamericana ni pagar la renta de sus almacenes. De alguna manera, los intercambios internacionales tienen que manejarse de manera
tal que los dos socios de la transacción se vayan con su moneda local.
Como sabe por el capítulo 20, la dirección del comercio entre dos países depende de los tipos
de cambio, que son el precio de la moneda de un país en términos de la moneda de otro. Si el yen
japonés fuera muy caro (y abaratara el dólar), tanto japoneses como estadounidenses comprarían a
productores de Estados Unidos. Si el yen fuera muy barato (y encareciera el dólar), nipones y
norteamericanos comprarían a los productores de Japón. En cierta banda de tipos de cambio, el comercio se mueve en ambas direcciones, cada país se especializa en la producción de los bienes en
los que tiene una ventaja comparativa y el comercio es benéfico para cada parte.
Como los tipos de cambio constituyen un factor que determina el movimiento del comercio internacional, es muy importante la manera en que se establecen. Desde 1900, el sistema
monetario internacional ha cambiado varias veces por obra de acuerdos y acontecimientos internacionales. En la primera parte del siglo XX, casi todas las monedas estaban respaldadas por
el oro. Sus valores se fijaban en determinado número de onzas de ese metal, que determinaban
su valor en el comercio internacional; es decir, los tipos de cambio.
En 1944, con el caos del sistema monetario internacional hacia el final de la Segunda Guerra Mundial, un copioso grupo de expertos que representaban en forma extraoficial a 44 países
se reunieron en Bretton Woods, New Hampshire, y firmaron varios acuerdos. Uno de ellos fue el
establecimiento de un sistema de tipos de cambio esencialmente fijos con los cuales cada país
aceptaba mediar la compraventa de moneda en el mercado de divisas, cuando fuera necesario, a
fin de mantener el valor acordado a su moneda.
Pero en 1971, la mayoría de los países, incluyendo Estados Unidos, renunciaron al propósito de fijar de manera formal los tipos de cambio y dejaron que los determinaran la oferta y la
demanda. Por ejemplo, sin la intervención del gobierno en el mercado, el precio en dólares de
la libra inglesa está decretado por la interacción de los que quieren cambiar dólares por libras
(quienes “demandan” libras) y los que quieren canjear libras por dólares (quienes “ofrecen” libras).
Si la cantidad de libras demandada excede el monto de libras ofrecida, el precio de la moneda
británica aumenta, así como sube el precio de los cacahuates o los sujetapapeles en circunstancias semejantes. En el apéndice del capítulo se da una explicación más detallada de los sistemas
monetarios que han estado vigentes desde 1900.
En este capítulo exploraremos con mayor detalle lo que ha dado en llamarse macroeconomía
de la economía abierta. En primer lugar, describiremos la balanza de pagos, que es el registro de las
transacciones de un país con el resto del mundo. Luego pasaremos a considerar cómo cambia el
análisis cuando damos cabida al intercambio internacional de bienes, servicios y capital.
LA BALANZA DE PAGOS
A veces reunimos todas las monedas (franco suizo, yen japonés, real brasileño, etc.) bajo el nombre de “divisas”. Las divisas son todas las monedas, aparte de la nacional; en el caso de Estados
Unidos, el dólar. La demanda de divisas en ese país es resultado de que sus ciudadanos quieren
comprar cosas que tienen precios cotizados en otras monedas, como joyería australiana, vacaciones en México y bonos o acciones de la corporación Sony de Japón. Cuando los ciudadanos
estadounidenses hacen esas compras, australianos, mexicanos y japoneses ganan dólares, los
cuales, desde su punto de vista, son divisas.
¿De dónde viene la oferta de divisas? Muy simple: Estados Unidos (en realidad, sus ciudadanos o empresas) ganan divisas cuando venden productos, servicios o activos a otro país. Así como México gana divisas cuando los turistas estadounidenses visitan Cancún, Estados Unidos
también (en este ejemplo, pesos mexicanos) cuando los paseantes mexicanos visitan Disney
World en Estados Unidos. Del mismo modo, las compras sauditas de acciones de General Motors o las adquisiciones colombianas de bienes raíces en Miami aumentan la oferta de divisas en
la Unión Americana.
El registro de las transacciones de bienes, servicios y activos de un país con el resto del mundo
es su balanza de pagos. Ésta también es el registro de un país de las fuentes (oferta) y los usos (demanda) de divisas.1
LA CUENTA CORRIENTE
La balanza de pagos se divide en dos rubros principales, la cuenta corriente y la cuenta de capital.
Se muestran en la tabla 21.1, donde se proporcionan los datos sobre la balanza de pagos de Estados Unidos en 2004. Comencemos con la cuenta corriente.
El primer rubro de la cuenta corriente es el comercio estadounidense de bienes. En esta categoría se incluyen las exportaciones de chips de computadora, papas fritas y discos compactos de
músicos estadounidenses y las importaciones de whisky escocés, calculadoras japonesas y petróleo mexicano. Las exportaciones estadounidenses ganan divisas para Estados Unidos y aparecen
en el haber (+) de la cuenta corriente. Las importaciones de ese país gastan divisas y son rubros de
débito (–). En 2004, Estados Unidos importó bienes por 665,500 millones de dólares más de los
que exportó.
El siguiente rubro de la cuenta corriente son los servicios. Como en los demás países, Estados
Unidos compra y vende servicios al extranjero. Por ejemplo, una empresa estadounidense que envía trigo a Inglaterra podría comprar una póliza a una aseguradora inglesa. Un florista holandés
puede enviar flores a Estados Unidos en un avión estadounidense. En el primer caso, Estados
Unidos importa servicios y luego gasta divisas; en el segundo, vende servicios a extranjeros y cobra divisas. En 2004, Estados Unidos exportó 48,400 millones de dólares más en servicios de lo
que importó.
La diferencia entre las exportaciones de bienes y servicios de un país y sus importaciones de
éstos es su balanza comercial. Si las exportaciones de bienes y servicios son menores que las importaciones, un país tiene un déficit comercial. El déficit comercial de Estados Unidos en 2004
fue enorme: 617,100 millones de dólares (esto es, 665,500 millones menos 48,400 millones).
El tercer rubro de la cuenta corriente abarca las entradas de inversiones. Los ciudadanos estadounidenses tienen títulos del extranjero (acciones, bonos y bienes raíces, como edificios y fábricas). Dividendos, intereses, rentas y utilidades pagadas a los estadounidenses que tienen los títulos
son una fuente de divisas. Por el contrario, cuando extranjeros ganan dividendos, intereses y utilidades por activos de Estados Unidos, se gastan divisas. En 2004, las entradas por inversiones recibidas
de extranjeros superaron los ingresos pagados a los extranjeros por 24,100 millones de dólares.
El cuarto rubro de la tabla 21.1 son los pagos netos de transferencias. Los pagos de transferencias estadounidenses a extranjeros representan otro uso de las divisas. Algunos de estos pagos
son de particulares y otros, del gobierno. Alguien puede mandar un donativo de ayuda para África.
Por el contrario, algunos extranjeros hacen pagos de transferencia a Estados Unidos. El adjetivo
neto se refiere a la diferencia entre los pagos de Estados Unidos a extranjeros y los liquidaciones
de extranjeros a este país.
Si sumamos las exportaciones netas de bienes, exportaciones netas de servicios, entradas netas por inversiones y pagos netos de transferencias, tenemos el saldo de la cuenta corriente, que
muestra cuánto gastó la nación en bienes y servicios foráneos, pagos de inversiones y transferencias, con respecto a cuánto ganó de otros países. Cuando el saldo es negativo, como pasó con Estados Unidos en 2004, la nación gastó más en bienes y servicios foráneos (además de las inversiones y transferencias pagadas) de lo que ganó por la venta de sus bienes y servicios al resto del
mundo (además de las inversiones y transferencias recibidas). Si una nación gastó más en bienes
y servicios foráneos, pagos de inversiones y transferencias de lo que ganó, mengua su posición de
riqueza en comparación con el resto del mundo. Por neto queremos decir los acticos de una
nación en el extranjero menos sus obligaciones. La cuenta de capital de la balanza de pagos registra los cambios de estos activos y obligaciones. Ahora pasamos a la cuenta de capital.
LA CUENTA DE CAPITAL
Por cada transacción registrada en la cuenta corriente, se asienta una transacción compensatoria en
la cuenta de capital. Pensemos en un ciudadano estadounidense que compra un automóvil japonés.
Digamos que el tipo de cambio de las divisas es de 100 yenes por un dólar y que el precio del vehículo es de dos millones de yenes, que equivale a 20,000 dólares. El ciudadano estadounidense
(probablemente una distribuidora) toma 20,000 dólares, compra dos millones de yenes y adquiere
el automóvil. En este caso, las importaciones de Estados Unidos aumentan 20,000 dólares y su
cuenta corriente de activos foráneos (en este caso, Japón tiene dólares) aumenta 20,000 dólares en la
cuenta de capital. La posición de riqueza neta de Estados Unidos en relación con el resto del mundo
disminuyó 20,000 dólares. El punto importante del que hay que tomar nota es que un aumento de
las importaciones estadounidenses da por resultado un incremento de los activos foráneos en
Estados Unidos. Este país tiene que “pagar” las importaciones y aquello con lo que pague (en este
caso, dólares) es un acrecentamiento de los activos foráneos en Estados Unidos. Por el contrario,
una ampliación de las exportaciones estadounidenses trae un incremento de los activos de Estados
Unidos en el exterior porque los extranjeros tienen que pagar las exportaciones estadounidenses.
En la tabla 21.1 se muestra que los activos foráneos de Estados Unidos se dividen en títulos
de particulares (línea 6) y del gobierno estadounidense (línea 8). Del mismo modo, los activos foráneos de Estados Unidos se dividen en privados (línea 7) y de otros gobiernos (línea 9). La suma
de las líneas 6, 7, 8 y 9 es el saldo de la cuenta de capital (línea 10). El siguiente rubro son las
transacciones netas de la cuenta de capital (línea 11). Tiene un monto muy pequeño y abarca cosas
como la cancelación de deudas del gobierno estadounidense. Estas transacciones afectan la cuenta
de capital pero no la cuenta corriente. Si se ignorara este rubro y si no hubiera errores de medición
en el acopio de datos, el saldo de la cuenta de capital sería igual al negativo del saldo de la cuenta
corriente, porque, como dijimos, por cada transacción de la cuenta corriente hay una transacción
compensatoria en la cuenta de capital. Otra manera de considerar el saldo de la cuenta de capital
es decir que se trata del cambio de posición de la riqueza neta del país en comparación con el resto
del mundo. Si el saldo de la cuenta de capital es positivo, significa que el cambio de los títulos
foráneos en el país es mayor que el de los títulos del país en el extranjero, lo cual equivale a una
reducción de la posición de riqueza neta nacional.
En la tabla 21.1 se muestra que en 2004 el saldo de la cuenta corriente de Estados Unidos fue
de -665,900 millones de dólares, lo cual significa que el país gastó mucho más de lo que ganó en
relación con el resto del mundo. Si el saldo de la cuenta corriente se mide de manera adecuada,
entonces la posición de riqueza neta de Estados Unidos en relación con el resto del mundo disminuyó 665,900 millones de dólares en 2004 más los 1,500 millones de la línea 11. En este caso,
el saldo de la cuenta de capital debería ser de 667,400 millones. El saldo de la cuenta de capital
(línea 10) es, de hecho, 615,500 millones de dólares y por eso el error de medición, llamado
discrepancia estadística, fue de 51,900 millones de dólares (línea 12) en 2004. La balanza de pagos
(línea 13) es la suma del saldo de la cuenta corriente, el saldo de la cuenta de capital, las transacciones netas de la cuenta de capital y la discrepancia estadística. Por construcción, siempre da cero.
Es importante notar en la tabla 21.1 que aunque la posición de riqueza neta de Estados Unidos
disminuyó en 2004, el cambio de sus títulos en el extranjero aumentó bastante (821,800 millones
privados menos 4,100 billones del gobierno). ¿Por qué ocurre así? Porque hubo un incremento todavía mayor de los activos foráneos en Estados Unidos (un billón 77,900 millones privados más
355,300 millones del gobierno). Es el cambio neto (es decir, el cambio de activos foráneos en Estados Unidos menos el cambio de los activos estadounidenses en el exterior) lo cual es igual al saldo
negativo en la cuenta corriente (aparte de la discrepancia estadística), no el cambio tan sólo en los
títulos estadounidenses en el extranjero. Buena parte del incremento de 355,300 millones en títulos
foráneos del gobierno fue la acumulación de dólares de Japón y China.
Hay muchas transacciones que asentar en la cuenta de capital que no atañen a la cuenta corriente. Pensemos en un residente de Estados Unidos que compra un título en Inglaterra. Para esto,
el norteamericano intercambia dólares por libras y usa las libras para comprar el título inglés.
Después de la transacción, los activos estadounidenses en el extranjero aumentaron (Estados
Unidos tiene más títulos de Inglaterra) y los activos foráneos de Estados Unidos se incrementaron
(los extranjeros tienen más dólares). La compra del título inglés se asienta como rubro menos en la
línea 6 de la tabla 21.1 y el aumento de las posesiones de dólares en el extranjero se anota como
rubro más en la línea 7. Con esto las cifras se cancelan. Así ocurre siempre que haya un cambio de
una clase por otro en comparación con el resto del mundo. Cuando una compañía japonesa
compró en 1990 el Centro Rockefeller de Nueva York, fue un acrecentamiento de los activos de
extranjeros en Estados Unidos (el centro fue propiedad de japoneses) y un incremento de los activos foráneos de Estados Unidos (porque Estados Unidos tenía los yenes que se usaron para pagar el
centro). Si después Estados Unidos hubiera comprado con los yenes títulos japoneses, se habría
tratado nada más de un cambio de una clase de títulos foráneos en Estados Unidos (yenes) por
otra (títulos japoneses).
ESTADOS UNIDOS COMO NACIÓN DEUDORA
Si un país tiene una posición de riqueza neta positiva con relación al resto del mundo, se dice que
es una nación acreedora. Por el contrario, si tiene una posición de riqueza neta negativa, se dice
que es una nación deudora. Recuerde que la posición de riqueza neta de un país aumenta si es positivo su saldo de la cuenta corriente y disminuye si es negativo. Es importante percatarse de que
la única manera de que varíe la posición de riqueza neta de un país es que su saldo actual en la
cuenta corriente sea distinto de cero. Nada más mudar de una forma de activo a otra, como intercambiar el Centro Rockefeller por títulos japoneses en el ejemplo anterior, no modifica la posición de riqueza. Otra manera de decirlo es que la posición de riqueza neta de un país es la suma
de todos sus saldos pasados de la cuenta corriente.
En la primera mitad de los años setenta, Estados Unidos tenía por lo regular un superávit en la
cuenta corriente y, por lo tanto, su posición de riqueza neta era positiva. Era una nación acreedora.
Esto comenzó a dar un giro hacia la mitad de esa década y para mediados de la siguiente, Estados
Unidos tenía grandes déficits en la cuenta corriente. En algún momento de este periodo, la Unión
Americana pasó de una posición de riqueza positiva en relación con el resto del mundo, a una negativa. En otras palabras, Estados Unidos se transformó de acreedor en deudor. Los déficits de la
cuenta corriente se extendieron a los años noventa y ahora este país es el mayor deudor del mundo.
En 2003, los activos foráneos de Estados Unidos sumaron 10.5 billones de dólares y los activos estadounidenses en el extranjero, 7.9 billones.2 La posición de riqueza neta de Estados Unidos fue de
–2.6 billones de dólares. Esta gran posición negativa refleja el hecho de que Estados Unidos gastó
mucho más en los años ochenta, noventa y primera década del nuevo siglo en bienes y serviciosde fuera (más pagos de inversiones y transferencias) de lo que ganó por la venta de sus bienes y
servicios al resto del mundo (más cobros de inversiones y transferencias).
PRODUCCIÓN DE EQUILIBRIO (INGRESO)
EN UNA ECONOMÍA ABIERTA
Todo lo que hemos dicho hasta aquí ha sido descriptivo. Ahora pasamos al análisis. ¿Cómo se
determinan estos movimientos de comercio y capital? ¿Qué efecto tienen en la economía de los
países participantes? Para simplificar aquí nuestro examen, vamos a suponer que los tipos de
cambio son fijos. Más adelante eliminaremos esta suposición.
EL SECTOR INTERNACIONAL Y EL GASTO TOTAL PLANEADO
En nuestras descripciones anteriores del multiplicador tomamos en cuenta los hábitos de consumo de hogares (C), la inversión planeada de las empresas (I) y el gasto del gobierno (G). Definimos
la suma de estos tres componentes como el gasto total planeado (GT).
Para estudiar el sector internacional, tenemos que incluir los bienes y servicios que exporta un
país al resto del mundo, así como lo que importa. Si designamos EX a nuestras exportaciones de
bienes y servicios, debe quedar claro que EX es un componente de la producción y del ingreso total. Una rasuradora estadounidense vendida a un comprador en México es parte de la producción
de Estados Unidos tanto como una rasuradora equivalente vendida en Pittsburgh. Las exportaciones representan la demanda de productos nacionales no por hogares y empresas nacionales, ni el
gobierno, sino por el resto del mundo.
¿Qué pasa con las importaciones? Recuerde que las importaciones no son parte de la producción nacional (Y). Por definición, no las produce el país que las adquiere. Recuerde también que
cuando consideramos el consumo total de los hogares, el gasto total en inversión de las empresas y
el gasto total del gobierno, se incluyen importaciones. Por lo tanto, para calcular en forma adecuada
la producción interna tenemos que restar las partes del consumo, inversión y gasto gubernamental
que constituyan importaciones. La definición del gasto total planeado es:
Gasto total planeado en una economía abierta:
GT ≡ C + I + G + EX − IM
Los dos últimos términos, juntos (EX – IM), comprenden las exportaciones netas de bienes y
servicios.
Determinación del nivel de importaciones ¿Qué determina el nivel de importaciones y exportaciones de un país? Por ahora, suponemos que el nivel de importaciones está en función del ingreso (Y). La argumentación es simple: cuando el ingreso en Estados Unidos aumenta,
sus ciudadanos compran más de todo, incluyendo automóviles y mantequilla de cacahuate estadounidense, televisores japoneses y acero y reproductores de discos coreanos. Cuando el ingreso
aumenta, las importaciones suben. De manera algebraica:
IM = mY
donde Y es el ingreso y m es algún número positivo (se supone que m es menor que 1; de otro
modo, un incremento de un dólar en el ingreso genera un aumento de las importaciones de más
de un dólar, lo cual no es lógico). Recuerde que en el capítulo 8 vimos que la propensión marginal a consumir (PMC) mide el cambio en el consumo que resulta de un cambio de un dólar en el
ingreso. Del mismo modo, la propensión marginal a importar, abreviada PMI o m, es el cambio en
las importaciones causado por un cambio de una unidad monetaria en el ingreso. Si m = 0.2 o 20
por ciento y el ingreso es de 1,000 dólares, las importaciones, IM, son igual a 0.2 × 1,000 = 200 dólares. Si el ingreso aumenta 100, a 1,100 dólares, el cambio en las importaciones será igual a
m × (el cambio en el ingreso) = 0.2 × 100 = 20 dólares.
Por ahora vamos a suponer que las exportaciones (EX) son dadas (es decir, que el estado de
la economía no las afecta, ni en forma indirecta). Más adelante retiramos esta suposición.
Solución de equilibrio Dada la suposición sobre cómo se determinan las importaciones,
podemos resolver el ingreso de equilibrio. En la figura 21.1 se ilustra el procedimiento. A partir de
la función de consumo (línea punteada) de la figura 21.1(a), acumulamos poco a poco los componentes del gasto total planeado (línea negra). Si en aras de la simplicidad suponemos que la inversión
planeada, las compras del gobierno y las exportaciones son constantes y no dependen del ingreso,
con facilidad pasamos de la línea punteada a la negra sumando las cantidades fijas de I, G y EX al
consumo en todos los niveles de ingreso. En este ejemplo, hacemos que I + G + EX sea igual a 80.
En cambio, C + I + G + EX incluye el gasto en las importaciones, que no son parte de la producción interna. Para obtener el gasto en bienes producidos en el país, tenemos que restar el monto
de lo que se importa en cada nivel de ingreso. En la figura 21.1(b), suponemos que m = 0.25, que es
la suposición de que 25 por ciento del ingreso total se gasta en bienes y servicios producidos en otros
países. Con esta suposición, las importaciones son una fracción constante del ingreso total; por lo
tanto, con niveles mayores de ingreso se gasta más en bienes y servicios de fuera. Por ejemplo, con
Y = 200, IM = 0.25Y o bien 50. Del mismo modo, con Y = 400, IM = 0.25Y o bien 100. En la figura 21.1(b) se muestra la curva del gasto total planeado interno.
Se alcanza el equilibrio cuando el gasto total planeado interno es igual a la producción total interna (ingreso). Esto es verdadero sólo en un nivel de la producción total, Y* = 200, en la figura 21.1(b).
Si Y estuviera por debajo de Y*, el gasto planeado excedería la producción, las existencias serían menores que lo planeado y la producción aumentaría. Con niveles por encima de Y*, la producción excedería el gasto planeado, las existencias serían mayores que lo planeado y la producción caería.
Multiplicador de la economía abierta Todo lo anterior tiene implicaciones para el tamaño del multiplicador. Recuerde el multiplicador, que presentamos en el capítulo 8, y considere un
aumento sostenido en las compras gubernamentales (G). Al principio, el aumento de G hace que el
gasto total planeado sea mayor que la producción total. Las empresas nacionales ven que sus existencias bajan más que lo planeado y entonces acrecientan su producción; pero la producción adicional
significa más ingresos. Se contratan más trabajadores y las utilidades son mayores. Una parte del
nuevo ingreso se ahorra y otra se gasta. El consumo agregado trae una segunda ronda de baja de existencias mayor que lo planeado y de elevación de la producción. La producción de equilibrio crece en
un múltiplo del aumento inicial de las compras gubernamentales. Éste es el multiplicador.
En los capítulos 8 y 9 mostramos que el multiplicador simple es igual a 1/(1 – PMC) o (1 – PMA).
Es decir, un incremento sostenido de las compras del gobierno igual a ΔG genera un aumento de la
producción total (ingreso) de ΔG[1/(1 – PMC)]. Si la PMC fuera de 0.75 y las compras gubernamentales ascendieran 10,000 millones de dólares, el ingreso de equilibrio se elevaría 4 × 10,000 millones o
40,000 millones de dólares. El multiplicador es [1(1 — 0.75)] = [1/0.25] = 4.0.
En una economía abierta, una parte del aumento causado por el incremento en G se gasta en importaciones, en lugar de bienes y servicios de producción nacional. La parte del ingreso que se gasta en
importaciones no acrecienta el ingreso nacional (Y) porque las importaciones las producen extranjeros. Para calcular el multiplicador necesitamos saber cuánto del incremento del ingreso se dedicó al
mayor consumo interno. (Suponemos que todas las importaciones son de bienes de consumo. En la
práctica, algunas importaciones son de bienes de inversión y algunos bienes los compra el gobierno.)
En otras palabras, necesitamos conocer la propensión marginal a consumir bienes nacionales. El consumo interno es C – IM. Así, la propensión marginal a consumir bienes nacionales es la propensión
marginal a consumir todos los bienes (la PMC) menos la propensión marginal a importar (la PMI).
La propensión marginal a consumir bienes nacionales es (PMC – PMI). Por consiguiente:
multiplicador de la economía abierta = −

Si la PMC es 0.75 y la PMI es 0.25, el multiplicador es 1/0.5, o sea, 2.0. Este multiplicador es menor
que el multiplicador en el que no se toman en cuenta las importaciones, que es 1/0.25 o 4.0. La lección del modelo del multiplicador de la economía abierta, dicho de manera sucinta, es:
El efecto de un aumento sostenido en el gasto del gobierno (o inversión) sobre el ingreso (es
decir, el multiplicador) es menor en una economía abierta que en una cerrada. La razón: cuando el gasto (o la inversión) del gobierno se incrementa y se elevan el ingreso y el consumo, parte
del consumo adicional se hace en productos foráneos y no en bienes y servicios nacionales.
IMPORTACIONES Y EXPORTACIONES Y EL EFECTO
DE REALIMENTACIÓN DEL COMERCIO
En aras de la simplicidad, hasta este punto hemos supuesto que el nivel de las importaciones depende sólo del ingreso y que el nivel de las exportaciones es invariable. En realidad, el monto del gasto
en importaciones depende de otros factores, aparte del ingreso, y las exportaciones no son fijas.
Ahora vamos a pintar un cuadro más realista.
Determinantes de las importaciones Los mismos factores que inciden en los hábitos
de consumo de los hogares y las inversiones de las empresas afectan la demanda de importaciones,
ya que algunos bienes importados son bienes de consumo y otros son de inversión. Por ejemplo, todo
lo que aumente el gasto de consumo podría incrementar la demanda de importaciones. Vimos en los
capítulos 8 y 12 que en el consumo influyen factores como el salario real después de descontar impuestos, ingresos que no son laborales (después de impuestos) y tasas de interés; por consiguiente,
también deben afectar el gasto en importaciones. Del mismo modo, todo lo que acreciente el gasto
en inversión podría elevar la demanda de importaciones. Por ejemplo, una reducción de las tasas de
interés debe alentar el gasto de productos nacionales y bienes extranjeros.
Sólo hay otra consideración sobre la determinación del gasto en importaciones: los precios relativos de los bienes nacionales y extranjeros. Si los precios de los bienes producidos en el exterior
bajan en relación con los precios nacionales, la gente consumirá más bienes importados que del
país. Si los automóviles japoneses son más baratos que los estadounidenses, su consumo debe ser
mayor y viceversa.
Determinantes de las exportaciones Ahora eliminaremos nuestra suposición de
que las exportaciones son fijas. La demanda de exportaciones estadounidenses en otros países es
idéntica a su demanda de importaciones de Estados Unidos. Alemania importa bienes, una parte
de los cuales se produce en Estados Unidos. Francia, España, etc., hacen lo mismo. El gasto total de
importaciones de Alemania está en función de los factores que acabamos de explicar, salvo que las
variables son alemanas, no estadounidenses. Lo mismo pasa con los demás países. La demanda de
exportaciones estadounidenses depende de la actividad económica en el resto del mundo (los salarios reales, riqueza, ingresos no laborales, tasas de interés, etc., del resto del mundo), así como de
los precios de los bienes estadounidenses en relación con el precio de los bienes del resto del mundo.
Si la producción foránea aumenta, las exportaciones estadounidenses crecen. Éstas también se incrementan cuando sus precios bajan en relación con el resto del mundo.
Efecto de realimentación del comercio Ahora podemos combinar lo que sabemos
sobre la demanda de importaciones y de exportaciones para examinar el efecto de la realimentación
del comercio. Supongamos que hay un aumento de las exportaciones de Estados Unidos, quizá porque en el mundo de repente se decide que son preferibles las computadoras estadounidenses a las
de cualquier otro lado. El crecimiento de las exportaciones trae un incremento de la producción
(ingreso) en Estados Unidos, lo cual aumenta las importaciones de ese país. Aquí comienza la realimentación del comercio. Como las importaciones estadounidenses son las exportaciones de otro país,
la demanda adicional de las importaciones de la Unión Americana acrecienta las exportaciones del
resto del mundo. Cuando suben las exportaciones de otros países a Estados Unidos, también aumentan su producción e ingreso, lo cual lleva a un incremento de la demanda de importaciones en el resto
del mundo. Una parte de estas importaciones adicionales demandadas por el resto del mundo vienen
de Estados Unidos, así que las importaciones de este país crecen. El incremento de las exportaciones
estadounidenses estimula aún más la actividad económica de Estados Unidos, lo cual propicia un
nuevo aumento de la demanda estadounidense de importaciones, y así sucesivamente.
Un aumento de las importaciones estadounidenses incrementa las exportaciones de otros
países, lo cual estimula la economía de éstos y aumenta sus importaciones, y esto acrecienta
las exportaciones de Estados Unidos, que a su vez estimula la economía estadounidense y
aumenta sus importaciones, etc. Se trata del efecto de realimentación del comercio. En otras
palabras, un incremento de la actividad económica de Estados Unidos lleva a un acrecentamiento mundial de la actividad económica, que “realimenta” Estados Unidos.
PRECIOS DE IMPORTACIÓN Y EXPORTACIÓN
Y EL EFECTO DE REALIMENTACIÓN DE LOS PRECIOS
Hemos hablado del precio de las importaciones, pero todavía no estudiamos los factores que influyen en
los precios de ellas. Es complicado el análisis de los precios de las importaciones porque hay más de una
moneda. Cuando hablamos de “precio de las importaciones”, ¿estamos hablando en términos de dólares,
yenes, libras inglesas, pesos mexicanos o qué moneda? Como las exportaciones de un país son las importaciones de otro, puede hacerse la misma pregunta respecto al precio de las exportaciones. Cuando México exporta refacciones de automóviles a Estados Unidos, los fabricantes mexicanos están interesados
en el precio de esas piezas en pesos, porque en México las transacciones se hacen en pesos. Los consumidores estadounidenses quieren saber el precio de las refacciones en dólares, porque en Estados Unidos se
usan dólares con ese fin. El vínculo entre los dos precios es el tipo de cambio de dólares y pesos.
Supongamos que México pasa por una inflación y que el precio de los radiadores sube de 1,000
a 1,200 pesos cada uno. Si el tipo de cambio permanece estable en, digamos, 0.10 dólares por peso,
el precio de exportación de los radiadores mexicanos, en dólares, también aumenta, de 100 a 120
dólares. Como las exportaciones de México a Estados Unidos son por definición las importaciones
estadounidenses de México, un alza de los precios en dólares de las exportaciones mexicanas a la
Unión Americana significa un incremento de los precios de las importaciones estadounidenses de
México. Por lo tanto, cuando los precios de exportación de México suben sin variaciones en el tipo
de cambio entre dólares y pesos, los precios de las importaciones estadounidenses suben.
Los precios de las exportaciones de otros países afectan los precios de las importaciones de
Estados Unidos.
Los precios de las exportaciones de un país se mueven muy a la par del nivel general de precios de
ese país. Si México tiene un incremento general de precios, es probable que este cambio se refleje en alzas de precios de todos los bienes de producción nacional, tanto de exportables como no exportables.
La tasa general de inflación en el extranjero es probable que afecte los precios de las importaciones de Estados Unidos. Si la tasa de inflación en el extranjero es alta, los precios de las
importaciones de Estados Unidos podrían subir.
Efecto de realimentación de los precios Acabamos de ver que cuando un país experimenta un aumento de sus precios internos, suben los precios de sus exportaciones. También
es verdad que cuando los precios de las importaciones de un país se incrementan, los precios de los
bienes nacionales crecen. Este efecto ocurre por lo menos de dos maneras.
En primer lugar, un incremento en los precios de los insumos importados desplaza a la izquierda
la curva de la oferta agregada de ese país. En el capítulo 13 estudiamos la respuesta de la macroeconomía a los choques de los costos. Recuerde que un movimiento a la izquierda de la curva de la oferta agregada, debida a un aumento de los costos, hace que la producción total baje y que suban los
precios (estanflación).
En segundo lugar, si los precios de las importaciones se elevan en relación con los precios nacionales, los hogares sustituyen con bienes y servicios importados los de su país. Esto es equivalente a un
desplazamiento a la derecha de la curva de la demanda agregada. Si la economía nacional opera en la
parte ascendente de la curva de la oferta agregada, el nivel general de precios internos aumentará en
respuesta a un incremento de la demanda agregada. Las empresas perfectamente competitivas verán
que suben los precios determinados por el mercado. Las imperfectas atestiguarán un aumento en la
demanda de sus productos. Por ejemplo, se ha demostrado en estudios que el precio de los automóviles producidos en Estados Unidos se mueve de manera muy semejante al precio de los importados.
Pero no es el final del cuento. Supongamos que un país (digamos que México) pasa por un crecimiento de su nivel interno de precios. Esto eleva el nivel de los precios de sus exportaciones a Canadá
(y a todos los demás países). El aumento en el precio de las importaciones canadienses de México causa un incremento de los precios internos de Canadá. Esta nación también exporta a México. El alza de
los precios canadienses genera un incremento del precio de las exportaciones canadienses a México, lo
cual hace subir más el nivel de precios en México.
A esto se llama efecto de realimentación de los precios, en el sentido de que la inflación es
“exportable”. Un aumento del nivel de precios de un país puede elevar los precios de otros, lo cual
incrementa más el nivel de precios del primer país. Con los precios de importaciones y exportaciones, un alza de precios nacionales puede “realimentarse”.
Es importante darse cuenta de que, hasta este punto, hemos basado nuestro análisis en la suposición de que los tipos de cambio son fijos. La vida es más complicada con tipos de cambio flexibles,
a los que pasamos ahora.
ECONOMÍA ABIERTA CON TIPOS
DE CAMBIO FLEXIBLES
En buena medida, los tipos de cambio fijos determinados por los acuerdos de Bretton Woods sirvieron como tratados monetarios internacionales hasta 1971. Más tarde, ese mismo año, Estados Unidos
y la mayoría de los demás países decidieron abandonar el sistema de tipo de cambio fijo en favor de
tipos de cambio o flotantes determinados por el mercado. Los gobiernos todavía intervienen para asegurarse de que los movimientos de los tipos de cambio son “ordenados”, pero en la actualidad los tipos cambiarios están determinados en buena medida por las fuerzas libres de la oferta y la demanda.
Entender cómo se relaciona una economía con el resto del mundo cuando los tipos de cambio no
son fijos, no es tan fácil como cuando suponemos que los tipos son invariables. Los tipos de cambio
determinan los precios de los bienes importados en relación con los bienes nacionales y pueden tener
efectos importantes en el nivel de las importaciones y exportaciones. Tomemos una caída de 20 por
ciento en el valor del dólar frente a la libra inglesa. Los dólares compran menos libras y éstas pueden
adquirir más dólares. Los residentes en Inglaterra, que ahora tienen más dólares por sus libras, y los
residentes en Estados Unidos, que tienen menos libras por sus dólares, encuentran más atractivos
los bienes y servicios estadounidenses. Los movimientos del tipo de cambio tienen efectos significativos en las importaciones, exportaciones y el movimiento de capital entre los países.
EL MERCADO DE DIVISAS
¿Qué determina los tipos de cambio en un sistema de tipos de cambio flotantes? Para explorar la pregunta, supongamos aquí que sólo hay dos países, Estados Unidos e Inglaterra. Es más fácil entender
un mundo con dos naciones y casi todo lo que digamos aquí puede generalizarse a un mundo con
muchos socios comerciales.
Oferta y demanda de libras Gobiernos, particulares, bancos y corporaciones cambian libras por dólares y dólares por libras todos los días. En nuestro caso de dos países, los que demandan
libras tienen dólares que quieren cambiar por libras. Quienes ofrecen libras tienen libras que desean
canjear por dólares. Es importante no confundir la oferta de dólares (o de libras) en el mercado de
divisas con el circulante en Estados Unidos o Inglaterra. El circulante es la suma de todo el dinero
que se mueve hoy. La oferta de dólares en el mercado de divisas es el número de dólares que sus poseedores quieren cambiar por libras en determinado periodo. La demanda y la oferta de dólares en
los mercados de divisas determinan los tipos de cambio; la demanda de dinero salda y el circulante
total nacional determina la tasa de interés.
El motivo habitual para cambiar dólares por libras es comprar algo producido en Inglaterra.
Los importadores estadounidenses que compran automóviles Jaguar o whisky escocés deben pagar
con libras. Los estadounidenses que viajan a Inglaterra y que quieren trasladarse en tren, alojarse en
un hotel o comer en un restaurante tienen que comprar libras con dólares para hacerlo. Si una corporación estadounidense construye una planta en Inglaterra, tiene que pagarla en libras.
Asimismo, algunas personas quieren comprar acciones o bonos ingleses. En forma implícita,
cuando los estadounidenses compran un bono emitido por el gobierno inglés o por una corporación inglesa, hacen un préstamo, pero la transacción requiere divisas. En última instancia, al
vendedor del bono hay que pagarle en libras.
La situación se invierte del lado de la oferta del mercado. Aquí tenemos personas (por lo regular, ciudadanos ingleses) que tienen libras con las que quieren comprar dólares. También en
este caso el motivo habitual es comprar cosas producidas en Estados Unidos. Si un importador
inglés decide adquirir coches de golf hechos en Georgia, al fabricante hay que pagarle en dólares.
Los turistas ingleses que visitan Nueva York se mueven en taxis, comen en restaurantes y pasean
por la isla Ellis. Hacer esto requiere dólares. Cuando una empresa inglesa construye un complejo
de oficinas en Los Ángeles, tiene que pagar al contratista en dólares.
Además de compradores y vendedores que cambian dinero para realizar transacciones, algunas personas e instituciones tienen saldos de divisas por motivos de especulación. Si pienso que el
dólar estadounidense va a perder valor frente a la libra, quizá opte por tener parte de mi riqueza
en la forma de libras. En la tabla 21.2 se resumen las principales categorías de solicitantes y proveedores de divisas en el caso dos países: Estados Unidos e Inglaterra.

FACTORES QUE AFECTAN LOS TIPOS DE CAMBIO
Ahora sabemos lo suficiente para analizar los factores que quizá ejerzan una influencia en los tipos
de cambio. Todo lo que modifique el comportamiento de las personas en la tabla 21.2 puede hacer
que las curvas de la demanda y la oferta se desplacen y que el tipo de cambio se ajuste de manera
correspondiente.
Paridad del poder de compra: ley del precio único Si los costos de transportar
los bienes entre dos países son pequeños, esperamos que el precio de tales productos en ambos
sea más o menos igual. Por ejemplo, el precio de las pelotas de baloncesto debe ser casi el mismo
en Canadá que en Estados Unidos.
No es difícil ver por qué. Si el precio de los balones de basquetbol es más bajo en Canadá, a
alguien le convendrá adquirirlas allí y venderlas más caras en Estados Unidos. Esto reduce la
oferta y hace subir los precios en Canadá, mientras que en el otro país se incrementa la oferta y
bajan los precios. El proceso continúa mientras haya un diferencial de precios y, por consiguiente,
perdure la oportunidad de obtener utilidades. Si se trata de un bien con costos ínfimos de transporte, esperamos que sea válida esta ley del precio único. El precio de un bien debe ser el mismo
dondequiera que se compre.
Si la ley del precio único es válida para todos los productos y si cada país consume la misma
canasta de bienes, el tipo de cambio entre las dos monedas estará determinado por los niveles de precios de ambos países. Si el precio de una pelota de baloncesto fuera de 10 dólares en Estados Unidos
y 12 en Canadá, el tipo de cambio entre los dos sería de un dólar estadounidense por 1.20 dólares canadienses. Si en cambio el tipo fuera de uno por uno, a la gente le convendría adquirir los balones en
Estados Unidos y venderlos en Canadá. Esto aumentaría la demanda de dólares estadounidenses en
Canadá, lo cual subiría su precio en términos de dólares canadienses a un dólar estadounidense por
1.20 dólares canadienses, punto en el cual nadie obtendría ninguna ganancia de embarcar pelotas de
baloncesto por las fronteras internacionales, y el proceso se detendría.4
La teoría de que los tipos de cambio se ajustan de modo que el precio de bienes semejantes sea
el mismo en diversos países se conoce como teoría de la paridad del poder de compra. De acuerdo
con ella, si para comprar una libra de sal en México se necesitan 10 veces tantos pesos mexicanos
cuantos dólares se requieren para comprar una libra de sal en Estados Unidos, entonces el tipo de
cambio de equilibrio debe ser de 10 pesos por un dólar.
En la práctica, los costos de transporte de muchos bienes son bastante grandes y la ley del precio
único no opera con estos bienes (suele citarse la peluquería como buen ejemplo: los costos de transporte de un estadounidense para cortarse el pelo en Inglaterra son muy grandes, a menos que sea
piloto de una línea comercial). Asimismo, muchos productos que son sustitutos potenciales unos
de otros no son idénticos. Por ejemplo, un Rolls Royce y un Honda son automóviles, pero no hay
motivos para esperar que el tipo de cambio entre la libra inglesa y el yen japonés se fije de tal manera
que los dos bienes se equiparen. Además, los países consumen diferentes artículos en la canasta de
su mercado, así que no esperaríamos que los niveles de precios agregados obedecieran la ley del precio único. De cualquier manera:
Una tasa elevada de inflación en un país, comparado con otro, presiona el tipo de
cambio entre ambos y hay una tendencia general a que se deprecie la moneda del país
con más inflación.
En la figura 21.5 se muestra el ajuste que podría darse luego de un incremento en el precio de Estados Unidos en relación con el precio de Inglaterra. Este cambio de los precios relativos afecta a los
ciudadanos de los dos países. Los precios más altos en Estados Unidos hacen que las importaciones
sean relativamente más baratas. Los estadounidenses aumentan su gasto de bienes importados de
Inglaterra, lo cual desplaza a la derecha la curva de la demanda de libras, de D0 a D1. Asimismo, los

Tasas de interés relativas Otro factor que influye en el tipo de cambio de un país es el
nivel de su tasa de interés en comparación con la de otros países. Si la tasa de interés es de 6 por
ciento en Estados Unidos y de 8 por ciento en Inglaterra, la gente con dinero para prestar tendrá
un incentivo para comprar mejor títulos ingleses que estadounidenses. Aunque a veces es difícil
que personas de un país compren títulos de otro, resulta fácil para lo bancos internacionales y las
compañías de inversiones. Si la tasa de interés es menor en Estados Unidos que en Inglaterra, habrá un movimiento de fondos de los títulos estadounidenses a los ingleses, cuando bancos y empresas pasen sus fondos a los títulos que dan más rendimientos.
¿Cómo compra un banco estadounidense títulos ingleses? Toma sus dólares, compra libras
inglesas y con éstas adquiere los títulos ingleses. La obtención de libras del banco aumenta el precio de las libras en el mercado de divisas. El aumento de la demanda de libras eleva su precio (y
reduce el del dólar). Una tasa de interés más alta en Inglaterra en relación con la de Estados Unidos devalúa el dólar.

tipo de cambio entre la libra y el dólar. Las tasas altas de interés en Estados Unidos atraen inversionistas ingleses. Para comprar títulos estadounidenses, los ingleses necesitan dólares. La oferta
de libras (la demanda de dólares) se desplaza a la derecha, de O0 a O1. Las mismas tasas de interésrelativas afectan la cartera de opciones de bancos, empresas y hogares estadounidenses. Con mayores tasas de interés en el país, los residentes en Estados Unidos tienen menos incentivos para

curva de la demanda se desplaza a la izquierda, de D0 a D1. El resultado neto es una devaluación
de la libra y una revaluación del dólar. El precio de la libra baja de 1.89 dólares a 1.25.
EFECTOS DE LOS TIPOS DE CAMBIO EN LA ECONOMÍA
Ahora estamos listos para analizar algunas implicaciones de los tipos de cambio flotantes. Recuerde
que si las tasas son fijas, los hogares gastan más de su ingreso en importaciones y el multiplicador es
menor. Las importaciones son una “fuga” del flujo circular, como los impuestos y los ahorros. Por el
contrario, las exportaciones son una “inyección” en el flujo circular; representan gasto del extranjero en bienes y servicios producidos en Estados Unidos y pueden estimular la producción.
El mundo es mucho más complicado cuando se permite que las tasas de interés floten. En primer lugar, el nivel de importaciones y exportaciones depende de los tipos de cambio, así como del ingreso y otros factores. Cuando los sucesos ajustan los tipos de cambio, los niveles de importaciones y
exportaciones varían. Por su parte, las variaciones en las importaciones y exportaciones repercuten
en los niveles del PIB real y de los precios. Además, los tipos de cambio se ajustan también a los cambios de la economía. Supongamos que el gobierno decide estimular la economía con una política
monetaria de expansión. Esto influye en las tasas de interés, las cuales afectan los tipos de cambio.
Efectos de los tipos de cambio en importaciones, exportaciones y PIB real
Como ya sabemos, cuando la moneda de un país se devalúa (baja de valor), los precios de sus importaciones aumentan y se reducen los precios de sus exportaciones (en divisas). Si el dólar estadounidense es barato, los productos de Estados Unidos compiten mejor con los artículos del resto
del mundo y los bienes hechos en el extranjero les parecen más caros a los estadounidenses.
Una devaluación de la moneda puede servir de estímulo a la economía. Supongamos que el
valor del dólar baja, como ocurrió entre 1985 y 1988. Si compradores del extranjero aumentan su
gasto en bienes de Estados Unidos y los compradores locales sustituyen las importaciones con
bienes estadounidenses, el gasto total de la producción nacional se incrementa, las existencias bajan y el PIB real (Y) aumenta.
La devaluación de la moneda de un país es probable que aumente su PIB.5
Tipos de cambio y balanza comercial: la curva J Como una moneda que se devalúa aumenta las exportaciones y disminuye las importaciones, quizá piense usted que también
reduce el déficit comercial de un país. De hecho, el efecto de una devaluación en la balanza comercial es ambiguo.
Muchos economistas creen que cuando una moneda empieza a devaluarse, la balanza comercial empeora durante los primeros trimestres (quizá de tres a seis) Después, puede mejorar. Este
efecto aparece en la gráfica de la figura 21.7. La curva de la figura asemeja una letra J y el movimiento de la balanza comercial que describe se llama efecto de la curva J. El punto de la forma en
J es que la balanza comercial empeora antes de mejorar luego de una devaluación de la moneda.
¿Cómo se produce la curva J? Recuerde que la balanza comercial es igual a los ingresos de las
exportaciones menos los costos de las importaciones, incluyendo exportaciones e importaciones
de servicios:
balanza comercial precio en dólares de la = s exportaciones cantidad de exportacione × s
− × precio en dólares de las importaciones cantidad de importaciones
Una devaluación de la moneda afecta los elementos del lado de la derecha de la ecuación como
sigue. En primer lugar, la cantidad de las exportaciones aumenta y el volumen de las importaciones disminuye; las dos tienen un efecto positivo en la balanza comercial (lo cual aminora el déficit
comercial o eleva el superávit comercial). Segundo, no es posible que cambie mucho el precio en
dólares de las exportaciones, por lo menos no al principio. El precio en dólares de las exportaciones cambia cuando varía el nivel de precios en Estados Unidos, pero quizá no sea demasiado
grande el efecto inicial de una devaluación en el nivel de precios nacional. En tercer lugar, aumenta
el precio en dólares de las importaciones. Las importaciones a Estados Unidos son más caras
porque un dólar estadounidense compra menos yenes, euros, etc., que antes. Un incremento del
precio de las importaciones tiene un efecto negativo en la balanza comercial.
Un ejemplo aclarará este último punto: el precio en dólares de un automóvil japonés que
cuesta 1,200,000 yenes sube de 10,000 a 12,000 dólares si el tipo de cambio se mueve de 120 a 100
yenes por dólar. Después de la devaluación de la moneda, Estados Unidos termina gastando más
que antes (en dólares) por el vehículo japonés. Desde luego, al final Estados Unidos compra menos
automóviles a Japón que antes. ¿El número de automotores baja lo suficiente para que el efecto de
la cantidad sea mayor que el efecto del precio o es al contrario? ¿El valor de las importaciones aumenta o disminuye?
El efecto neto sobre la balanza comercial de una devaluación puede ir en un sentido o en el
otro. La devaluación estimula las exportaciones y reduce las importaciones, pero también incrementa el precio en dólares de las importaciones. Parece ser que primero domina un efecto negativo.
En general, las consecuencias de una devaluación sobre el precio de las importaciones se resienten
enseguida, pero toma tiempo para que las cantidades de exportaciones e importaciones respondan
a los cambios de precios. En lo inmediato, el valor de las importaciones crece más que el valor de
las exportaciones, así que la balanza comercial empeora. El efecto inicial quizá sea negativo; pero
cuando exportaciones e importaciones hayan tenido tiempo de reaccionar, el efecto neto se vuelve
positivo. Cuanto más elástica sea la demanda de exportaciones e importaciones, mayor será la
mejora final de la balanza de pagos.
Tipos de cambio y precios La devaluación de la moneda de un país aumenta el nivel de
precios. Hay dos razones para este efecto. En primer lugar, cuando la moneda de un país es menos
cara, sus productos son más competitivos en los mercados mundiales, así que las exportaciones
crecen. Además, los compradores nacionales reemplazan con productos locales las importaciones,
que ahora son más caras. Esto significa que aumenta el gasto total planeado en bienes y servicios
nacionales y que la curva de la demanda agregada se desplaza a la derecha. El resultado es un nivel
de precios más alto, mayor producción o ambos (si quiere, trace un diagrama OA/DA para verificar este resultado). Si la economía está cerca de su capacidad plena, podrían obtenerse precios más
altos. En segundo lugar, una depreciación hace que las materias primas importadas sean más caras.
Si los costos aumentan, la curva de la oferta agregada se desplaza a la izquierda. Si la demanda
agregada permanece sin cambios, el resultado es un incremento en el nivel de precios.
Política monetaria con tipos de cambio flexibles Vamos ahora a reunir todo lo
de este capítulo para considerar lo que ocurre cuando la política monetaria se usa primero para
estimular la economía y después para contraerla.
Supongamos que la economía está por debajo del pleno empleo y que la Reserva Federal decide ampliar el circulante. El volumen de las reservas del sistema se expande, quizá mediante compras de la Fed, en los mercados abiertos, de títulos del gobierno estadounidense. El resultado es
una disminución de la tasa de interés. La tasa de interés más baja estimula el gasto en inversión
planeada y el de consumo.
Este gasto adicional hace que las existencias sean menores que lo planeado y aumenta la producción total (ingreso, Y), pero hay otros dos efectos. Uno, que la tasa de interés más baja influye en
el mercado de divisas. Una tasa de interés reducida significa que los extranjeros demanden menos
títulos estadounidenses, así que la demanda de dólares decrece. Segundo, que es probable que los
gerentes de inversiones estadounidenses compren títulos del extranjero (que ahora pagan tasas más
altas), así que aumenta la oferta de dólares. Los dos acontecimientos menguan el valor del dólar.
Un dólar más barato es bueno si la meta de la expansión monetaria es estimular la economía nacional, ya que el dólar barato significa más exportaciones de Estados Unidos y menos
importaciones. Si los consumidores cambian las importaciones por bienes hechos en Estados
Unidos, tanto el aumento de las exportaciones como la reducción de las importaciones significan
que haya más gasto en productos domésticos, así que, de hecho, el multiplicador se acrecienta.
Ahora supongamos que la inflación causa problemas y que la Fed quiere bajarla estrechando
el control monetario. Aquí también sirve el tipo de cambio flotante. Una política monetaria estrecha funciona mediante la elevación de la tasa de interés. Una tasa alta aminora la inversión y el
gasto de consumo, reduce el gasto total, disminuye la producción y baja el nivel de precios. La tasa más alta también atrae compradores foráneos a los mercados financieros de Estados Unidos;
así se eleva el valor el dólar y se reduce el precio de las importaciones. La contracción del precio
de las importaciones desplaza a la derecha la curva de la oferta agregada, lo cual ayuda a combatir la inflación.
Política fiscal con tipos de cambio flexibles La apertura de la economía y los tipos de cambio flexibles no siempre operan para ventaja de las autoridades. Pensemos en una
iniciativa para reducir los impuestos y estimular la economía. Supongamos que el Congreso
promulga una importante contracción fiscal destinada a aumentar la producción. El gasto de
los hogares crece, pero no todo este gasto adicional es en productos nacionales: una parte se fuga de la economía estadounidense, lo cual reduce el multiplicador.
Cuando el ingreso se incrementa, lo mismo pasa con la demanda de dinero (Md); no la demanda de dólares en el mercado de divisas, sino las cantidades que quiere la gente para realizar
sus transacciones. Salvo que la Fed lo asimile por completo, la tasa de interés subirá. Una tasa
de interés más alta atrae demanda foránea de títulos estadounidenses. Este aumento de la demanda eleva el precio del dólar, lo cual mella todavía más la eficacia de la reducción de impuestos. Si el
valor del dólar se acrecienta, las exportaciones estadounidenses son menos competitivas en los
mercados mundiales y se reduce su volumen. Del mismo modo, un dólar fuerte abarata los bienes importados y los ciudadanos estadounidenses gastan más en artículos de fuera y menos en
bienes nacionales, un efecto que reduce más el multiplicador.
Hay otra advertencia con respecto a la explicación del multiplicador de los capítulos 8 y 9.
Si la Fed no asimila del todo los cambios, tres factores reducen el multiplicador: 1) una tasa de
interés más alta por el aumento de la demanda de dinero puede tener el efecto de expulsar la
inversión privada y el consumo, 2) una parte del aumento del ingreso debido a la expansión se
gastará en importaciones, y 3) una tasa de interés más elevada hace que el dólar se revalúe, lo
que desalienta las exportaciones y estimula más las importaciones.
Política monetaria con tipos de cambio flexibles Aunque la mayoría de los países tienen un tipo de cambio flexible (para nuestros propósitos aquí se cuentan los de la zona
del euro como uno solo), es interesante preguntarse qué función puede cumplir la política monetaria si un país tiene un tipo de cambio fijo. La respuesta es: ninguna. Para que un país tenga
un tipo de cambio invariable con respecto, digamos, al dólar estadounidense, su tasa de interés
no puede cambiar en relación con las tasas de Estados Unidos. Si las autoridades monetarias
del país redujeran la tasa de interés porque quisieran estimular la economía, su moneda se devaluaría (suponiendo que las tasas de interés en Estados Unidos no cambiaran). La gente querría
vender la moneda del país para comprar dólares e invertir en títulos estadounidenses, porque la
tasa de interés nacional bajó en relación con la de Estados Unidos. En otras palabras, las autoridades monetarias no pueden cambiar la tasa de interés en relación con la tasa estadounidense
sin tener que cambiar el tipo de cambio. Las autoridades monetarias están a merced de Estados
Unidos y no tienen medios independientes de cambiar su tasa de interés, si quieren mantener
su tipo de cambio fijo con respecto al dólar.
Esta restricción significa que cuando varios países europeos adoptaron en 1999 una moneda común, el euro, renunciaron a su política monetaria. Ahora sólo hay una política monetaria para los países incluidos en la zona del euro y es la que se decide en el Banco Central de
Europa.
El único caso en que un país puede modificar su tasa de interés y mantener fijo su tipo de
cambio consiste en que imponga controles al capital. Esto significa que el país limita o impide
que la gente compre o venda su moneda en el mercado de divisas. Por ejemplo, se prohíbe a un
ciudadano que compre dólares con su moneda nacional. El problema de los controles de capital
es que es difícil hacerlos valer, sobre todo en naciones grandes y durante mucho tiempo.
UNA ECONOMÍA MUNDIAL INTERDEPENDIENTE
La creciente interdependencia de los países en la economía mundial ha vuelto más difíciles los
problemas que enfrentan las autoridades. Acostumbrábamos pensar que Estados Unidos es una
región relativamente autosuficiente. Hace 40 años, los acontecimientos económicos fuera de
las fronteras estadounidenses tenían pocos efectos en su economía. La situación ya no es así.
Los hechos de las últimas cuatro décadas nos han enseñado que el desempeño de la economía
estadounidense depende mucho de lo que sucede al otro lado de sus fronteras.
En este capítulo y en el anterior dimos apenas los elementos fundamentales de la macroeconomía de mercados abiertos. Si usted continúa el estudio de la economía, hay más que agregar
a la historia básica que hemos presentado.
Acerca del campo internacional, en el capítulo 22 nos ocupamos de los problemas de los
países en desarrollo y las economías en transición de no capitalistas a capitalistas.

### Resumen

1. La principal diferencia entre una transacción internacional y una
operación interna estriba en el cambio de divisas: cuando las
personas de países diferentes compran y venden entre ellas,
también tiene que verificarse un intercambio de monedas.
2. El tipo de cambio es el precio de la moneda de un país en
términos de la moneda de otro.
LA BALANZA DE PAGOS
3. Las divisas son todas las monedas aparte de la moneda nacional.
El registro de las transacciones de bienes, servicios y activos de
un país con el resto del mundo es su balanza de pagos. Ésta es
también el registro de las fuentes (insumos) de un país y los
usos (demanda) de las divisas.
PRODUCCIÓN DE EQUILIBRIO (INGRESO) EN UNA
ECONOMÍA ABIERTA
4. En una economía abierta, una parte del ingreso se gasta en
bienes manufacturados en el exterior, en lugar de artículos
nacionales. Para medir el gasto total planeado interno en una
economía abierta, sumamos las exportaciones totales y restamos
las importaciones totales: C + I + G + EX – IM. La economía
abierta está en equilibrio cuando la producción total interna
(ingreso, Y) es igual al gasto total planeado interno.
5. En una economía abierta, el multiplicador es igual a 1/[1 – (PMC –
PMI)], donde PMC es la propensión marginal a consumir. La
propensión marginal a importar es el cambio en las importaciones
causado por una variación de una unidad monetaria en el ingreso.
6. Además del ingreso, otros factores que afectan el nivel de las
importaciones son la tasa salarial real después de impuestos, el
ingreso no laboral después de impuestos, las tasas de interés y los
precios de los productos nacionales en relación con los extranjeros.
La demanda de exportaciones está determinada por la actividad
económica del resto del mundo y sus precios relativos.
7. Un aumento de la actividad económica estadounidense lleva
a un incremento de la actividad económica mundial, que
luego “realimenta” a Estados Unidos. Una expansión de las
importaciones estadounidenses acrecienta las exportaciones
de otros países, lo cual estimula su economía e impulsa sus
importaciones; esto, a su vez, eleva las exportaciones de Estados
Unidos, lo que estimula la economía estadounidense y amplía sus
importaciones, etc. Es el efecto de realimentación del comercio.
8. Los precios de exportación de otros países afectan los precios de
importación de Estados Unidos. La tasa general de inflación en el
extranjero podría influir también en los precios de las importaciones estadounidenses. Si la inflación en el extranjero es alta, los
precios de las importaciones podrían incrementarse.
9. Como las exportaciones de un país son las importaciones de
otro, un aumento en el precio de las exportaciones incrementa
los precios de las importaciones de otros países. Un alza de los
precios de las importaciones de otros países genera un
crecimiento en sus precios internos y en sus precios de
exportación. En síntesis, los precios de las exportaciones
influyen en los de las importaciones y viceversa. El efecto de
realimentación de los precios demuestra que la inflación es
“exportable”: un aumento del nivel de precios de un país puede
hacer subir los precios de otros, lo cual empeora la inflación en
el primero.
ECONOMÍA ABIERTA CON TIPOS DE CAMBIO FLEXIBLES
10. El tipo de cambio de equilibrio ocurre cuando la cantidad
demandada de una moneda en el mercado de dividas es igual
a la cantidad ofrecida de esa moneda en el mismo mercado.
11. La devaluación de una moneda ocurre cuando la moneda de
una nación baja de valor en relación con la de otro país.
La revaluación de una moneda ocurre cuando el valor de la
moneda de una nación aumenta en relación con la moneda
de otra.
12. De acuerdo con la ley del precio único, si los costos de transporte
son pequeños, el precio del mismo bien en diferentes países
debe ser casi el mismo. La teoría de que el tipo de cambio se fija,
de modo que el precio de bienes similares en diferentes países
sea igual, se llama teoría de la paridad del poder de compra. En la
práctica, los costos de transporte de muchos bienes son cuantiosos
y la ley del precio único no es válida con éstos.
13. Una tasa elevada de inflación en un país respecto de otro
presiona el tipo de cambio entre ambos. Hay una tendencia
general a devaluar las monedas de los países con mayor inflación.
14. Una devaluación del dólar tiende a aumentar el PIB de Estados
Unidos porque abarata sus exportaciones (lo hace más competitivo
en el exterior) y porque las importaciones son más caras en ese
país (lo cual alienta a los consumidores a cambiar por los bienes
y servicios de producción nacional.
15. No está claro el efecto de una devaluación de la moneda de un
país en su balanza comercial. En el corto plazo, la devaluación
de la moneda aumenta el déficit de la balanza comercial, porque
incrementa el precio de las importaciones. Aunque este efecto
reduce la cantidad de bienes importados, la secuela de una
devaluación en el precio de las importaciones se resiente enseguida,
pero pasa tiempo para que las cantidades de exportaciones e
importaciones respondan a las variaciones de precios. El
resultado inicial quizá sea negativo, pero cuando exportaciones
e importaciones han tenido tiempo para responder, el efecto
neto se vuelve positivo. Se conoce como efecto de la curva J a la
tendencia a ensancharse y reducirse del déficit de la balanza
comercial a resultas de una devaluación monetaria.
16. La devaluación de la moneda de un país aumenta el nivel de
precios por dos causas. En primer lugar, la devaluación incrementa
el gasto total planeado, un resultado que desplaza la curva de la
demanda agregada a la derecha. Si la economía está cerca de
su plena capacidad, es probable que resulten precios más altos.
En segundo lugar, una devaluación encarece los insumos
importados. Si los costos aumentan, la curva de la oferta
agregada se desplaza a la izquierda. Si la demanda agregada
permanece sin cambios, el resultado es un incremento del nivel
de precios.
17. Cuando los tipos de cambio son flexibles, una política monetaria
de expansión reduce la tasa de interés y estimula la inversión
planeada y el gasto de consumo. La menor tasa de interés reduce
la demanda de títulos estadounidenses entre los extranjeros y
acrecienta la demanda de títulos foráneos por parte de los
gerentes estadounidenses de fondos de inversión. Como resultado,
el dólar se devalúa. Una política monetaria de contracción
revalúa el dólar.
18. Los tipos de cambio flexibles no siempre funcionan como desean
las autoridades. Una política fiscal de expansión puede revaluar el
dólar y reducir el multiplicador.

## Globalización

Globalización es una palabra que se
usa con frecuencia en la actualidad . Una búsqueda revela que el término no se mencionó nunca en las páginas del New York Times en la década de
1970; sin embargo, ahora aparece en los
artículos del Times casi a diario. El motor de búsqueda Google devuelve 55
millones de enlaces en los que se usa la
palabra. Pese al hecho de que la palabra
se use tan a menudo, rara vez se define
con claridad. La globalización es el proceso de aumentar la interdependencia
entre los países y sus ciudadanos.
Definida así, la globalización tiene dimensiones sociales, culturales,
políticas y económicas. Incluye intensificar el comercio internacional, el
turismo, la divulgación de películas y
música, la increíble disponibilidad de la información en todo el mundo a través de Internet, el problema de las enfermedades, el cambio climatológico global, el terrorismo, la religión, la guerra, etcétera.
Cualquier intento por abordar todos los asuntos relacionados con la creciente interdependencia en un
solo capítulo, o incluso en un solo libro, sería inútil.
Un tema incluso más candente en 2005 y 2006 es el outsourcing, palabra que figura en 83 millones
de enlaces en Google. El outsourcing se refiere por lo general a la práctica de obtener bienes o servicios
de fuentes externas. En la mayoría de los casos, la palabra se usa para describir la contratación por parte de empresas estadounidenses de trabajadores extranjeros que se quedan en sus propios países, pero
que se comunican y realizan tareas por teléfono y por Internet. Los críticos argumentan de manera
contundente que esta práctica equivale a exportar los empleos a otros países y pagar salarios muy bajos a trabajadores extranjeros en lugar de contratar trabajadores bien pagados en Estados Unidos.
El propósito aquí es explorar el proceso de la globalización económica. La globalización económica es el proceso de aumentar la interdependencia económica entre los países y sus ciudadanos. Aunque
los temas que se analizarán tienen dimensiones políticas y culturales, nos centraremos en las causas y
consecuencias del creciente comercio internacional de bienes y servicios, la gran cantidad de movimientos entre fronteras de la mano de obra y el aumento de los flujos financieros internacionales con
el fin de apegarnos lo más posible a la economía.
EL FLUJO CIRCULAR GLOBAL
La figura 22.1 inserta el resto del mundo en el diagrama de flujo circular introducido en el capítulo 3.
Evidentemente, es más complejo. El diagrama básico de una economía cerrada sencilla incluye la oferta y la demanda de bienes y servicios, la oferta y la demanda de mano de obra y la oferta y la demanda
de capital (ahorro e inversión). El ingreso familiar proviene del trabajo y de la propiedad de activos; las
empresas demandan mano de obra y capital y ofrecen productos; las familias demandan bienes y servicios y ofrecen mano de obra y capital.
La apertura de la economía para incluir al resto del mundo añade diez nuevos flujos al diagrama.
Éstos se muestran en la figura 22.1: (A.) Las familias nacionales (en este caso, estadounidenses) ahora
pueden comprar bienes y servicios del exterior: importaciones. Esto produce supuestamente mayor
variedad y precios potencialmente más bajos. (B.) Las empresas nacionales ahora pueden vender en los
mercados exteriores: exportaciones. Esto abre el potencial de oportunidades para obtener utilidades.
(C.) Las empresas pueden contratar trabajadores de otros países, ya sea mediante el empleo de inmi
grantes o el outsourcing de tareas o responsabilidades a trabajadores extranjeros, en esencia, a través
de la importación de mano de obra. (D.) Las empresas estadounidenses ahora pueden conseguir capital en el resto del mundo. Una empresa estadounidense puede financiar una nueva fábrica con un
préstamo de un banco alemán o la venta de bonos en Londres. (E.) Hoy, los ciudadanos estadounidenses pueden buscar empleo en los mercados exteriores de mano de obra. Un trabajador de alta tecnología despedido tiene la posibilidad de buscar trabajo en Canadá. (F.) Las familias nacionales pueden invertir sus ahorros en acciones y bonos extranjeros. Durante periodos de tasas de interés bajas en
Estados Unidos, las tasas más altas en otros países podrían parecer atractivas.
Pero eso no es todo: (G.) Los trabajadores extranjeros pueden trabajar en Estados Unidos. Miles de
inmigrantes mexicanos, muchos legales y muchos ilegales, entran en Estados Unidos cada semana en
busca de trabajo. (H.) Los ciudadanos extranjeros pueden comprar acciones y bonos estadounidenses o
depositar dinero en bancos de Estados Unidos. Una de las razones del enorme auge del mercado de valores a finales de los años noventa fue que los extranjeros querían comprar acciones de empresas tecnológicas de Estados Unidos. (I.) Las empresas extranjeras pueden demandar mano de obra y ofrecer empleos a los trabajadores de Estados Unidos. Por último, (J.) las empresas extranjeras pueden conseguir
fondos en Estados Unidos para invertirlos como nuevo capital. Una compañía de teléfonos italiana podría financiar la adquisición de nuevo equipo de conmutación con un préstamo de Citibank.
El argumento básico a favor de la globalización económica se encuentra en las decisiones en las
que se basan cada uno de estos flujos. Un consumidor estadounidense compra un bien o servicio de un
productor extranjero si el bien que va a comprar es más barato de lo que tendría que pagar por él en
Estados Unidos o si es de mayor calidad. Asimismo, los ciudadanos estadounidenses invierten sus ahorros en bonos extranjeros si éstos les dan un rendimiento superior al de un bono parecido en Estados
Unidos. Estas decisiones se toman en función del interés personal.
Piense que Estados Unidos es un conjunto de 50 estados independientes que están integrados por
completo en términos económicos. La gente, el capital financiero y los bienes y servicios fluyen con libertad entre fronteras en respuesta a las preferencias de la gente. Si no se permitiera que la gente se
moviera de manera libre entre fronteras, o si no se permitiera que los ciudadanos de Nueva Jersey
compraran acciones de empresas de California, o si se prohibiera la exportación de pescado de las costas al interior, el sistema sería sin lugar a dudas ineficiente. La eficiencia en cualquier economía, incluso
la mundial, se logra si el capital y la mano de obra pueden moverse con libertad adonde su productividad sea mayor… si los factores de producción se ponen a trabajar en sus “mejores y más altos usos”.
Pero, un momento. Si fuera así de sencillo, ¿por qué tanto alboroto? ¿Por qué hay protestas tan
violentas contra el proceso de globalización? Algunos sostienen que el comercio provoca desempleo
nacional y salarios más bajos, ya que los trabajadores extranjeros compiten con los nacionales. Otros
aseguran que la industria estadounidense no puede competir con los productos producidos en países
que tienen mano de obra muy barata. Otros más, argumentan que la apertura económica conduce a
normas de trabajo injustas y condiciones de explotación en muchos países o que la apertura origina
una feroz competencia y degradación ambiental. A algunos tan sólo les disgusta la expansión del capitalismo, que consideran de por sí injusta. Exploraremos una a una de las objeciones después de estudiar brevemente la historia de la globalización.
BREVE HISTORIA DE LA GLOBALIZACIÓN ECONÓMICA
El historiador económico Jeffrey Williamson clasifica el periodo de 1820 a 1914 como el primer gran
periodo de globalización y el periodo desde la Segunda Guerra Mundial como el segundo. El comercio
como porcentaje del producto interno bruto (PIB) del mundo aumentó de 5 por ciento durante los
primeros años del siglo XIX a 20 por ciento a principios de la Primera Guerra Mundial. Las barreras comerciales que se impusieron después de 1914, las guerras mundiales y la Gran Depresión, conspiraron
para reducir de nuevo el comercio como un porcentaje del PIB mundial a alrededor de 5 por ciento.
Sin embargo, el establecimiento de un nuevo orden financiero en Bretton Woods, New Hampshire, en
1946, marcó el inicio de una nueva era y el comercio volvió a aumentar al 20 por ciento del PIB mundial hacia 1975.
En 1919, el joven John Maynard Keynes dijo:
¡Qué extraordinario episodio en el progreso económico del hombre fue aquella era que llegó
a su fin en agosto de 1914!… Un habitante de Londres podía ordenar por teléfono, mientras
bebía su té matutino en la cama, los diversos productos de toda la Tierra, en la cantidad que
considerara conveniente, y esperar, dentro de lo razonable, la entrega pronta y oportuna en
su puerta; en el mismo momento y por el mismo medio, podía arriesgar su riqueza en los recursos naturales y las nuevas empresas en cualquier confín del mundo y participar, sin esfuerzo o siquiera dificultad, de sus posibles frutos y ventajas.1
Aunque es difícil obtener los datos, muchos argumentarían que con respecto a la movilidad internacional de la mano de obra, estamos menos globalizados ahora de lo que estábamos a principios del siglo XX. En esa época, aproximadamente 15 por ciento de la población de Estados Unidos había nacido
en otros países. La Oficina del Censo estableció esta cifra en 10.4 por ciento en 2000, lo que representa
un aumento con respecto a 7.9 por ciento en 1990. Como es evidente, la inmigración ha aumentado en
los últimos años, pero no queda claro si nos estamos aproximando a las tasas de los años veinte.
Se responsabilizó a los flujos internacionales de capital por los dramáticos sucesos de 1997 en
Asia. Los tipos de cambio volátiles, ocasionados por el rápido retiro de préstamos y créditos a Indonesia, Corea, Tailandia y otros países de Asia Oriental provocaron fluctuaciones drásticas en los tipos de
cambio y recesiones muy graves en esos países.
Muchas dimensiones de la globalización son nuevas en la actualidad. En los últimos años se han observado reducciones bruscas en las barreras comerciales con la firma del Tratado de Libre Comercio de
América del Norte (TLCAN) y el Acuerdo General sobre Aranceles y Comercio (GATT), mencionados en
capítulos anteriores. Los aumentos espectaculares en los flujos de información y comercio por Internet y
la mayor velocidad y menor costo de viajar han convertido el mundo en un lugar mucho más pequeño
y mucho más consciente de las diferencias culturales, políticas y religiosas. Uno de los grandes temas de
la elección presidencial de 2004, que también lo es de un libro que ha sido todo un éxito de ventas, escrito
por Thomas Friedman en 2005, el outsourcing, se ha convertido en un asunto político muy importante.
LOS BENEFICIOS Y COSTOS DE LA GLOBALIZACIÓN
Se ha criticado a la globalización económica desde muchos frentes. Para poner en claro los argumentos, primero revisaremos el debate sobre el libre comercio en bienes y servicios. En seguida atenderemos el debate sobre la movilidad de la mano entre países: la inmigración. Por último, analizaremos los
problemas únicos de los mercados financieros globales.
RECONSIDERACIÓN DEL DEBATE SOBRE EL LIBRE COMERCIO
Los posibles beneficios del comercio se han examinado varias veces en este texto. La historia sencilla de
la ventaja comparativa se analizó brevemente en el capítulo 2 y más adelante con mayor detalle en el
capítulo 20. El argumento, formalizado por primera vez por David Ricardo en el siglo XIX, es simple,
pero poderoso, y ha resistido la prueba del tiempo. El argumento se basa en dos intuiciones. Primero,
si usted tiene algo que yo quiero y yo tengo algo que usted quiere, podemos hacer un intercambio y
ambos salimos ganando. El intercambio voluntario es eficiente. Segundo, si se aplica el concepto a las
naciones, dos naciones pueden beneficiarse del libre comercio si se especializan en la producción de
los bienes en los que disfrutan de una ventaja comparativa. Un país disfruta de una ventaja comparativa en la producción de un bien si la producción de dicho bien tiene un costo de oportunidad menor
del que tendría si se produjera en otro país.
En el capítulo 2 y de nuevo en el capítulo 20, demostramos cómo el comercio entre dos países
permitía a ambos liberarse de las restricciones de sus fronteras de posibilidades de producción. Demostramos que esto sigue aplicándose incluso si un país es más eficiente en la producción de todos los
bienes. Para que resulte aún más sencillo de entender, si los consumidores pueden adquirir el mismo
producto que compran en su país a un menor costo en los mercados internacionales, ¿cuál es la lógica
de comprar en el mercado nacional o de “proteger” los bienes y servicios producidos en el país contra
la competencia del exterior?
Como hemos visto, quienes se oponen al comercio plantean una serie de argumentos. El primero
y más importante es que los bienes producidos en el país se producen con trabajadores nacionales y
que comprar importaciones o practicar el outsourcing simplemente exporta los empleos a otros países. Además, ¿cómo se puede competir con países donde se pagan salarios muy bajos o los obreros trabajan en condiciones de explotación? No sólo se pierden empleos nacionales, sino que se contribuye a
que continúen las condiciones laborales por debajo de la norma en los países pobres.
Los defensores del libre comercio contraponen varios argumentos:
1. No podemos comprar a otros países a menos que ellos nos compren a la vez. Hace varios años, el
candidato presidencial Ross Perot se opuso al TLCAN aduciendo que se oiría el “ensordecedor
ruido de una enorme aspiradora” a medida que todos los buenos empleos en Estados Unidos
eran absorbidos por México, donde los trabajadores ganan salarios muy bajos. Piense en lo que
pasaría si se produjera este cambio.
Si los ciudadanos estadounidenses quisieran comprar todo en México, pero los mexicanos
no quisieran comprar nada en Estados Unidos, los mercados de cambio de divisas impedirían
que esto ocurriera. A los trabajadores mexicanos se les paga en pesos. Para comprar un bien pro-
ducido en México, el comprador tiene que obtener pesos. Pero si los mexicanos no quisieran
comprar nada en Estados Unidos, no habría nadie dispuesto a intercambiar dólares por pesos.
Dicho en términos sencillos, no habría oferta de pesos en los mercados de cambio de divisas, pero habría una gran demanda de esta divisa. Como resultado, el precio del peso (es decir, el tipo de
cambio) aumentaría.
Ahora piense detenidamente en lo que significa un peso más caro. Puesto que los compradores
estadounidenses tienen que comprar pesos para comprar productos mexicanos, los productos mexicanos se encarecen más. Asimismo, el aumento en el precio de los pesos (dólares por peso) hace que
los productos estadounidenses resulten más baratos para los mexicanos. Esto continúa hasta que la
oferta y la demanda de pesos sean iguales. Eso significa que los mexicanos encontrarán algunos productos más baratos en Estados Unidos. De hecho, las exportaciones a México aumentaron de 46,000
millones de dólares en 1995, poco después de que el TLCAN entró en vigor, a 111,000 millones de
dólares en 2000. El tipo de cambio se ajusta para compensar el diferencial en salarios.
En este caso hay una excepción importante. Un país puede comprar más productos a otro
durante un periodo prolongado si dichas compras se equilibran con las compras de acciones y
bonos. China, por ejemplo, vendió más de 10,000 millones de dólares más cada mes en bienes y
servicios a Estados Unidos durante 2005 que lo que Estados Unidos compró a China. Para equilibrar eso y mantener constante el valor de la moneda china, China acumuló dólares y los usó para comprar acciones y bonos estadounidenses. El resultado fue un enorme déficit comercial para
Estados Unidos. Esto ha sido tema de numerosas charlas sobre comercio entre Estados Unidos y
China en los últimos años. Por último, los ciudadanos chinos deben poder comprar bienes y servicios para disfrutar de los beneficios de sus enormes exportaciones.
2. Proteger a una industria contra la competencia extranjera o tratar de prohibir a las empresas que
practiquen el outsourcing para salvar empleos costará otros empleos en los sectores que se expandirían con el libre comercio o los costos reducidos. Desde luego, cuando el comercio se abre, algunos sectores se ven afectados por las importaciones más baratas. De hecho, ése es precisamente el
punto: conseguir productos a precios más bajos. Si gastamos menos en verduras de América del
Sur, podemos gastar más en otros productos en nuestro país. Además, la demanda de nuestros socios comerciales estimulará el crecimiento en la producción de nuestros productos de exportación.
Si se reduce el comercio con otro país, también se reduce en número de empleos en los sectores de
exportación. Del mismo modo, si las empresas pueden reducir los costos con el outsourcing, la
competencia trasladará esos ahorros a los consumidores en la forma de precios más bajos. Los consumidores, a su vez, podrán comprar bienes y servicios de otros sectores, un beneficio que expandirá y creará nuevos empleos para mantener el empleo de la población económicamente activa.
3. Proteger una industria puede provocar ineficiencia y falta de capacidad para competir después
en los mercados mundiales. Las industrias que están sujetas a la competencia rigurosa del exterior tienden a adoptar la tecnología más nueva y eficiente y a mantener los costos al mínimo. La
industria del acero de Estados Unidos estuvo fuertemente protegida durante muchos años y, a la
larga, la vigorosa competencia de los productores de otros países le quitó la mayoría de la clientela mundial del acero.
4. Mantener el índice de desempleo en un nivel bajo es un problema macroeconómico. Cuando la
economía se expande, el desempleo disminuye y puede ser bajo sin importar la composición del
comercio. El índice de desempleo más bajo en Estados Unidos durante décadas fue en el año
2000. En ese año, luego de la expansión del comercio gracias al TLCAN y el GATT, el desempleo
promedió apenas 4 por ciento, el índice más bajo desde 1969. Las herramientas correctas para
combatir el desempleo son las políticas fiscales y monetarias y no las políticas contra el comercio.
5. Si el objetivo es reducir la pobreza, ¿en qué ayuda impedir el comercio? Abrir un país al comercio
aumenta la demanda de mano de obra en los sectores de exportación. En muchos países la demanda interna es escasa porque el ingreso es bajo. Si la demanda de exportaciones logra aumentar el empleo, el ingreso aumentará, no bajará.
Un segundo argumento en contra del libre comercio es que daña el medio ambiente. El principal
fundamento de este argumento es que los países pobres no están sujetos a las mismas normas ambientales que los países ricos. Las normas ambientales poco rigurosas reducen el costo de hacer negocios
para los productores de un país. Así, los países tienen un incentivo para hacer caso omiso del ambiente y ser más competitivos en los mercados mundiales.
Los partidarios de la globalización responden con dos argumentos. Primero, dicen, la regulación
ambiental es problema de todos los países, sin importar si tienen o no relaciones comerciales. Los países pobres, donde la gente gana salarios de subsistencia, no cuentan con los recursos necesarios para
dedicarlos a la protección ambiental. No hace falta más que visitar una ciudad importante en un país
pobre para convencerse de que el medio ambiente es, en general, un problema. Pero la verdadera espe-
ranza de un mejor ambiente es el crecimiento y un gobierno responsable. Sin lugar a dudas, el mejoramiento ambiental es un bien suntuario. Alimentar a la ciudadanía es primero y mejorar el medio ambiente viene después. Segundo, el comercio puede ser incluso una palanca para aumentar la protección
ambiental. La Organización Mundial de Comercio (OMC) y el TLCAN tienen negociaciones en curso
destinadas a presionar a los países para que mejoren las leyes ambientales.
Un tercer argumento contra el libre comercio es que las reglas de éste y el poder de las organizaciones como la OMC pueden debilitar la soberanía nacional porque obligan a los países a sacrificar varias cosas. Esto, por supuesto, es verdad. El ejemplo clásico es un subsidio a la agricultura. Los cabildeos agrícolas en muchos países son fuertes. Con los adelantos en la productividad, es cada vez más
difícil, en especial para los pequeños agricultores, ganarse la vida trabajando la tierra. En muchos países, incluido Estados Unidos, se pagan subsidios considerables a algunos agricultores, lo que les permite producir a costo más bajo. Puesto que los agricultores pueden producir a menor costo, disfrutan de
una ventaja significativa sobre los agricultores que producen en países donde no hay subsidios. Esto se
ha convertido en un enorme problema entre los países desarrollados y en vías de desarrollo.
Un último problema es el debate sobre los alimentos modificados genéticamente (MG) o los organismos modificados genéticamente. Muchas empresas están experimentando con variedades de alimentos que se han modificado genéticamente por diversas razones. Tales cultivos se producen a escala
comercial en más de 40 países. Los principales cultivos que se han modificado genéticamente incluyen
soya, maíz y algodón resistentes a plagas y herbicidas. Entre otros cultivos MG se encuentran el arroz
con más hierro y vitaminas y una batata resistente a un virus que podría diezmar buena parte de las cosechas africanas. Se han modificado genéticamente algunos microorganismos para que produzcan enzimas que ayudan a fabricar mejores vinos.
El problema es que no se sabe todo acerca de los posibles peligros de los organismos recién creados. Pueden ser nocivos. Varios países europeos han prohibido en el pasado la importación de cultivos
MG a pesar de su bajo costo. Como una concesión, en 2003 la Comunidad Europea optó por cancelar
la prohibición, pero insistió en la etiqueta obligatoria de cualquier producto que contenga aunque sea
una cantidad mínima de un ingrediente modificado de manera genética.
Los productores de Estados Unidos han llevado el caso ante la OMC, aduciendo que es sólo otra forma de impedir que los productos estadounidenses, menos costosos, ingresen al mercado y de subsidiar de
nuevo a los agricultores europeos. Los partidarios del libre comercio en productos modificados genéticamente aseguran que aunque los beneficios son claros y manifiestos, los costos son especulativos. A la fecha, no hay ninguna prueba de que los cultivos MG hayan causado efectos secundarios perjudiciales.
COMERCIO, CRECIMIENTO Y POBREZA
Buena parte de lo se ha dicho hasta el momento es teoría. Sin duda, hay argumentos teóricos que pueden exponerse a favor de eliminar las barreras comerciales. Pero, ¿y las pruebas? ¿Las naciones que están más abiertas al comercio crecen con mayor rapidez que los países que están menos abiertos? ¿La
pobreza ha disminuido a ritmo más acelerado en los países que comercian de manera más abierta?
Existen ejemplos claros de economías abiertas que han crecido con rapidez y donde la pobreza se ha
reducido, como China. También hay ejemplos de países que comercian abiertamente que han registrado poco crecimiento e índices crecientes de pobreza, como Haití. Sin embargo, si se controlan otros
determinantes de la pobreza y el crecimiento, ¿el comercio es una ventaja o una desventaja?
La Oficina Nacional de Investigación Económica y el Banco Mundial patrocinaron una serie de
estudios de la liberalización del comercio en las décadas de 1970 y 1980. La mayoría de ellos mostró
que los países que estaban más integrados a la economía mundial crecieron con mayor rapidez que los
que estaban menos integrados2. Un estudio citado a menudo de Jeffrey Sachs y Andrew Warner, concluye que los países abiertos crecieron 2 por ciento anual más rápido que las economías cerradas, mediante el control de una serie de factores3. Aunque algunos estudios ponen en tela de juicio la solidez
de la relación,4 la mayoría de las pruebas respaldan la aseveración de que la liberalización del comercio
fue un componente importante de las políticas reformadas que condujeron a un crecimiento más acelerado. Incluso uno que cuestionó la solidez de la relación admite: “Ningún país se ha desarrollado de
forma satisfactoria al darle la espalda al comercio internacional y los flujos de capital a largo plazo”.5
Pero, ¿el crecimiento reduce necesariamente la pobreza o todos los beneficios del crecimiento van
a parar a los bolsillos de los ricos? Por supuesto, como el economista Stanley Fischer afirma en un ensayo reciente sobre la globalización: “La lógica indica que no hay modo de sacar de la pobreza a las poblaciones de los países pobres sin un crecimiento sostenido. A nivel global, la disminución de la pobreza ha sido más rápida donde el crecimiento ha sido también más rápido (en los países en vías de
desarrollo de Asia) y más lenta donde el crecimiento registrado ha sido muy escaso (en África)”.6 Los
estudios empíricos también verifican que, en promedio, cuando los países crecen, el ingreso de quintil
más bajo de la distribución del ingreso aumenta aproximadamente a la misma tasa que el ingreso agregado, incluyendo cuando se induce el crecimiento con más comercio abierto.7
LA GLOBALIZACIÓN DE LOS MERCADOS DE TRABAJO:
LA ECONOMÍA DE LA INMIGRACIÓN Y EL OUTSOURCING
La globalización también se refiere a la mayor movilidad de la mano de obra entre fronteras internacionales. Siempre que el desempleo aumenta, el problema de la inmigración ocupa las primeras planas
de los periódicos. Además, pocos temas suscitan respuestas más exaltadas. En 1994, los electores en California aprobaron por un margen de 3 a 2 la Propuesta 187, un referendo sumamente controvertido
que impide que los inmigrantes ilegales reciban del estado educación, asistencia social y prestaciones
médicas no de emergencia. En el mismo año, el gobernador Lawton Chiles de Florida declaró una
emergencia en todo el estado y exigió que se detuviera a las decenas de miles de inmigrantes cubanos
que llegaban a Florida. El Presidente Clinton respondió enviando a la Guardia Costera para interceptarlos y conducirlos a un campo de detención en la Base Naval de Guantánamo en Cuba. Poco después, algunos fueron enviados a Panamá y a otros se les permitió la entrada a Estados Unidos. Estos
acontecimientos despertaron recuerdos del éxodo del Mariel, en 1980, como resultado del cual más de
125,000 refugiados llegaron a Florida. Muchos de ellos, delincuentes que aún se encuentran en cárceles de Estados Unidos.
¿Qué impacto tiene la inmigración en la economía de un país? ¿Todo es malo, como muchos quisieran hacernos creer? Resulta interesante que uno de los elementos fundamentales del acuerdo entre
los miembros de la Unión Europea es la abolición de los controles fronterizos para estimular el libre
flujo de la mano de obra entre los países miembros en respuesta a los diferenciales en salarios y las condiciones económicas. La relajación de las leyes de inmigración en Europa se implementó en aras de fomentar el crecimiento económico y la prosperidad.
Breve historia de la inmigración en Estados Unidos La inmigración hacia Estados
Unidos se ha producido en oleadas irregulares. La primera “Gran migración” ocurrió entre 1880 y
1924, cuando 25.8 millones de inmigrantes ingresaron al país, cifra que representó más de 40 por ciento del aumento demográfico total del periodo. Sin embargo, durante la década de 1920, el Congreso estableció un sistema de cuotas de origen nacional que limitó el flujo anual de los países del Hemisferio
Oriental a 150,000 personas. De conformidad con las nuevas leyes, se expidieron visas en proporción
con la composición étnica de Estados Unidos en 1920. El resultado fue que 60 por ciento de las visas se
otorgaron a inmigrantes alemanes e ingleses.
El flujo de inmigrantes se redujo significativamente en los años treinta, pero ha aumentado desde
entonces. Entre 1990 y 2000 la inmigración promedió poco más de 1 millón al año. Muchos de los inmigrantes son ilegales en la actualidad. Hay estudios que indican que a finales de los años ochenta entre
dos y tres millones de personas vivían ilegalmente en Estados Unidos y de 200,000 a 300,000 nuevos
ilegales llegaban cada año. Aunque la Patrulla Fronteriza intercepta y devuelve aproximadamente 1.3
millones de ilegales al año, muchos inmigrantes ilegales ingresan al país legalmente. Entran legalmente por las aduanas y tan sólo se quedan en el país.
En 1986 el Congreso promulgó la Ley de Reforma y Control de la Inmigración, que otorgaba amnistía a alrededor de 3 millones de ilegales e imponía un conjunto de sanciones severas a los empleadores, con la intención de desacelerar el flujo de inmigrantes hacia Estados Unidos. La Ley de Inmigración de 1990 aumentó en 150,000 el número de inmigrantes legales admitidos en el país.
Antes de 1960, el grupo más grande de inmigrantes que llegaba a Estados Unidos venía de Europa. Entre 1991 y 2000 el grupo más grande de inmigrantes provino de América, en especial de México. A
partir de 1970, el número de inmigrantes de Asia comenzó a crecer con rapidez. La tabla 22.1 muestra el
número de inmigrantes de los países de origen principales en 2004.
Argumentos económicos a favor de la libre inmigración ¿Debe un país permitir la
inmigración completamente libre dentro de sus fronteras? El argumento a favor de la libre inmigración es que aumenta la producción mundial. La mano de obra fluye entre fronteras en respuesta a diferenciales en salarios. Considere el caso de México y Estados Unidos. Los trabajadores mal pagados de
México emigran a Estados Unidos porque los salarios son más altos en ese país. Si los mercados son en
esencia competitivos, los salarios reflejan la productividad de los trabajadores. En otras palabras, debido a que Estados Unidos tiene más capital y usa tecnología más avanzada que México, la productividad
de los trabajadores de bajos salarios es mayor en Estados Unidos que en México. Así, la misma mano de
obra lleva a cabo más producción total después de la inmigración y la producción mundial aumenta.
Ahora considere el caso de Francia e Italia. Si existiera escasez de mano de obra en Francia porque
la demanda de vino francés aumentó, los salarios franceses se incrementarían y atraerían trabajadores
de otros países europeos. Si, al mismo tiempo, la demanda de productos de cuero producidos en Italia
disminuyera, los salarios italianos se reducirían y los trabajadores italianos se mudarían a Francia,
donde su productividad sería mayor.
El argumento a favor del libre movimiento de la mano de obra entre países es exactamente el mismo que el argumento a favor del libre movimiento de la mano de obra entre los sectores de la economía interna. Suponga que una economía produce sólo dos bienes, X y Y. Si la demanda del bien X repunta, la demanda de la mano de obra que se utiliza para producir X aumenta, porque las ganancias de
contratar más mano de obra en la producción de X están en aumento. La mano de obra saldrá de la
producción del bien Y sólo si su productividad es mayor en X en términos del valor de la producción.
Este movimiento garantiza la eficiencia. Recuerde la definición sencilla de que una economía eficiente
es la que produce lo que la gente quiere al menor costo.
Quienes están a favor de una política más liberal hacia los inmigrantes creen que éstos no desplazan a los trabajadores estadounidenses, sino que más bien aceptan empleos que los estadounidenses
simplemente no quieren. Los inmigrantes sirven como trabajadores domésticos y agrícolas que perciben salarios bajos y producen cosas que Estados Unidos necesita. Además, la economía de Estados
Unidos ha absorbido oleada tras oleada de inmigrantes, al tiempo que ha mantenido casi el empleo
pleno. Cerca de la totalidad de los ciudadanos de Estados Unidos, excepto los nativos americanos, tienen antepasados relativamente recientes que llegaron a ese país como inmigrantes.
El argumento en contra de la libre inmigración Ningún economista discute la idea de
que es probable que la distribución del ingreso cambie entre países y entre grupos dentro de cada país
en respuesta a la inmigración. Suponiendo que los inmigrantes son trabajadores que perciban salarios
bajos, los salarios de equilibrio en el mercado de la mano de obra escasamente calificada aumentarán
en el país de destino y con ellos aumentarán las utilidades, mientras que el ingreso de capital se reducirá en el país de origen.
El argumento a favor de la libre inmigración supone que todos los trabajadores obtienen empleo.
Sin embargo, la impresión popular es, sin lugar a dudas, que los inmigrantes (que por lo general trabajan por salarios muy bajos) les quitan empleos a los estadounidenses de bajos ingresos y aumentan los
índices de desempleo. Además, muchos creen que los inmigrantes acaban con frecuencia viviendo de la
asistencia social y se convierten en una carga para los contribuyentes. Los oponentes también señalan
la delincuencia en los vecindarios étnicos y las rivalidades entre grupos raciales como prueba de los
costos adicionales para la sociedad.
Las pruebas: los costos netos de la inmigración Para determinar si los beneficios de
la inmigración superan los costos netos, es preciso plantearse una pregunta importante: ¿hasta qué
punto la inmigración reduce los salarios nacionales y aumenta el desempleo? Una serie de estudios recientes ha concluido que las zonas metropolitanas que tienen mayores cantidades de inmigrantes parecen tener salarios apenas inferiores e índices de desempleo un poco más altos.
Un influyente estudio del economista David Card de la Universidad de California en Berkeley observa con detalle los salarios y las oportunidades de empleo en la zona metropolitana de Miami durante y después la crisis del Mariel en 1980. Casi de la noche a la mañana, alrededor de 125,000 cubanos llegaron a Florida y aumentaron la fuerza laboral de Miami en más de 7 por ciento. Card estudió las
tendencias en los salarios y el desempleo entre los trabajadores de Miami entre 1980 y 1985 y no encontró prácticamente ningún efecto. Además, los datos que examinó reflejaban la experiencia de los trabajadores en Los Ángeles, Houston, Atlanta y ciudades similares que no sufrieron la misma conmoción.8
Sin embargo, un estudio más reciente de Borjas, Freeman y Katz discrepa de buena parte del trabajo realizado hasta la fecha. Ellos argumentan que los inmigrantes no se quedan en las ciudades a las
que llegan, sino que más bien se desplazan dentro de Estados Unidos en respuesta a las oportunidades
de trabajo y los diferenciales en salario. Así, sostienen que los efectos de la inmigración en los salarios
y el desempleo deben analizarse a nivel nacional y no a nivel de cada ciudad. Su estudio señala la baja
pronunciada en los salarios de las personas que abandonaron el bachillerato sin graduarse en relación
con los trabajadores que tenían más educación durante la década de 1980. Sus resultados indican que
una tercera parte de la caída en los salarios relativos de las personas que no completaron el bachillerato puede atribuirse a los inmigrantes poco calificados.9
Respecto al asunto de los efectos de la inmigración en los costos del gobierno, también existen datos mixtos. Es evidente que generaciones anteriores de inmigrantes han tenido un efecto positivo tanto
en la economía en su conjunto como en los presupuestos gubernamentales de manera más específica.
Estudios de las pautas salariales de los primeros inmigrantes parecen mostrar que los salarios rebasaban
en promedio los salarios de los trabajadores nativos después de 15 años. Por tanto, los inmigrantes de
primera generación, como grupo, podrían pagar más en impuestos de lo que cobran en prestaciones
que se otorgan previa comprobación de los recursos de una persona, como la asistencia social.
Sin embargo, los datos demuestran que a través del tiempo ha ocurrido un descenso radical en el
nivel de educación, experiencia y competencias entre los inmigrantes. Asimismo, la participación en los
programas de asistencia pública entre los inmigrantes se ha disparado. Borjas calcula que en 1990, las familias de inmigrantes en Estados Unidos contribuyeron entre 7,600 y 10,100 millones de dólares en impuestos y cobraron 23,800 millones de dólares en prestaciones de los programas de asistencia social.
¿La inmigración es buena o mala? La inmigración es otra de esas cuestiones económicas
en las que no se impone ninguna respuesta clara. Las pruebas sobre los efectos de la inmigración son
mixtas y la teoría nos da argumentos tanto en pro como en contra de la cuestión. Sólo el tiempo dirá si
oleada reciente de inmigrantes se asimilará tan bien como las generaciones anteriores. Mientras tanto,
la inmigración seguirá siendo un tema “sensible” en el ámbito político y Estados Unidos tendrá que tomar algunas decisiones importantes acerca del trato que se les da a los inmigrantes recientes.
En 2001, después de los ataques del 11 de septiembre, la inmigración se asoció con los temas del
terrorismo internacional. Con el traslado del Servicio de Inmigración y Naturalización al recién creado Departamento de Seguridad Nacional, los solicitantes de visas, incluso las de estudiante, se someten
a un escrutinio mucho más riguroso. Esta nueva política ha suscitado temores de que se perderán algunos de los beneficios que la inmigración trae consigo.
Outsourcing Ahora analizaremos el outsourcing, uno de los mayores temas en el debate de la
globalización de Estados Unidos en la actualidad. La inmigración lleva a los trabajadores a los empleos
en Estados Unidos. El outsourcing manda los empleos a los trabajadores que perciben bajos salarios en
lugares tan distintos como Bulgaria, India y Filipinas.
Al principio, el número de contratos de outsourcing era reducido. La palabra outsourcing no apareció de hecho en ningún libro de texto sino hasta hace pocos años. Además, el outsourcing se concentraba en trabajos que requerían relativamente pocas competencias, como la atención a clientes. (Cuando un usuario llama para pedir asistencia con el fin de entender sus cuentas bancarias, configurar un
puerto inalámbrico en su computadora o usar un nuevo teléfono celular, es posible que hable con
un representante de atención a clientes en Bangalore, India). Sin embargo, a medida que el número de
oportunidades de outsourcing crecía, el fenómeno se extendió a áreas insólitas, como la medicina y el
derecho (véase el apartado en la p. 453).
Hace algunos años, los pronosticadores dijeron que las empresas estadounidenses subcontratarían de forma externa varios millones de empleos y algunos cálculos recientes apuntan a cifras incluso
más altas. En realidad, es difícil determinar la cifra correcta. Las empresas no tienen la obligación de
informar el número de trabajadores empleados por contratistas externos.
El outsourcing se convirtió, como es natural, en un tema político muy delicado. Durante la primera administración de George W. Bush, el jefe de economistas del Presidente, Gregory Mankiw, sostuvo que el outsourcing producía los beneficios de costos menores de producción y, como resultado,
precios más bajos de los productos. Recibió críticas muy duras de los medios de comunicación y los
políticos de ambos bandos.
Lo que posibilitaba el outsourcing era el cambio tecnológico y la enorme inversión en tecnología
que tuvo lugar en la década de 1990. El advenimiento de Internet y un increíble aumento en la calidad
y velocidad de la transferencia de información echaron a andar este proceso. Después, el mundo invirtió centenares de millones de dólares en la conectividad de banda ancha y cables submarinos con enorme capacidad. El costo del equipo de computación bajó de forma drástica al mismo tiempo que la capacidad de almacenamiento aumentaba con rapidez.
¿Adónde llevará el fenómeno del outsourcing y cómo terminará esta historia? En este momento,
nadie lo sabe. Sin embargo el ritmo del cambio es vertiginoso. De hecho, los argumentos a favor y en
contra del outsourcing son exactamente los mismos que se han ofrecido respecto al libre comercio y
éstos se remontan a la obra de David Ricardo en el siglo XIX (véase el capítulo 20). El hecho de que el
señor Mankiw fuera criticado en todas partes con tanta severidad por argumentar, en esencia, a favor
del libre comercio indica que el debate sobre el libre comercio dista mucho de terminar.
MOVILIDAD DEL CAPITAL
Otro importante aspecto de la globalización es la movilidad del capital. Examine de nuevo el diagrama
de flujo circular en la figura 22.1. Seis de los dieciséis flujos representados en la figura se refieren a la
oferta o la demanda de capital. En la economía global, las familias nacionales ponen su riqueza a trabajar en los mercados financieros nacionales o internacionales; las empresas nacionales acuden a los
mercados financieros nacionales o internacionales para financiar proyectos de inversión. Lo mismo
aplica a las empresas y familias de otros países.
En la actualidad, casi todos las carteras bien diversificadas de acciones y bonos de las familias o las
instituciones contienen algunos activos internacionales. Es tan fácil comprar o vender acciones de una
compañía textil italiana como de General Electric. Además, la mayoría de las grandes corporaciones de
renombre en Estados Unidos son multinacionales. Es decir, cuentan con instalaciones de producción,
operaciones o ventas en varios países. Las compras y ventas de acciones y bonos del exterior por parte
de ciudadanos estadounidenses y las compras y ventas de acciones y bonos estadounidenses por parte de
ciudadanos de otros países ascendieron en total a billones de dólares en 2003. Una tercera parte de todas las tenencias extranjeras de valores estadounidenses están en manos de japoneses. La inversión
extranjera directa en Estados Unidos, es decir, la propiedad de capital físico por parte de empresas extranjeras en Estados Unidos se elevó a un total de alrededor de 1.2 billones de dólares en 2000, mientras que
las empresas estadounidenses eran propietarias de aproximadamente el mismo capital en otros países.
El argumento a favor de la movilidad libre y abierta de los mercados financieros es que el capital debe
fluir hacia su mejor y más alto uso. Los mercados financieros deben asignar capital a los usos que producen
los rendimientos más altos, controlando los riesgos asociados con esas inversiones. Si invertir en una fábrica en Nigeria produce un rendimiento de 20 por ciento anual, mientras que una inversión parecida, de ries
go comparable, en Estados Unidos produce un rendimiento de 14 por ciento, los mercados financieros deben financiar en primer lugar el proyecto nigeriano sin importar de dónde proceda el financiamiento.
Sin embargo, los flujos rápidos y libres de capital financiero han tenido consecuencias negativas para algunas naciones. A mediados de la década de 1990, un formidable volumen de capital financiero fluyó de Estados Unidos hacia Tailandia, Indonesia y Corea. Las sociedades de inversión compraron acciones de empresas extranjeras, los bancos y las compañías de seguros financiaron grandes transacciones de
bienes raíces y los fondos de cobertura cambiaria especularon al apostar sobre los mercados de valores y
los movimientos del tipo de cambio en estos países en los años inmediatamente anteriores a 1997.
Todo esto parecía bien para todos los interesados. Era probable que las compras de capital físico
aumentaran la productividad y los salarios en esos países. La construcción de edificios de oficinas creó
empleos directos y proporcionó espacio para las empresas en crecimiento y nuevas fábricas suministraron capital (máquinas) con el que los obreros trabajaban. Un efecto secundario consistió en que las
monedas de estos países se fortalecieron en alto grado. Por ejemplo, el won coreano aumentó considerablemente de valor. El efecto fue que las importaciones se abarataron mucho. La combinación entre
importaciones baratas, alta demanda de bienes de inversión y mercados financieros prósperos, incluidos los mercados bursátiles en plena expansión, se alimentó de sí misma y los flujos se incrementaron.
Sin embargo, lo que sucede a menudo en estos casos es que los mercados de capital “se salen de
curso”. Debido a que nadie coordina las inversiones realizadas por los inversionistas individuales y a
que suele haber demoras considerables entre la decisión de invertir y la creación real de nuevo capital,
a veces se presenta un exceso de inversión. Así ocurrió en Tailandia, Corea e Indonesia en 1997. Cuando las cosas empezaron a salir mal en esos países, todos los acreedores corrieron a la puerta al mismo
tiempo. Es decir, las sociedades de inversión, bancos y otras entidades que tenían activos en esos países
trataron al mismo tiempo de liquidar sus posiciones. El resultado fue una caída drástica del valor del
won (Corea), el baht (Tailandia) y la rupia (Indonesia), que hicieron subir como la espuma los precios
de las importaciones y lesionaron el gasto de consumo. Los mercados bursátiles quebraron, la inversión empresarial se agotó, el gasto agregado cayó en picada y los tres países entraron en una severa recesión. Auges y desplomes similares en Brasil (1998), México (1994), Rusia (1998), Turquía (2000) y
Argentina (2001) parecen tener su origen en tipos parecidos de flujos financieros rápidos.
En parte como reacción ante esas crisis, algunos países tienen “controles de capital”. Dichos controles imponen límites estrictos sobre el volumen de los flujos financieros que entran y salen de un país
en periodo determinado. En su ensayo reciente sobre la globalización, Stanley Fischer plantea el siguiente argumento:
¿Qué puede hacerse para reducir la volatilidad de los flujos de capital hacia países de mercados
emergentes? La primera respuesta sería que los países se aislaran de los flujos de capitales internacionales. Hay que destacar que, a pesar de las crisis y los argumentos de muchos de los críticos de la globalización, casi ningún país ha tomado este camino; la preferencia revelada de los
países de mercados emergentes es seguir participando en el sistema financiero internacional.10
POLÍTICA PÚBLICA Y GLOBALIZACIÓN
La creciente interdependencia de naciones y sus ciudadanos provoca varios debates específicos sobre
política que van más allá de las cuestiones del libre comercio y movilidad de capital y la mano de obra
que ya hemos analizado. En primer lugar, hay bienes públicos y externalidades globales. ¿Cómo deben
responder las naciones del mundo a estos problemas de coordinación a gran escala, como el calentamiento global? En segundo término, varias instituciones grandes y poderosas, a menudo llamadas organizaciones no gubernamentales (ONG), desempeñan una función decisiva en la aplicación de los
tratados monetarios y reglas de comercio internacionales. Las más famosas son el Banco Mundial y el
Fondo Monetario Internacional (FMI). La pregunta es: ¿qué función deben desempeñar las instituciones en promover el crecimiento y eliminar la pobreza en el mundo?
EXTERNALIDADES Y BIENES PÚBLICOS GLOBALES
La economía, en general, estudia el funcionamiento eficiente de los mercados tanto nacionales como
internacionales. Pero a menudo se presentan problemas que los mercados privados no manejan con
eficiencia. Uno de estos problemas se relaciona con una clase de bienes llamados bienes públicos.
Los bienes públicos, a veces llamados bienes sociales, son bienes o servicios que confieren beneficios colectivos a los miembros de la sociedad. En general, nadie puede ser excluido del disfrute de sus
beneficios una vez que se producen. Los ejemplos clásicos son el aire limpio y la defensa nacional.
Los bienes privados, como las hamburguesas, se producen en el sector privado porque una empresa puede excluir a los que no pagan. Si la exclusión es imposible, se presentan dos problemas en los
mercados: el “problema del polizón” y el “problema del grano de arena en el desierto”. ¿Por qué debo
pagar por un bien si obtengo los beneficios aunque no pague? Debido a que muchas personas se benefician de la producción, mi aportación es tan pequeña que no importa. Como resultado, el sector privado no puede hacer nada porque conviene a los intereses de la gente no pagar y recae en el gobierno
la responsabilidad de ocuparse del abasto de estos bienes. Los gobiernos pueden destinar dinero y recaudar impuestos para cosas como la defensa nacional, un sistema judicial y la salud pública, obligando así a los ciudadanos a pagar por ellos. Supuestamente, los gobiernos producen los bienes públicos
que sus ciudadanos quieren.
Pero, ¿qué sucede con los bienes públicos cuyos beneficios son mundiales o por lo menos se comparten entre dos países fronterizos? En este caso, incluso los gobiernos y las naciones enfrentan los problemas del “polizón” y del “grano de arena en el desierto”. Considere el problema del calentamiento global. Adoptar medidas para desacelerar el calentamiento global supuestamente produciría un bien
público mundial. Como es evidente, no hay modo de excluir a las naciones de los beneficios si no contribuyen, y el impacto de una sola nación es pequeño. Como resultado, las naciones tienen un incentivo
para no contribuir. Esta situación es especialmente cierta en los países en vías de desarrollo, como China e India, donde la limpieza sería costosa y el sacrificio grande en función del ingreso nacional.
Otra forma de pensar en el calentamiento global es que es resultado de una externalidad. Una externalidad es un costo o un beneficio resultante de alguna actividad o transacción que se impone o
confiere a alguna parte ajena a la actividad o la transacción. El ejemplo clásico es la contaminación, pero hay muchos otros.
Por eficiencia, quienes toman en consideración los costos y beneficios de sus decisiones de producción o consumo deben sopesar todos los costos y beneficios. Una de las funciones del gobierno es
“internalizar” las externalidades. En esencia, la idea es forzar a quienes generan las externalidades a tomar en cuenta esas externalidades en sus decisiones. Un ejemplo de una política destinada a internalizar una externalidad es un cargo o impuesto que grava a una empresa contaminadora en proporción
con el daño que causa. Por supuesto, la situación se vuelve muy compleja cuando un país impone un
costo externo a otro. Si el número de países implicados es pequeño, las negociaciones y tratos entre
ellos pueden resolver simplemente el problema. Pero cuando interviene un gran número de jurisdicciones, se presentan problemas relacionados con los bienes públicos.
Las Naciones Unidas y otros organismos internacionales existen para promover cosas que ofrecen
beneficios al mundo, pero que requieren la cooperación y colaboración internacionales. Una interpretación de su propósito es que existen para garantizar la producción de bienes públicos internacionales.
Sin embargo, su índice de éxito ha sido irregular. La política internacional es, sin lugar a dudas, más
compleja y problemática que la política nacional. Además, los organismos internacionales no tienen
casi ninguna facultad para imponer gravámenes.
El concepto de bienes públicos también contribuye a explicar la pandemia del VIH/SIDA. En
2003, más de 60 millones de personas estaban infectadas con el virus. Todos los días ocurrían aproximadamente 14,000 infecciones. Más de 20 millones de personas han muerto. El VIH/SIDA es la causa
principal de muerte en África subsahariana, donde la enfermedad amenaza con revertir los logros del
desarrollo de los últimos 50 años. El problema del VIH/SIDA es más que sólo una cuestión de salud
pública. Diezma a la población trabajadora, empeora la pobreza y la desigualdad, crea grandes cantidades de huérfanos y cuesta a los países más pobres miles de millones de dólares en gastos médicos. El
tratamiento básico de un paciente de sida puede costar tres veces el ingreso nacional bruto per cápita.
Para tener impacto en la epidemia, las naciones deben cooperar y estar dispuestas a pagar. Por supuesto, éste es un problema que el mercado no solucionará. Se requerirá un esfuerzo multinacional si se desea controlar la pandemia.
ORGANIZACIONES NO GUBERNAMENTALES Y ECONOMÍA
INTERNACIONAL: EL CONSENSO DE WASHINGTON
Ningún tema de la economía internacional suscita debates más acalorados que la postura de la política
adoptada por el FMI hace algunos años cuando estableció las “condiciones” que debían satisfacerse para que los países tuvieran derecho a préstamos y subvenciones para el desarrollo.
Aunque existe desacuerdo considerable respecto a quién exactamente lo formó o la rigurosidad
con que se pensaba aplicar, se estableció un conjunto de objetivos o metas para los países que el FMI
estaba financiando. Estos objetivos, conocidos como Consenso de Washington, tenían 10 elementos:
(1) disciplina fiscal: déficit presupuestario modesto o presupuestos equilibrados; (2) prioridades de
gasto público en salud y educación; (3) reforma fiscal: la base tributaria debía ser amplia y las tasas tributarias marginales, bajas; (4) tasas de interés determinadas por el mercado, positivas, pero moderadas; (5) un tipo de cambio competitivo (idealmente flotante) como el “primer elemento esencial de
una política económica orientada hacia el exterior”; (6) liberalización de las importaciones; en esencia, una política de libre comercio y aranceles reducidos; (7) apertura a la inversión extranjera; (8) privatización, “con base en la idea de que la industria privada se administra con mayor eficiencia que las
empresas públicas”; (9) desregulación y (10) protección de los derechos de propiedad.
Desde luego, había margen considerable para discrepancias respecto al grado hasta el que estos
elementos debían o podían aplicarse. Muchos creyeron que el FMI se extralimitó y fue muy estricto en
sus tratos con muchos países. Tomar un país, como uno de los países ex comunistas de Europa Oriental, y exigir un presupuesto equilibrado, apertura y privatización en rápida sucesión era demasiado para tan poco tiempo.
De las malas experiencias de la década de 1990 ha surgido un nuevo consenso de gradualismo.
Aunque el espíritu del Consenso de Washington sigue intacto, el rigor de su aplicación parece ser bastante más moderado.
GLOBALIZACIÓN, CAPITALISMO Y DEMOCRACIA
Es importante mencionar un último tema antes de dejar el análisis de la globalización. Con mucha frecuencia, el problema de la apertura y la globalización se entremezcla con cuestiones referentes a los sistemas políticos y económicos. Los defensores de la globalización son partidarios acérrimos del capitalismo del laissez-faire. Desde luego, los diez puntos del Consenso Washington lo llevan a uno a la
conclusión de que sus defensores creen firmemente en el sistema de mercado.
Sin embargo, ¿la cuestión de la apertura y la deseabilidad de aumentar la interdependencia entre
las economías nacionales dependen del tipo de sistema económico o político que un país decide establecer? Probablemente no.
Ante todo, es importante separar el debate sobre los sistemas políticos del debate sobre los sistemas económicos. Los términos democracia y dictadura se refieren a las instituciones de gobierno y al
proceso de elección pública. ¿Hasta qué punto se toman las decisiones del gobierno a través de instituciones democráticas que expresan la voluntad del pueblo?
Los términos socialismo y capitalismo se refieren, por otra parte, a las instituciones económicas
que determinan la asignación de los recursos. Una economía puramente socialista es aquella en que el
gobierno es dueño de la tierra y el capital y en la cual los recursos se asignan, en esencia, por medio de
un plan del gobierno central. Una economía capitalista laissez-faire es aquella en que el gobierno no
desempeña prácticamente ninguna función en la dirección de la economía. Más bien, las decisiones
sobre la asignación de los recursos las toman las familias y empresas, en lo individual, viendo por sus
intereses particulares, con poca o ninguna participación del gobierno. Las decisiones esenciales se toman a través del mercado.
De hecho, la mayoría de las economías son mixtas, y el debate no consiste en realidad en si el gobierno debe intervenir o no. En cambio, se trata de la función que el gobierno debe desempeñar en la
economía. Incluso los partidarios más conservadores de la libre empresa están convencidos de que el
gobierno debe desempeñar alguna función en la economía. En microeconomía, se habló de la función
del gobierno en abastecer bienes públicos, regular el poder de los monopolios, internalizar los costos y
beneficios externos y asegurar que todos los agentes económicos estén bien informados. También se
habló del grado hasta el cual los ciudadanos pueden querer que el gobierno intervenga en la redistribución del ingreso. En macroeconomía, se habla de la posible función del gobierno en estabilizar la
economía para mantener al mínimo los índices de desempleo e inflación.
Incluso en el sistema económico de libre empresa por excelencia de Estados Unidos, 30 por ciento del ingreso total se paga en impuestos para sostener al gobierno, y éste desempeña una función significativa en las decisiones de asignación de los recursos.
Hay algunos países que han tenido sistemas políticos muy democráticos, pero que deciden tener
economías con inclinación socialista y una considerable dirección del gobierno. Los ejemplos incluyen a
Suecia y Dinamarca. Por otra parte, hay países que han tenido dictaduras severas, pero que deciden
implementar sistemas económicos muy al estilo laissez-faire. Los ejemplos incluyen Singapur y Chile
bajo el régimen de Augusto Pinochet.
Aunque los economistas en su conjunto tienden a favorecer la globalización, hay una amplia gama de opiniones sobre la función que el gobierno debe desempeñar en la economía. Aunque muchos
están a favor de una participación sustancial del gobierno, muchos economistas creen que la asignación de los recursos debe dejarse al mercado en general y que debe permitirse que las empresas y las familias que maximizan las utilidades respondan a los incentivos del mercado. Esos mismos economistas
tienden en general a creer que debe permitirse, con la menor interferencia posible, que las familias y
empresas se dediquen al comercio, inviertan en el extranjero o trabajen en otros países.
ADVERTENCIA FINAL
¿Cuál es el resultado de todo esto? ¿Hay algún consenso sobre la deseabilidad de la globalización? No lo
hay. Es probable que nada de lo que se ha dicho aquí convenza a los críticos de que es algo bueno. Sin
embargo, es importante entender que existe una lógica poderosa que respalda la apertura económica:
la idea de que es posible que el libre flujo de recursos, bienes y servicios entre fronteras nacionales, impulsado por incentivos económicos eficientes, incluido el deseo de maximizar las utilidades, mejore el
bienestar de los ciudadanos más que si las fronteras estuvieran cerradas y las economías vueltas hacia
el interior.

## Crecimiento económico en economías en vías de desarrollo y en transición

El polo de atención de este texto
se ha centrado en los países con
economías industrializadas modernas
que dependen en gran medida de los
mercados para la distribución de recursos; sin embargo, ¿qué sucede con
los problemas económicos que enfrentan países como Somalia o Haití?
¿Se pueden aplicar los mismos principios económicos que se han estudiado
a estas naciones menos desarrolladas?
Sí. Todos los análisis económicos
tratan del problema de tomar decisiones en condiciones de escasez, y el
problema de satisfacer los deseos y necesidades de la gente es tan real en Somalia y Haití como lo es en Estados
Unidos, Alemania y Japón. La universalidad de la escasez es lo que hace que el análisis económico sea relevante en todas las naciones, independientemente del nivel de bienestar material o ideología política predominante.
Las herramientas básicas de la oferta y la demanda, las teorías sobre los consumidores y las empresas, y las teorías sobre la estructura de los mercados contribuyen a la comprensión de los problemas
económicos que enfrentan las naciones en vías de desarrollo en el mundo. Sin embargo, estas naciones
suelen hacer frente a los problemas económicos de manera muy diferente que los países más ricos y desarrollados. En las naciones en vías de desarrollo, es posible que el economista tenga que preocuparse
por la escasez crónica de alimentos, el explosivo crecimiento demográfico y la hiperinflación, que alcanza cifras de tres y hasta cuatro dígitos. Estados Unidos y otras economías industrializadas rara vez
tienen esas dificultades.
Los instrumentos de la administración económica también varían de una nación a otra. Las instituciones del mercado financiero en Estados Unidos están bien desarrolladas y cuenta con un banco
central fuerte (la Reserva Federal) por medio del cual el gobierno puede controlar la macroeconomía
hasta cierto punto. Incluso la intervención limitada es imposible en algunos de los países en vías de desarrollo. En Estados Unidos, las leyes fiscales pueden cambiar para fomentar el ahorro, estimular tipos
específicos de inversiones o para redistribuir el ingreso. En la mayoría de los países en vías de desarrollo,
no hay impuestos personales sobre la renta ni políticas fiscales eficaces que sean significativos.
A pesar de que los problemas económicos y los instrumentos de política disponibles para atacarlos varían entre las naciones, el pensamiento económico para considerar estos problemas puede aplicarse con facilidad en un ambiente u otro. En este capítulo se hablará de varios de los problemas económicos específicos de los países en vías de desarrollo para tratar de captar algunas de las percepciones
que el análisis económico puede ofrecer.

LA VIDA EN LAS NACIONES EN VÍAS DE
DESARROLLO: POBLACIÓN Y POBREZA
En el año 2006, la población mundial era de más de 6,500 millones de personas. La mayoría de las más
de 200 naciones del mundo pertenecen al mundo en vías de desarrollo, en el que viven cerca de tres
cuartas partes de la población mundial.
A principios de la década de 1960, los países del mundo podían clasificarse con facilidad en categorías: los países desarrollados incluían la mayor parte de Europa, América del Norte, Japón, Australia
y Nueva Zelanda; los países en vías de desarrollo incluían al resto del mundo. A menudo se dice que las
naciones en vías de desarrollo pertenecen al Tercer Mundo para distinguirlas de las naciones industrializadas de Occidente (el Primer Mundo) y el ex bloque socialista de países de Europa Oriental (el Segundo Mundo).
En 2006, el mundo no se dividía sin dificultad en tres partes bien definidas. El rápido progreso económico ha acercado a algunos países en vías de desarrollo a las economías desarrolladas. Países como
Argentina y Corea, todavía consideradas “en vías de desarrollo”, a menudo se les conoce como países de
medianos ingresos, o recién industrializados. Otros países, como los que se encuentran en buena parte
del África subsahariana y algunos del sur de Asia, se han estancado y rezagado tanto con respecto a los
adelantos económicos del resto del mundo, que se ha utilizado el término Cuarto Mundo para referirse
a ellos. Aún no está claro dónde acabarán ubicándose las repúblicas de la ex Unión Soviética y otros países ex comunistas de Europa Oriental. La producción ha disminuido de manera notable en muchos de
ellos. Por ejemplo, entre 1990 y 1997, el producto interno bruto (PIB) real se redujo alrededor de 40 por
ciento en las economías en transición y más de 50 por ciento en Rusia y Asia Central. Una estimación
del PIB per cápita de Rusia en 2004 lo ubica en casi 3,500 dólares. Algunas de las nuevas repúblicas tienen ahora más en común con los países en vías de desarrollo que con los países desarrollados.
Aunque los países del mundo en vías de desarrollo son muy diversos, tanto en sus niveles de vida
como en sus experiencias particulares de crecimiento, existen diferencias marcadas que siguen separándolos de las naciones desarrolladas. Los países desarrollados tienen un nivel promedio más alto de
bienestar material (las cantidades de alimentos, ropa, vivienda y otros productos consumidos por la
persona típica). Las comparaciones del ingreso nacional bruto a menudo se usan como un índice rudimentario del nivel del bienestar material entre las naciones. El INB es una medición del ingreso de
un país, que se calcula mediante un método más preciso de convertir el poder adquisitivo en dólares.
Véase la tabla 23.1, donde el INB per cápita en las economías de mercado industriales supera de manera significativa el INB de las economías en vías de desarrollo, de bajos y medianos ingresos.
Otras características del desarrollo económico incluyen las mejoras en la educación y salud básicas. El grado de libertad política y económica que disfrutan los ciudadanos particulares también puede ser parte de lo que significa ser una nación desarrollada. Algunos de estos criterios son fáciles de
cuantificar; la tabla 23.1 presenta datos de diferentes tipos de economías, de acuerdo con algunos
de los índices de desarrollo que se miden con mayor facilidad. Como podrá ver, las economías de mercado industrializadas disfrutan de niveles más altos de vida, con base en cualquiera de los indicadores
de desarrollo que se elija.
Detrás de estas estadísticas yace la realidad de la vida tan difícil que enfrenta la gente del mundo
en desarrollo. Para la mayoría, los escasos ingresos que perciben sólo cubren las necesidades más elementales. La mayoría de las comidas son las mismas y consisten en el alimento básico que se produce
en la región: arroz, trigo o maíz. La vivienda es primitiva. Muchas personas comparten una habitación
pequeña, por lo general de piso de tierra y sin instalaciones sanitarias. La gran mayoría de la población vive en zonas rurales donde el trabajo agrícola es duro y consume mucho tiempo. La productividad
(producción por trabajador) es baja porque las parcelas familiares son pequeñas y sólo se pueden con-seguir los implementos agrícolas más rudimentarios. La baja productividad implica que la producción
agrícola por persona es apenas suficiente para alimentar a la familia del agricultor y no le queda nada
para vender a otros. Los niños en edad escolar pueden recibir alguna educación formal, pero el analfabetismo sigue siendo crónico tanto entre los jóvenes como entre los adultos. La mortalidad infantil llega a
ser 10 veces mayor que en Estados Unidos. Aunque las infecciones parasitarias son comunes y debilitadoras, sólo hay un médico por cada 5,000 personas. Además, en muchas naciones en vías de desarrollo se
libran guerras civiles y externas.
La vida en las naciones subdesarrolladas es una lucha continua contra las circunstancias de la pobreza y las perspectivas de mejoras sustanciales en los niveles de vida de la mayoría de la gente son muy
desalentadoras. Como sucede con todas las generalizaciones, hay excepciones. Algunos países se encuentran en mejor situación que otros y en cualquier país determinado siempre hay un grupo de elite
que vive con grandes lujos.
La pobreza, y no la abundancia, domina el mundo subdesarrollado. Estudios recientes indican
que 40 por ciento de la población de los países subdesarrollados percibe un ingreso anual insuficiente
para cubrir las necesidades de una nutrición adecuada.
Aunque las naciones desarrolladas representan sólo una cuarta parte de la población mundial,
se calcula que consumen tres cuartas partes de la producción mundial. Esto deja a las naciones
en vías de desarrollo con alrededor de tres cuartas partes de la población mundial, pero sólo una
cuarta parte del ingreso mundial. El resultado simple es que la mayoría de la población de nuestro planeta es pobre.
En Estados Unidos, la quinta parte más pobre (el 20 por ciento inferior) de las familias recibe poco menos de 5 por ciento del ingreso total; la quinta parte más rica recibe alrededor de 46 por ciento.
Sin embargo, la desigualdad en la distribución mundial del ingreso es mucho mayor. Cuando se estudia la población mundial, la quinta parte más pobre de las familias gana cerca de .5 por ciento y la
quinta parte más rica gana 79 por ciento del total del ingreso mundial.
DESARROLLO ECONÓMICO: ORÍGENES
Y ESTRATEGIAS
Los economistas han tratado de comprender el crecimiento y el desarrollo económico desde Adam
Smith y David Ricardo en los siglos XVIII y XIX, pero el estudio de la economía del desarrollo, en lo que
aplica a las naciones en vías de desarrollo, tiene una historia mucho más breve. Las luchas geopolíticas
después de la Segunda Guerra Mundial llamaron la atención hacia las naciones en vías de desarrollo y
sus problemas económicos. Durante este periodo, el nuevo campo de la economía del desarrollo tan
sólo preguntó: ¿por qué algunas naciones son pobres y otras ricas? Si los economistas pudieran entender los obstáculos del crecimiento económico que impiden que las naciones se desarrollen y los requisitos esenciales que les ayudarían a desarrollarse, podrían prescribir estrategias para lograr el progreso
económico.
LOS ORÍGENES DEL DESARROLLO ECONÓMICO
Aunque no ha aparecido una teoría general de desarrollo económico aplicable a todas las naciones y
quizá nunca surja, se han señalado algunos factores básicos que limitan el crecimiento económico de
un país pobre. Éstos incluyen formación insuficiente de capital, escasez de recursos humanos y espíritu emprendedor, falta de capital para gastos generales sociales y las restricciones impuestas por la
dependencia de las naciones desarrolladas.
Formación de capital Una explicación de los bajos niveles de producción en las naciones subdesarrolladas son las cantidades insuficientes de insumos necesarios. Las naciones en vías de desarrollo tienen diversas dotaciones de recursos: Congo, por ejemplo, posee recursos naturales en abundancia, en tanto que Bangladesh es pobre en recursos. Casi todos los países subdesarrollados tienen
escasez de capital físico en relación con los demás recursos, en especial la mano de obra. El pequeño
acervo de capital físico (fábricas, maquinaria, equipo agrícola y demás capital productivo) limita la
productividad de la mano de obra y ocasiona rezagos en la producción nacional.
No obstante, decir que la escasez de capital es la causa de la baja productividad no explica gran cosa. Es necesario saber por qué el capital es tan escaso en los países en vías de desarrollo. Hay muchasexplicaciones. Una, la hipótesis del círculo vicioso de la pobreza, indica que una nación pobre tiene
que consumir la mayoría de su ingreso para mantener a duras penas su ya de por sí bajo nivel de vida.
Consumir la mayoría del ingreso nacional implica ahorro limitado, y esto, a su vez, implica bajos niveles de inversión. Sin inversión, el acervo de capital no crece, los ingresos siguen siendo bajos y el círculo vicioso se completa. La pobreza se autoperpetúa.
La dificultad que plantea el argumento del círculo vicioso es que si fuera verdad, ninguna nación podría desarrollarse. Por ejemplo el PIB per cápita de Japón en 1990 era muy inferior al de muchas de las naciones subdesarrolladas de la actualidad. El argumento del círculo vicioso no reconoce que cada nación tenga cierto excedente por encima de las necesidades de consumo que está disponible para la inversión. A
menudo, este excedente es más notorio en los notorios hábitos de consumo de las familias más ricas del país.
La pobreza por sí misma no puede explicar la escasez de capital, y la pobreza no necesariamente
se autoperpetúa.
En una economía en vías de desarrollo, la escasez de capital puede tener más que ver con la falta
de incentivos para que los ciudadanos ahorren e inviertan de forma productiva que con la escasez absoluta de ingreso disponible para acumulación del capital. Muchos de los ricos en los países en vías de
desarrollo invierten sus ahorros en Europa o en Estados Unidos en lugar de hacerlo en su propio país,
en el que puede haber un clima político mucho más riesgoso. Los ahorros transferidos a Estados Unidos no producen crecimiento del capital en los países en vías de desarrollo. El término fuga de capitales se refiere al hecho de que tanto el capital humano como el financiero (el ahorro interno) salen de
los países subdesarrollados en busca de tasas más altas de rendimiento en otras partes o de rendimientos que ofrezcan menos riesgo. Además, las políticas gubernamentales en las naciones subdesarrolladas
(incluidos los techos de precios, el control de las importaciones e incluso la expropiación directa de la
propiedad privada) tienden a desalentar la inversión.
Sean cuales fueren las causas de la escasez de capital, es evidente que la ausencia de capital productivo impide que el ingreso aumente en cualquier economía. La disponibilidad de capital es una
condición necesaria, pero no suficiente, para el crecimiento económico. El panorama de los países en
vías de desarrollo está plagado de fábricas inactivas y maquinaria abandonada. Se requieren otros ingredientes para lograr el progreso económico.
Recursos humanos y espíritu emprendedor El capital no es el único factor de producción que se requiere para elaborar productos. La mano de obra es igualmente importante. Primero que
nada, para ser productiva la población trabajadora debe estar sana. La enfermedad, en la actualidad, es
la principal amenaza para el desarrollo en buena parte del mundo. El problema de salud más devastador en el mundo de la actualidad es la pandemia del VIH/SIDA. Sólo en 2005, 3.1 millones de personas
murieron de sida y 4.9 millones más se infectaron con el VIH. A finales de 2005, había más de 40 millones
de personas infectadas con el virus. En total, más de 25 millones han muerto. Es la principal causa de
muerte en el África subsahariana, donde la enfermedad amenaza con revertir los logros de desarrollo
alcanzados en los últimos 50 años. Aparte del sida, la salud y la nutrición son esenciales para el desarrollo de los trabajadores. Los programas de nutrición y salud pueden verse como inversiones en capital humano que generarán una mayor productividad e ingresos más altos.
Pero la salud no es el único problema. Para ser productivos, los trabajadores deben educarse y capacitarse. Las formas más conocidas de inversión en capital humano, que incluyen la educación formal
y la capacitación práctica en el trabajo, son esenciales. Por ejemplo, la educación básica, así como la capacitación especializada en administración agrícola pueden producir rendimientos altos tanto para el
trabajador en lo individual como para la economía. La educación se ha convertido en la mayor categoría del gasto gubernamental en muchas naciones en vías de desarrollo, en parte gracias a la convicción
de que los recursos humanos son el factor determinante por excelencia del progreso económico.
Al igual que el capital financiero busca el rendimiento más alto y seguro, también el capital humano. Miles de estudiantes de los países en vías de desarrollo, muchos de los cuales fueron apoyados por
sus gobiernos, se gradúan todos los años de las universidades e instituciones de educación superior de
Estados Unidos como ingenieros, doctores, científicos, economistas y otros profesionales. Después
de la graduación, estas personas enfrentan una difícil decisión: quedarse en Estados Unidos y ganar un
salario alto o regresar a casa y aceptar un empleo con un salario mucho menor. Muchos se quedan en
Estados Unidos. Esta fuga de cerebros aleja a muchas de las mentes más talentosas de los países en vías
de desarrollo.
Los empresarios innovadores que están dispuestos a correr riesgos son un recurso humano esencial en toda economía. En una nación subdesarrollada, rara vez es necesario inventar nuevas técnicas
de producción porque por lo general pueden adaptarse de la tecnología ya desarrollada en los paísestecnológicamente avanzados. Sin embargo, parece que los empresarios que pueden y están dispuestos
a organizar y llevar a cabo actividades económicas escasean. Los lazos familiares y políticos a menudo
son más importantes que la capacidad cuando se trata de asegurar posiciones de autoridad. Sin importar cuál sea la explicación:
El desarrollo no puede darse sin los recursos humanos capaces de iniciar y administrar la actividad económica.
Capital para gastos generales sociales Cualquiera que haya pasado un tiempo en una
nación subdesarrollada sabe lo difícil que puede ser enviar una carta, hacer una llamada telefónica local o viajar dentro del país. Además, si a esto se suman los problemas con el suministro de agua, las interrupciones frecuentes en el suministro de corriente eléctrica (en las pocas zonas donde hay electricidad) y el control de plagas y mosquitos, a menudo ineficaz, pronto se pone de manifiesto lo deficientes
que pueden ser incluso los bienes y servicios más sencillos y básicos que proporciona el gobierno.
En cualquier economía, en vías de desarrollo o no, el gobierno tiene la considerable oportunidad
y responsabilidad de intervenir donde las condiciones alientan los monopolios naturales (como en las
industrias de los servicios públicos) y donde es preciso proporcionar bienes públicos (como los caminos y el control de plagas). En una economía en vías de desarrollo, el gobierno tiene que poner énfasis
en la creación de infraestructura básica: carreteras, generación de energía y sistemas de irrigación. A
menudo hay buenas razones por las que el sector privado no puede llevar a cabo de manera satisfactoria estos proyectos, llamados capital para gastos generales sociales. Primero, muchos de estos proyectos funcionan con economías de escala, lo que significa que sólo pueden ser eficientes si son muy
grandes. En ese caso, pueden ser demasiado grandes para que una empresa o grupo de empresas privadas los lleve a cabo.
Segundo, el sector privado no puede emprender muchos proyectos que son útiles en el aspecto social porque no hay modo de que los agentes privados capturen una parte suficiente de los rendimientos para que esos proyectos resulten rentables. Este así llamado problema del polizón es común en las
economías del mundo desarrollado. Piense en la defensa nacional: todas las personas de un país se benefician de la defensa nacional, ya sea que hayan pagado por ella o no. Cualquiera que intentara incursionar en el negocio privado de proporcionar defensa nacional se iría a la quiebra. ¿Por qué debo comprar defensa nacional si tu compra de defensa también me protege? ¿Por qué debes comprarla tú si mi
compra también te protege?
Los gobiernos de los países en vías de desarrollo pueden hacer cosas importantes y útiles para
fomentar el desarrollo, pero muchos de sus esfuerzos deben concentrarse en áreas que el sector
privado nunca tocaría. Si el gobierno no interviene en esos campos, el desarrollo económico
puede verse coartado por la falta de capital para gastos generales sociales.
ESTRATEGIAS DE DESARROLLO ECONÓMICO
Al igual que ninguna teoría explica la falta de progreso económico, ninguna estrategia de desarrollo
tiene probabilidades de tener éxito en todas las naciones. En años pasados, se han propuesto muchas
estrategias alternativas de desarrollo. Aunque estas estrategias han sido muy diferentes, todas reconocen que una economía en vías de desarrollo enfrenta intercambios compensatorios básicos. Una cantidad insuficiente de recursos humanos y físicos implica que debe elegirse entre la agricultura y la industria, las exportaciones y la sustitución de importaciones y la planeación central y los mercados libres,
entre otras decisiones.
¿Agricultura o industria? La mayoría de los países del Tercer Mundo empezaron a adquirir
independencia política poco después de la Segunda Guerra Mundial. La tradición de promover la industrialización como la solución a los problemas del mundo subdesarrollado data de esta época. Los
primeros planes quinquenales de desarrollo en India preveían el fomento de las actividades manufactureras; el gobierno actual de Etiopía (un país sumamente pobre) tiene intenciones parecidas.
La industria tiene varios atractivos aparentes sobre la agricultura. Primero, si es verdad que la escasez de capital restringe el crecimiento económico, la construcción de fábricas es un paso obvio para
aumentar el acervo de capital de una nación. Segundo, y tal vez más importante, una de las características primarias de las economías más desarrolladas es su transición estructural de la agricultura hacia
la industria manufacturera y los servicios modernos. Como muestra la tabla 23.2, la participación dela agricultura en el PIB se reduce de manera considerable a medida que aumenta el ingreso per cápita. La
participación de los servicios aumenta en consecuencia, en particular en las primeras etapas del desarrollo económico.
Muchas economías han tratado de desarrollar la industria a costa de la agricultura. Sin embargo,
en muchos países la industrialización no ha tenido éxito o ha sido decepcionante; es decir, no ha traído
los beneficios esperados. La experiencia indica que tan sólo tratar de replicar la estructura de las economías desarrolladas no es en sí mismo garantía de desarrollo exitoso o ni siquiera lo promueve.
Desde principios de la década de 1970, el sector agrícola ha recibido mucha mayor atención. Las
estrategias de desarrollo agrícola han tenido numerosos beneficios. Aunque algunos proyectos agrícolas (como la construcción de presas importantes y redes de irrigación) requieren mucho capital,
muchas otras (como los servicios que contribuyen a enseñar mejores técnicas agrícolas y programas
de fertilización a pequeña escala) tienen pocas necesidades de capital e importación. Los programas
como éstos pueden afectar a grandes cantidades de familias y como sus beneficios están dirigidos a las
zonas rurales, es muy probable que ayuden a las familias más pobres del país.
La experiencia de las últimas tres décadas indica que un cierto equilibrio entre estos métodos conduce al mejor resultado; es decir, es importante y eficaz prestar atención tanto a la industria como a la
agricultura. Los chinos se han referido a este doble enfoque en el desarrollo como “caminar con dos
piernas”.
¿Exportaciones o sustitución de importaciones? A medida que las naciones en vías de
desarrollo expanden sus actividades industriales, deben decidir qué tipo de estrategia comercial seguir,
por lo general una de dos alternativas: sustitución de importaciones o promoción de las exportaciones.
La sustitución de importaciones es una estrategia comercial e industrial para desarrollar las industrias locales que pueden fabricar productos que reemplacen a las importaciones. Por ejemplo, si se
importa fertilizante, la sustitución de importaciones requerirá una industria nacional de fertilizantes
que produzca los sustitutos de los fertilizantes importados. Esta estrategia adquirió prominencia en
América del Sur en la década de 1950. En esa época, la mayoría de los países en vías de desarrollo exportaban productos agrícolas y minerales, bienes que soportaban mercados internacionales inciertos y
a menudo inestables. En esas condiciones, era comprensible la necesidad de políticas de sustitución de
importaciones. Se aplicaron medidas gubernamentales especiales, como la protección de aranceles y
cuotas y las importaciones subsidiadas de maquinaria, para estimular las nuevas industrias nacionales.
También se invitó a empresas multinacionales a que iniciaran operaciones en muchos países.
La mayoría de los economistas creen que las estrategias de sustitución de importaciones han fracasado casi en todas partes donde se han puesto en práctica. Con industrias nacionales protegidas de la
competencia internacional por los elevados aranceles (a menudo de hasta 200 por ciento), se crearon
importantes ineficiencias económicas. Por ejemplo, Perú tiene una población de 26.7 millones de personas, y sólo una pequeña parte de esta población puede costear la compra de un automóvil. Sin embargo, en cierto momento, el país tenía cinco o seis diferentes fabricantes de automóviles, cada uno de
los cuales producía sólo unos cuantos miles de autos al año. Debido a que hay economías de escala sustanciales en la producción de automóviles, el costo por automóvil era mucho más alto de lo necesario,
y recursos valiosos que podían haberse dedicado a otra actividad más productiva se despilfarraban en
la producción de automóviles.Además, las políticas diseñadas para promover la sustitución de importaciones a menudo estimulaban métodos de producción intensivos en capital, que limitaban la creación de empleos y perjudicaban las actividades de exportación. Un país como Perú no podía exportar automóviles, porque los producía a un costo mucho más alto que el precio en el mercado mundial. Peor aún, las políticas de
sustitución de importaciones alentaron el uso de productos nacionales caros, como los tractores y fertilizantes, en lugar de importaciones más baratas. Estas políticas gravaban los sectores que podrían haber competido con éxito en los mercados mundiales. Hasta el punto en que la industria azucarera peruana tuvo que depender del fertilizante producido en el país a un costo muy elevado, por ejemplo, su
capacidad de competir en los mercados internacionales disminuía porque sus costos de producción se
elevaban de manera artificial.
Como alternativa a la sustitución de importaciones, algunas naciones han seguido estrategias de
promoción de las exportaciones. La promoción de las exportaciones es tan sólo la política de fomentarlas. Como una economía de mercado industrializada, Japón es un ejemplo sorprendente para el mundo
en vías de desarrollo del éxito económico que las exportaciones pueden ofrecer. Con un índice de crecimiento anual promedio del PIB real per cápita de alrededor de 6 por ciento anual desde 1960, los logros
de Japón se basan en parte en la producción industrial orientada hacia los consumidores extranjeros.
Varios países del mundo en vías de desarrollo han tratado de emular el éxito de Japón. A partir de alrededor de 1970, Hong Kong, Singapur, Corea y Taiwán (los “cuatro pequeños dragones” entre los dos
grandes dragones, China y Japón) empezaron a promover la exportación de productos manufacturados. En
la actualidad, sus índices de crecimiento han superado al de Japón. Otras naciones, como Brasil, Colombia y Turquía, también han tenido cierto éxito con su política comercial orientada hacia el exterior.
El respaldo gubernamental a la promoción de las exportaciones a menudo ha adoptado la forma
de mantener un tipo de cambio lo bastante favorable para permitir que las exportaciones compitan
con productos manufacturados de las economías desarrolladas. Por ejemplo, mucha gente cree que Japón mantuvo bajo el valor del yen de manera artificial en los años setenta. Debido a que el yen “barato” significa productos japoneses baratos en Estados Unidos, las ventas de productos japoneses (en especial de automóviles) presentaron un gran aumento. Los gobiernos también han proporcionado
subsidios a las industrias de exportación.
Un gran problema para los países que crecen o tratan de crecer con la venta de exportaciones en
los mercados mundiales es el libre comercio. En 2003, Estados Unidos y Europa fueron acusados de
proteger a sus productores agrícolas con grandes subsidios que conferían a los productores nacionales
una gran ventaja al vender en los mercados mundiales. Los países africanos en particular expusieron el
problema durante una gira por el continente que realizó el Presidente Bush en 2003. (Véase el apartado Análisis de las noticias titulado “Comercio y desarrollo en África en 2003”).
¿Planeación central o el mercado? Como parte de esta estrategia para lograr el desarrollo económico, una nación debe decidir cómo dirigirá su economía. Las opciones básicas están entre
un sistema económico orientado al mercado y otro planeado a nivel central.
En la década de 1950 y parte de los años sesenta, las estrategias de desarrollo que exigían la planeación nacional recibieron un amplio apoyo. El rápido crecimiento económico de la Unión Soviética,
una economía planeada de manera central, proporcionó un ejemplo de la rapidez con que una nación
agraria menos desarrollada podía transformarse en una moderna potencia industrial. (Los costos a
menudo pasmosos de esta estrategia —disciplina severa, violación flagrante de los derechos humanos
y daños al medio ambiente— eran menos conocidos en general.) Además, el subdesarrollo de muchos
mercados de productos básicos y activos en el mundo en vías de desarrollo llevó a muchos expertos a
creer que las fuerzas del mercado no podían dirigir una economía de manera confiable y que, por lo
tanto, era necesaria una intervención considerable del gobierno. Incluso Estados Unidos, con su compromiso con la libre empresa en el mercado, apoyó las primeras campañas de planeación central en
muchas naciones en vías de desarrollo.
Hoy en día, la planeación asume muchas formas en las naciones en vías de desarrollo. En algunas,
la planeación central ha sustituido los resultados basados en el mercado con controles directos de las
variables económicas, administrativamente determinados, como los precios, la producción y el empleo. En otras, la planeación nacional equivale a poco más que la formulación de metas a 5 o 10 años
como planes rudimentarios del futuro económico de una nación.
El atractivo económico de la planeación reside, en teoría, en su capacidad para canalizar los ahorros hacia la inversión productiva y para coordinar las actividades económicas que los actores privados
de la economía no podrían emprender. Sin embargo, la realidad de la planeación central es que es una
pesadilla muy politizada y difícil de administrar desde el punto de vista técnico. Dada la escasez de recursos humanos y el ambiente político inestable en muchas naciones en vías de desarrollo, la planeación por sí sola, por no mencionar la ejecución del plan, se vuelve una tarea formidable.
El fracaso de muchos esfuerzos de planeación central ha provocado que cada vez más se pida menos intervención del gobierno y más orientación hacia el mercado en las economías en vías de desarrollo. La eliminación de los controles de precios, la privatización de las empresas estatales y las reducciones en las restricciones a las importaciones son ejemplos de reformas orientadas al mercado que
recomiendan los organismos internacionales como el Fondo Monetario Internacional (FMI), cuyos objetivos principales son estabilizar los tipos de cambio internacionales y prestar dinero a los países que
tienen problemas para financiar sus transacciones internacionales, y el Banco Mundial, que presta dinero a los países para proyectos que promueven el desarrollo económico.
Las aportaciones de los miembros de ambas organizaciones se determinan con base en el tamaño
de sus economías. Sólo 20 por ciento de los fondos del Banco Mundial proviene de las aportaciones; 80
por ciento procede de las ganancias retenidas y las inversiones en los mercados de capital. El mundo en
desarrollo reconoce cada vez más el valor de las fuerzas de mercado en determinar la asignación de los
recursos escasos. No obstante, el gobierno sigue teniendo una función muy importante que desempeñar. En las décadas por venir, los gobiernos de las naciones en vías de desarrollo necesitarán determinar en qué situaciones la planeación es superior al mercado y dónde el mercado es superior a la planeación.
CRECIMIENTO FRENTE A DESARROLLO: EL CICLO DE LA POLÍTICA
Hasta este momento, se han utilizado los términos crecimiento y desarrollo como si significaran lo mismo, pero esto no siempre es así. Es fácil imaginar casos en los que un país ha alcanzado niveles más altos de ingreso (crecimiento) con pocos o ningún beneficio para la mayoría de sus ciudadanos (desarrollo). Así, la pregunta es si el crecimiento económico necesariamente produce el desarrollo
económico.
En el pasado, la mayoría de las estrategias de desarrollo estaban dirigidas a incrementar el índice de
crecimiento del ingreso per cápita. Muchas aún lo están, con base en la teoría de que los beneficios
del crecimiento económico “llegarán poco a poco” a todos los miembros de la sociedad. Si esta teoría
es correcta, el crecimiento debería promover el desarrollo.
A principios de los años setenta, la relación entre crecimiento y desarrollo se cuestionaba cada vez
más. Un estudio del Banco Mundial en 1974 concluyó:
Ahora queda claro que más de una década de crecimiento rápido en los países subdesarrollados ha redundado en pocos o ningún beneficio para tal vez una tercera parte de la población… De manera paradójica, aunque las políticas de crecimiento han superado con creces
las expectativas de la primera década de desarrollo, la mera idea del crecimiento agregado como un objetivo social se ha cuestionado cada vez más.
El estudio del Banco Mundial indicaba que los aumentos en el PIB per cápita no garantizaban mejoras
significativas en los indicadores del desarrollo, como nutrición, salud y educación. Aunque el PIB per
cápita aumentaba, sus beneficios llegaban sólo a una pequeña minoría de la población. Este éxito muy
limitado indujo a formular nuevas estrategias de desarrollo que abordaran directamente los problemas de la pobreza. Estas nuevas estrategias favorecieron la agricultura sobre la industria, exigieron la
redistribución interna del ingreso y la riqueza (en especial la tierra) y alentaron programas para satisfacer necesidades básicas, como alimentación y albergue.
A finales de los años setenta y principios de los ochenta, las crisis macroeconómicas internacionales de los altos precios del petróleo, la recesión mundial y la deuda del Tercer Mundo hicieron que la
atención se desviara de los programas planeados para eliminar la pobreza de manera más directa. Entonces, durante los años ochenta y noventa, el enfoque de la política dio un viraje de 180 grados. El
Banco Mundial y Estados Unidos empezaron a exigir “ajustes estructurales” en los países en vías de desarrollo como requisito esencial para enviarles ayuda. Los programas de ajuste estructural suponían
reducir el tamaño del sector público mediante la privatización o las reducciones del gasto, disminuir
de forma sustancial el déficit presupuestario, poner freno a la inflación y estimular el ahorro y la inversión privados con reformas fiscales. Estas exigencias a favor del mercado eran un intento por estimular
el crecimiento; las consecuencias de la distribución pasaron a segundo plano.
PROBLEMAS DEL DESARROLLO ECONÓMICO
Toda nación en vías de desarrollo tiene una historia cultural, política y económica propia y, por lo tanto, enfrenta un conjunto único de problemas. A pesar de esto, es posible analizar los problemas económicos comunes que cada nación debe enfrentar según su manera particular. Estos problemas incluyen
el rápido crecimiento demográfico y las crecientes cargas de endeudamiento.
CRECIMIENTO DEMOGRÁFICO
Se calcula que las poblaciones de los países en vías de desarrollo crecen a alrededor de 1.7 por ciento
anual. (Compárese esta cifra con un índice de crecimiento demográfico de sólo .5 por ciento anual en
las economías de mercado industrializadas). Si el índice de crecimiento demográfico del Tercer Mundo sigue siendo de 1.7 por ciento, dentro de 41 años la población se duplicará con respecto al nivel de
4,100 millones en 1990 a más de 8,000 millones de personas en el año 2031. Por otro lado, las naciones
industrializadas tardarán 139 años en duplicar sus poblaciones. Lo que alarma de inmediato respecto
a estas cifras es que dados los problemas económicos actuales de las naciones en vías de desarrollo, es
difícil imaginar cómo podrán absorber a tanta gente en un periodo relativamente tan corto.
La preocupación por el crecimiento demográfico mundial no es nueva. El reverendo Thomas Malthus (que fue el primer profesor de economía política de Inglaterra) expresó sus temores por los aumentos
demográficos que observó hace 200 años. Malthus creía que las poblaciones crecían de una manera geométrica a una tasa de crecimiento constante —por eso, el tamaño absoluto del aumento es cada vez mayor
cada año—, pero que las provisiones de alimentos crecen mucho más despacio debido a la productividad
marginal decreciente de la tierra.1 Estos dos fenómenos indujeron a Malthus a pronosticar el creciente empobrecimiento de la población mundial, a menos que pudiera detenerse el crecimiento demográfico.
Los temores de Malthus respecto a Europa y Estados Unidos resultaron infundados. No previó los
cambios tecnológicos que revolucionarían la productividad agrícola ni el posterior decremento en
los índices de crecimiento demográfico en Europa y América del Norte. No obstante, la predicción de
Malthus puede ser acertada, aunque prematura. ¿Las circunstancias en el mundo en vías de desarrollo
se ajustan a sus predicciones? Aunque algunos observadores contemporáneos creen que el punto de
vista malthusiano es correcto y que la población de la Tierra aumentará finalmente a un nivel que los
recursos del planeta no podrán sustentar, otros piensan que el cambio tecnológico y las transiciones
demográficas (a índices de crecimiento demográfico desacelerados) permitirán otras mejoras en el
bienestar global.
Las consecuencias del rápido crecimiento demográfico Sabemos mucho menos
sobre las consecuencias económicas del rápido crecimiento demográfico de lo que podría esperarse. La
opinión ortodoxa advierte de las terribles consecuencias económicas que tendrá la “explosión demográfica” de los países en vías de desarrollo, pero es difícil sustanciar estas predicciones con la información disponible. El rápido crecimiento económico de Estados Unidos, por ejemplo, llegó acompañado
de un crecimiento demográfico relativamente rápido desde el punto de vista histórico. La desaceleración del crecimiento demográfico no ha sido necesaria para el progreso económico logrado por muchos de los países recién industrializados. No obstante, la expansión poblacional en muchas de las naciones más pobres de la actualidad es de una magnitud sin precedentes en la historia mundial, como
muestra con claridad la figura 23.1. Del año 1 d.C. hasta mediados del siglo XVII, las poblaciones crecieron despacio, a tasas de apenas 0.04 por ciento al año aproximadamente. Desde entonces, y en especial desde 1950, las tasas se han disparado. Hoy en día, las poblaciones crecen a tasas de entre 1.5 y 4.0
al año en todo el mundo en vías de desarrollo.
Debido a que estos índices de crecimiento nunca se presentaron antes del siglo XX, nadie sabe el
impacto que tendrán en el desarrollo económico futuro. Sin embargo, una preocupación económica
básica es que un crecimiento demográfico así de rápido puede limitar la inversión y restringir los aumentos en la productividad de la mano de obra y el ingreso. El rápido crecimiento demográfico cambia la composición de edades de una población y genera muchos hijos dependientes en relación con el
número adultos trabajadores productivos. Tal situación puede disminuir las tasas de ahorro, y por tanto de inversión, a medida que las necesidades inmediatas de consumo de los pequeños adquieren prioridad sobre el ahorro para el futuro.
Incluso si los índices bajos de ahorro no son una consecuencia necesaria del rápido crecimiento
demográfico, como algunas autoridades sostienen, quedan otros problemas económicos. La capacidad de
mejorar el capital humano por medio de una amplia variedad de programas, desde la nutrición infantil hasta la educación secundaria formal, pueden verse seriamente limitados si la explosión demográfica continúa. Estos programas suelen ser responsabilidad del estado, y no se puede esperar que los gobiernos que ya son débiles mejoren sus servicios bajo la carga de las presiones demográficas que
aumentan con rapidez la demanda de todo tipo de bienes y servicios públicos.
Por ejemplo, el índice de crecimiento demográfico de Mozambique (3.7 por ciento) es uno de los
más altos del mundo. Su población de más de 17 millones de personas en 1998 aumentó cerca de 3.4
millones en 2003 y es probable que aumente otros 7 millones en 2008. Es una perspectiva sobrecoge-dora, y es difícil imaginar cómo, en tan poco tiempo, Mozambique, que tiene un PIB per cápita de menos de 100 dólares, podrá proporcionar a sus habitantes el capital físico y humano necesario para
mantener, ya no digamos mejorar, los ya de por sí bajos niveles de vida.
Causas del rápido crecimiento demográfico El crecimiento demográfico se determina
por la relación entre nacimientos y muertes, es decir, entre los índices de fertilidad y los índices de
mortalidad. El índice natural de crecimiento demográfico se define como la diferencia entre la tasa de nacimientos y la tasa de defunciones. Si la tasa de nacimientos es de 4 por ciento, por ejemplo, y la de defunciones es de 3 por ciento, la población crece a una tasa de 1 por ciento anual.
Desde un punto de vista histórico, los índices bajos de crecimiento demográfico se mantenían debido a los altos índices de mortalidad, a pesar de los altos niveles de fertilidad. Es decir, las familias tenían muchos hijos, pero las expectativas de vida eran muy bajas en promedio, y muchos niños (y adultos) morían a temprana edad. En Europa y América del Norte, las mejoras en nutrición, en los
programas de salud pública (en especial los relacionados con el agua potable y los servicios sanitarios)
y en las prácticas médicas han producido un descenso en el índice de mortalidad y, por tanto, un crecimiento demográfico más acelerado. A la larga, los índices de fertilidad también se redujeron, y el crecimiento demográfico volvió a ubicarse en una tasa baja y estable.
Los programas de salud pública y la mejor nutrición en los últimos 30 años también han provocado descensos importantes de los índices de mortalidad en las naciones en vías de desarrollo. Sin embargo, los índices de fertilidad no han disminuido con la misma rapidez y el resultado son tasas naturales altas de crecimiento demográfico. El crecimiento reducido de la población depende hasta cierto
punto de las tasas de natalidad más bajas, pero los intentos por disminuir los índices de fertilidad deben tomar en cuenta cómo sienten y actúan las diferentes culturas con respecto a la fertilidad.
La planeación familiar y las modernas formas de control de la natalidad son mecanismos importantes para reducir la fertilidad, pero por sí mismos han tenido éxito muy limitado en la mayoría de los
países donde se han intentado. Si se pretende que las estrategias de planeación familiar tengan éxito,
deben tener sentido para la gente que supuestamente se beneficiará de ellas. Los planificadores de tales
estrategias deben entender por qué las familias en las naciones en vías de desarrollo tienen tantos hijos.
En gran medida, en los países en vías de desarrollo la gente quiere familias grandes porque cree
que las necesita. Los economistas han intentado entender las pautas de fertilidad en los países en vías
de desarrollo y para ello se han enfocado en los factores determinantes de la demanda de niños. En lassociedades agrarias, los niños son fuente de mano de obra agrícola y realizan contribuciones significativas al ingreso familiar. En las sociedades que no tienen programas públicos de seguridad social para
los ancianos, los hijos también pueden proporcionar una fuente de ingresos para los padres que son
demasiado ancianos para sostenerse. Con el alto valor de los niños, aumentado por los altos índices de
mortalidad infantil, no es de extrañar que las familias traten de tener muchos hijos para asegurarse de que
un número suficiente sobrevivirá y llegará a la edad adulta.
Los valores culturales y religiosos también afectan el número de hijos que las familias desean
tener, pero los incentivos económicos para tener familias grandes son muy poderosos. Sólo cuando la
relación entre los costos y beneficios de tener hijos cambie, los índices de fertilidad disminuirán. La expansión de oportunidades de empleo para las mujeres en una economía aumenta los costos de oportunidad de la crianza de los hijos (porque brinda a las mujeres una alternativa más valiosa que tener
hijos) y a menudo produce índices de natalidad más bajos. Los incentivos gubernamentales para las familias pequeñas, como la educación subsidiada para familias que tienen menos de tres hijos, pueden
tener un efecto parecido. En general, los ingresos crecientes parecen disminuir los índices de fertilidad,
lo que indica que el desarrollo económico reduce los índices de crecimiento demográfico.
Las teorías económicas del crecimiento demográfico indican que las decisiones relacionadas con
la fertilidad que toman las familias pobres no deben verse como desinformadas o sin control. Una familia puede pensar que tener muchos hijos es una estrategia racional para la supervivencia económica, dadas las condiciones en que se encuentra. Sin embargo, esto no significa que tener muchos hijos
sea un beneficio neto para la sociedad en su conjunto. Cuando una familia decide tener un gran número de hijos, impone costos al resto de la sociedad; es necesario educar a los niños, atender sus necesidades de salud, etcétera. En otras palabras, lo que conviene a una familia en lo individual puede crear
efectos negativos para el país en su conjunto.
Toda nación que desee desacelerar su índice de crecimiento demográfico quizá necesite ofrecer
incentivos económicos para que las familias tengan menos hijos, así como programas de planeación familiar.
CARGAS DE ENDEUDAMIENTO DE LOS PAÍSES
EN VÍAS DE DESARROLLO
En la década de 1970, los especialistas en desarrollo se preocupaban por las numerosas crisis que enfrentaban los países en vías de desarrollo, pero no por las crisis de endeudamiento. En menos de una década,
el panorama cambió de forma radical. Las dificultades financieras de naciones como Brasil, México y Filipinas ocuparon la primera plana de todos los diarios. Lo que alarmaba a quienes estaban familiarizados
con la situación de endeudamiento no era su posible impacto en las naciones en vías de desarrollo, sino
la convicción de que ésta amenazaba también el bienestar económico de las naciones desarrolladas.
Entre 1970 y 1984, los países en vías de desarrollo pidieron prestado tanto dinero a otros países
que su deuda combinada aumentó 1,000 por ciento, a casi 700,000 millones de dólares. Sólo tres países
—Brasil, México y Venezuela— tenían préstamos pendientes de pago con tres importantes bancos estadounidenses (Citibank, Chase Manhattan y Manufacturer’s Hanover, ahora parte de Chemical) que
ascendían a más del doble del valor neto de esas instituciones financieras. A medida que los países económicamente avanzados entraron en una recesión severa a principios de los años ochenta, el crecimiento de las exportaciones de los países deudores se redujo y muchos se vieron en la situación de que
ya no podían pagar el dinero que debían.
Mientras la situación seguía deteriorándose, muchos temieron que las naciones deudoras pudieran repudiar de manera abierta las deudas y dejar de pagar sus préstamos pendientes. Cuando se presenta incumplimiento (falta de pago) en los préstamos nacionales, por lo general siempre hay una garantía que cubre todo o parte de la deuda restante. En el caso de los préstamos a otro país, es casi
imposible conseguir dicha garantía. Dada la extensa relación con los prestatarios de los países en vías
de desarrollo, los bancos occidentales no querían poner en marcha una pauta de incumplimiento internacional y los prestatarios no deseaban caer en incumplimiento. Los líderes de las naciones en vías
de desarrollo reconocieron que el incumplimiento en los pagos podía traer como consecuencia que se
les negara el acceso a las facilidades bancarias para países desarrollados y a los mercados en los países
industrializados, lo que planteaba obstáculos importantes para futuros esfuerzos de desarrollo.
Varios países reprogramaron su deuda como una solución provisional. De conformidad con un
acuerdo de reprogramación de la deuda, los bancos y los prestatarios negociaron un nuevo programa
para el pago de la deuda existente, que a menudo preveía la cancelación de parte de la deuda y periodos de pago extendidos. A cambio, se esperaba que los países deudores firmaran un convenio con el
FMI para reformar sus políticas económicas con el fin de proporcionar incentivos para ganancias deexportación más altas y menores importaciones. Este tipo de convenio, llamado programa de estabilización, exige por lo general medidas de austeridad, como devaluaciones de la moneda, reducción del
gasto gubernamental y aumentos en la recaudación de impuestos.
A principios de los años noventa, la crisis de la deuda no había terminado, pero se había amortiguado, sobre todo como resultado de los sucesos macroeconómicos que provocaron la reducción de
las tasas de interés. La economía internacional ha revivido un poco, ayudando a algunas naciones a aumentar sus ganancias de las exportaciones. Otros países se han beneficiado de nuevas políticas internas. Sin embargo, otras más, que incluyen a Panamá y muchas naciones africanas, continúan enfrentando cargas de endeudamiento que son inmanejables en el corto plazo.
En 1998, varios países tuvieron problemas económicos graves que requirieron garantías de préstamos de emergencia del FMI. Las economías asiáticas, como Indonesia, Tailandia y Corea, sufrieron colapsos repentinos de sus monedas cuando los acreedores revocaron los préstamos a corto plazo y los inversionistas salieron huyendo. La caída drástica del valor de las divisas dificultó aún más el pago de los
intereses sobre la deuda denominada en dólares y de las importaciones. Circunstancias similares se presentaron en Rusia y Brasil en 1998. Rusia dejó de pagar una serie de convenios de préstamos en agosto
de ese año. En cada caso, el FMI otorgó garantías de créditos de emergencia a cambio del ofrecimiento de disciplina fiscal y austeridad económica. En 2003, estas economías se estaban recuperando y creciendo a ritmo más o menos rápido, pero cada una aún tenía la carga de un pesado endeudamiento.
Entre 2000 y 2003, las tasas de interés bajaron de manera considerable en todo el mundo y, por
fin, se adoptó una medida importante para reducir el problema en septiembre de 2005, cuando el Banco Mundial y el Fondo Monetario Internacional se comprometieron a cancelar hasta 55,000 millones
de dólares en deudas de algunos de los países más pobres del mundo. Un total de 18 países (14 de África y 4 de América Central y del Sur) resultarán beneficiados. Se esperaba que el Fondo para el Desarrollo Africano hiciera lo mismo a principios de 2006. Aunque la cancelación de la deuda no resuelve el
problema del desarrollo por sí misma, sí elimina una importante restricción.
Una lección económica de las últimas dos décadas es que la administración adecuada del capital
extranjero en los países en vías de desarrollo es esencial. Se desperdiciaron muchos préstamos del exterior en proyectos que tenían pocas probabilidades de generar los rendimientos necesarios para pagar
los costos iniciales. En otros casos, las políticas internas que usaban la deuda como sustituto del ajuste
a las nuevas circunstancias económicas resultaron perjudiciales a la larga. En general, el exceso de optimismo sobre las perspectivas de las economías en vías de desarrollo fue inapropiado. Independientemente de todo lo demás que hayamos podido aprender de estos errores, la crisis de la deuda subrayó la
creciente interdependencia de todas la economías, ricas y pobres, grandes y pequeñas.
ECONOMÍAS EN TRANSICIÓN
Durante 40 años, entre el fin de la Segunda Guerra Mundial y mediados de los años ochenta, existió
una poderosa rivalidad entre la Unión Soviética y Estados Unidos. Esta “Guerra Fría” enfrentó a las dos
superpotencias en una lucha de influencias y exacerbó la carrera del armamentismo nuclear. En cierto
momento, la desconfianza mutua entre estos dos países era tan fuerte que el concepto de “destrucción
mutua asegurada” dominaba las relaciones internacionales.
Sin embargo, el mundo comenzó a cambiar a mediados de los años ochenta cuando las estructuras políticas y económicas de la Unión Soviética y los países comunistas de Europa Oriental se empezaron a derrumbar. En 1989, se suscitaron revoluciones relativamente pacíficas en rápida sucesión en
Polonia, Hungría y Checoslovaquia (ahora República Checa). Una revolución sangrienta en Rumania
derrocó a Nicolae Ceausescu, que había gobernado con mano de hierro desde hacía 24 años. El Muro
de Berlín, que había dividido Berlín desde 1961, se vino abajo y Alemania se reunificó. Entonces, en
agosto de 1991, después de un fallido intento de golpe de estado por los comunistas de línea dura, la
propia Unión Soviética empezó a desmoronarse. Hacia finales de 1991, la Unión Soviética (URSS) se
había desintegrado en 15 estados independientes, el más grande de los cuales es la República Rusa.
Diez de estas 15 repúblicas formaron la Mancomunidad de Estados Independientes (MEI) en diciembre de 1991. La Guerra Fría había terminado.
Reflexionamos sobre las rivalidades políticas históricas en un texto de economía por dos razones.
Primera, la lucha que duró 40 años entre Estados Unidos y la Unión Soviética era fundamentalmente
una lucha entre dos sistemas económicos: el capitalismo basado en el mercado (el sistema estadounidense) y el socialismo planeado de forma central (el sistema soviético). Segunda, la Guerra Fría terminó
de manera tan abrupta a finales de los años ochenta porque la economía soviética y las de los países de
Europa Oriental prácticamente se colapsaron durante ese periodo. En cierto sentido, se podría decir
que 1991 fue el año en que el mercado triunfó.¿Y ahora qué? Los estados independientes de la ex Unión Soviética y las otras economías antes comunistas de Europa Oriental luchan por realizar la transición del socialismo planeado de forma central a alguna forma de capitalismo basado en el mercado. En algunos países, como Serbia y Bosnia
Herzegovina, las reformas económicas han quedado relegadas a un segundo plano ante las violentas y
crueles rivalidades étnicas y políticas que habían estado latentes por décadas. En otros países, como
Polonia y Rusia, el problema más importante sigue siendo la transformación económica.
El éxito o el fracaso de esta transición del socialismo planeado de forma central al capitalismo basado en mercado determinarán el curso de la historia. Aunque muchos países han hecho la transición
de un sistema basado en el mercado a un sistema planeado de forma central, lo contrario nunca ha
ocurrido. El proceso ha sido y continuará siendo doloroso y lleno de altibajos. Entre 1989 y 1997, la
producción industrial disminuyó más de 40 por ciento en países como la ex Alemania del Este, Albania, Polonia y Rumania. En Rusia, la producción disminuyó alrededor de 30 por ciento. En todas estas
naciones, personas relativamente prósperas se encontraron de pronto percibiendo ingresos reales
anuales cercanos a los de los habitantes de países subdesarrollados. Para mucha gente, el problema fue
de supervivencia: cómo conseguir suficiente comida y combustible para pasar el invierno.
Hacia 1995, las cosas habían mejorado y aunque persistía cierta incertidumbre y algunos problemas, la producción aumentó en gran parte de Europa Central y Oriental. Un creciente optimismo parecía propagarse. La historia de éxito más importante se presentó en Alemania Oriental, donde la producción real en 1994 creció más de 9 por ciento, el índice de crecimiento más rápido de todas las
regiones de Europa. El auge en la industria de la construcción, el rápido desarrollo de la infraestructura, la baja inflación y el aumento en las exportaciones contribuyeron al éxito de la región. Sin embargo, la situación de Alemania Oriental es única porque fue absorbida por Alemania Occidental, una nación próspera, moderna y completamente desarrollada, que se ha fijado como objetivo principal el
desarrollo de la región oriental.
Europa Central, que incluye a Hungría, Polonia, República Checa, Bulgaria y Rumania, también
alcanzó la estabilidad macroeconómica y empezó a crecer entre 1993 y 1994. Polonia disfrutó del crecimiento económico más rápido del grupo (alrededor de 4.5 por ciento). Estimulado por la inversión
extranjera, la privatización y el espíritu emprendedor, el sector privado polaco representaba en 1992
más de la tercera parte de la producción total de la nación, aunque persisten muchos problemas. Rusia
y las ex repúblicas de la Unión Soviética habían logrado menos hasta 2000. No obstante, las condiciones han mejorado y las perspectivas de éxito son mayores que hace sólo unos cuantos años.
SISTEMAS POLÍTICOS Y SISTEMAS ECONÓMICOS:
SOCIALISMO, CAPITALISMO Y COMUNISMO
Toda sociedad tiene tanto un sistema político como un sistema económico. Por desgracia, las dimensiones políticas y económicas de una sociedad a menudo se confunden.
Los términos democracia y dictadura se refieren a sistemas políticos. Una democracia es un sistema
de gobierno en el que el poder último reside en el pueblo, que toma decisiones gubernamentales ya sea
de manera directa a través del voto o indirecta por medio de sus representantes. Una dictadura es un
sistema político en el que el poder último se concentra ya sea en un pequeño grupo de elite o en una
sola persona.
Desde tiempos históricos, han existido dos sistemas económicos principales: socialismo y capitalismo. Una economía socialista es aquella en que la mayoría del capital (fábricas, equipo, edificios,
ferrocarriles, etcétera) es propiedad del gobierno en lugar de estar en manos de ciudadanos particulares. Propiedad social es otro término que se utiliza para describir este tipo de sistema. Una economía
capitalista es aquella en que la mayor parte del capital es de propiedad privada. Más allá de estos sistemas está un sistema económico puramente teórico llamado comunismo.
El comunismo es un sistema económico en el que el pueblo controla de manera directa los medios
de producción (tierra y capital), sin la intervención de un gobierno o estado. En el mundo concebido
por los comunistas, el estado desaparecería poco a poco y la sociedad planearía la economía de la misma manera que una sociedad colectiva. Aunque algunos países aún se consideran comunistas (como
China, Corea del Norte, Cuba y Tanzania), el gobierno se ocupa de la planeación económica en todos
ellos.
Al comparar las economías de la actualidad, la distinción real está entre el socialismo planeado
de manera central y el capitalismo y no entre el capitalismo y el comunismo.No existen economías socialistas ni economías capitalistas puras. Incluso en la Unión Soviética,
que en esencia era socialista, había un gran sector privado. Una cuarta parte completa de la producción
agrícola de lo que fue la URSS se producía de forma legal en parcelas privadas y se vendía. A la inversa,
un sistema capitalista fuerte, como el de Estados Unidos, apoya a muchas empresas gubernamentales,
entre otras, el sistema postal de Estados Unidos. No obstante, la propiedad pública es la excepción en
Estados Unidos y la propiedad privada era la excepción en la Unión Soviética.
Se debate si algunos tipos específicos de sistemas políticos tienden a asociarse con determinados
sistemas económicos. Estados Unidos y Japón son países que tienen sistemas económicos en esencia
capitalistas e instituciones políticas básicamente democráticas. China y Corea del Norte tienen economías en esencia socialistas donde el poder político se concentra en un solo partido político. Estas observaciones no implican que todos los países capitalistas tengan instituciones políticas democráticas o
que todos los países socialistas estén sujetos al gobierno de un partido totalitario. Algunos países (Singapur, por ejemplo) tienen economías fundamentalmente capitalistas sin sistemas políticos democráticos. Otros países que se ubican mucho más cerca del extremo socialista del espectro económico también mantienen fuertes tradiciones democráticas. Suecia es un ejemplo de un país democrático que
tiene instituciones socialistas fuertes.
Sin embargo, ¿ciertos tipos de sistemas económicos conducen a gobiernos represivos? El economista austriaco Friedrich von Hayek asegura que sí:
Las reformas económicas y la coerción gubernamental son el camino hacia la servidumbre… Las
libertades personal y económica son inseparables. Una vez que se empieza a transitar por el
camino de la regulación gubernamental y la planeación de la economía, la libertad de expresión y de elegir a los líderes políticos se pone en peligro.2
Los recientes acontecimientos en Europa Oriental y Rusia parecen sustentar la tesis de Hayek. Ahí, las
reformas económicas y políticas se están llevando a cabo al mismo tiempo, y se acumulan pruebas de
que la esencia tanto del sistema de mercado como de la democracia es la libertad individual.
No obstante, algunos refutan el argumento de Hayek aduciendo que la reforma social y la participación activa del gobierno en la economía son las únicas maneras de impedir el advenimiento de un
estado totalitario. Sostienen que los mercados libres y sin regulación provocan desigualdad y acumulación del poder económico. El poder económico acumulado, a su vez, conduce al poder político que
se utiliza de manera inevitable para favorecer los intereses de los pocos ricos, y no los intereses de todos.
PLANEACIÓN CENTRAL FRENTE AL MERCADO
Además de la propiedad del capital, los sistemas económicos también difieren en el grado hasta el cual
las decisiones económicas se toman a través de la planeación central en lugar del sistema de mercado.
En algunas economías socialistas, la asignación de los recursos, la mezcla de productos y la distribución de la producción se determinan de forma central de acuerdo con un plan. La ex Unión Soviética,
por ejemplo, generaba planes anuales y quinquenales en los que se establecían objetivos específicos de
producción en casi todos los sectores de la economía. En las economías de mercado, las decisiones se
toman de manera independiente por compradores y vendedores que responden a las señales del mercado. Los productores producen sólo lo que esperan vender. La mano de obra es atraída a diversas ocupaciones por los salarios que determinan las fuerzas de la oferta y la demanda.
Así como no hay economías capitalistas o socialistas puras, tampoco hay economías de mercado
puras ni economías planeadas puras. Incluso en la ex Unión Soviética existían los mercados y éstos determinaban, en buena medida, la asignación de los recursos. Muchas entidades, entre ellas el Pentágono, establecen los objetivos de producción en Estados Unidos.
En general, las economías socialistas favorecen la planeación central sobre la asignación del mercado, en tanto que las economías capitalistas dependen en mucha mayor medida del mercado. No obstante, existe cierta variedad. La ex Yugoslavia era un país socialista que hacía gran uso del mercado. La
propiedad del capital y la tierra correspondía al gobierno, pero las empresas determinaban en lo individual sus precios y sus propios niveles de producción, y elaboraban sus planes de inversión. Las empresas yugoslavas pedían préstamos a los bancos para financiar inversiones y pagaban intereses sobre
sus préstamos. Este tipo de sistema, que combina la propiedad del mercado con la asignación del mercado se llama economía socialista de mercado.EL FIN DE LA UNIÓN SOVIÉTICA
Ningún debate serio sobre asuntos económicos tuvo lugar en la Unión Soviética sino hasta después de
la muerte de Josef Stalin en 1953. En 1965, el gobierno de Alexei Kosygin introdujo reformas oficiales.
Mikhail Gorbachev anunció una serie de reformas en 1986 y 1987, pero la estructura de la economía
no cambió de manera radical.
Aunque las ideas de Gorbachev parecían prometedoras, la situación de la Unión Soviética se deterioró de forma severa en 1987. El intento de transición de la planeación central a un sistema parcial
de libre mercado ocasionó problemas serios. El crecimiento de la producción se redujo a prácticamente nada en 1989 y 1990, y en 1991 el sistema económico se vino abajo. La producción industrial
cayó en picada, la escasez de alimentos empeoró, la inflación se agravó y la deuda externa aumentó
con rapidez.
A Gorbachev se le acababa el tiempo en agosto de 1991 cuando la lucha entre los partidarios de la
línea dura y los reformadores radicales llegó a ser frontal. Los partidarios de la línea dura encarcelaron
a Gorbachev y asumieron el control del gobierno. El golpe de estado duró sólo tres días. El pueblo tomó las calles de Moscú y opuso resistencia a los tanques, el ejército soviético se negó a obedecer órdenes y los partidarios de la línea dura tuvieron que retirarse.
No obstante, el fin estaba cerca tanto para Gorbachev como para la Unión Soviética. En diciembre de 1991, la Unión Soviética se desintegró, diez de las ex repúblicas soviéticas formaron la MEI y
Boris Yeltsin llegó a la presidencia de la República Rusa cuando Gorbachev pasó a formar parte de la
historia. Desde el principio, Yeltsin demostró ser un reformador comprometido con convertir de manera expedita la economía rusa en un sistema de mercado y mantener, al mismo tiempo, las libertades
políticas del pueblo, ganadas con tantos trabajos. Su plan de reforma preveía la desregulación de los
precios, la privatización de empresas públicas y la estabilización de la macroeconomía.
Boris Yeltsin fue presidente de la República Rusa y paladín de la reforma de 1991 a 1999. Yeltsin liberalizó la mayoría de los precios, inició el proceso de privatización e intentó estabilizar la macroeconomía. En 1995, el progreso había sido lento, pero significativo. La privatización había logrado avances
continuos y se llegó al punto en 1994 en que el sector privado generaba 60 por ciento del ingreso personal. La inflación había disminuido y una nueva “constitución económica”, en forma de leyes revisadas para establecer los derechos de propiedad y estimular la actividad económica, entró en vigor en
1995. No obstante, las cosas no iban bien en lo general.
La mayoría de los observadores calculan que el PIB ruso empezó a crecer en la primavera de 1997,
después de caer alrededor de 50 por ciento con respecto a su nivel de 1989. Sin embargo, los problemas
seguían siendo graves en 1998. El mayor problema residía en atraer inversiones y el mayor obstáculo
era la delincuencia y la corrupción. Al gobierno ruso le estaba resultando muy difícil hacer valer el imperio de la ley. El término capitalismo vaquero se utiliza a menudo para describir la situación actual en
Rusia.

LA TRANSICIÓN A UNA ECONOMÍA DE MERCADO
Las reformas que se pusieron en marcha en la República Rusa y en otros países ex comunistas de Europa Oriental han tomado forma muy despacio y en medio del debate sobre cómo sería mejor proceder.
Recuerde que no hay en lo absoluto algún precedente histórico que ofrezca lecciones. A pesar de esta
falta de precedentes, existe un acuerdo considerable entre los economistas respecto a lo que es necesario hacer.
SEIS REQUISITOS BÁSICOS DE LA TRANSICIÓN EXITOSA
Los economistas por lo general están de acuerdo en que existen seis requisitos básicos para realizar una transición exitosa del socialismo a un sistema basado en el mercado: (1) estabilización
macroeconómica; (2) desregulación de los precios y liberalización del comercio; (3) privatización de las empresas propiedad del estado y desarrollo de nueva industria privada; (4) establecimiento de instituciones que apoyen al mercado, como las leyes de propiedad y contratos, los sistemas de contabilidad, etcétera; (5) una red de seguridad social para hacer frente al desempleo y
la pobreza; y (6) asistencia externa.Ahora se analizará cada componente. Aunque nos centraremos en la experiencia de la República
Rusa, estos principios aplican a todas las economías en transición.
Estabilización macroeconómica Prácticamente todos los países en transición han tenido
problemas con la inflación, pero en ninguna parte ha sido peor que en Rusia. A medida que las condiciones económicas empeoraban, el gobierno comenzó a tener problemas presupuestarios graves. Los
flujos de ingresos se desaceleraron y los compromisos de gastos aumentaron, lo que provocó un enorme
déficit presupuestario. Al mismo tiempo, cada una de las nuevas repúblicas estableció su propio banco
central. Cada banco central empezó a emitir “créditos en rublos” para mantener a flote a las empresas
importantes y pagar las cuentas del gobierno. La emisión de estos créditos, que por lo general se aceptaban como medio de pago en todo el país, provocó un crecimiento considerable de la oferta de dinero.
Casi desde el principio, la mayor oferta de dinero significó que se usaba demasiado dinero para
pagar muy pocos bienes. Esta situación empeoró con los precios controlados por el gobierno, que se fijaron muy por debajo de los niveles de equilibrio del mercado. La combinación de expansión monetaria y control de precios fue mortal. Las tiendas operadas por el gobierno, que vendían productos a precios controlados, estaban vacías. La gente hacía largas filas durante días y a menudo se violentaba
cuando sus esfuerzos por comprar bienes a precios oficiales bajos se veían frustrados. Asimismo, los
proveedores descubrieron que podían cobrar precios mucho más altos por sus productos en el mercado negro, que crecía día a día y exacerbaba aun más la escasez de productos en las tiendas gubernamentales. Con el tiempo, el valor del rublo se reducía de forma paulatina mientras los precios en el
mercado negro aumentaban con mayor rapidez. Rusia se encontraba cerca de la hiperinflación en
1992. Para lograr un sistema de mercado que funcione de la manera correcta, es necesario estabilizar
los precios. Para hacerlo, el gobierno debe encontrar la forma de equilibrar el presupuesto y controlar
la oferta de dinero.
Desregulación de precios y liberalización del comercio Para pasar con éxito de la
planeación central a un sistema de mercado, los precios deben desregularse. Un sistema de precios que
se mueven con libertad constituye la columna vertebral de un sistema de mercado. Cuando la gente
quiere más de un producto de lo que se produce en la actualidad, el precio aumenta. Este precio más
alto aumenta las utilidades de los productores y ofrece un incentivo para que las empresas existentes
expandan la producción y para que nuevas empresas entren en la industria. A la inversa, si una industria produce un bien para el cual no hay mercado o un producto que la gente ya no quiere en la misma
cantidad, el resultado será una oferta excedente y el precio de ese producto se reducirá. Este resultado
disminuye las utilidades o crea pérdidas, lo que ofrece un incentivo para que algunas empresas existentes reduzcan la producción y otras cierren. En resumen, un mecanismo de precios sin regulación asegura la distribución eficiente de los recursos entre las industrias. Hasta que los precios se desregulen,
este mecanismo no podrá funcionar.
Las barreras comerciales también deben eliminarse. Los países con mentalidad reformista deben
poder importar capital, tecnología e ideas. Además, no tiene sentido seguir subsidiando industrias que
no pueden ser competitivas en los mercados mundiales. Si es más barato comprar acero a una fábrica
eficiente en Alemania Occidental que producirlo en una fábrica rusa anticuada y subsidiada, la fábrica
rusa debe modernizarse o cerrar. En última instancia, como indica la teoría de la ventaja comparativa,
el comercio liberalizado presionará a cada país para que produzca los bienes que produce mejor.
La desregulación de los precios y la eliminación de los subsidios pueden provocar graves problemas
políticos. Muchos productos en Rusia y el resto del mundo socialista tenían precios por debajo de los niveles de equilibrio del mercado, por razones de equidad. Muchos consideraban que la vivienda, comida y
ropa constituían derechos. Es improbable que sea una medida popular hacerlos más caros, por lo menos
en relación con sus precios en épocas anteriores. Además, obligar a empresas ineficientes a operar sin
subsidios hará que muchas cierren y se perderán empleos. Así, aunque la desregulación de los precios y la
liberación del comercio son necesarias, también son muy difíciles de lograr en términos políticos.
Privatización Uno de los problemas que plantea un sistema de propiedad central es la falta de
responsabilidad. En un sistema de propiedad privada, los dueños cosechan las recompensas de sus éxitos y sufren las consecuencias de sus fracasos. La propiedad privada ofrece un fuerte incentivo para la
operación eficiente, la innovación y el trabajo arduo, que brillan por su ausencia cuando la propiedad
es centralizada y las utilidades se distribuyen entre la gente.
La historia clásica para ilustrar este punto se llama tragedia de los comunes. Suponga que una comunidad ganadera tiene 10,000 hectáreas de pastizales. Si la tierra fuera de propiedad común, de modoque todos los rancheros tuvieran derechos ilimitados para alimentar a su ganado, cada ranchero tendría
un incentivo para utilizar la tierra de forma excesiva para el pastoreo. El ranchero cosecharía los beneficios plenos de llevar a pastar terneros adicionales, mientras que los costos de alimentar a los terneros se cubrirían de manera colectiva. El sistema no ofrece ningún incentivo para administrar la tierra
con eficiencia. Asimismo, si la eficiencia y los beneficios de mi trabajo esforzado y competencias administrativas se acumulan para otros o para el estado, ¿qué incentivo tengo para trabajar duro o para
ser eficiente?
Una solución de la tragedia de los comunes que se probó en Inglaterra en el siglo XVIII fue dividir
la tierra en parcelas de propiedad privada. En la actualidad, sostienen muchos economistas, la solución
del problema de los incentivos que se encuentra en las empresas propiedad del estado consiste en privatizarlas y dejar que los dueños compitan.
Además de una mayor responsabilidad, la privatización significa crear un ambiente propicio para
que las nuevas empresas prosperen. Si hay demanda en el mercado de un producto que en la actualidad no se produce, los empresarios deben gozar de la libertad de poner un negocio y obtener utilidades.
En los últimos meses de existencia de la Unión Soviética, las empresas privadas, como los servicios de
taxi, servicios de reparación de automóviles, restaurantes e incluso los hoteles empezaron a aparecer
por todo el país.
Al igual que la desregulación de los precios, la privatización es difícil en el aspecto político. La privatización implica cerrar muchas empresas protegidas porque no pueden competir a los precios mundiales,
lo que provoca pérdida de empleos, cuando menos de manera temporal.
Instituciones que apoyan al mercado Entre 1991 y 1997, las empresas estadounidenses se
apresuraron a ir a Europa Oriental en busca de mercados y oportunidades, y de inmediato se dieron
cuenta de que había un enorme obstáculo. Las instituciones que hacían que el mercado funcionara relativamente sin contratiempos en Estados Unidos no existían en Europa Oriental. Por ejemplo, el mercado
de capitales, que canaliza el ahorro privado hacia la inversión productiva de capital en las economías capitalistas desarrolladas, se compone de cientos de instituciones diferentes. El sistema bancario, los fondos
de capital de riesgo, el mercado de valores, el mercado de bonos, las bolsas de comercio, las casas de corretaje, los bancos de inversión, etcétera, se han desarrollado en Estados Unidos a lo largo de cientos de
años y simplemente no pueden replicarse de la noche a la mañana en el mundo ex comunista.
Muchas instituciones que apoyan al mercado son tan elementales que los estadounidenses las dan
por descontadas. La institución de la propiedad privada, por ejemplo, es un conjunto de derechos que deben protegerse por leyes que el gobierno debe estar dispuesto a hacer cumplir. Suponga que la cadena
francesa de hoteles Novotel decide construir un nuevo hotel en Moscú. Novotel debe adquirir el terreno,
primero que nada. Luego, construirá un edificio basado en la expectativa de alquilar cuartos a los clientes. Estas inversiones se realizan con la confianza de que el dueño tiene el derecho a usarlas y el derecho a
las utilidades que producen. Para que tales inversiones se realicen, estos derechos deben estar garantizados por un conjunto de leyes que rijan la propiedad. Esto es igualmente válido para las grandes empresas
mercantiles y para los empresarios rusos que desean establecer sus propias empresas.
Asimismo, la ley debe contener disposiciones para hacer respetar los contratos. En Estados Unidos, un vasto cuerpo de leyes determina lo que le ocurrirá a uno si rompe una promesa formal hecha
de buena fe. Las empresas existen bajo la premisa de producir y la promesa de pagar. Si no se puede recurrir a las leyes cuando se incumple un contrato, no se celebrarán contratos, no se fabricarán productos y
no se prestarán servicios.
Otro asunto en apariencia sencillo que resulta ser bastante complejo es el establecimiento de un
conjunto de principios de contabilidad. En Estados Unidos, las normas de contabilidad están plasmadas en un conjunto de principios de contabilidad generalmente aceptados (GAAP) que tienen fuerza y
efecto legal. Se requiere que las empresas lleven el control de sus ingresos, gastos y pasivos para que los
accionistas, las autoridades fiscales y otros interesados en la compañía puedan observar y evaluar su
desempeño. Si usted ha tomado un curso de contabilidad, sabrá que estas normas se han vuelto muy detalladas. Imagine tratar de hacer negocios en un país que funciona con cientos de conjuntos de reglas
diferentes. Eso es lo que ha estado pasando en Rusia.
Otra institución son los seguros. Siempre que una empresa lleva a cabo una actividad de alto riesgo, compra un seguro para protegerse. Hace varios años, Amnistía Internacional (una organización sin
fines de lucro que trabaja para proteger las libertades civiles en todo el mundo) patrocinó una gira
mundial de conciertos con una serie de grupos de rock e intérpretes famosos. La parte más difícil de
organizar la gira fue obtener seguros para los artistas y su equipo cuando tocaban en los países entonces comunistas de Europa Oriental.Red de seguridad social En una economía socialista planeada de forma central, el mercado
de mano de obra no funciona con libertad. Se garantiza un empleo en alguna parte a todo aquel que
quiera trabajar. El número de empleos está determinado por un plan central que coincide con el número de trabajadores. En esencia, no hay desempleo. Se ha argumentado que esto es una de las grandes
ventajas de un sistema planeado. Además, un sistema de planeación central ofrece vivienda básica, alimentación y ropa a niveles muy accesibles para todos. Sin desempleo y con los productos de primera
necesidad disponibles a muy bajos precios, no hay necesidad del seguro de desempleo, asistencia social
u otros programas sociales.
La transición a un mercado libre de mano de obra y la liberalización de los precios implican que
algunos trabajadores acabarán desempleados y que todos pagarán precios más altos por los productos
de primera necesidad. En efecto, durante las primeras etapas del proceso de transición, el desempleo es
alto. Las empresas estatales ineficientes tienen que cerrar; algunos sectores se contraen en tanto que
otros se expanden. A medida que cada vez más personas experimentan el desempleo, es probable que el
apoyo popular a las reformas disminuya, a menos que se establezca alguna especie de red de seguridad
social para facilitar la transición. Esta red de seguridad social podría incluir el seguro de desempleo,
ayuda a los pobres y asistencia para alimentación y vivienda. Las experiencias del mundo desarrollado
han demostrado que dichos programas son costosos.
Asistencia externa Muy pocos creen que la transición a un sistema de mercado puede lograrse sin apoyo externo y cierto financiamiento del exterior. El conocimiento y la experiencia de las instituciones capitalistas que existen en Estados Unidos, Europa Occidental y Japón son de interés vital para las naciones de Europa Oriental. Las competencias básicas de contabilidad, administración y
desarrollo empresarial pueden enseñarse en Europa Oriental; muchos dicen que conviene a todos hacerlo. Otros más argumentan también que la mayor pesadilla del mundo es una Rusia débil desde el
punto de vista económico o desesperada, equipada con armamento nuclear, que sacrifique las reformas o caiga en manos de un dictador.
Sin embargo, existe poco acuerdo respecto al grado de apoyo financiero que debe brindarse. Estados Unidos ha presionado para que se haga un esfuerzo mundial por proporcionar miles de millones
de dólares en ayuda. Esta ayuda, muchos argumentan, ayudará a Rusia a estabilizar su macroeconomía
y comprar en el extranjero bienes que necesita con desesperación. No obstante, los críticos en Estados
Unidos y otros posibles países donadores dicen que invertir dinero en Rusia en este momento es como
tirarlo en un pozo sin fondo: sin importar cuánto se done, tendrá poco impacto en el éxito o fracaso
último de las reformas.
¿Terapia de choque o gradualismo? Aunque los economistas por lo general están de
acuerdo en lo que las antiguas economías socialistas deben hacer, debaten la secuencia y la oportunidad de reformas específicas.
La prensa popular describe el debate como si éste se desarrollara entre quienes creen en la “terapia
de choque” (a veces llamada método del Big Bang) y quienes prefieren un método más gradual. Los
partidarios de la terapia de choque creen que las economías en transición deben actuar de inmediato
en todos los frentes. Deben dejar de imprimir dinero, desregular los precios y liberalizar el comercio,
privatizar, crear instituciones de mercado, construir una red de seguridad social y obtener ayuda externa: todo tan rápido como sea posible. Según el argumento, el dolor será atroz, pero al final quedará en
el olvido cuando la transición eleve los niveles de vida. Los partidarios del método gradualista creen
que el mejor camino es construir primero las instituciones de mercado, suprimir de manera gradual el
control de los precios y privatizar primero sólo las empresas del gobierno más eficientes.
Los que están a favor de actuar con rapidez señalan el aparente éxito de Polonia, que avanzó con
rapidez por las primeras etapas de la reforma. La experiencia de Rusia en los primeros años de la transición ha demostrado que, por lo menos en ese país, el cambio debe ser gradual hasta cierto punto. En
teoría, la estabilización y la liberación de los precios pueden alcanzarse al instante. Para disfrutar de los
beneficios de la liberalización, tienen que realizarse muchas privatizaciones y esto tarda más tiempo.
Un analista ha dicho que privatización significa “vender activos sin valor a gente sin dinero”. Algunos
cálculos indican que la mitad de las empresas rusas de propiedad estatal son incapaces de obtener utilidades a los precios mundiales. Abandonarlas simplemente crearía un caos. En cierto sentido, Rusia
no ha tenido más remedio que actuar despacio.

### Resumen 

1. Los problemas económicos que enfrentan los países en vías de
desarrollo a menudo son muy diferentes de los que enfrentan las
naciones industrializadas. Las opciones de política disponibles
para los gobiernos también pueden diferir. No obstante, las herramientas del análisis económico son útiles tanto para comprender las
economías de los países menos desarrollados como la economía de
Estados Unidos.

pobreza. Aunque existe una diversidad considerable entre las naciones en vías de desarrollo, la mayoría de las personas que viven en

3. En casi todas las naciones en vías de desarrollo hay escasez de capital físico en relación con los demás recursos, en especial la mano de
obra. La hipótesis del círculo vicioso de la pobreza dice que los países
pobres no pueden escapar de la pobreza porque no pueden darse el

de lo que muchos países en vías de desarrollo son en la actualidad.
Sin embargo, a menudo es difícil movilizar los ahorros con eficiencia en muchas naciones subdesarrolladas.
4. El capital humano (el acervo de educación y competencias personificado en los trabajadores) desempeña una función vital en el desarrollo económico.
5. Los países en vías de desarrollo a menudo se ven agobiados por la
insuficiencia del capital para gastos generales sociales, que va desde
malas instalaciones sanitarias y de salud pública hasta caminos, teléfonos y sistemas judiciales inadecuados. Con frecuencia, es difícil
proporcionar este capital para gastos generales sociales y muchos
gobiernos tan sólo no están en posición de emprender muchos proyectos útiles porque éstos son demasiado costosos.
6. Debido a que las economías desarrolladas se caracterizan por una
participación importante de la producción y el empleo en el sector
industrial, muchos países en vías de desarrollo parecen creer que el
desarrollo y la industrialización son sinónimos. En muchos casos,
los países en vías de desarrollo han privilegiado la industria a costa
de la agricultura, con resultados diversos. La información reciente
indica que cierto equilibrio entre la industria y la agricultura produce el mejor resultado.
7. Las políticas de sustitución de las importaciones, una estrategia comercial que favorece el desarrollo de las industrias locales que pueden fabricar productos para reemplazar las importaciones, fueron
muy comunes en alguna época en las naciones en vías de desarrollo. En general, tales políticas no han alcanzado el mismo éxito que
las que promueven las abiertas, orientadas hacia las exportaciones.
8. El fracaso de muchos esfuerzos de planeación central ha suscitado
que cada vez más se pida menos intervención del gobierno y más
orientación hacia el mercado en las economías en vías de desarrollo.
PROBLEMAS DEL DESARROLLO ECONÓMICO
9. El rápido crecimiento demográfico es característico de muchos países en vías de desarrollo. Las familias grandes pueden ser racionales
en términos económicos para los padres que necesitan sustento en
la vejez o porque los hijos ofrecen una importante fuente de mano
de obra. Sin embargo, tener muchos hijos no significa un beneficio
neto para la sociedad en su conjunto. El rápido crecimiento demográfico puede ejercer más presión sobre los servicios públicos, ya
sobrecargados, como la educación y la salud.
10. Entre 1970 y 1984, la deuda de los países en vías de desarrollo creció 10 veces. A medida que la recesión se afianzaba en los países
avanzados durante los primeros años de la década de 1980, el crecimiento de las exportaciones de los países deudores se desaceleró y
muchos no pudieron pagar el dinero que debían. La perspectiva de
que los países en vías de desarrollo incumplieran el pago de los
préstamos amenazó a todo el sistema financiero internacional y
transformó la crisis de la deuda en un problema global.
ECONOMÍAS EN TRANSICIÓN
11. En una economía socialista la mayoría del capital es propiedad del
gobierno y no de los ciudadanos particulares. En una economía capitalista la mayoría del capital es de propiedad privada. El comunismo es un sistema económico teórico en el que el pueblo controla de
manera directa los medios de producción (el capital y la mano
de obra) sin la intervención de un gobierno o estado.
12. Las economías difieren hasta el grado al cual las decisiones se toman a través de la planeación central en lugar del sistema de mercado. En general, las economías socialistas favorecen la planeación
central sobre la distribución del mercado y las economías capitalistas dependen en una medida mucho mayor del mercado. No obstante, hay mercados en todas las sociedades y la planeación tiene
lugar en todas las economías.
EL FIN DE LA UNIÓN SOVIÉTICA
13. La Unión Soviética creció con rapidez hasta mediados de la década
de 1970. A finales de los años cincuenta, la economía de la Unión
Soviética creció mucho más rápido que la de Estados Unidos. La
clave de los primeros éxitos soviéticos fue la rápida acumulación de
capital planeado. A finales de los años setenta, las cosas empezaron
a deteriorarse. Mikhail Gorbachev introdujo finalmente reformas
drásticas, después de que llegó al poder en 1985. No obstante, la
economía soviética se derrumbó en 1991. La Unión Soviética se desintegró y el nuevo presidente de la República Rusa, Boris Yeltsin,
tuvo que emprender la difícil tarea de realizar la transición a un
sistema de mercado.
LA TRANSICIÓN A UNA ECONOMÍA DE MERCADO
14. Los economistas por lo general están de acuerdo en los seis requisitos de una transición exitosa del socialismo a un sistema basado en
el mercado: (1) estabilización macroeconómica; (2) desregulación
de los precios y liberalización del comercio; (3) privatización; (4)
establecimiento de instituciones que apoyen al mercado; (5) una
red de seguridad social; y (6) asistencia externa.
15. Existe un gran debate sobre la secuencia y la oportunidad de reformas específicas. La idea de la terapia de choque es actuar de inmediato en los seis frentes, incluida la rápida desregulación de los precios y la privatización. El método gradualista consiste en crear
primero las instituciones de mercado, suprimir de manera gradual
el control de los precios y privatizar primero sólo las empresas más
eficientes del gobierno.
