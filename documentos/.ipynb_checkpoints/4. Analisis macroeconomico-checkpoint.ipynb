### 1.3 Las conexiones entre los mercados de bienes y de dinero

Hay dos conexiones esenciales entre los mercados de bienes y de dinero:

- **Primera conexión: ingreso y demanda de dinero** La primera conexión entre el mercado
de bienes y el mercado de dinero se debe a que la demanda de dinero depende del ingreso.
Cuando aumenta la producción total (ingreso, Y), se incrementa el número de transacciones que requieren dinero (acabamos de ver este efecto en el capítulo 11). Un crecimiento de
la producción, con la tasa de interés constante, origina un acrecentamiento en la demanda
de dinero.

- **Segunda conexión: gasto en inversión planeada y tasa de interés** La segunda conexión
entre el mercado de bienes y el mercado de dinero obedece a que el gasto en inversión planeada (I) depende de la tasa de interés (r). En los capítulos 8 y 9 supusimos que el gasto en
inversión planeada es fijo en cierto nivel, pero lo hicimos tan sólo para simplificar. En la
práctica, la inversión no es fija, sino que depende del número de las principales variables económicas. Una es la tasa de interés. Cuanto mayor sea la tasa de interés, menor será el gasto en
inversión planeada.

Conexiones entre los mercados de bienes y de dinero

**El ingreso, que se determina en el mercado de bienes, tiene una influencia considerable sobre la demanda en el mercado de dinero:**
>
>> La demanda de dinero depende del ingreso.
>
> **Mercado de bienes:** 
>
>> $Y = C + I + G$ (determina $Y$) 
>>
>> $\downarrow$ _Primer conexión_ $\downarrow$ 
>
> **Mercado de dinero:** 
>
>> La demanda de dinero depende de $Y$ (determina $r$)

**La tasa de interés, que se determina en el mercado de dinero, tiene efectos importantes sobre la inversión planeada en el mercado de bienes:**
>
>> La inversión planeada depende de la tasa de interés. 
>
> **Mercado de bienes:** 
>
>> $I$ depende de $r$
>>
>> $\uparrow$ _Segunda conexión_ $\uparrow$ 
>
> **Mercado de dinero:** 
>
>> La demanda de dinero depende de $Y$

#### 1.3.1 Inversión, tasa de interés y mercado de bienes

No debe causarle ninguna sorpresa que haya una relación inversa entre el nivel de la inversión
planeada y la tasa de interés.

> **Cuando la tasa de interés baja, la inversión planeada aumenta.**
>
> **Cuando la tasa de interés sube, la inversión planeada disminuye.**

Para ver la razón de estos efectos, recuerde que inversión se refiere a las compras que hace una
empresa de capital fresco: máquinas y fábricas nuevas. Que una empresa decida invertir en un
proyecto depende de que las utilidades esperadas justifiquen los costos. Por lo regular, un gasto
grande para un proyecto de inversión es el costo de los intereses.
Pensemos en una empresa que abre una planta nueva o en la inversión que se requiere para
poner una heladería nueva. Cuando una firma de manufactura construye una fábrica nueva, hay
que pagar al contratista cuando termine la edificación. Cuando un emprendedor decide abrir una
nevería, necesita refrigeradores, mesas, sillas, iluminación y letreros. Por todo esto hay que pagar
cuando lo instalan.
Por lo común, el dinero necesario para realizar tales proyectos se toma a préstamo y se paga
en un plazo extenso. En buena medida, el costo real de un proyecto de inversión depende de la tasa de interés: el costo de tomar un préstamo. Cuando la tasa de interés aumenta, se encarece el
préstamo y es probable que se emprendan menos proyectos; si aumenta la tasa de interés, ceteris
paribus, se reduce el gasto en inversión planeada. Cuando la tasa de interés baja, se abarata el
préstamo y es posible que se lleven a cabo más proyectos de inversión; si se reduce la tasa de interés, ceteris paribus, es probable que aumente el gasto en inversión planeada.
La curva de la demanda de la figura 12.2, que tiene pendiente en descenso, ilustra la relación
entre la tasa de interés y la inversión planeada. Cuanto mayor sea la tasa de interés, menor será el nivel de la inversión planeada. Con una tasa de interés de 3 por ciento, la inversión planeada es I
0
.
Cuando la tasa de interés aumenta de 3 a 6 por ciento, la inversión planeada disminuye de I
0 a I
1
. Sin
embargo, al bajar la tasa más proyectos se vuelven rentables, así que se emprenden más inversiones.

Esquema de la inversión planeada

<img src="4. Analisis macroeconomico\Inflación, demanda y oferta agregadas\1.3.1 Inversión, tasa de interés y mercado de bienes-1.PNG" width=600 height=400 />

_El gasto en inversión planeada es una función negativa de la tasa de interés. Un aumento en la tasa de interés de 3 a 6 por ciento reduce la inversión planeada de $I_0$ a $I_1$._

Ahora podemos tomar en cuenta el hecho de que la inversión planeada depende de la tasa de
interés para considerar el efecto de esta relación en el gasto total planeado (GT). Recuerde que el
gasto total planeado es la suma de consumo, inversión planeada y compras del gobierno. Es decir:

$GT ≡ C + I + G$

Ya sabemos que en realidad hay muchos niveles posibles de I, cada uno correspondiente a una
tasa de interés. Cuando la tasa de interés cambia, la inversión planeada varía. Por consiguiente, un
cambio en la tasa de interés (r) produce también un cambio en el gasto total planeado (C + I + G).1
En la figura 12.3 se muestra lo que ocurre al gasto total planeado cuando la tasa de interés
aumenta de 3 a 6 por ciento. Con una tasa mayor, la inversión planeada es menor; así, el gasto total planeado se desplaza hacia abajo. Recuerde que en los capítulos 8 y 9 vimos que un descenso
en cualquier componente del gasto total tiene un efecto todavía mayor (o “multiplicador”) sobre
el ingreso de equilibrio (Y). Cuando la tasa de interés se incrementa, la inversión planeada (y el
gasto total planeado) cae, y la producción (ingreso) de equilibrio baja más, incluso, que la inversión planeada. En la figura 12.3, Y de equilibrio baja de Y0 a Y1 cuando la tasa de interés crece de
3 a 6 por ciento.
Resumamos los efectos de un cambio en la tasa de interés en el nivel de equilibrio de la producción.

> **Los efectos de un cambio en la tasa de interés son:**
>
> **■ Una tasa de interés ($r$) elevada desalienta la inversión planeada ($I$).**
>
> **■ La inversión planeada es parte del gasto total planeado ($GT$).**
>
> **■ Así, cuando la tasa de interés sube, baja el gasto total planeado ($GT$) en todos los niveles de ingreso.**
>
> **■ Por último, una reducción del gasto total planeado aminora la producción (ingreso) de equilibrio ($Y$) en un múltiplo de la disminución inicial de la inversión planeada.**

Efecto de un aumento de la tasa de interés en el gasto total planeado

<img src="4. Analisis macroeconomico\Inflación, demanda y oferta agregadas\1.3.1 Inversión, tasa de interés y mercado de bienes-2.PNG" width=600 height=400 />

_Un incremento en la tasa de interés de 3 a 6 por ciento disminuye el gasto total planeado y así se reduce el ingreso de equilibrio de $Y_0$ a $Y_1$._

Si acudimos a un esquema simbólico cómodo, tenemos:

> **$r \uparrow \rightarrow I \downarrow \rightarrow GT \downarrow \rightarrow Y \downarrow$**
>
> **$r \downarrow \rightarrow I \uparrow \rightarrow GT \uparrow \rightarrow Y \uparrow$**

Como se aprecia, el nivel de equilibrio de la producción (Y) no está determinado exclusivamente por los acontecimientos del mercado de bienes, como habíamos supuesto en explicaciones
previas simplificadas. El motivo es que el mercado de dinero afecta el nivel de la tasa de interés, que
también repercute en la inversión planeada en el mercado de dinero. Hay un nivel diferente de equilibrio de Y por cada nivel posible de la tasa de interés (r). El nivel final de equilibrio de Y depende de
cuál resulte ser la tasa de interés, lo cual depende de los acontecimientos del mercado de dinero.

#### 1.3.2 Demanda de dinero, producción total (ingreso) y mercado de dinero

Acabamos de ver que la tasa de interés (que se determina en el mercado de dinero) influye en el
nivel del gasto en inversión planeada y, por consiguiente, en el mercado de bienes. Ahora veremos
el otro lado de la historia: la manera en que el mercado de bienes incide en el mercado de dinero.
En el capítulo 11 exploramos la demanda de dinero de hogares y empresas y explicamos por
qué depende en forma negativa de la tasa de interés. Un aumento de la tasa de interés eleva el costo de oportunidad de quedarse con el dinero, hecho que no genera intereses (a diferencia de los
bonos que sí los pagan), lo cual alienta a la gente a tener una parte mayor de sus fondos en bonos
y menos en los saldos de las cuentas de cheques. En la figura 12.4 se muestra la curva de la demanda de dinero, con pendiente a la baja (Md
).
También vimos en el capítulo 11 que la demanda de dinero depende del nivel de ingreso de
la economía. Más ingreso deriva en más transacciones y un volumen mayor de éstas representa
una demanda más grande de dinero. Como hay más personas con mejores ingresos y que adquieren más bienes y servicios, se demanda más dinero para satisfacer el aumento en el volumen de
las transacciones. Por consiguiente, un incremento del ingreso desplaza la curva de la demanda
de dinero a la derecha (consulte la figura 11.5).
Si, como suponemos, la decisión de la Reserva Federal (Fed) sobre la cantidad del circulante no depende de la tasa de interés, entonces la curva de la oferta de dinero es una línea vertical.

> **El nivel de equilibrio de la tasa de interés no se determina únicamente en el mercado de dinero. Las variaciones en la producción total (ingreso, Y), que ocurren en el mercado de bienes, desplazan la curva de la demanda y propician cambios en la tasa de interés. Para una cantidad dada del circulante, niveles mayores de Y traen niveles más altos de equilibrio de r. Niveles menores de Y determinan niveles de equilibrio de r más bajos. Esto se representa con los siguientes símbolos:**
>
> **$Y \uparrow \rightarrow M^d \uparrow \rightarrow r \uparrow$**
>
> **$Y \downarrow \rightarrow M^d \downarrow \rightarrow r \downarrow$**

### 1.4 Unión de los mercados de bienes y de dinero

Ahora que estamos al tanto de las conexiones entre los mercados de bienes y de dinero, podemos
examinarlos al mismo tiempo. Para ver cómo se relacionan, será conveniente investigar los efectos sobre la economía de cambios en las políticas fiscales y monetarias. Queremos analizar lo que
sucede con los niveles de equilibrio de la producción total (ingreso, Y) y la tasa de interés (r)
cuando bajan o suben ciertas variables fundamentales, sobre todo el gasto gubernamental (G), los
impuestos netos (I) y el circulante (Ms).

#### 1.4.1 Efectos de una política expansiva

Se dice de toda política pública destinada a estimular la producción total (ingreso, Y) es expansiva. Una política fiscal expansiva es un aumento del gasto gubernamental (G) o una reducción
de los impuestos netos (T) dirigida a incrementar la producción total (ingreso, Y). Una política
monetaria expansiva es un acrecentamiento del circulante con la finalidad de elevar la producción total (ingreso, Y).
**Política fiscal expansiva: un aumento del gasto gubernamental (G) o una
reducción de los impuestos netos (T)** Como aprendimos en el capítulo 9, las compras gubernamentales (G) y los impuestos netos (T) son las dos herramientas de la política fiscal
del gobierno. Éste puede estimular la economía, es decir, puede intensificar la producción total
(ingreso, Y), ya sea con un aumento de sus compras o con una reducción de los impuestos netos.
El efecto de una rebaja de impuestos es algo menor que el de una ampliación de G, pero los dos
tienen un efecto multiplicador en el nivel de equilibrio de Y.
Tomemos por ejemplo un incremento de las compras gubernamentales (G) de 10,000 millones. Este crecimiento del gasto hace que las existencias de las empresas sean menores que lo
planeado. Las reducciones imprevistas del inventario estimulan la producción (Y), que las empresas aumentan. No obstante, como la producción agregada significa más ingreso, parte del
cual se gasta a continuación, también crece el gasto de consumo (C). Otra vez las existencias serán
menores que lo planeado y la producción se elevará aún más. El nivel de equilibrio definitivo es
mayor por un múltiplo del incremento inicial en las compras del gobierno.
Pero la historia del multiplicador está incompleta. Hasta este capítulo hemos supuesto que la
inversión planeada (I) es fija en cierto nivel, pero ahora sabemos que depende de la tasa de interés. Asimismo, podemos examinar lo que pasa con el multiplicador cuando la inversión varía
porque tenemos conocimientos del mercado de dinero, y allí se determina la tasa de interés.
Volvamos a la historia de nuestro multiplicador en el momento en que las empresas comienzan a ampliar la producción en respuesta a un crecimiento de las compras del gobierno. Cuando
se incrementa la producción total (ingreso, Y), se resiente un impacto en el mercado de dinero: el
aumento del ingreso (Y) ensancha la demanda de dinero (Md
). (Por el momento, supongamos
que la Fed mantiene fijo el circulante, Ms
). El desequilibrio que se aproxima, en el que la cantidad
de dinero demandada es mayor que el volumen del circulante, eleva la tasa de interés. El incremento de G hace que Y y r aumenten.
El aumento de r tiene un efecto colateral: una tasa de interés más costosa abate el gasto en inversión planeada (I). Como este gasto es un componente del gasto total planeado (C + I + G), la
disminución de I obra en contra del aumento de G. Un incremento del gasto gubernamental (G)
acrecienta el gasto total planeado y la producción total, pero una reducción de la inversión planeada aminora el gasto total planeado y disminuye la producción total.
Esta tendencia a que el gasto del gobierno genere reducciones del gasto de la inversión privada se llama efecto de expulsión. Sin una expansión del circulante que asimile el aumento del
ingreso y el crecimiento de la demanda de dinero, el gasto en inversión planeada es desplazado,
en parte porque la tasa de interés es más alta. El gasto adicional creado por el incremento de las
compras gubernamentales se compensa, en alguna medida, por la caída del gasto de la inversión
planeada. De cualquier forma, el ingreso crece, aunque el efecto del multiplicador del aumento
en G se minimiza por el impacto negativo sobre la inversión planeada de una tasa de interés más
alta.
Este efecto de expulsión se ilustra con la gráfica de la figura 12.6. Un aumento de las compras gubernamentales de G0 a G1 desplaza en sentido ascendente la curva del gasto total planeado de (C + I
0 + G0
). El incremento en (Y) de Y0 a Y1 eleva la demanda de dinero, lo cual da por
resultado un desequilibrio en el mercado de dinero. El exceso de la demanda de dinero eleva la
tasa de interés y rebaja I de I
0 a I
1
. La caída de I hace descender la curva del gasto planeado, y
esto abate el nivel de equilibrio del ingreso a Y *. (Recuerde que se alcanza el equilibrio cuando
Y = GT.)
Observe que la magnitud del efecto de expulsión y el tamaño final del multiplicador del
gasto gubernamental dependen de varias cosas. En primer lugar, supusimos que la Fed no cambió la cantidad del circulante. Si, por el contrario, creyéramos que la Fed amplía el circulante
para asimilar el aumento de G, el multiplicador sería más grande. En este caso, la mayor demanda de dinero quedaría satisfecha con una cantidad superlativa de circulante y la tasa de interés
no aumentaría. Sin una tasa más alta, no se daría este efecto de expulsión.
En segundo lugar, el efecto de expulsión depende de la sensibilidad o la insensibilidad a la
tasa de interés de la inversión planeada. El efecto de expulsión ocurre porque una tasa de interés
más alta reduce el gasto en inversión planeada. Ahora bien, la inversión depende de otros factores,
aparte de la tasa de interés, y muchas veces es insensible a los cambios de ésta. Si la inversión planeada no baja cuando sube la tasa de interés, no se produce el efecto de expulsión. A continuación se resumen estos efectos:

El efecto de expulsión

<img src="4. Analisis macroeconomico\Inflación, demanda y oferta agregadas\1.4.1 Efectos de una política expansiva-1.PNG" width=600 height=400 />

_Un aumento del gasto gubernamental, G, de G0 a G1 desplaza el esquema del gasto total planeado de
1 a 2. A continuación, el efecto de expulsión de la baja de la inversión planeada (causado por el aumento de la tasa de interés) mueve el esquema del gasto total planeado de 2 a 3._

> **Efectos de una política fiscal expansiva:**
>
> **$G \uparrow \rightarrow Y \uparrow \rightarrow M^d \uparrow \rightarrow r \uparrow \rightarrow I \downarrow$** 
>
> **$Y$ aumenta menos que si $r$ no se incrementara**

El mismo razonamiento se aplica con puntualidad a los cambios en los impuestos netos. El
efecto definitivo de una reducción de impuestos sobre el nivel de equilibrio de la producción depende de cómo reaccione el mercado de dinero. La expansión de Y que origina una rebaja de impuestos
lleva a un aumento de la tasa de interés y, así, a una disminución del gasto en inversión planeada. Por
consiguiente, el aumento decisivo de Y es menor de lo que sería si la tasa de interés no subiera.
**Política monetaria expansiva: un aumento del circulante** Ahora consideremos lo que pasa si la Fed decide aumentar el circulante mediante operaciones de mercado abierto.
Al principio, las operaciones en mercados abiertos inyectan nuevas reservas al sistema y extienden
la cantidad del circulante (la curva de la oferta de dinero se desplaza a la derecha). Como la cantidad
del circulante es mayor de lo que quieren conservar los hogares, la tasa de interés de equilibrio cae.
El gasto de inversión planeada, que es un componente del gasto total programado, crece cuando
la tasa de interés baja.
El aumento del gasto en inversión planeada significa que entonces el gasto total proyectado es
mayor que la producción total. Las empresas pasan por reducciones imprevistas de existencias y elevan la producción (Y). Una extensión del circulante abate la tasa de interés e incrementa Y. Sin embargo, el nivel más alto de Y acrecienta la demanda de dinero (la curva de la demanda se desplaza a
la derecha) y esto impide que la tasa de interés baje tanto como ocurriría en otras circunstancias.
Si repasamos la secuencia de acontecimientos que siguen a la expansión monetaria, se ve la relación entre la inyección de reservas en la economía hecha por la Reserva Federal y el incremento
de la producción. En primer lugar, el aumento del circulante rebaja la tasa de interés. En segundo,
una tasa disminuida eleva el gasto en inversión planeada. Tercero, la ampliación del gasto en inversión planeada significa que hay más gasto total programado, a lo que corresponde un crecimiento
de la producción cuando las empresas reaccionan a la mengua imprevista de las existencias. Cuarto, el acrecentamiento de la producción (ingreso) hace crecer la demanda de dinero (la curva de la
demanda de dinero se desplaza a la derecha), lo cual significa que la tasa de interés se reduce menos de lo que habría hecho si la demanda de dinero no hubiera aumentado.

> **Efectos de una política fiscal expansiva:**
>
> **$M^s \uparrow \rightarrow r \downarrow \rightarrow I \uparrow \rightarrow Y \uparrow \rightarrow M^d \uparrow$** 
>
> **$r$ disminuye menos que si $M^d$ no se incrementara**

La fuerza de la política monetaria para influir en el mercado de bienes depende del grado de
la reacción en cada eslabón de la cadena. Quizá el paso más relevante sea la conexión entre I y r.
La política monetaria es eficaz sólo cuando I reacciona a los cambios de r. Si de repente las empresas
incrementan el número de proyectos de inversión que emprenden cuando la tasa de interés cae, la
política monetaria expansiva consigue estimular la economía. En cambio, si se muestran reacias
a invertir aun con una tasa baja, la política rinde pocos frutos. En otras palabras, la eficacia de la
política monetaria depende de la pendiente de la función de la inversión. Si es casi vertical (lo cual
indica muy poca sensibilidad de la inversión a la tasa de interés), el eslabón central de la cadena es
débil y vuelve ineficaz a la política monetaria.
**La política expansiva en la práctica: las recesiones de 1974-1975,
1980-1982, 1990-1991 y 2001** Estados Unidos ha sufrido cuatro recesiones desde
1970. Entre 1974 y 1975 y de 1980 a 1982 el gobierno efectuó reducciones de impuestos que tuvieron el efecto de estimular el gasto de consumo (C). Como C es un componente del gasto total
planeado, estas rebajas fiscales hicieron posible un aumento de la producción total (ingreso, Y).

Consideremos la recesión de 1974 a 1975. La Ley de Reducción de Impuestos de 1975 trajo por
resultado una disminución de impuestos de 8,000 millones de dólares, en 1974, que pagaron los
consumidores en el segundo trimestre de 1975. Ésta y otras rebajas de impuestos aumentaron el
gasto de consumo, lo cual contribuyó a la recuperación económica que se inició poco después
de que entraran en vigor las nuevas leyes fiscales.
¿Pero qué pasó con el efecto de expulsión? ¿La política fiscal expansiva de 1975 elevó las tasas
de interés y desplazó el gasto privado? En este caso, no. Al mismo tiempo que el Congreso recortaba
los impuestos para fomentar el gasto, la Fed trataba de estimular la economía aumentando el
circulante. Aunque el incremento de la producción durante la expansión hizo que creciera la
demanda de dinero, la Fed ampliaba simultáneamente la oferta de dinero y las tasas de interés no
cambiaron mucho. En la figura 12.7 se muestra la situación.
Una concatenación semejante de acontecimientos tuvo lugar durante la recesión del periodo
1980-1982. Por recomendación del presidente Ronald Reagan, el Congreso aprobó un recorte
enorme de impuestos en el verano de 1981. Ésta, como la reducción de 1975, trajo un incremento
del gasto de consumo que sirvió para sacar a la economía de la recesión.

La recuperación de la recesión de 1980-1982 también recibió el impulso de la Fed, que comenzó a incrementar vigorosamente el circulante en la primavera de 1982. Así, aunque la producción y el ingreso crecían a finales de ese año, lo cual incrementaba la demanda de dinero, las tasas

La recesión de 1990 a 1991 comenzó poco después de la invasión de Irak a Kuwait, a finales
del verano de 1990. La recesión fue efímera y somera, en comparación con las dos recesiones previas. El PIB real comenzó a aumentar en el segundo trimestre de 1991, pero este restablecimiento

grandes siguieron recortando la nómina (aunque la producción mejoraba), la tasa de desempleo
se mantuvo elevada hasta las elecciones presidenciales de 1992.
El primer presidente Bush debatió si debía aplicarse una reducción de impuestos para activar la
economía. Pero preocupado por el déficit gubernamental, que ya era grande, y presionado por la Reserva Federal, se convenció de esperar. La Fed bajó las tasas de interés con la intención de reactivar la
economía. Con todo, para la época de las elecciones pocas respuestas se habían hecho patentes.
El presidente Clinton defendió un modesto estímulo fiscal cuando asumió el cargo, pero el
Congreso se opuso. Luego, en el verano de 1993, el Congreso aprobó el paquete de Clinton para reducir el déficit, en el que se aumentaban los impuestos y se reducía el gasto del gobierno. Entre tanto,
la política monetaria siguió siendo expansiva. Con el tiempo, las tasas de interés llegaron a sus niveles más bajos en 30 años. Durante buena parte de 1993, la tasa de los certificados de la Tesorería
a tres meses fueron inferiores a 3 por ciento por primera vez desde 1962 y la tasa del bono a 30
años bajó a menos de 6 por ciento por primera ocasión desde que el gobierno empezó a vender
esos bonos.
A finales de 1994 terminó la lenta recuperación del crecimiento y comenzó una expansión verdadera, alimentada en gran medida por el auge del mercado accionario que comenzó en 1995. El
apogeo llegó a su fin en 2000 y la economía estadounidense entró en una recesión en 2001. La política monetaria respondió pronto a la recesión. En ese año, la Fed redujo la tasa de interés de corto plazo en casi cuatro puntos porcentuales. Fue acaso la mayor medida de política monetaria
expansiva de la historia de aquel país. La recesión fue moderada y a finales del cuarto trimestre de
2001, la economía volvió a crecer. No obstante, la recuperación no fue muy firme y la Reserva Federal mantuvo baja la tasa de interés de corto plazo en 2002, 2003 y 2004. Además, la política fiscal fue
muy expansiva entre 2001 y 2004, como veremos en el capítulo 15.

#### 1.4.2 Efectos de una política restrictiva

Se dice que toda estrategia gubernamental dirigida a reducir la producción total (ingreso, Y) es
restrictiva. Si la política expansiva se aplica para fomentar la economía, la restrictiva se usa para
desacelerarla.
Si pensamos que una de las cuatro principales metas de la economía es el crecimiento económico (capítulo 1), ¿por qué un gobierno adopta políticas destinadas a reducir el gasto total?
Como veremos en los dos capítulos siguientes, una manera de combatir la inflación es reducir el
gasto total. Cuando la inflación es alta, el gobierno llega a sentirse obligado a ejercer sus facultades
para contraer la economía. Antes de que analicemos las políticas restrictivas que el gobierno ha
emprendido en los últimos años, tenemos que estudiar el funcionamiento de las políticas fiscal
y monetaria restrictiva.
**Política fiscal restrictiva: disminución del gasto gubernamental (G) o aumento de impuestos netos (T)** Una política fiscal restrictiva es una reducción del gasto gubernamental (G) o un incremento de los impuestos netos (T) con la finalidad de reducir la
producción total (ingreso, Y). Los efectos de esta política son opuestos a los resultados de una
política fiscal expansiva.
Una disminución de las compras del gobierno o un incremento de los impuestos netos lleva
a una reducción de la producción total (ingreso, Y), de la demanda de dinero (Md
) y de la tasa de
interés (r). La restricción de Y que viene con una política fiscal restrictiva es menor de lo que sería si no tomáramos en cuenta el mercado de dinero, porque la disminución de r también hace
que la inversión planeada (I) aumente. Este incremento de I compensa parte de la reducción del
gasto total planeado causado por el descenso de G. (Esto también significa que el efecto del multiplicador es menor de lo que sería si no tomáramos en cuenta el mercado de dinero.) Los efectos
de una reducción de G o un aumento de T se representan como sigue:

> **Efectos de una política fiscal restrictiva:**
>
> **$G \downarrow$ o $T \uparrow \rightarrow Y \downarrow \rightarrow M^d \downarrow \rightarrow r \downarrow \rightarrow I \uparrow$** 
>
> **$Y$ disminuye menos que si $r$ no se incrementara**

**Política monetaria restrictiva: disminución del circulante** Una política monetaria restrictiva es una disminución del circulante con la finalidad de reducir la producción
agregada (e ingreso, Y). Como recordará, el nivel del gasto en inversión planeada es una función
negativa de la tasa de interés: mientras más alta sea la tasa de interés, menor será la inversión planeada. Cuanto menor sea la inversión planeada, menor será el gasto total planeado y menor el
nivel de equilibrio de la producción (ingreso, Y). El ingreso de equilibrio más bajo da por resultado
una reducción de la demanda de dinero, lo cual significa que el aumento de la tasa de interés será
menor de lo que habría sido si no hubiéramos tomado en cuenta el mercado de bienes.

> **Efectos de una política monetaria restrictiva:**
>
> **$M^s \downarrow \rightarrow r \uparrow \rightarrow I \downarrow \rightarrow Y \downarrow \rightarrow M^d \downarrow$** 
>
> **$r$ aumenta menos que si $M^d$ no se incrementara**

**La política restrictiva en la práctica: 1973-1974 y 1979-1981** La Fed ha tomado medidas restrictivas vigorosas dos veces, en la década de los setenta y la siguiente: primero en el periodo 1973-1974 y de nuevo entre 1979 y 1981. En 1974, las tasas de interés de corto
plazo fueron superiores a 12 por ciento y en 1981 algunas pasaron de 20 por ciento. Estas tasas
elevadas tuvieron un efecto negativo sobre el gasto total planeado y contribuyeron a las recesiones que siguieron. El objetivo de la Fed al seguir una política monetaria estricta era desacelerar la inflación. (En el capítulo siguiente veremos por qué una política restrictiva puede reducir la
inflación.)

#### 1.4.3 La combinación de políticas macroeconómicas

Hemos tratado por separado la política fiscal y monetaria, pero debe quedar claro que pueden
usarse de manera simultánea. Por ejemplo, es posible aumentar al mismo tiempo las compras gubernamentales (G) y el circulante (Ms
). Hemos visto que un incremento de G eleva tanto Y como r,
mientras que un acrecentamiento de Ms eleva Y pero disminuye r. Por lo tanto, si el gobierno
quiere incrementar Y sin cambiar r, puede hacerlo subiendo G y Ms en los montos apropiados.
El paquete de políticas es el grupo de políticas monetarias y fiscales que se siguen en determinado periodo. Un paquete de políticas que consiste en una disminución del gasto del gobierno
y un aumento del circulante favorecería más al gasto de inversión que al gubernamental. La explicación estriba en que el incremento del circulante y la baja de las compras gubernamentales
reducirían la tasa de interés, lo cual conduciría a un crecimiento de la inversión planeada. Lo
opuesto es cierto de un paquete que consiste en una política fiscal expansiva y una política monetaria restrictiva. Este paquete favorece más al gasto gubernamental que al de inversión. Tal estrategia tendrá el efecto de elevar el gasto gubernamental y reducir el circulante. La contracción del
circulante y la expansión del gasto gubernamental suben la tasa de interés y bajan la inversión
planeada.
No hay reglas sobre qué constituye el “mejor” paquete de políticas o la “mejor” composición
de la producción. Sobre este tema, como sobre muchos más, no se ponen de acuerdo los economistas (y otros). En parte, la preferencia de una persona por cierta composición de la producción
(digamos, una muy inclinada al gasto privado con poco gasto gubernamental) depende de su
postura con respecto a temas como la función apropiada del gobierno en la economía.
En la tabla 12.1 se resumen los efectos de diversas combinaciones de políticas sobre distintas
variables macroeconómicas importantes. Si usted es capaz de explicar los motivos de todos los
efectos indicados en la tabla, debe sentirse satisfecho de que tiene una buena comprensión de las
conexiones que hay entre los mercados de bienes y de dinero.

### 1.5 Otros determinantes de la inversión planeada

En este capítulo hemos supuesto que la inversión planeada depende sólo de la tasa de interés. En
realidad, la inversión planeada depende de otros factores. Vamos a estudiarlos más en el capítulo
17, pero aquí daremos una breve descripción.
**Expectativas y espíritus animales** Las expectativas de las empresas acerca de sus
ventas futuras cumplen una importante función en sus decisiones de inversión. Cuando una
empresa invierte, acrecienta sus existencias de capital (y éste se usa en el proceso de producción).

Efectos del paquete de políticas macroeconómicas

<img src="4. Analisis macroeconomico\Inflación, demanda y oferta agregadas\1.5 Otros determinantes de la inversión planeada.PNG" width=600 height=400 />

Si una empresa espera que sus ventas aumenten en el futuro, ahora puede comenzar a acumular
sus existencias de capital (es decir, a invertir) para que pueda producir más en el futuro con el fin
de cubrir el aumento de las ventas. El optimismo o pesimismo de los emprendedores sobre el curso
futuro de la economía puede tener un efecto importante sobre la inversión planeada actual.
Keynes acuñó la expresión espíritus animales para referirse a los sentimientos de los emprendedores
y decía que influyen en las decisiones de inversión.
**Tasa de utilización del capital** El grado de utilización de las existencias de capital de
una empresa también podría incidir en la inversión planeada. Si la demanda de lo que produce
una compañía ha estado disminuyendo y como respuesta decide aminorar su producción, quizá
tenga una tasa baja de utilización del capital. Puede ser costoso deshacerse pronto del capital que
ya se tiene, de modo que cuando cae la producción, a veces las empresas conservan el capital pero
lo utilizan menos; por ejemplo, las máquinas trabajan pocas horas al día o a menor velocidad.
Las empresas invierten menos en capital nuevo cuando sus tasas de utilización del capital son bajas, que cuando son altas.
**Costos relativos de mano de obra y capital** El costo del capital (del que la tasa de
interés es el principal componente) en relación con el costo de la mano de obra puede afectar la inversión planeada. Si la mano de obra es cara en relación con el capital (tasas salariales altas), las
empresas tienden a sustituir trabajo con capital. Tienen el objetivo de retener más capital, en comparación con la mano de obra, cuando las tasas salariales son elevadas que cuando son bajas.

> **Los determinantes de la inversión planeada son:**
>
> **■ Tasa de interés**
>
> **■ Expectativas sobre ventas futuras**
>
> **■ Tasas de utilización del capital**
>
> **■ Costos relativos de capital y mano de obra**

de dinero: el nivel de la producción real (ingreso, Y), que se establece en el mercado de bienes, determina el volumen de las transacciones de cada periodo y así afecta la demanda del mercado de
dinero, y la tasa de interés (r), que está determinada en el mercado de dinero, influye en el nivel del gasto en inversión planeada
en el mercado de bienes.
3. Hay una relación inversa entre la inversión planeada y la tasa de
interés, ya que ésta determina el costo de los proyectos de inversión. Cuando la tasa de interés aumenta, la inversión planeada
disminuye; cuando la tasa baja, la inversión planeada sube.
4. Para cada valor de la tasa de interés hay un nivel diferente de gasto
en inversión planeada y un nivel distinto de producción de
equilibrio. El nivel definitivo de la producción de equilibrio depende de cuál sea la tasa de interés, la cual es resultado de los acontecimientos del mercado de dinero.
5. Para una cantidad dada de circulante, la tasa de interés depende
de la demanda de dinero. Ésta, a su vez, depende del nivel de la
producción (ingreso). Con determinado circulante, las altas y bajas de Y inciden en la demanda de dinero, lo cual repercute en la
tasa de interés de equilibrio.

6. Una política fiscal expansiva es un aumento del gasto del
gobierno (G) o una reducción de los impuestos netos (T) destinado a aumentar la producción total (ingreso, Y). Una política
fiscal expansiva basada en incrementos del gasto gubernamental
genera un efecto de expulsión: como la ampliación del
gasto gubernamental significa que hay más transacciones en la
economía y, por ello, una demanda mayor de dinero, la tasa de
interés sube. La disminución del gasto en inversión planeada que
viene con el aumento de la tasa de interés se compensa en parte
con el incremento del gasto total causado por la expansión de G.
7. La magnitud del efecto de expulsión, que influye en la medida del
multiplicador del gasto gubernamental, depende de dos cosas: la
suposición de que la Fed no cambia la cantidad del circulante y
la sensibilidad o insensibilidad de la inversión planeada a los cambios de la tasa de interés.
8. Una política monetaria expansiva es un aumento del
circulante dirigido a incrementar la producción total (ingreso, Y).
Un crecimiento del circulante rebaja la tasa de interés, aumenta
la inversión planeada y el gasto total planeado y, en última instancia,
se llega a un nivel de equilibrio superior de la producción total
(ingreso, Y). Se aplican políticas expansivas para sacar a la
economía de recesiones.
9. Una política fiscal restrictiva es una disminución del gasto
gubernamental o un incremento de los impuestos netos con el fin
de reducir la producción total (ingreso, Y). Una disminución
del gasto gubernamental o un alza de los impuestos netos trae
una reducción de la producción total (ingreso, Y), de la demanda
de dinero y de la tasa de interés. Ahora bien, la disminución de Y
se compensa en parte por la inversión planeada adicional que es
resultado de la baja de la tasa de interés.
10. Una política monetaria restrictiva es una disminución del
circulante destinada a reducir la producción total (ingreso, Y).
La tasa de interés más alta generada por la reducción del circulante
rebaja el gasto en inversión planeada, lo mismo que el nivel de
equilibrio de la producción. Pero este nivel inferior de la producción
causa una restricción de la demanda de dinero, lo cual significa
que el incremento de la tasa de interés es menor de lo que sería
si no tomáramos en cuenta el mercado de bienes. Se aplican
políticas restrictivas para combatir la inflación.
11. El paquete de políticas es el grupo de estrategias fiscales y
monetarias que se aplican en determinado periodo. No hay una
regla sobre qué constituya el mejor paquete de políticas o la mejor
composición de la producción. En parte, las preferencias de una
persona por cierta composición de la producción dependen de
su postura con respecto a temas como la función apropiada del
gobierno en la economía.

### 2.3 Curva de la demanda agregada

El lugar para comenzar nuestra exploración del nivel de precios es el mercado de dinero. (Si ha
olvidado los detalles sobre cómo se calcula el nivel general de precios o qué significa, repase los
capítulos 6 y 7.) Como vimos en el capítulo 11, la demanda de dinero de la gente depende del ingreso (Y), la tasa de interés (r) y el nivel de precios (P).
No es difícil entender por qué el nivel de precios afecta la demanda de dinero. Supongamos
que quiere comprar medio kilo de chocolate, una bolsa de papas fritas y un pastelito Hostess. Si
esos artículos cuestan siete dólares, dos y uno, usted necesita 10 dólares en efectivo o en su cuenta
para hacer obtenerlos. Supongamos que el precio de estos bienes se duplica; entonces, para adquirirlos necesita 20 dólares.
En general, el dinero requerido para hacer cierto número de transacciones depende en forma
directa y proporcional del precio promedio de esas transacciones. Cuando precios y salarios aumentan, los hogares quieren tener más dinero en la bolsa y la cuenta del banco, las empresas necesitan más efectivo en la caja chica, etc. Si precios y salarios aumentan 6 por ciento al año, anticipamos
que la demanda de dinero aumente alrededor de 6 por ciento anual, ceteris paribus.

> **La demanda de dinero está en función de tres variables: la tasa de interés (r), el nivel del ingreso real (Y) y el nivel de precios (P). (Recuerde que Y es la producción real, o el ingreso. Mide el volumen real de la producción, sin consideraciones sobre el cambio de precios.) La demanda de dinero aumenta si se incrementan los niveles real de la producción (ingreso) y de precios, o si baja la tasa de interés.**

#### 2.3.1 Derivación de la curva de la demanda agregada

Recuerde que la demanda agregada es la demanda total de bienes y servicios de la economía. Para derivar la curva de la demanda agregada, examinamos lo que sucede con la producción total
(ingreso, Y) cuando cambia el nivel de precios (P). ¿La producción crece, disminuye o se mantiene constante si aumentan los precios? Nuestro estudio de los mercados de bienes y dinero nos
ofrece las herramientas para responder esta pregunta.
La curva de la demanda agregada se deriva suponiendo que no cambian las variables de la
política fiscal [compras gubernamentales (G) e impuestos netos (T)] ni de la política monetaria
(Ms
). En otras palabras, pensamos que el gobierno no emprende ninguna medida para influir en
la economía a fin de responder a los cambios del nivel de precios.
Como sabemos, un aumento del nivel de precios incrementa la demanda de dinero y desplaza
su curva a la derecha, según se ilustra en la figura 13.1(a). Con la tasa de interés inicial de 6 por

mayor de los precios, hogares y empresas tienen que conservar saldos mayores de dinero que antes.
Sin embargo, la cantidad del circulante es la misma. (Recuerde que suponemos que la Reserva Federal no toma ninguna medida para cambiar el circulante.) El mercado de dinero se desequilibra.
El equilibrio se restablece con una tasa de interés superior, de 9 por ciento.

> **Cada par de valores de P y Y en la curva de la demanda agregada corresponden a un punto donde los mercados de bienes y de dinero están en equilibrio.**

#### 2.3.2 Advertencia sobre la curva de la demanda agregada

Es muy importante no perder de vista lo que representa la curva de la demanda. Como se estableció en el capítulo 5, la curva de la demanda agregada es mucho más compleja que la curva de la
demanda de un solo individuo o un mercado. La curva DA no es la correspondiente al mercado,
tampoco es la suma de todas las curvas de demanda del mercado en la economía.
Para entender por qué, recuerde el razonamiento en que se basa la curva simple de la demanda
de un hogar con pendiente en descenso. Una curva de la demanda muestra la cantidad de producción demandada (por un hogar o un mercado único) a todos los precios posibles, ceteris paribus.
Al trazar una curva de demanda simple, suponemos que los otros precios y el ingreso son fijos.
De estas suposiciones se deduce que una causa de que baje la cantidad demandada de un bien
cuando su precio aumenta, es que los otros precios no se incrementan. Por lo tanto, el bien de que se trata se vuelve más caro en relación con otros artículos, y la reacción de los hogares es sustituir el
producto encarecido con otros. Además, si el ingreso no mejora con el precio de un bien, el ingreso real baja. Esto también abate la demanda de la mercancía cuyo precio se elevó.

#### 2.3.3 Otros motivos de una curva de la demanda agregada con pendiente descendente

Además de los efectos de la oferta y la demanda de dinero en la tasa de interés, otros dos factores explican la pendiente descendente de la curva DA: la conexión con el consumo y el efecto de la riqueza real.
**Conexión con el consumo** Dijimos en el capítulo 12 (y veremos con mayor detalle en
el capítulo 17) que el consumo (C) y la inversión planeada (I) dependen de la tasa de interés.
En igualdad de circunstancias, los gastos de consumo tienden a aumentar cuando la tasa de
interés baja y se reducen cuando ésta crece; hace lo mismo que la inversión planeada. Esta tendencia es otra conexión entre los mercados de bienes y de dinero. Si algún suceso modifica la
tasa de interés en el mercado de dinero, en el de bienes se afectan el consumo y la inversión
planeada.
La conexión con el consumo es otra causa de que la pendiente de la curva DA descienda. Un
aumento en el nivel de precios acrecienta la demanda de dinero, lo cual lleva a un incremento de
la tasa de interés y esto trae una reducción del consumo (y de la inversión planeada), y esto es
causa de una disminución de la producción total (ingreso). La baja inicial del consumo (ocasionada por el alza de la tasa de interés) contribuye a la mengua general de la producción

> **La inversión planeada no lleva todo el peso de ser el eslabón entre una tasa de interés más alta y un nivel menor de la producción. El consumo aminorado por la tasa de interés más alta también contribuye a este efecto.**

**Efecto de la riqueza real** También anotamos en el capítulo 8 (y detallaremos en el capítulo 17) que el consumo depende de la riqueza. En igualdad de circunstancias y mientras más
riqueza tengan los hogares, más consumirán. La riqueza incluye posesiones de dinero, acciones,
bonos y bienes raíces, entre otros. Si este patrimonio mengua, habrá menos consumo ahora y en
el futuro.
El nivel de precios tiene un efecto en algunas clases de riqueza. Supongamos que usted tiene
1,000 dólares en una cuenta o en un fondo del mercado de dinero y que el nivel de precios sube
10 por ciento. Su posesión ahora vale 10 por ciento menos porque los bienes que podría comprar con 1,000 dólares se encarecieron 10 por ciento. El poder de compra (o “valor real”) de su dinero
disminuyó 10 por ciento.
Un incremento del nivel de precios también puede disminuir el valor real de acciones y bienes raíces, aunque todo depende de qué pase con las cotizaciones de éstos cuando sube el nivel
general de precios. Si los precios de las acciones y las casas aumentan en la misma proporción
que el nivel general de precios, su valor real no cambia. El punto es:

> **Un aumento del nivel de precios reduce el valor real de algunas formas de riqueza.**

El hecho de que el nivel de precios disminuye el valor real de la riqueza es otra causa de la
pendiente a la baja de la curva DA. Un aumento del nivel de precios reduce el valor real de la riqueza. Esto genera una merma del consumo que, a su vez, da lugar a un menoscabo de la producción total (ingreso). Así, hay una relación negativa entre el nivel de precios y la producción a través
de este **efecto de la riqueza real** o **efecto del saldo real**.

#### 2.3.4 Gasto total y demanda agregada

Hasta este punto de nuestro estudio de la macroeconomía, nos hemos referido a la inversión total planeada como el gasto total planeado de hogares ($C$), empresas ($I$) y gobierno ($G$). En equilibrio, el gasto total planeado ($GT ≡ C + I + G$) y la producción total (ingreso, $Y$) son iguales:

> **Condición de equilibrio: $C + I + G = Y$**

¿Cómo se relaciona el gasto total planeado con la demanda total?

> **En todo punto de la curva de la demanda total, la cantidad total demandada es exactamente igual al gasto total planeado, $C + I + G$.**

Lo anterior puede apreciarse en las figuras 13.1 y 13.2. Cuando el nivel de precio aumenta, lo que
baja es el gasto total planeado, y con esto ascendemos en la curva de la demanda agregada.
No obstante, la curva de la demanda agregada representa más que el puro gasto total planeado. Cada punto de la curva DA representa el nivel particular del gasto total planeado que
corresponde al equilibrio de los mercados de bienes y de dinero al precio dado. En la figura
13.2, observe que la variable sobre el eje horizontal de la curva de la demanda agregada es Y. En
todo punto de la curva DA, Y = C + I + G.

### 2.4 Curva de la oferta agregada

La oferta agregada es la oferta total de bienes y servicios de la economía. Aunque hay pocos desacuerdos entre los economistas sobre la lógica de la curva de la demanda agregada, hay grandes
discrepancias sobre el razonamiento que respalda la curva de la oferta agregada. También hay
divergencias sobre su forma.

#### 2.4.1 Advertencia sobre la curva de la oferta agregada

La curva de la oferta agregada (OA) indica la relación entre la suma de la producción de todas
las empresas de una economía y el nivel general de precios. Para entender la curva de la oferta
agregada, tenemos que saber algo sobre el comportamiento de las empresas que componen la
economía.
Parecería lógico derivar la curva de la oferta agregada sumando las curvas de todas las empresas de la economía. Sin embargo, la lógica de la relación entre el nivel general de precios de la
economía y el nivel de la producción total (ingreso) —es decir, de la curva OA— es muy diferente
de la lógica de la curva de la oferta de una empresa individual. La curva de la oferta agregada no
es una curva de oferta del mercado, tampoco es la simple suma de todas las curvas individuales
de la oferta en la economía. (Recuerde que hicimos una advertencia semejante respecto a la curva de la demanda agregada.)
Para entender el motivo, recuerde la razón de la curva de la oferta simple que presentamos
en el capítulo 3. Una curva de la oferta muestra la cantidad de producción que arrojaría una empresa a cada precio posible, ceteris paribus. Cuando trazamos la curva de la oferta de una firma,
suponemos que los precios de los insumos, incluyendo tasas salariales, son fijos. La curva de la
oferta de una compañía indica lo que pasaría con su producción si el precio de lo que hace cambia
sin un aumento ajustado a los costos. Dicha premisa es razonable para una empresa porque ésta
es pequeña en comparación con el total de la economía. (Es poco probable que una compañía que
aumenta el precio de su producción ocasione alzas significativas en los precios de los insumos de la
economía.) Si el precio de la producción más rentable de una empresa aumenta sin incrementos
en los costos de ningún insumo, es probable que la producción se acreciente.
Ahora bien, ¿qué pasaría si hubiera un acrecentamiento en el nivel general de precios? Por
dos razones, es poco realista suponer que los costos son constantes para las empresas individuales
si el nivel general de precios está aumentando. En primer lugar, los productos de unas empresas
son los insumos de otras; por consiguiente, si los precios de los productos se elevan, habrá un incremento en por lo menos algunos precios de insumos. Segundo, es poco realista suponer que
las tasas salariales (un importante costo de insumos) suben en absoluto cuando se incrementa el
nivel general de precios. Como los precios de todos los insumos (incluyendo tasas salariales) no
son constantes cuando cambia el nivel general de precios, la curva de la oferta de las empresas se
desplaza conforme cambia el nivel general de precios, así que no podemos sumarlas para obtener
una curva de la oferta agregada.
Otra razón por la que la curva de la oferta agregada no puede ser la suma de las curvas de la
oferta de todas las empresas de la economía, es que muchas compañías (algunos dirían que la
mayoría de ellas) no se reducen a reaccionar a los precios determinados por el mercado, sino que
de hecho fijan los precios. Sólo en los mercados de competencia perfecta las empresas se concretan
a reaccionar a los precios determinados por las fuerzas del mercado. Empresas de otros sectores
industriales (para ser exactos, en las industrias de competencia imperfecta) toman decisiones de
producción y precios sobre la base de su idea de la demanda y los costos. Las empresas que fijan
precios no tienen curvas individuales de la oferta porque escogen al mismo tiempo la producción
y los precios. Para derivar una curva individual de la oferta, tenemos que imaginarnos que proponemos un precio a una empresa y le pedimos que nos diga cuánto producto ofrecería a ese
precio. No podemos hacerlo si las empresas también fijan los precios. Si no hay curvas de la oferta
de empresas en competencia imperfecta, está de más decir que no podemos sumarlas para obtener una curva de la demanda agregada.
¿Qué cabe decir de la relación entre producción total y nivel general de precios? Como los
precios de los insumos cambian con las variaciones del nivel general de precios y ya que muchas
empresas deciden los precios, además de la producción, está claro que no existe una “curva de la
oferta agregada” en la acepción tradicional de la palabra oferta. Lo que hay podría llamarse “curva de respuesta de precio y producto”, que representa las decisiones de precios y producción de
todos los mercados y empresas en determinadas circunstancias.
¿Cómo sería esa curva?

#### 2.4.2 La oferta agregada en el corto plazo

Muchos dicen que la curva de la oferta agregada (o la curva de respuesta de precio y producto)
tiene pendiente positiva, por lo menos en el corto plazo. (Más adelante detallaremos la distinción entre corto y largo plazo.) Además, muchos afirman que con una producción total muy escasa
(por ejemplo, cuando la economía está en recesión) la curva de la oferta agregada es bastante plana,
y en altos niveles de producción (como cuando está en auge), la curva es vertical, o casi. Ésta se
muestra en la figura 13.6.
Para entender la forma de la curva OA de la figura 13.6, consideremos la respuesta a la producción y los precios de mercados y firmas cuando hay un aumento en la demanda agregada
causado por una política fiscal, o monetaria, de expansión. La reacción de las empresas en tal ensanchamiento podría depender de dos factores: 1) qué tan cerca está la economía de colmar su
capacidad en el momento de la expansión, y 2) con qué prontitud reaccionan los precios de los
insumos (como las tasas salariales) a incrementos en el nivel general de precios.

**Limitaciones de la capacidad** En microeconomía, el “corto plazo” se refiere al tiempo
en que las decisiones de las empresas están limitadas por algún factor fijo de la producción. Los
agricultores están constreñidos en el corto plazo por la extensión de sus propiedades: la cantidad
de tierras que poseen es el factor fijo de su producción. Las decisiones de producción de corto
plazo de las empresas manufactureras están acotadas por el tamaño de sus instalaciones fabriles.
En un plazo más amplio, las empresas superan estas restricciones invirtiendo en más capacidad;
por ejemplo, adquieren más terrenos o construyen una fábrica nueva.

Curva de la oferta agregada de corto plazo

<img src="4. Analisis macroeconomico\Inflación, demanda y oferta agregadas\2.4.2 La oferta agregada en el corto plazo.PNG" width=400 height=200 />

_En el corto plazo, la curva de la oferta agregada (curva de respuesta de precio y producto) tiene una pendiente positiva. Con una producción total escasa, la curva es bastante llana. Cuando la economía se acerca a su capacidad, la curva es casi vertical. A toda su capacidad, es vertical._

La idea de una capacidad fija de corto plazo también cumple una función en la macroeconomía. Los especialistas en la materia se centran en si las empresas individuales producen a toda
su capacidad o se aproximan a esta meta. Una firma produce a capacidad plena si aprovecha por
completo el capital y la mano de obra que tiene. Como veremos a conciencia en el capítulo 17,
en ocasiones las empresas cuentan con excedentes de capital y de mano de obra, recursos que no
necesitan para alcanzar el nivel actual de producción. Por ejemplo, si hay costos por deshacerse
del capital instalado, quizá una compañía decida retener una parte de este capital, aun si la economía está a la baja y la empresa ha reducido su producción. En este caso, la firma no utiliza todas sus existencias de capital. Las compañías tienden a actuar de esta manera si esperan que la
baja sea efímera y creen que luego necesitarán el capital para producir a niveles mayores. Las
empresas tienen razones parecidas para retener el excedente de mano de obra. Puede resultar
caro, tanto para la moral de los trabajadores como en costos administrativos, despedir a muchos
empleados.
La Reserva Federal informa cada mes sobre la “tasa de utilización de la capacidad” de la nación. Por ejemplo, en diciembre de 1990, durante la recesión de 1990 a 1991, la tasa de utilización
de la capacidad de las firmas manufactureras fue de 79.3 por ciento. Esto indica que casi 20 por
ciento de la capacidad fabril de la nación estaba ociosa. En las recesiones de 1974 a 1975 y de 1980
a 1982, el empleo de la capacidad bajó a menos de 75 por ciento. En la recesión de 2001, también
fue inferior a 75 por ciento. Este parámetro siguió bajo durante 2002 y 2003. En julio de 2003 era
de apenas 72.8 por ciento.
Los estudiosos de la macroeconomía se centran en si la economía en conjunto opera o no a
plena capacidad. Si hay desempleo cíclico (el que es superior a las cifras de fricción y estructural),
la economía no aprovecha toda su fuerza de trabajo. Hay personas que quieren trabajar pero, a las
tasas salariales actuales, no encuentran ocupación.

> **Aun si las empresas no tienen excedentes de trabajadores y capital, puede ser que la economía opere a menos de su capacidad si hay desempleo cíclico.**

**Niveles de producción y respuestas de precio y producto** Con una producción escasa en la
economía, es probable que haya exceso de capacidad tanto de las empresas en lo individual como
de la economía en su conjunto. Las compañías producen a niveles que están por debajo de los límites de su capacidad. Es decir, quizá tengan excedentes de capital y mano de obra. Asimismo, es
probable que haya desempleo cíclico en la economía en periodos de poca producción. Si éste fuera el caso, las firmas responderían a un incremento de la demanda elevando la producción mucho
más que los precios. Las empresas están por debajo de su capacidad, así que pueden reducir el
costo adicional de producir más. Además, pueden conseguir más trabajadores (entre las filas de
los desempleados) sin que aumenten mucho, o nada, las tasas salariales.

> **Un aumento de la demanda agregada, cuando la economía opera a niveles bajos de producción, trae por resultado un incremento de la producción con poco o ningún incremento del nivel general de precios. Es decir, la curva de la oferta agregada (respuesta de precio y producto) es bastante llana a niveles mínimos de producción total.**

Consulte la figura 13.6. La producción total es mucho más alta en B que en A, pero el nivel de precios de B es apenas mayor que en A.
Si la producción total continúa su expansión, las cosas cambian. A medida que las empresas
y la economía en su conjunto se acercan cada vez más a su capacidad plena, la respuesta de las
empresas a un aumento de la demanda pasa del incremento preferencial de la producción al
crecimiento preferencial de los precios. ¿Por qué? Como las empresas no dejan de ampliar su producción, comienzan a toparse con sus límites de capacidad de corto plazo. Además, el desempleo
baja si las empresas contratan más trabajadores para que realicen la mayor producción, de modo
que el conjunto de la economía se aproxima a colmar su capacidad. Si la producción total aumenta, los precios de la mano de obra y el capital (costos de los insumos) comienzan a elevarse más
deprisa, lo cual lleva a las compañías a subir los precios de sus productos.
En cierto nivel de producción, es de hecho imposible que las empresas se expandan más. En
ese punto, todos los sectores utilizan a plenitud la capacidad de sus fábricas y equipos. Las plantas
operan en dos turnos y muchos trabajadores laboran horas extra. Además, en la economía hay poco o ningún desempleo cíclico. Asimismo, las empresas responden a los incrementos de la demanda sólo con aumentos de precios, puesto que ya no pueden expandir la producción

> **Cuando la economía produce a su máximo nivel (es decir, a su entera capacidad), la curva de la oferta agregada se vuelve vertical.**

Entre C y D en la figura 13.6, la curva OA es vertical. Si se pasa de C a D no hay aumento de la producción total, aunque sí un gran incremento del nivel de precios.

**La respuesta de los precios de insumos a los cambios del nivel general de precios** Produzca o no la economía cerca de su capacidad plena, debe haber alguna demora entre los cambios de los precios de los insumos y las variaciones en los precios de los productos para
que la pendiente de la curva de la oferta agregada (respuesta de precio y producto) tienda a subir. Si
los precios de los insumos se modificaran en el mismo monto que los precios de los artículos, la curva OA sería vertical.
Es fácil ver por qué. En general, se da por hecho que las empresas toman decisiones con el objetivo de maximizar sus utilidades. Si todos los precios de productos e insumos aumentan 10 por
ciento, a ninguna empresa le parecería ventajoso cambiar su nivel de producción. ¿Por qué? Porque el nivel de producción que maximiza las utilidades antes del incremento de 10 por ciento será
igual al que consiga ese efecto después del aumento.1 Así, si los precios de los insumos se ajustaran
de inmediato a los precios de los productos, la curva de la oferta agregada (respuesta de precio y
producto) sería vertical.
Las tasas salariales pueden aumentar a la misma velocidad que el nivel general de precios si esta
alza se anticipa por completo. Si se espera que la inflación sea de 5 por ciento este año, este incremento previsto podría incorporarse en sueldos y salarios contractuales. Sin embargo, la mayoría de
los trabajadores no reciben aumentos automáticos cuando sube el nivel general de precios y a veces
éste se incrementa de manera imprevista. Los precios de los insumos (en particular las tasas salariales) se demoran con respecto a los aumentos de los precios de los productos por varias razones (que
veremos en el capítulo 14). Por lo menos en el corto plazo, las tasas salariales son lentas para ajustarse a los cambios macroeconómicos generales. Es, con exactitud, este punto el que ha llevado a
una importante distinción entre la curva OA de largo plazo y la curva OA de corto plazo.2 Volveremos enseguida a esta distinción, pero por ahora supondremos que la curva OA tiene la forma de la
figura 13.6.

#### 2.4.3 Cambios de corto plazo en la curva de la oferta agregada

Así como la curva de la demanda agregada puede desplazarse, también se mueve la curva de la
oferta agregada (respuesta de precio y producto). Recuerde el comportamiento de la empresa individual que acabamos de considerar al describir la forma de la curva OA de corto plazo. Las
compañías que tienen el poder de fijar precios optan por las combinaciones de precios y producción
que maximizan sus utilidades. Las firmas en sectores industriales de competencia perfecta eligen
montos de producción para ofrecer a determinados niveles de precios. La curva OA traza estas
respuestas de precios y productos a las condiciones económicas.
Todo lo que afecte las decisiones de estas empresas individuales puede repercutir en la curva
OA. Entre estos factores se cuentan los choques de costos, crecimiento económico, estancamiento,
políticas públicas y desastres naturales.

> **El hecho de que todos los precios suban en el mismo porcentaje es análogo a cambiar la unidad monetaria de cuenta, por
ejemplo, de dólares verdes a dólares rojos, de modo que 1.1 dólares rojos equivalgan a un dólar verde. Un cambio de unidad monetaria no tiene efecto en las decisiones para maximizar las utilidades de las empresas. Si el valor nominal de todos los precios de
insumos y productos aumenta 10 por ciento, no ocurre nada real. Cuando todos los valores nominales aumentan 10 por ciento,
las decisiones de las empresas sobre la producción real no cambian. 2 En algunos libros se deriva la curva de la demanda agregada de corto plazo con la premisa de que todos los precios de los insumos son fijos. “Precios de insumos fijos” significa que tales precios no cambian con las variaciones del nivel general de precios.
Esta premisa no es realista porque los productos de algunas empresas (como los bienes intermedios y los de capital) son los insumos de otras firmas. También es poco objetivo suponer que las tasas salariales no responden a todos los cambios del nivel general de precios. Es más razonable suponer que las tasas salariales no responden por completo en el corto plazo, que suponer que
no reaccionan para nada**

**Choque de los costos** Las decisiones de las empresas están muy influidas por los costos.
Algunos costos cambian al mismo tiempo que el nivel general de precios; otros se atrasan con
respecto a este nivel y unos más no varían en absoluto. Las modificaciones de los costos que ocurren de forma paralela a las alteraciones del nivel de precios están incorporadas en la forma de la
curva OA de corto plazo. Por ejemplo, cuando el nivel de precios aumenta, las tasas salariales
pueden subir cuando mucho la mitad de ese monto en el corto plazo. (Este efecto ocurriría si la mitad de todos los contratos salariales de la economía tuvieran cláusulas de costo de la vida y la otra
mitad no.) La forma de la curva OA de corto plazo reflejaría esta respuesta.
A veces ocurren cambios de costos que no resultan de las variaciones del nivel general de precios; por ejemplo, el costo de la energía. En el verano de 2005, los precios mundiales del petróleo se
duplicaron, de 35 a 70 dólares por barril. Cuando se aclaró que los fuertes huracanes de ese año no
iban a causar grandes trastornos al suministro de crudo, los precios bajaron a menos de 60 dólares.
Por el contrario, entre 1973 y 1974, y de nuevo en 1979, el precio del petróleo aumentó sustancialmente y se mantuvo en niveles altos. El petróleo es un insumo importante para muchas empresas e
industrias, y cuando el precio de las materias primas se incrementa, éstas responden elevando los
precios y reduciendo la producción. En conjunto, esto significa que un incremento en el precio
del petróleo (o un alza equivalente de costos) desplaza la curva OA a la izquierda, como en la figura 13.7(a). Un movimiento a la izquierda de la curva OA representa un nivel mayor de precios para determinado nivel de producción. Una reducción de los costos traslada la curva OA a la derecha,
como en la figura 13.7(b). Un desplazamiento a la derecha de la curva OA equivale a un nivel de
precios más bajo para determinada producción. Los movimientos de la curva OA causados por un
cambio en los costos se llaman **choques de los costos** o **choque de la oferta**.
**Crecimiento económico** El crecimiento económico gira la curva OA a la derecha. Recuerde que la parte vertical de una curva OA de corto plazo representa la producción máxima
(capacidad plena) de la economía. Esta producción máxima está determinada por los recursos de
la economía y el estado actual de la tecnología. Si la oferta de mano de obra aumenta o si las existencias de capital se acrecientan, la curva OA también se desplaza a la derecha. Se espera que la
fuerza laboral crezca de manera natural con los incrementos de la población en edad de trabajar,
pero también aumenta por otras razones. Por ejemplo, desde los años sesenta el porcentaje de
mujeres integradas a la fuerza laboral ha crecido en forma notable. Este acrecentamiento de la
oferta de trabajadoras ha desplazado a la derecha la curva OA.
La inmigración también desplaza la curva OA. En la década de los setenta, Alemania enfrentaba una escasez grave de trabajadores; entonces, abrió sus fronteras a numerosos “trabajadores invitados”, sobre todo de Turquía. Estados Unidos ha tenido una inmigración importante, legal e ilegal, de México y países de Centro y Sudamérica y Asia. Con el tiempo, la acumulación de existencias de
capital y los adelantos tecnológicos también desplazan a la derecha la curva OA. En el capítulo 20
estudiaremos con mayor detalle el crecimiento económico.

**Estancamiento y falta de inversión** Lo opuesto al crecimiento económico son el estancamiento y el decaimiento de la economía. Con el paso del tiempo, el capital se deteriora y al final se acaba por completo si no recibe el mantenimiento apropiado. Si una economía no invierte en
capital público (a veces llamada infraestructura) y capital privado (plantas y equipo) a un ritmo suficiente, las existencias de capital menguan. Si esto sucede, la curva OA se desplaza a la izquierda.
**Políticas públicas** Las políticas públicas pueden desplazar la curva OA. Por ejemplo, en
los años ochenta el gobierno de Ronald Reagan puso en vigor una forma de política pública basada en una economía de la oferta. La idea era desregular la economía y reducir los impuestos para

curva OA. Una sequía grave reduce la oferta de productos agrícolas; la mezcla perfecta de sol y
lluvia producen una cosecha abundante. Si una economía queda dañada por una guerra o un desastre natural, la curva OA se desplaza a la izquierda. Cuando parte del suministro de un recurso
en una economía se reduce o destruye, la curva OA se traslada a la izquierda.

### 2.5 Nivel de precios de equilibrio

El nivel de precios de equilibrio de la economía se alcanza en el punto donde se intersecan las curvas OA y DA, como se muestra en la figura 13.9, en la que el nivel de precios de equilibrio es P0 y
el nivel de equilibrio de la producción total (ingreso) es Y0.
La figura 13.9 parece simple, pero es un medio poderoso para estudiar varias preguntas macroeconómicas. Consideremos primero qué ocurre verdaderamente en la intersección de las curvas OA y DA. Cada punto de la curva DA corresponde al equilibrio de los mercados de bienes y de
dinero. Cada punto de la curva OA representa las respuestas de precios y productos de todas las
empresas de la economía. Esto significa que:

> **El punto donde se intersecan las curvas DA y OA corresponde al equilibrio de los mercados de bienes y servicios y a un grupo de decisiones de precios y productos de todas las empresas de la economía.**

### 2.6 La curva de la oferta agregada de largo plazo

Como ya lo dijimos, para que la curva OA no sea vertical, algunos costos tienen que retrasarse con
respecto a los incrementos del nivel general de precios. Si todos los precios (de insumos y productos) cambian al mismo ritmo, el nivel de la producción total no cambia. Hemos supuesto que en el
corto plazo por lo menos algunos costos se demoran con respecto a los cambios del nivel de precios,
¿pero qué pasa en el largo plazo?
Muchos economistas creen que los costos se retrasan con respecto a los precios en el corto plazo,
pero que en última instancia se mueven junto con el nivel de precios. Por ejemplo, las tasas salariales
se acercan mucho a éste con el paso del tiempo. Si el nivel de precios aumenta a un ritmo constante, la
inflación se anticipa por completo y se puede incorporar a la mayoría de los contratos laborales.
Si los costos y el nivel de precios se mueven juntos durante un lapso bastante prolongado para
que se hagan estos ajustes, en el largo plazo, la curva OA se traza, sobre todo, como una vertical. El
porqué y se ve en la figura 13.10. Al principio, la economía está en equilibrio a un nivel de precios
de P0 y una producción total de Y0 (el punto A donde se intersecan DA0 y OA0). Ahora imaginemos
un desplazamiento de la curva DA de DA0 a DA1. En respuesta a este movimiento, tanto el nivel de
precios como la producción total aumentan en el corto plazo, a P1 y Y1, respectivamente (el punto
B en el que se intersecan DA1 y OA0). Pero recuerde que el movimiento sobre la curva OA0, que
tiende hacia arriba, cuando aumenta Y de Y0 a Y1 supone que parte de los costos se retrasan con respecto al incremento del nivel general de precios.
Supongamos ahora que los costos se ajustan por completo en el largo plazo, cuando, por ejemplo, los sindicatos renegocian los contratos laborales para compensar los aumentos de precios. Estos
incrementos de los costos hacen que la curva OA se desplace a la izquierda, de AO0 a AO1. Si al final
los costos y los precios aumentaron el mismo porcentaje, la producción total vuelve a Y0 (el punto
C en el que se intersecan DA1 y OA1).

> **Si en el largo plazo las tasas salariales y otros costos se ajustan completamente a los cambios de los precios, la curva OA de largo plazo es vertical.**

#### 2.6.1 PIB Potencial

Recuerde que incluso la curva OA de corto plazo se vuelve vertical en algún nivel particular de
producción. La parte vertical de la curva OA de corto plazo se debe a que hay límites físicos para lo que una economía puede producir en determinado periodo. En el límite físico, todas las plantas
operan las 24 horas del día, muchos trabajadores laboran horas extra y no hay desempleo cíclico.

Curva de la oferta agregada de largo plazo

<img src="4. Analisis macroeconomico\Inflación, demanda y oferta agregadas\2.6.1 PIB Potencial.PNG" width=400 height=200 />

_Cuando la curva $DA$ se desplaza de $DA_0$ a $DA_1$, el nivel de precios de equilibrio asciende primero de $P_0$ a $P_1$ y la producción sube también, de $Y_0$ a $Y_1$. Los costos responden en el largo plazo, de modo que la curva $OA$ se desplaza de $OA_0$ a $OA_1$. En definitiva, si los costos aumentan el mismo porcentaje que el nivel de precios, la cantidad ofrecida volverá a $Y_0$, que también se llama PIB potencial._

Observe que las partes verticales de las curvas OA de corto plazo de la figura 13.10 están a la
derecha de Y0. Si estas partes verticales de las curvas representan “capacidad”, ¿cuál es la naturaleza
de Y0, el nivel de producción que corresponde a la curva OA de largo plazo?
Y0 representa el nivel de producción total que puede sostenerse en el largo plazo sin inflación.
También se llama producción potencial o PIB potencial. La producción se impulsa a Y0 en diversas
circunstancias y cuando lo hace, hay una presión por aumentar los costos. A medida que la economía se acerca a la capacidad de corto plazo, las tasas salariales crecen porque las empresas intentan
atraer más personas a la fuerza de trabajo y alientan a más trabajadores a laborar horas extra. El incremento de los costos desplaza la curva OA de corto plazo a la izquierda (figura 13.10 de OA0 a
OA1) y devuelven la producción a Y0.

**quilibrio de corto plazo debajo del PIB potencial** Hasta aquí hemos dicho que
si las curvas de la oferta y la demanda agregadas de corto plazo se intersecan a la derecha de Y0 en
la figura 13.10, los salarios y otros precios de insumos suben, lo cual causa que la curva OA de
corto plazo se desplace a la izquierda y se impulsa el PIB de vuelta a Y0. Diversos economistas tienen opiniones diferentes sobre cómo determinar si la economía opera en su PIB potencial o más,
pero hay un acuerdo general sobre que hay un nivel máximo de producción (debajo de la parte
vertical de la curva de la oferta agregada de corto plazo) que puede sostenerse sin inflación.
¿Qué sucede con los equilibrios de corto plazo que se alcanzan a la izquierda de Y0? ¿Qué ocurre si las curvas de la oferta y la demanda agregadas de corto plazo se intersecan en un nivel de producción por debajo del PIB potencial? Aquí tampoco se ponen de acuerdo los economistas. Quienes piensan que la curva de la oferta agregada es vertical, a la larga creen que si hay equilibrios de
corto plazo debajo de Y0, el PIB aumenta, y disminuye cuando se alcanza un equilibrio de corto
plazo sobre Y0. El argumento es que cuando la economía opera por debajo del pleno empleo con
exceso de capacidad y alto desempleo, los precios de los insumos (incluyendo salarios) bajan. Una
reducción de precios de los insumos desplaza la curva de la oferta agregada a la derecha, lo cual hace
que el nivel de precios se reduzca y que el PIB real vuelva a Y0. Este ajuste automático sólo funciona
si los precios de los insumos caen cuando hay exceso de capacidad y desempleo. En el capítulo 14 veremos con atención los ajustes salariales en periodos de desempleo.

### 2.7 Demanda y oferta agregadas, política monetaria y fiscal

Ahora estamos listos para usar el marco OA/DA y considerar los efectos de la política monetaria
y fiscal. Primero valoremos los efectos de corto plazo.
Recuerde que las dos variables de la política fiscal son las compras del gobierno (G) y los
impuestos netos (T). La variable de la política monetaria es la cantidad del circulante (Ms
). Una
política de expansión pretende estimular la economía mediante un incremento de G o Ms
, o
bien, una reducción de T. Una política de contracción se dirige a desacelerar la economía mediante
una disminución de G o Ms
, o bien, un aumento de T. Ya vimos que una política de expansión desplaza la curva DA a la derecha y que una política de contracción la traslada a la izquierda. ¿Qué
efecto tienen estas políticas en los valores de equilibrio del nivel de precios (P) y el nivel de la
producción total (ingreso)?
Cuando se consideran los efectos de un cambio de políticas, hay que tener el cuidado de
constatar en qué parte de la curva OA de corto plazo se encuentra la economía al momento del
cambio. Si la economía está en la parte llana de la curva OA, como en el punto A de la figura
13.11, entonces una política de expansión —que desplaza la curva DA a la derecha— da por resultado un pequeño aumento de precios en relación con el crecimiento de la producción: el incremento del equilibrio de Y (de Y0 a Y1) es mucho mayor que la ampliación del equilibrio de P (de
P0 a P1). Es el caso en que una política de expansión funciona bien. Hay un ensanchamiento de la
producción con un aumento ligero del nivel de precios.
Si al principio la economía se encuentra en la parte empinada de la curva OA, como el punto
B de la figura 13.12, una política de expansión trae como resultado un aumento reducido de la
producción de equilibrio (de Y0 a Y1) y un incremento grande en el nivel de precios de equilibrio
(de P0 a P1). En este caso, una política de expansión no funciona bien, porque origina un nivel de
precios mucho mayor con un crecimiento pequeño de la producción. Por consiguiente, el multiplicador está cerca de cero: al principio la producción está cerca de colmar su capacidad y los esfuerzos
por aumentarla se traducen, sobre todo, en un aumento del nivel de precios.
En las figuras 13.11 y 13.12 se demuestra que es importante saber dónde está la economía
antes de emprender un cambio de políticas. La economía produce cerca de la parte llana de la
curva OA si la mayoría de las empresas producen a menos de su capacidad. Cuando ocurre así,
las empresas responden a un aumento de la demanda incrementando la producción mucho más
que los precios. Si la economía produce en la parte inclinada de la curva OA, las empresas están
cerca de toda su capacidad y responden a un aumento de la demanda con un incremento de los
precios mayor que el crecimiento de la producción.

Desplazamiento de la curva de la demanda agregada cuando la economía está en la parte casi llana de la curva OA

_La demanda agregada se desplaza a la derecha por varias razones, como un aumento del circulante, una reducción de impuestos o un crecimiento del gasto gubernamental. Si el movimiento ocurre cuando la economía está cerca de la parte llana de la curva $OA$, el resultado es un acrecentamiento de la producción con poco incremento del nivel de precios del punto $A$ al punto $A′$._

Desplazamiento de la curva de la demanda agregada cuando la economía opera a toda, o casi toda, su capacidad

_Si ocurre un desplazamiento de la curva agregada cuando la economía opera a casi su capacidad total, el resultado será un aumento del nivel de precios con un ligero incremento de la producción del punto $B$ al $B′$._

Para ver lo que pasa cuando la economía está en la parte inclinada de la curva OA, tomemos
el efecto de un aumento de G sin ningún cambio del circulante. Lo que pasa es que cuando G crece, prácticamente no hay incrementos en Y. En otras palabras, la política fiscal de expansión no
estimulará la economía. Para considerarlo, necesitamos retroceder al capítulo 12 y repasar el razonamiento en que se funda la curva DA.
Lo primero que sucede cuando G aumenta es una reducción inesperada de las existencias de las
empresas. Como las compañías están muy cerca de su capacidad de producción cuando la economía está en la parte empinada de la curva OA, no pueden elevar mucho su producción. El resultado,
como se observa en la figura 13.12, es un incremento sustancial del nivel de precios. Éste aumenta la
demanda de dinero, la cual (con un circulante fijo) origina un alza de la tasa de interés y una baja de
la inversión planeada. Hay un efecto de expulsión casi completa de la inversión. Si las empresas producen a toda su capacidad, precios y tasas de interés siguen subiendo hasta que el incremento de G se
equipara del todo con una reducción de la inversión planeada y un efecto completo de expulsión.

#### 2.7.1 Oferta agregada de largo plazo y efectos en las políticas

Hasta aquí hemos considerado los efectos de la política monetaria y fiscal en el corto plazo. En lo
que respecta al largo plazo, es importante advertir que:

> **Si la curva OA es vertical en el largo plazo, ni la política monetaria ni la fiscal tienen ningún efecto duradero en la producción total.**

Volvamos a la figura 13.10. La política monetaria y fiscal desplaza la curva DA. Si la curva OA
de largo plazo es vertical, la producción siempre vuelve a Y0. En este caso, la política afecta sólo el
nivel de precios de largo plazo y el efecto multiplicador de un cambio del gasto gubernamental
sobre la producción total de largo plazo es nulo. En las mismas circunstancias, el multiplicador de
impuestos también es cero.
La conclusión de que las políticas no tienen efecto en la producción total de largo plazo es sorprendente. ¿Los economistas están de acuerdo en que la curva de la oferta agregada es vertical en el
largo plazo?
La mayoría de los economistas aceptan que los precios de los insumos se demoran con respecto a los precios de la producción en el corto plazo, lo cual da a la curva OA una pendiente poc positiva. También casi todos aceptan que la curva OA tiende a ser más empinada en el largo plazo,
¿pero qué tan prolongado? Cuanto mayor sea la demora, más grande será el efecto potencial de la
política monetaria y fiscal en la producción total. Si el largo plazo es apenas de tres a seis meses, es
poco probable que la política afecte la producción; si el largo plazo es de tres o cuatro años, la política puede tener efectos significativos. Un gran cúmulo de investigaciones en macroeconomía se
enfoca en la duración de la demora entre los precios de insumos y productos. En cierto sentido, la
duración del largo plazo es una de las preguntas abiertas más importantes de la macroeconomía.
Otro motivo de desacuerdo se centra en si los equilibrios debajo del PIB potencial, Y0 en la figura 13.10, se corrigen solos (es decir, sin intervención del gobierno). Recuerde que quienes creen
en una curva OA de largo plazo vertical, piensan que una desaceleración de la economía presionará
los precios de los insumos a la baja (incluyendo salarios), lo cual hace que la curva OA de corto plazo se desplace a la derecha y lleve el PIB de vuelta a Y0. No obstante, algunos afirman que los salarios
y otros precios de los insumos no bajan en periodos de desaceleración y que la economía puede
“atorarse” en un equilibrio por debajo del PIB potencial. En este caso, se necesitarían políticas monetarias y fiscales para restaurar el pleno empleo. En el capítulo 14 volveremos a este debate

### 2.9 Causas de la inflación

Ahora pasamos a la inflación y nos valemos del marco de referencia OA/DA para considerar sus causas.

#### 2.9.1 Recordatorio sobre inflación e inflación sostenida

Antes de que analicemos las causas específicas de la inflación, recuerde la distinción que hicimos
en el capítulo 7. Como sabe, la inflación es el incremento del nivel general de precios. Todo lo que
desplace la curva DA a la derecha o la curva OA a la izquierda causa inflación, pero suele ser útil
distinguir entre un aumento único del nivel de precios (de una ocasión) y una inflación sostenida.
Una inflación sostenida ocurre cuando el nivel general de precios no deja de crecer durante un periodo prolongado. Por ejemplo, cuando hablamos de una inflación sostenida de 3 por ciento,
en general queremos decir que el nivel de precios ha aumentado a una tasa de 3 por ciento anual
durante varios años.
En general se acepta que hay muchas causas posibles de un incremento único en el nivel de
precios (enseguida comentaremos las principales causas). La mayoría de los economistas creen
que para que el nivel de precios aumente en cada periodo, debe quedar “acomodado” por un aumento del circulante. Esto lleva a la afirmación de que la inflación sostenida, cualquiera que sea la
causa primera del incremento del nivel de precios, es en esencia un fenómeno monetario.

#### 2.9.2 Inflación de la demanda

La inflación iniciada por un aumento de la demanda se llama inflación de la demanda. En las figuras 13.11 y 13.12 se aprecia cómo actúa la inflación de la demanda. En las dos, la inflación comienza con un desplazamiento del esquema de la demanda agregada, de DA0 a DA1, lo cual hace
que el nivel de precios aumente de P0 a P1 (la producción también se incrementa, de Y0 a Y1). Si la
economía opera en la parte inclinada de la curva OA en el momento del incremento de la demanda
agregada, como en la figura 13.12, casi todo el efecto será un aumento del nivel de precios, en lugar
de un crecimiento de la producción. Si la economía opera en la parte llana de la curva OA, como
en la figura 13.11, la mayor proporción del efecto será un ensanchamiento de la producción, en
lugar de que suba el nivel de precios.
Recuerde que, en el largo plazo, el aumento inicial del nivel de precios hace que la curva OA
se desplace a la izquierda pues el importe de los insumos (los costos) responde al crecimiento de
las tarifas de los productos. Si la curva OA de largo plazo es vertical, como en la figura 13.10, el incremento de los costos desplaza la curva OA de corto plazo (OA0) a la izquierda, a OA1, lo cual sube
aún más el nivel de precios, a P2. Si la curva OA de largo plazo es vertical, un giro de la curva de la
demanda agregada de DA0 a DA1 da por resultado, a la larga, que no haya incremento de la producción y que el nivel de precios suba de P0 a P1.

#### 2.9.3 Inflación de los costos o la oferta

La inflación también puede ser causada por un incremento de los costos; es la llamada inflación
de los costos o de la oferta. En las últimas dos décadas, varias veces subieron de manera repentina los precios del petróleo en los mercados mundiales. Como el petróleo se usa en casi todos los
negocios, los costos se dispararon.
Un aumento de los costos (un choque de los costos) desplaza la curva OA a la izquierda, como
se muestra en la figura 13.13. Si suponemos que el gobierno no reacciona a este desplazamiento
de OA cambiando las políticas fiscales o monetarias, la curva DA no se desplaza. El movimiento de la oferta hace que el nivel de equilibrio de los precios suba (de P0 a P1) y que el nivel de la producción baje (de Y0 a Y1). Recuerde que en el capítulo 5 dijimos que la estanflación ocurre cuando la
producción baja a la vez que los precios suben; en otras palabras, cuando la economía pasa por
una contracción y una inflación, al mismo tiempo. En la figura 13.13 se muestra que una causa
posible de la estanflación es un aumento de los costos.

_Un aumento de los costos desplaza la curva OA a la izquierda. Si se supone que el gobierno no reacciona a este movimiento, la curva DA no se mueve, el nivel de precios sube y la producción baja._

#### 2.9.4 Expectativas e inflación

Cuando las empresas toman sus decisiones de precios y productos, en ellas influyen sus expectativas
de los precios futuros. Si una compañía espera que la competencia aumente sus precios, de antemano puede subir los propios.
Tomemos el caso de una firma que fabrica tostadoras. Tiene que decidir cuánto cobrar a las
tiendas por sus aparatos. Si exagera el precio y cobra mucho más que otros fabricantes, perderá
demasiados clientes. Si subestima el precio y cobra mucho menos, ganará clientes pero a cambio
de una pérdida considerable de utilidades por venta. Es probable que el precio óptimo de la empresa, la tarifa en que maximiza sus utilidades, no esté muy lejos del promedio de los precios de la
competencia. Si la empresa no conoce los precios proyectados por la competencia antes de fijar
sus precios, como ocurre a menudo, debe basarlos en cuáles supone que serán los precios de sus
rivales.

Los choques de los costos son malas noticias para las autoridades

<img src="4. Analisis macroeconomico\Inflación, demanda y oferta agregadas\2.9.4 Expectativas e inflación.PNG" width=400 height=200 />

_Un choque de los costos sin cambio en las políticas fiscal y monetaria desplaza la curva de la oferta
agregada de OA0 a OA1, abate la producción de Y0 a Y1 y eleva el nivel de precios de P0 a P1. La política
monetaria o fiscal podría cambiarse lo suficiente para que la curva DA se gire de DA0 a DA1. Esta 
estrategia aumentaría de nuevo la producción Y, pero subiría todavía más los precios a P2._

Supongamos que la inflación ha sido de casi 10 por ciento anual. Quizá nuestra compañía anticipe que los competidores aumentarán sus precios más o menos 10 por ciento este año, así que
quizá el precio de su tostadora se eleve 10 por ciento. Esta respuesta manifiesta cómo las expectativas “se incorporan en el sistema”. Si las empresas esperan que sus competidoras incrementen 10
por ciento los precios, entonces subirán los suyos alrededor de 10 por ciento. Todas terminan con
el alza de precios que se esperaba.
El hecho de que las expectativas inciden en el nivel de precios es engorroso. Asimismo, se
pueden generar inercias que hacen difícil detener una espiral inflacionaria. Si los precios han estado subiendo, y las perspectivas de la gente son moldeables (es decir, si se forman expectativas de
acuerdo con el comportamiento anterior de los precios), las empresas siguen subiendo sus precios aun si la demanda se desacelera o contrae. En términos del diagrama OA/DA, un aumento de
las expectativas de la inflación que lleva a las empresas a subir sus precios desplaza la curva OA a
la izquierda. Recuerde que ésta representa las respuestas de precios y productos de las empresas.
Si las firmas elevan sus precios por un cambio en las expectativas con respecto a la inflación, el resultado es un desplazamiento a la izquierda de la curva OA.

#### 2.9.5 Dinero e inflación

Es fácil ver que un aumento del circulante puede llevar a un incremento del nivel general de precios. Como se indica en las figuras 13.11 y 13.12, un aumento del circulante (Ms
) desplaza la curva
DA a la derecha y da como resultado un nivel de precios más alto. Se trata de una inflación de la
oferta y nada más.
Sin embargo, el circulante tiene también su parte en el origen de una inflación sostenida.
Consideremos un aumento inicial del gasto gubernamental (G) sin variaciones de circulante
(Ms
). Como éste no cambia, es un incremento de G que no “asimila” la Fed. El aumento de G desplaza la curva DA a la derecha y da por resultado un nivel de precios mayor. Esto se muestra en la
figura 13.15 como un giro de DA0 a DA1 (en la figura 13.15 se supone que la economía opera en
la parte vertical de la curva OA).
Recuerde lo que ocurre cuando sube el nivel de precios. El nivel de precios más alto hace
que se incremente la demanda de dinero. Sin cambios del circulante y con un aumento en la demanda de dinero, la tasa de interés sube y el resultado es una disminución del gasto en inversión
planeada (I). El nuevo equilibrio corresponde a un G mayor, menor I, tasa de interés más alta y
nivel de precios más elevado.

Inflación sostenida de un aumento inicial de G y la adaptación de la Fed

<img src="4. Analisis macroeconomico\Inflación, demanda y oferta agregadas\2.9.5 Dinero e inflación.PNG" width=400 height=200 />

_Un aumento de G con el circulante fijo desplaza la curva DA de DA0 a DA1. Aunque no se muestra en la figura, esto lleva a un incremento de la tasa de interés y a un efecto de expulsión de la inversión planeada.
Si la Fed trata de mantener inalterable la tasa de interés aumentando el circulante, la curva DA se desplaza cada vez más a la derecha. El resultado es una inflación sostenida, quizá hasta una hiperinflación._

Ahora llevemos nuestro ejemplo un paso adelante. Supongamos que la Fed simpatiza con la
política fiscal de expansión (el aumento de G que acabamos de comentar) y decide expandir el
circulante para mantener constante la tasa de interés. Como este nivel superior de precios impulsa la demanda de dinero, la Fed amplía el circulante con la finalidad de que la tasa de interés no
cambie y se elimine el efecto de expulsión de una tasa más alta.
Cuando aumenta el circulante, la curva DA se expande de nuevo a la derecha, de DA1 a DA2.
Este movimiento de la curva DA, causado por el incremento del circulante, hace subir más los
precios. Por su parte, los precios altos aumentan más la demanda de dinero, lo que requiere un
nuevo incremento del circulante, etcétera.
¿Qué pasaría si la Fed tratara de mantener constante la tasa de interés cuando la economía opera en la parte empinada de la curva OA? La situación podría desembocar en una hiperinflación, un
periodo de incrementos muy rápidos del nivel general de precios. Si no puede sacarse más producción de la economía y si no se permite que la inflación planeada baje (porque la tasa de interés se
mantiene sin cambios), no es posible que aumente G. Como la Fed inyecta más y más dinero a la
economía para mantener invariable la tasa de interés, el nivel de precios no deja de aumentar.

#### 2.9.6 Inflación sostenida como fenómeno puramente monetario

La mayoría de los economistas está de acuerdo en que todo lo que incite un movimiento de la
curva DA a la derecha o de la curva OA a la izquierda puede dar lugar a un incremento del nivel
de precios. Entre estas causas se encuentran las políticas fiscales de expansión, expansión monetaria, choques de los costos, cambios de expectativas, etc. También se acepta en general que para
que ocurra una inflación sostenida, la Fed debe hacer adaptaciones. En este sentido, una inflación
sostenida puede considerarse como un fenómeno puramente monetario.
Este argumento, propuesto por los monetaristas (capítulo 19), ha ganado gran aceptación.
Es fácil demostrar, como acabamos de hacer, que ampliar el circulante puede desplazar de manera continua la curva DA. No es fácil dar con otras razones de los movimientos constantes de la
curva DA si el circulante es fijo. Una posibilidad es que el gobierno aumente el gasto en forma
permanente sin incrementar los impuestos, pero este proceso no puede extenderse por siempre.
Para financiar el gasto sin impuestos, el gobierno debe pedir prestado. Sin una expansión del circulante, la tasa de interés tiene un aumento drástico por el crecimiento de la oferta de bonos del
gobierno. Ahora, el público tiene que estar dispuesto a comprar los bonos del gobierno expedidos para financiar los aumentos de gastos. En algún momento, el público ya no quiere comprar
más bonos aunque la tasa de interés sea muy alta.3 En ese momento, el gobierno ya no puede aumentar el gasto que no financia con impuestos sin la cooperación de la Fed. Si esto es cierto, no
puede haber una inflación sostenida sin la participación de la Fed.

CURVA DE LA DEMANDA AGREGADA
1. La demanda de dinero está en función de tres variables: 1) la tasa
de interés (r), 2) el nivel de ingreso real (Y) y 3) el nivel de precios
(P). La demanda de dinero aumenta si se incrementa el nivel real
de la producción (ingreso) y el nivel de precios, o si baja la tasa de
interés.
2. Con un nivel de precios más alto, hogares y empresas necesitan
conservar más saldos de dinero. Si el circulante no cambia, la 
demanda mayor de dinero hace que la tasa de interés suba y que
el gasto en inversión planeada baje. Como resultado, el gasto total
planeado es menor, las existencias son mayores que lo planeado,
las empresas reducen la producción y Y baja. Un aumento del 
nivel de precios abate el nivel de la producción total (ingreso).
Una reducción del nivel de precios hace subir el nivel de la 
producción total (ingreso).
3. La demanda agregada es la demanda total de bienes y servicios de la
economía. La curva de la demanda agregada (DA) ilustra la relación
negativa entre la producción total (ingreso) y el nivel de precios. En
cada punto de la curva DA tanto el mercado de bienes como de dinero se encuentran en equilibrio. La curva DA no es la suma de todas las curvas de la demanda del mercado en la economía.
4. En todos los puntos de la curva de la demanda, la cantidad 
demandada agregada en la economía es exactamente igual 
al gasto total planeado.
5. Un aumento del circulante, así como de las compras gubernamentales, o una reducción de los impuestos netos a un nivel 
dado de precios desplazan la curva de la demanda agregada a 
la derecha. Una reducción en la cantidad del circulante, una 
disminución de las compras gubernamentales o un alza de los 
impuestos netos traslada la curva de la demanda agregada a la 
izquierda.
LA CURVA DE LA OFERTA AGREGADA
6. La oferta agregada es la oferta total de bienes y servicios de la
economía. En la curva de la oferta agregada (OA) se muestra la
relación entre la producción total que ofrecen todas las empresas
de la economía y el nivel general de precios. La curva OA no es
la curva de la oferta del mercado; tampoco es la simple suma de
todas las curvas de la oferta de la economía. Por esta razón, es
útil pensar en la curva OA como una curva de “respuesta de
precios y productos”; es decir, una curva que representa las 
decisiones de precios y de producción de todos los mercados y
empresas de la economía dadas las circunstancias prevalecientes.
7. La forma de una curva OA de corto plazo es causa de muchos 
debates en la macroeconomía. Muchos economistas creen que
con una producción muy baja la curva OA es bastante llana, y a
niveles altos de producción total es vertical, o casi. De este modo,
la curva OA tiene una pendiente en ascenso y se vuelve vertical
cuando la economía llega a su producción máxima o de capacidad plena.
8. Todo lo que tenga un efecto en las decisiones de una empresa individual puede desplazar la curva OA. Entre estos factores están
los choques de costos, crecimiento económico, estanflación, políticas públicas y desastres naturales.
NIVEL DE PRECIOS DE EQUILIBRIO
9. El nivel de precios de equilibrio de la economía se alcanza en el
punto donde se intersecan las curvas OA y DA. La intersección 
de estas curvas corresponde al equilibrio de los mercados de 
bienes y servicios, y también al grupo de decisiones de precios 
y productos de todas las empresas de la economía.
LA CURVA DE LA OFERTA AGREGADA DE 
LARGO PLAZO
10. Para que la curva OA tenga una pendiente ascendente, algunos
precios de los insumos deben demorarse luego de incrementos
del nivel general de precios. Si las tasas salariales y otros costos 
se ajustan por completo a los cambios de los precios en el largo
plazo, la curva OA de largo plazo es vertical.
11. El nivel de la producción total que puede sostenerse en el largo 
plazo sin inflación se llama producción potencial o PIB potencial.
DEMANDA Y OFERTA AGREGADAS Y POLÍTICA 
MONETARIA Y FISCAL
12. Si la economía produce en la parte llana de la curva OA una 
política de expansión (que desplaza la curva DA a la derecha) 
da por resultado un incremento menor del nivel de equilibrio 
de los precios en relación con el aumento de la producción de
equilibrio. Si al comienzo la economía produce en la parte
abrupta de la curva OA, una política de expansión trae un 
aumento menor de la producción de equilibrio y un incremento
grande del nivel de equilibrio de los precios.
13. Si la curva OA es vertical en el largo plazo, ni las políticas 
monetarias ni las fiscales tienen ningún efecto en la producción
total de largo plazo. Por esta razón, la duración del largo plazo
es una de las preguntas más apremiantes de la macroeconomía.
CAUSAS DE LA INFLACIÓN 
14. La inflación es un incremento del nivel general de precios. Una
inflación sostenida ocurre cuando el nivel general de precios 
sigue subiendo durante un periodo prolongado. La mayoría de
los economistas piensan que ocurren inflaciones sostenidas sólo
si la Fed aumenta de manera continua el circulante.
15. La inflación de la demanda es la que comienza con un incremento
de la demanda agregada. La inflación de los costos o de la oferta es
la que se genera por un aumento de los costos. Un alza de los 
costos también desemboca en estanflación: la situación en que 
la economía pasa por una contracción y una inflación al mismo
tiempo.
16. La inflación “se incorpora en el sistema” como resultado de 
las expectativas. Si los precios han estado subiendo y la gente 
se forma expectativas de acuerdo con el comportamiento 
anterior de los precios, las empresas siguen subiéndolos aun 
si la demanda se desacelera o contrae.
17. Cuando el nivel de precios sube, también aumenta la demanda 
de dinero. Si la economía opera en la parte abrupta de la curva 
OA y la Fed trata de mantener constante la tasa de interés 
expandiendo el circulante, el resultado podría ser una hiperinflación: un periodo de incrementos muy rápidos del nivel de precios.

### 3.3 Conceptos básicos del mercado de trabajo

Vamos a revisar de manera acuciosa lo que mide la tasa de desempleo (véase el capítulo 7). El primer viernes de cada mes, el Departamento del Trabajo de Estados Unidos publica los resultados de
una encuesta de hogares que proporciona una estimación del número de personas con empleo, los
empleados (E), así como de la cantidad de personas que buscan trabajo pero no lo encuentran, los
desempleados (U). La fuerza laboral (FL) es la suma de empleados más desempleados:

$FL = E + U$

La tasa de desempleo es el número de desempleados, como porcentaje de la fuerza laboral:

$tasa de desempleo = \frac{U}{FL}$

Repetimos, para ser desempleada, una persona debe carecer de empleo y estar buscando una colocación. Si deja de solicitar trabajo, se considera que está fuera de la fuerza laboral y ya no se le
cuenta como desempleada.
Es importante darse cuenta de que aun si la economía marcha a toda su capacidad o está
cerca de lograrlo, la tasa de desempleo nunca es de cero. La economía es dinámica. Los estudiantes
se titulan en las universidades y egresan de programas de capacitación; algunas empresas generan
utilidades y crecen, mientras que otras sufren pérdidas y quiebran; la gente entra y sale de la
fuerza de trabajo y cambia de profesión. Pasa tiempo para que las personas encuentren el trabajo
correcto y para que los patrones cubran las vacantes que tienen con los trabajadores indicados.
Este desempleo friccional y estructural es inevitable y, en muchos sentidos, deseable (repase el
capítulo 7 si estos términos le parecen oscuros).
En este capítulo nos interesa el desempleo cíclico, el aumento de la desocupación que ocurre
en recesiones y depresiones. Cuando la economía se contrae, el número de desocupados y la tasa
de desempleo crecen. Estados Unidos ha sufrido varios periodos de alto desempleo. Durante la
Gran Depresión, la tasa de trabajadores en paro se mantuvo en más de 17 por ciento durante casi una década. En diciembre de 1982, más de 12 millones de personas estaban sin ocupación, lo
cual situó la tasa de desempleo en 10.8 por ciento.
En cierto sentido, es obvia la razón de que el empleo decrezca cuando la economía pasa por
una baja. Cuando las empresas reducen la producción, necesitan menos trabajadores, así que despiden gente.

> **El empleo se reduce cuando la producción total mengua y aumenta cuando ésta se eleva.**

Sin embargo, una reducción de la demanda de mano de obra no significa por fuerza que el desempleo
aumente. Si los mercados operan como lo explicamos en los capítulos 3 y 4, una disminución de
la demanda de trabajo genera primero un excedente en la oferta de mano de obra. Por consiguiente, la tasa salarial baja hasta que la cantidad de mano de obra ofrecida sea igual a la que se
demanda, lo cual restituye el equilibrio al mercado de trabajo. A la nueva tasa salarial más baja,
quien quiera un trabajo lo consigue.
Si se hacen subir y bajar las tasas salariales para equilibrar la cantidad de mano de obra demandada y la cantidad ofrecida, no debe haber un desempleo persistente sobre las cifras estructurales y por fricción. Tal es la idea que sostenían los economistas clásicos que antecedieron a
Keynes. Asimismo, en la actualidad es la opinión de muchos especialistas en la materia.

### 3.4 Noción clásica del mercado de trabajo

En la figura 14.1 se ilustra la noción clásica del mercado de trabajo. Los economistas clásicos suponen que la tasa salarial se ajusta para equiparar la cantidad de mano de obra demandada con la
cantidad ofrecida, lo cual implica que no haya desempleo. Para ver cómo se verifica este ajuste,
supongamos que hay una reducción de la demanda de mano de obra que desplaza la curva de la
demanda de la figura 14.1 de D0 a D1. Esta disminución de la demanda hace que bajen la tasa salarial, de W0 a W1, y la cantidad de mano de obra demandada, de L0 a L1. La merma de la cantidad
de mano de obra ofrecida es un movimiento por la curva de la oferta de mano de obra.
En la figura 14.1, cada punto de la curva de la oferta de mano de obra representa la cantidad
de trabajo que quieren ofrecer los hogares a cada nivel salarial. La decisión de cada hogar sobre
cuánto trabajo ofrecer es parte del problema general de la elección de consumo de una familia.
Cada integrante de un hogar observa la tasa salarial del mercado, los precios de los productos y
el valor del tiempo libre (por ejemplo, el valor de quedarse en casa a trabajar en el jardín o cuidar
a los niños) y escoge qué cantidad de trabajo ofrece (si lo hiciera). Un integrante de un hogar que no está en la fuerza de trabajo ha decidido que su tiempo es más valioso en actividades que no
son del mercado.

Mercado clásico de trabajo

<img src="4. Analisis macroeconomico\Inflación, demanda y oferta agregadas\3.4 Noción clásica del mercado de trabajo.PNG" width=400 height=200 />

_Los economistas clásicos creen que el mercado de trabajo siempre se equilibra. Si la demanda de
mano de obra se desplaza de D0 a D1, el salario de equilibro baja de W0 a W1. Quien quiera un trabajo
en W1 lo conseguirá._

Es fácil ver por qué. Si decide salir de la fuerza laboral, el motivo es que usted, un miembro
de la sociedad, le confiere mayor valor al uso de su tiempo que el otorgado por la sociedad al
producto que usted haría si trabajara. Pensemos en los hogares de países menos desarrollados.
En ellos, la alternativa a trabajar por un salario es la agricultura de supervivencia. Si la tasa salarial del mercado de trabajo es demasiado baja, muchas personas prefieren cultivar por su
cuenta. En este caso, el valor de lo que producen en el campo debe ser mayor que la cotización
establecida por la sociedad a lo que producirían si trabajaran por un salario. Si no fuera así, los
salarios aumentarían y más personas se unirían a la fuerza laboral.
En la figura 14.1, cada punto de la curva de la demanda de trabajo (mano de obra) representa
la cantidad de trabajadores que quiere emplear una empresa a cada nivel salarial. La decisión de
cada compañía sobre cuántos operarios demandar es parte de la decisión general de maximizar
las utilidades. Una firma gana utilidades vendiendo su producción a las familias. Contrata trabajadores si el valor de su producción es suficiente para justificar el salario que paga. Así, la cantidad de
mano de obra que contrata una empresa depende del valor de la producción que entregan los
empleados.
Los economistas clásicos consideraban que el funcionamiento del mercado de trabajo (el
comportamiento de la oferta y la demanda de mano de obra) era óptimo desde el punto de
vista de hogares y empresas individuales, así como desde la perspectiva de la sociedad. Si las
familias quieren más producción de la que se realiza, aumenta la demanda de ésta, sube la tasa salarial y más trabajadores son atraídos a la fuerza laboral. (Algunos de los que preferían no
trabajar con la tasa salarial baja son alentados a incorporarse a la fuerza laboral con una tasa
más alta.) En el equilibrio, precios y salarios reflejan la compensación entre el valor que confieren los hogares a la producción y el valor del tiempo dedicado al ocio y al trabajo no remunerado. En el equilibrio, la gente que no trabaja decidió no laborar a la tarifa del mercado. En
este sentido, siempre hay pleno empleo. Los economistas clásicos creían que el mercado lograría resultados óptimos si se dejaba libre a sus propios medios y que el gobierno no podría hacer nada para mejorarlo.

#### 3.4.1 El mercado clásico de trabajo y la curva de la oferta agregada

Ahora podemos relacionar la noción clásica del mercado de trabajo con la teoría de la curva OA
vertical del capítulo 13. La idea clásica de que los salarios se ajustan para equilibrar el mercado laboral es congruente con la noción de que los salarios responden pronto a los cambios de precios.
Recuerde que el argumento de que la curva OA es vertical a la larga requiere ajustes de los precios de
los insumos. Si las curvas de la oferta OA y la demanda DA agregadas de corto plazo se intersecan
por arriba del PIB potencial (Y0 en la figura 13.10), los precios de los insumos, incluyendo salarios,
aumentan; asimismo, desplazan la curva OA a la izquierda y revierten Y a Y0. Del mismo modo, si
las curvas OA y DA de corto plazo se intersecan a la izquierda de Y0, el desempleo y el exceso de capacidad bajarían los precios de los insumos, incluyendo los salarios, lo cual desplazaría la curva
OA a la derecha y devolvería Y a Y0. Recuerde también que si la curva OA es vertical, la política
monetaria y fiscal no afecta el nivel de la producción ni el empleo en la economía. Por lo tanto, se
sigue que es probable que quienes piensan que la tasa salarial se ajusta deprisa para equilibrar el
mercado laboral, crean también que la curva OA es vertical (o casi vertical) y que la política monetaria y fiscal tiene poco efecto o ninguno sobre la producción y el empleo.

#### 3.4.2 La tasa de desempleo y la noción clásica

Si, como suponen los economistas clásicos, el mercado de trabajo funciona bien, ¿cómo se explica el
hecho de que a veces la tasa de desempleo parece alta? Hay ocasiones en que millones de personas
quieren trabajo con las tasas salariales prevalecientes y no lo encuentran. ¿Cómo reconciliamos esta
situación con la premisa clásica sobre el mercado laboral?
Algunos economistas responden que la tasa de desempleo no es una medida correcta del buen
funcionamiento del mercado laboral. Sabemos que la economía es dinámica y en todo momento algunas industrias se expanden y otras se contraen. Por ejemplo, consideremos a un carpintero que es
despedido por causa de una mengua en el sector de la construcción. Quizá adquirió habilidades específicas relacionadas con esta industria que no siempre serán útiles en otras partes. Si ganaba
40,000 dólares al año como carpintero, acaso gana sólo 30,000 en un área distinta. Tal vez ascienda
de nuevo a un salario de 40,000 dólares en el nuevo sector en la medida en que aprenda nuevas destrezas, aunque esto es tardado . ¿El carpintero aceptaría un trabajo de 30,000 dólares? Hay por lo
menos dos motivos por los que podría rechazarlo. El primero, que a lo mejor piense que la baja del
sector de la construcción es temporal y que pronto recuperará su empleo. En segundo lugar, tal vez
piense que ganaría más de 30,000 dólares en otro sector si siguiera buscando un mejor trabajo.
Si nuestro carpintero se decide a seguir en la búsqueda de un trabajo que pague más de
30,000 dólares por año, se consideraría desempleado porque busca ocupación. Esto no significa
de manera automática que el mercado de trabajo no funcione como es debido. El carpintero decidió no trabajar por un pago de 30,000 dólares anuales, pero si su valor para cualquier empresa
ajena al sector de la construcción no supera los 30,000 dólares anuales, no esperaríamos que encuentre un trabajo que le pague más. La tasa de desempleo, como la mide el gobierno, no es un
indicador exacto sobre la buena o mala marcha del mercado laboral.
Si el grado al que los sectores industriales cambian en la economía fluctúa con el tiempo, en
algunas épocas habrá más personas como nuestro carpintero que en otras. Esta variación hace
que oscile la tasa de desempleo. Algunos economistas aseveran que la tasa medida del desempleo
parece más alta, aunque el mercado de trabajo funcione bien. La cantidad del trabajo ofrecido a la
tasa salarial presente es igual a la cantidad demandada con esa tasa. El hecho de que haya personas
dispuestas a trabajar sólo por un salario mayor que el actual no significa que el mercado laboral
no funcione. Siempre que la curva de la oferta tiene una pendiente ascendente en un mercado
(como suele ocurrir en el mercado de trabajo), la cantidad ofrecida a un precio mayor que el de
equilibrio siempre es superior a la cantidad ofrecida al precio de equilibrio.
Los economistas que ven el desempleo de este modo no se percatan de que es un grave problema. Todavía recordamos las imágenes terribles de las colas para el pan en la década de los treinta y a
muchos les costaba trabajo creer que todo estaba bien a finales de 1982, cuando 12 millones de personas buscaban una ocupación. Como veremos enseguida, hay otras opiniones sobre el desempleo.
Algunos economistas piensan que el desempleo no es un problema grave. Sin embargo, todavía
recordamos las imágenes de la década de los treinta y durante la recesión de 1982, cuando
más de 12 millones de personas buscaban trabajo.

### 3.5 Explicación del desempleo

Si el desempleo es un grave problema macroeconómico (y muchos economistas piensan que lo
es, en efecto), tenemos que indagar entre los motivos propuestos como su causa. Entre ellos se
cuentan los salarios rígidos, la teoría de los sueldos de eficiencia, información imperfecta y leyes
del salario mínimo.

#### 3.5.1 Salarios rígidos

Una explicación del desempleo (más allá del desempleo estructural y por fricción) es que los salarios son rígidos para bajar. Es decir, el salario de equilibrio se atora en un nivel particular y no
baja cuando se reduce la demanda de mano de obra. Esta situación se ilustra en la figura 14.2,
donde el salario de equilibrio se detiene en W0 (el salario original) y no baja a W* cuando la demanda se reduce de D0 a D1. El resultado es desempleo por la cantidad de L0 – L1, en la que L0 es
la cantidad de mano de obra que los hogares quieren suministrar a la tasa salarial W0, y L1 es el
monto de trabajo que las empresas quieren contratar a esa tasa salarial de W0. L0 – L1 es el número de personas que quisieran trabajar a W0, pero no encuentran colocación.
La explicación del desempleo por los salarios rígidos hace una petición de principio. ¿Por qué
son rígidos los salarios, si lo fueran, y por qué los salarios no bajan para equilibrar el mercado de
trabajo en periodos de mucho desempleo? Se han propuesto muchas respuestas, pero a la fecha ninguna ha sido aceptada. Esta falta de consenso es una razón de que la macroeconomía haya pasado
tanto tiempo sin consolidarse. La existencia del desempleo sigue siendo un acertijo. Aunque vamos
a estudiar las principales teorías que han propuesto los economistas para explicar por qué los salarios no equilibran el mercado de trabajo, no podemos ofrecer conclusiones. El tema sigue abierto.
**Contratos sociales o implícitos** Una explicación de la lentitud de los salarios para
descender es que las empresas firman contratos sociales o implícitos con los trabajadores, a fin
de no reducir el pago. Para bajar los salarios, parece que se necesitan acontecimientos graves: recesión profunda, desregulación o amenaza de bancarrota. No hubo reducciones salariales durante la Gran Depresión, ni en el sector de la aviación civil después de la desregulación de 1980; tampoco a últimas fechas, cuando algunas empresas estadounidenses de manufactura corrieron el
peligro de quebrar ante la dura competencia foránea. Son excepciones a la regla general. Por razones que bien pueden ser más sociológicas que económicas, reducir los salarios es casi un tabú.
Un argumento relacionado, la explicación del desempleo por el salario relativo, sostiene que
los trabajadores están preocupados por su salario en relación con las remuneraciones de otros trabajadores en otras empresas y sectores económicos y no aceptan de buena gana reducciones a su paga,a menos que sepan que otros operarios sufren menguas semejantes. Como es difícil asegurarle a un
grupo de trabajadores que otros se encuentran en la misma situación, se resisten a cualquier rebaja
de su pago. Quizá hay un entendido tácito entre empresas y trabajadores de que las primeras no harán nada que empeore a los segundos, en relación con los trabajadores de otras partes.

Salarios rígidos

<img src="4. Analisis macroeconomico\Inflación, demanda y oferta agregadas\3.5.1 Salarios rígidos.PNG" width=400 height=200 />

_Si los salarios “se atoran” en W0, en lugar de bajar al nuevo salario de equilibrio de W*, luego de un
cambio en la demanda de D0 a D1, el resultado es un desempleo igual a L0 – L1._

**Contratos explícitos** Muchos trabajadores (en particular los sindicalizados) firman contratos laborales de uno a tres años con las empresas. En estos convenios se estipulan sus salarios
de cada año. Los pagos establecidos de tal manera no fluctúan con las condiciones de la economía, ni en ascenso ni en descenso. Si la economía se desacelera y las empresas demandan menos
trabajadores, el salario no baja, sino que más bien se despiden empleados.
Aunque los contratos explícitos dilucidan la lentitud de algunos salarios, también hay que
considerar una cuestión más profunda. Sin duda, en el momento de firmar un contrato, trabajadores y empresas saben que ciertos eventos pueden hacer que los salarios establecidos en el
contrato sean demasiado altos o muy bajos. ¿Por qué ambas partes se obligan de esta manera?
Una explicación es que es costoso pactar los salarios. Las negociaciones entre sindicatos y compañías tardan mucho tiempo, que bien podría dedicarse a la producción, y sería muy caro acordar los
salarios cada semana o cada mes. Los contratos son una manera de incurrir en estos costos en
plazos de no más de uno, dos o tres años. Hay un equilibrio entre los costos de afianzar a empresas
y trabajadores en contratos tan dilatados y los costos de las negociaciones salariales. Al parecer, la
duración de los contratos que reduce al mínimo los costos de las negociaciones (dicho a partir
de lo que observamos en la práctica) es de entre uno y tres años.
Algunos contratos de varios años se adaptan a los sucesos imprevistos mediante la previsión
de ajustes por el costo de la vida (ACV). Estas correcciones indexan los salarios a los cambios en
el costo de la vida: cuanto mayor sea la inflación, más subirán los salarios. Así, estas rectificaciones protegen a los trabajadores de una inflación inesperada, aunque muchos suben los pagos en
un porcentaje menor que el del aumento de los precios.

#### 3.5.2 Teoría de los salarios de eficiencia

Otra explicación del desempleo se centra en la teoría de los salarios de eficiencia, la cual sostiene que la productividad de los trabajadores aumenta con la tasa salarial. De ser así, las empresas
deberán tener un incentivo para dar salarios superiores a la paga en que la cantidad de trabajo
ofrecido sea igual a la cantidad de trabajo demandado.
Una empresa tiene un estímulo para contratar trabajadores siempre que el valor de lo que
produzcan sea igual o mayor que la tasa salarial. Sin los efectos de la eficiencia, el mercado de la
figura 14.2 arrojaría un salario de equilibrio de W*. Pero supongamos que la compañía pudiera
aumentar la productividad de todos sus trabajadores elevando la tasa salarial a más de W*. La
demanda de trabajo de la empresa no sería menor, pero con una tasa salarial más alta, la cantidad
de la mano de obra ofrecida aumentaría; además, superaría la cantidad demandada con la nueva
tasa salarial más alta (el salario de eficiencia) y el resultado sería el desempleo.
En estudios empíricos de los mercados laborales se han identificado varios beneficios potenciales que obtienen las empresas por pagar a sus trabajadores más del salario de equilibrio del
mercado. Algunas de esas ventajas son rotación mínima, aumento de la moral y menor “encogimiento” del trabajo. Aunque la teoría del salario de eficiencia anticipa algún desempleo, es poco
probable que el comportamiento que describe explique mucho de las grandes variaciones cíclicas
del desempleo al paso del tiempo.

#### 3.5.2 Información imperfecta

Hasta aquí hemos supuesto que las empresas saben con precisión qué tasas salariales tienen que establecer para equilibrar el mercado laboral. Quizá decidan no fijar sus salarios en este nivel, pero
por lo menos saben cuál es ese monto. No obstante, en la práctica las empresas no siempre tienen
suficiente información para saber cuál es el sueldo de equilibrio del mercado. En este caso, se dice
que tienen información imperfecta. Si las empresas tienen información imperfecta o incompleta,
determinan mal los salarios; es decir, designan salarios que no equilibran el mercado de trabajo.
Si una empresa fija los salarios demasiado altos, más operarios querrán trabajar para ella de los
que quiera contratar, y algunos trabajadores potenciales serán rechazados. Desde luego, el resultado es desempleo. Una objeción de esta idea es que explica el desempleo sólo en el más breve plazo. En
cuanto una empresa ve que ha cometido un error, ¿por qué no lo corrige de inmediato y ajusta
sus salarios al nivel correcto, el que despeja al mercado? ¿Por qué persiste el desempleo?
Si la economía fuera simple, no se necesitarían más que unos cuantos meses para que las empresas corrigieran sus errores pero la economía es compleja. Aunque estén al tanto de sus equivocaciones y traten de rectificarlas, todo el tiempo pasan cosas nuevas. Como el cambio constante es
característico de la economía, incluyendo que varíe de manera continua el nivel de equilibrio de los
salarios, a las empresas les resulta difícil ajustar sus salarios al nivel de equilibrio del mercado. El mercado de trabajo no es como el mercado accionario o el del trigo, donde los precios se determinan en
intercambios organizados todos los días. Por el contrario, miles de empresas fijan salarios y millones
de trabajadores responden a ellos. Llega a pasar mucho tiempo para que se determinen los salarios
que aclaran el mercado, después de que se ha alterado su estado de equilibrio.

#### 3.5.2 Leyes de salario mínimo

Las leyes del salario mínimo explican por lo menos una pequeña parte del desempleo. Estas legislaciones determinan el límite inferior de los niveles salariales, es decir, la tarifa mínima por hora
o por día para cualquier género de labor. En 2005, el sueldo mínimo federal era de 5.15 dólares
por hora. Si el salario de equilibrio del mercado de algunos grupos de trabajadores está debajo de
este monto, el grupo estará desempleado. En la figura 14.2, si el salario mínimo es de W0 y el salario de equilibrio del mercado es W*, el número de desempleado es de L0 – L1.
Los adolescentes, que tienen poca experiencia laboral, suelen ser los más dañados por las leyes del salario mínimo. Si algunos adolescentes pueden generar un valor de producción de tan sólo 4.50 dólares por hora, ninguna empresa estaría dispuesta a contratarlos por un salario de 5.15,
pues de hacerlo incurrirían en pérdidas de 0.65 por hora. En un mercado no regulado, estos jóvenes podrían encontrar trabajo al salario de saldo de 4.50 dólares por hora. Si las leyes del salario
mínimo impiden que el pago baje de 5.15 dólares, estos trabajadores no encontrarán ocupación y
estarán desempleados. Otros que también salen perjudicados son los que tienen muy pocas habilidades y algunos inmigrantes recientes.
En respuesta a este argumento contrario al salario mínimo, el Congreso estadounidense estableció un apartado del salario mínimo especial para adolescentes. La nueva ley permite a los
patrones contratar jóvenes con un “sueldo de oportunidad” de 4.25 dólares hasta por 90 días.

#### 3.5.2 Una pregunta abierta

Como hemos visto, hay muchas explicaciones sobre por qué el mercado de trabajo no se despeja.
Las teorías que acabamos de proponer no por necesidad se excluyen unas a otras y habrá elementos de verdad en todas. El mercado total del trabajo es muy complejo y no hay respuestas simples
a la pregunta por el desempleo. Muchas de las investigaciones actuales de la macroeconomía atañen en forma directa o indirecta a esta pregunta y es un campo estimulante de estudio. Qué argumentos prevalezcan al final está por verse.

### 3.6 Relaciones de corto plazo entre tasa de desempleo e inflación

La relación entre la tasa de desempleo y la inflación (dos de las variables más importantes de la
macroeconomía) ha sido un tema muy debatido. Ahora tenemos suficientes conocimientos macroeconómicos para explorar esta conexión.
Tenemos que empezar por considerar el vínculo entre la producción total (ingreso, Y) y la
tasa de desempleo (U). Un aumento de Y significa que las empresas producen más. Para lograr
esto, se necesitan más trabajadores en el proceso de producción. Por consiguiente, un incremento
en Y conduce a un incremento del empleo. Esto significa que más personas trabajan (hay menos
desempleados) y baja la tasa de desempleo. A un crecimiento de Y corresponde una disminución
de U. Así, U y Y guardan una relación inversa:

> **Cuando Y aumenta, la tasa de desempleo baja y cuando Y desciende, la tasa de desempleo sube.**

Curva de la oferta agregada

<img src="4. Analisis macroeconomico\Inflación, demanda y oferta agregadas\3.6 Relaciones de corto plazo entre tasa de desempleo e inflación-1.PNG" width=400 height=200 />

_La curva OA muestra una relación positiva entre el nivel de precios (P) y la producción total (ingreso, Y)._

Ahora consideremos una curva OA con pendiente ascendente, como se muestra en la figura
14.3. Esta curva representa la relación entre Y y el nivel general de precios (P). La relación es positiva: cuando P aumenta, también lo hace Y; cuando P baja, Y disminuye.
Como recordará del capítulo anterior, la forma de la curva OA está determinada por el
comportamiento de las empresas y cómo reaccionan a un incremento de la demanda. Si la demanda agregada se desplaza a la derecha y la economía opera en la parte casi llana de la curva
OA (lejos de la capacidad plena), la producción aumenta pero el nivel de precios no cambia demasiado. Por el contrario, si la economía opera en la parte empinada de la curva OA (cerca de la
capacidad plena), una ampliación de la demanda eleva el nivel de precios, aunque la producción
está limitada por la capacidad y no crece mucho.
Piense en lo que ocurre luego de un acontecimiento que traiga un incremento en la demanda
agregada. En primer lugar, las empresas sufren una mengua imprevista de sus existencias. Responden aumentando la producción (Y) y contratando trabajadores (la tasa de desempleo baja).
Si la economía no está cerca de su capacidad plena, el nivel de precios sube poco. Pero si la demanda agregada sigue creciendo, el poder de la economía de incrementar la producción acaba
por llegar a su tope. A medida que la demanda agregada se desplaza cada vez más a la derecha
sobre la curva OA, el nivel de precios aumenta y la producción comienza a tocar sus límites. En el
punto en que la curva OA se vuelve vertical, la producción ya no se acrecienta. Si la producción no
puede crecer, la tasa de desempleo no puede bajar más.

> **Hay una relación inversa entre la tasa de desempleo y el nivel de precios. Cuando la tasa de desempleo baja porque la economía se acerca en forma progresiva al límite de capacidad de producción, el nivel general de precios aumenta cada vez más, como se muestra en la figura 14.4.**

La curva de la figura 14.4 no ha sido objeto de gran atención en la macroeconomía. Más bien,
la curva que ha sido muy estudiada se muestra en la figura 14.5, donde se da la gráfica de la tasa
de inflación en el eje vertical y la tasa de desempleo en el horizontal. La tasa de inflación es el
cambio porcentual del nivel de precios, no el nivel de precios en sí.
Las figuras 14.4 y 14.5 tienen significados diferentes. En la figura 14.4 se muestra que el nivel
de precios es el mismo si no cambia la tasa de desempleo. La curva de la figura 14.5 se llama curva
de Phillips, por el economista inglés A. W. Phillips, que fue el primero en estudiarla con datos del
Reino Unido. Dicho en los términos más simples, la curva de Phillips es una gráfica que muestra
la relación entre las tasas de inflación y de desempleo.
En el resto del capítulo nos concentramos en la curva de Phillips de la figura 14.5 porque
es la relación macroeconómica más estudiada. Pero recuerde que no es fácil pasar de la curva
OA a la de Phillips. Hemos ido de las gráficas donde el nivel de precios aparece en el eje vertical (figuras 14.3 y 14.4) a gráficas donde el eje vertical está ocupado por el cambio porcentual del nivel de precios (figura 14.5). Dicho de otra manera, la teoría que sustenta la curva de Phillips es
distinta que en el caso de la curva OA. Por fortuna, casi todos los conocimientos adquiridos en el
análisis OA/DA referentes al procedimiento del nivel de precios se aplican también al comportamiento de la tasa de inflación.

En los años cincuenta y sesenta se dio una relación bastante uniforme entre las tasas de desempleo y de inflación, como indica la figura 14.6 con respecto a la década de los sesenta. Como se

tenemos que aceptar una tasa de desempleo mayor, y para abatir la tasa de desempleo, tenemos
que aceptar una tasa de inflación más alta.
Los manuales escolares escritos en los años sesenta y a comienzos de la siguiente década recurrían a la curva de Phillips como la principal explicación de la inflación. Las cosas parecíansimples: la inflación parecía responder de manera muy previsible a los cambios de la tasa de desempleo. Por este motivo, en los sesenta las discusiones sobre políticas públicas giraban en torno
de la curva de Phillips. Se pensaba que el papel de las autoridades era escoger un punto de la curva.
Los conservadores proponían que se escogiera un punto con una tasa de inflación baja y estaban
dispuestos a aceptar a cambio una tasa de desempleo más alta. Los liberales pedían que se admitiera más inflación para mantener el desempleo en un nivel bajo.

#### 3.6.2 Análisis de la oferta y la demanda agregadas y curva de Phillips

¿Cómo se explica la estabilidad de la curva de Phillips en los años cincuenta y sesenta y la falta de
estabilidad en lo sucesivo? Para responder, tenemos que volver al análisis OA/DA.
Si la curva DA se desplaza de un año al otro pero la curva OA no lo hace, los valores de P y
Y cada año se disponen sobre la curva OA (figura 14.8(a)). La gráfica de la relación entre P y Y tendrá una pendiente hacia arriba. En correspondencia, la gráfica de la correspondencia entre la tasa
de desempleo (que disminuye conforme la producción crece) y la tasa de inflación será una curva
con pendiente en descenso. En otras palabras, esperaríamos ver una relación inversa entre las tasas
de desempleo y de inflación.
Ahora bien, la relación entre las tasas de desempleo y de inflación tendrá otro aspecto si la
curva OA se mueve cada año sin que lo haga la curva DA. Un desplazamiento a la izquierda de
la curva OA genera un incremento del nivel de precios (P) y una disminución de la producción
total (Y; figura 14.8(b)). Cuando la curva OA se traslada a la izquierda, la economía sufre inflación y también un incremento de la tasa de desempleo (porque la reducción de la producción
significa que el desempleo crece). En otras palabras, si la curva OA se desplaza año con año, esperaríamos ver una relación positiva entre las tasas de empleo y de inflación.
Si las dos curvas OA y DA se desplazan al mismo tiempo, no hay una relación sistemática entre
P y Y (figura 14.8(c)) ni, por consiguiente, entre las tasas de desempleo y de inflación.
**La función de los precios de las importaciones** Uno de los principales factores
que hacen que la curva OA se desplace es el precio de las importaciones. (Recuerde que la curva OA
se desplaza cuando cambian los precios de los insumos, los cuales son afectados por los costos de
las importaciones, sobre todo del petróleo importado.) En la figura 14.9 se muestra la gráfica de los
precios de las importaciones en el periodo 1960 I-2005 II. Como se observa, el precio de las importaciones cambió muy poco entre 1960 y 1970. No hubo desplazamientos grandes de la curva OA
en los años sesenta debidos a cambios en los precios de las importaciones. En esa década tampoco
hubo cambios acusados en los precios de los insumos, así que en general la curva OA se movió
muy poco en dicho lapso. La principal variación durante los sesenta ocurrió en la demanda agregada, así que el desplazamiento de la curva DA recorrió puntos de la curva OA.
En la figura 14.9 también se muestra que el precio de las importaciones aumentó considerablemente en los setenta. Este incremento llevó a grandes movimientos de la curva OA durante la década, pero la curva DA también se desplazó en el mismo periodo. Al moverse las dos curvas, los
puntos correspondientes a los datos de P y Y se dispersaron en toda la gráfica y la relación observada entre P y Y no resultó sistemática en absoluto.

La historia de los precios de las importaciones y las curvas OA y DA en los años sesenta y setenta se extiende a la curva de Phillips. Ésta fue estable en los sesenta porque la principal fuente
de variación de la economía era la demanda, no los costos. En la década de los setenta variaban
tanto la demanda como los costos, así que no se manifestó ninguna relación evidente entre las
tasas de desempleo y de inflación. En cierta medida, lo notable de la curva de Phillips no es que
no haya sido uniforme después de los sesenta, sino que lo haya sido antes.

#### 3.6.3 Expectativas y curva de Phillips

Otra causa de que la curva de Phillips no sea estable atañe a las expectativas.Vimos en el capítulo 13
que si una empresa espera que otras aumenten sus precios, sube el precio de su producto. Si todas las firmas se conducen de esta manera, los precios se elevarán porque se espera que lo hagan.
En este sentido, las expectativas se cumplen por sí mismas. De igual manera, si se espera que la
inflación sea elevada, los salarios negociados son más altos que si se anticipara poca inflación.
Así, la inflación de los salarios sufre la influencia de lo que se espera sobre la inflación futura de
los precios. Como los salarios son costos de insumos, los precios se acrecientan cuando las empresas responden a los costos salariales mayores. Las expectativas sobre los precios que inciden
en los contratos salariales al final repercuten en los propios precios.
Si la tasa de inflación depende de las expectativas, la curva de Phillips se desplaza con el
cambio de éstas. Por ejemplo, si aumentan las expectativas de inflación, el resultado será un alza
de la tasa de inflación aun si la tasa de desempleo no cambia. En este caso, la curva de Phillips se
desplaza a la derecha. Si bajan las expectativas de inflación, la curva de Phillips se desplaza a la
izquierda (hay menos inflación para cualquier tasa de desempleo).
En los años cincuenta y sesenta las expectativas inflacionarias fueron bastante estables. La tasa
de inflación fue moderada durante la mayor parte del periodo y la gente esperaba que siguiera
siendo así. Como estas expectativas no cambiaron demasiado, no hubo grandes movimientos de
la curva de Phillips, una situación que explica su estabilidad en esa época.
Hacia el final de los años sesenta las expectativas de la inflación comenzaron a crecer, sobre
todo como reacción al incremento real de la inflación que ocurría por la economía austera causada por la Guerra de Vietnam. Las expectativas de inflación crecieron todavía más en los años
setenta, como resultado de grandes alzas en los precios del petróleo. Estas nuevas expectativas
trajeron cambios en la curva de Phillips y causaron que ésta no fuera estable en esa década.

#### 3.6.4 ¿Hay un equilibrio a corto plazo entre inflación y desempleo?

¿El hecho de que la curva de Phillips se desmoronara en la década de los setenta significa que no
hay un equilibrio entre inflación y desempleo en el corto plazo? En absoluto: más bien significa que hay otras influencias en la inflación, aparte del desempleo. Así como se desplaza la relación entre el precio y la cantidad demandada en una curva normal de la demanda cuando el ingreso u otros
factores cambian, la relación entre desempleo e inflación se modifica cuando varían otros factores.
Por ejemplo, en 1975 la inflación y el desempleo eran altos. Como ya lo explicamos, esta estanflación fue causada en parte por un aumento de los precios del petróleo que desplazó a la izquierda
la curva de la oferta agregada y en parte por expectativas de que continuaría la inflación que mantuvo los precios al alza a pesar de los grandes niveles de desempleo. En respuesta a esta situación,
la Reserva Federal aplicó una política monetaria de contracción que trasladó la curva DA a la izquierda y elevó todavía más el desempleo. En 1977, la tasa de inflación había bajado de más de
11 por ciento a casi 6 por ciento. Entonces, la tasa de desempleo llevó a una reducción de la inflación, resultado que refleja un equilibro.

> **Hay una relación de corto plazo entre inflación y desempleo, pero otros factores, aparte de la cesantía, repercuten en la inflación. Las políticas consisten en mucho más que escoger un punto en una curva uniforme y limpia.**

En el capítulo 7 dijimos que acaso las recesiones son el precio que la economía paga por eliminar la inflación. Ahora podemos entender mejor esta declaración. Cuando el desempleo sube,
en igualdad de circunstancias, la inflación baja.

### 3.7 Curva de oferta agregada de largo plazo, PIB potencial y tasa natural de desempleo

Recuerde que en el capítulo 13 vimos que muchos economistas creen que la curva OA es vertical a la
larga. En el corto plazo, sabemos que los precios de algunos insumos (que son costos para las empresas) se retrasan con respecto a los aumentos del nivel general de precios. Si el nivel de precios sube sin
un ajuste completo de los costos, las utilidades de las empresas son mayores y la producción crece.
Pero en el largo plazo los precios de los insumos alcanzan a los incrementos de los precios de los productos. Si las cotizaciones de los insumos suben en periodos subsiguientes e impulsan alzas de costos,
la curva de la oferta agregada de corto plazo se desplaza a la izquierda y la producción total baja.
Esta situación se ilustra en la figura 14.10. Supongamos que el equilibrio inicial se encuentra en
la intersección de DA0 y la curva de la oferta agregada de largo plazo. Ahora consideremos un desplazamiento de la curva de la demanda agregada de DA0 a DA1. Si los precios de los insumos se demoran con respecto a los cambios del nivel general de precios, la producción total aumenta de Y0
a Y1 (se trata de un movimiento sobre la curva OA de corto plazo, OA0). A la larga, los precios de
los insumos se recuperan. Por ejemplo, los contratos laborales del año siguiente prevén el hecho
de que este año los incrementos salariales no correspondieron al costo de la vida. Si los precios de
los insumos se recuperan en el largo plazo, la curva OA se desplaza de OA0 a OA1 y devuelve la
producción total a Y0. Si al final los precios de los insumos suben en el mismo porcentaje que los
precios de los productos, las empresas dan el mismo nivel de producción que antes del aumento
de la demanda agregada.

Curva de Phillips de largo plazo: tasa natural de desempleo

<img src="4. Analisis macroeconomico\Inflación, demanda y oferta agregadas\3.7 Curva de oferta agregada de largo plazo, PIB potencial y tasa natural de desempleo.PNG" width=600 height=400 />

_Si la curva OA es vertical en el largo plazo, lo mismo pasa con la curva de Phillips. A la larga, la curva de Phillips corresponde a la tasa
natural de desempleo; es decir, la tasa de desempleo que es congruente con la noción de una producción fija de largo plazo al PIB potencial. U* es la tasa natural de desempleo._

En el capítulo 13 dijimos que Y0 también se llama PIB potencial. La producción total puede
impulsarse sobre Y0 en el corto plazo. Pero si la producción total excede Y0, hay una presión para
que suban los precios de los insumos y los costos. La tasa de desempleo ya es bastante baja, las
empresas comienzan a tocar los límites de capacidad de sus plantas, etc. En niveles de producción
total mayores que Y0, los costos crecen, la curva OA se desplaza a la izquierda y el nivel de precios
aumenta. Así, el PIB potencial es el nivel de producción total que puede sostenerse en el largo plazo sin inflación.
Esta historia se relaciona directamente con la curva de Phillips. Quienes piensan que en el
largo plazo la curva OA es vertical al PIB potencial, también creen que la curva de Phillips es vertical a cierta tasa natural de desempleo en el mismo término. Recuerde que en el capítulo 7 dijimos
que la tasa natural de desempleo se refiere a la cesantía que se presenta como parte normal del
funcionamiento de la economía. A veces se toma como la suma del desempleo por fricción y el
desempleo estructural. La lógica en la que se funda la curva de Phillips dicta que cuando la tasa de
desempleo baja de la tasa natural, los salarios comienzan a aumentar, lo cual lleva los costos hacia
arriba. Esto causa un nivel menor de producción que hace volver el desempleo a la tasa natural.
En la tasa natural, puede considerarse que la economía está en situación de pleno empleo.

#### 3.7.1 Tasa de inflación no acelerada del desempleo (TINAD)

En la figura 14.10, la curva de Phillips vertical en el largo plazo es una gráfica donde se mide la inflación sobre el eje vertical y la tasa de desempleo en el horizontal. La tasa natural de desempleo
es U*. A la larga, de acuerdo con los defensores de la curva de Phillips vertical en el largo plazo, la
tasa real de desempleo se mueve a U* por el funcionamiento natural de la economía.
Otra gráfica de interés se muestra en la figura 14.11, que ilustra el cambio de la tasa de inflación sobre el eje vertical y la tasa de desempleo en el eje horizontal. Muchos economistas piensan que la relación entre el cambio de la inflación y la tasa de desempleo es como la representa la curva PP de la figura. El valor de la tasa de desempleo donde la curva PP cruza el origen se llama tasa
de inflación no acelerada del desempleo (NAIRU, por sus siglas en inglés). Si la tasa real de desempleo está a la izquierda de la TINAD, el cambio de la tasa de inflación es positivo. Según se ilustra
en la figura, en U1 el cambio de la tasa de inflación es 1. Por el contrario, si la tasa de desempleo
real está a la derecha de la TINAD, el cambio de la tasa de inflación es negativo: en U2 el cambio
es de –1.

El diagrama TINAD

<img src="4. Analisis macroeconomico\Inflación, demanda y oferta agregadas\3.7.1 Tasa de inflación no acelerada del desempleo (TINAD).PNG" width=600 height=400 />

_A la izquierda de la TINAD, el nivel de precios se acelera (cambios positivos de la tasa de inflación) y a la
derecha se desacelera (variaciones negativas de la tasa de inflación). Sólo cuando la tasa de desempleo
es igual a la TINAD, el nivel de precios cambia a una tasa constante (sin cambio de la tasa de inflación)._

Consideremos lo que ocurre si la tasa de desempleo baja de la TINAD a U1 y se mantiene allí
durante varios periodos. Supongamos también que la tasa de inflación de la TINAD fue de 2 por
ciento. Entonces, en el primer periodo la tasa de inflación aumenta de 2 a 3 por ciento. Pero la tasa
de inflación se queda en esa cifra superior de 3 por ciento. En la siguiente etapa la tasa de inflación
aumenta de 3 a 4 por ciento, y así sucesivamente. El nivel de precios se acelera (es decir, el cambio
de la inflación es positivo) cuando la tasa real de desempleo está debajo de la TINAD. Por el contrario, el nivel de precios se desacelera (es decir, el cambio de la tasa de inflación es negativo) si la
tasa real de desempleo está arriba de la TINAD.1
La curva PP de la figura 14.11 es como la curva OA de la figura 14.3: los mismos factores
que desplazan la curva OA, como los choques de los costos, también desplazan esa curva PP. En
la figura 13.8 se resumen los diversos factores que desplazan la curva OA y que también atañen
a la curva PP. Un movimiento favorable de la curva PP es a la izquierda, porque la curva PP cruza
el origen a una tasa de desempleo más baja, lo cual indica que la TINAD es menor. Algunos dicen
que una posible fuente nueva de desplazamientos favorables es un aumento de la competencia extranjera, que mantiene bajos los costos de los salarios y de otros insumos.
Antes de 1995, más o menos, los defensores de la teoría de la TINAD afirmaban que el valor de
ésta en Estados Unidos era de casi 6 por ciento. A finales de 1995 la tasa de desempleo descendió a
5.6 por ciento y en 2000 bajó más a 3.9 por ciento. Si la TINAD fuera de 6 por ciento, hubiera
debido verse un incremento continuo de la tasa de inflación desde alrededor de 1995. De hecho,
durante el periodo de 1995 a 2000 se vio una ligera mengua de la inflación. La inflación, pues, no
aumentó; ni siquiera subió de una vez a una cifra superior para mantenerse ahí. A medida que
disminuía la tasa de desempleo durante ese periodo, los defensores de la TINAD revisaron sus
cálculos para situarla más o menos al parejo de la baja real de la tasa de desempleo. Esta recalibración se justifica diciendo que ha habido desplazamientos favorables continuos de la curva PP, como
una competencia foránea quizá mayor. Sin embargo, los críticos dicen que este procedimiento casi
equivale a despojar de sentido la teoría de la TINAD. ¿Es posible poner a prueba una teoría si el
cálculo de la TINAD se cambia cada vez que no concuerda con los datos? ¿Qué tan confiable es
apelar a los desplazamientos favorables?
En la actualidad, los expertos en macroeconomía debaten si las ecuaciones calculadas según
la teoría de la TINAD son buenas aproximaciones. Se necesita más tiempo para dar con respuestas definitivas.

### 3.8 Conclusiones

CONCEPTOS BÁSICOS DEL MERCADO
DE TRABAJO
1. Como la economía es dinámica, el desempleo por fricción y el
estructural son inevitables y en cierto sentido deseables. Las
épocas de desempleo cíclico causan preocupación a las autoridades
encargadas de la macroeconomía.
2. En general, el empleo baja cuando la producción total mengua y
sube cuando ésta se incrementa.
NOCIÓN CLÁSICA DEL MERCADO DE TRABAJO
3. Lo economistas clásicos creen que la interacción de la oferta y la
demanda en el mercado de trabajo produce un equilibrio y que
el desempleo (aparte del estructural y por fricción) no existe.
4. La noción clásica del mercado de trabajo es congruente con la
teoría de la curva vertical de la oferta agregada.
EXPLICACIÓN DEL DESEMPLEO
5. Algunos economistas piensan que el desempleo no es un indicador
adecuado de si un mercado laboral funciona apropiadamente.
Los desempleados que se consideran parte de la población
activa quizá se rehúsan a aceptar los puestos que se les ofrecen a
causa de los salarios que pagan. Algunos desempleados también
optan por no trabajar, pero este resultado no significa que el
mercado de trabajo funcione mal.
6. Quienes no se adhieren a la noción clásica del mercado laboral
proponen varias causas del desempleo. Los salarios rígidos para
bajar son quizá producto de contratos sociales (implícitos) o
explícitos que no permiten reducir los pagos. Si la tasa salarial de
equilibrio baja pero se impide que los salarios caigan también,
el resultado es desempleo.
7. La teoría del salario de eficiencia dice que la productividad de
los trabajadores aumenta con la tasa salarial. Si fuera verdad,
las empresas tendrían un incentivo para pagar salarios mayores
al monto al que la cantidad de trabajo ofrecido es igual a la
cantidad demandada. Si todos los salarios están arriba del nivel
de equilibrio, habrá un excedente de mano de obra y, por lo
tanto, desempleo.
8. Si las empresas operan con información incompleta o imperfecta,
quizá no sepan cuál es el salario de saldo del mercado. Como
resultado, es posible que fijen mal sus salarios y generen
desempleo. Como la economía es tan compleja, pasa mucho
tiempo para que las empresas corrijan estos errores.
9. Las leyes del salario mínimo, que fijan una base para las tasas
salariales, son un factor que contribuye al desempleo de
adolescentes y trabajadores muy poco calificados. Si el salario
de saldo de determinados grupos de trabajadores está debajo
del mínimo, algunos integrantes de esos grupos estarán
desempleados.
RELACIONES DE CORTO PLAZO ENTRE TASA
DE DESEMPLEO E INFLACIÓN
10. Hay una relación negativa entre la tasa de desempleo (U) y
la producción total (ingreso, Y): cuando Y aumenta, U baja;
cuando Y baja, U crece.
11. La relación entre la tasa de desempleo y el nivel de precios es negativa: cuando la tasa de desempleo baja y la economía se acerca
a toda su capacidad, el nivel de precios aumenta cada vez más.
12. La curva de Phillips representa la relación entre la tasa de
inflación y la tasa de desempleo. En los años cincuenta y sesenta,
esta relación era estable y parecía que había un equilibrio
previsible entre inflación y desempleo. Como resultado de
aumentos de los precios de las importaciones (que causan
desplazamientos de la oferta agregada) y las variaciones de
la demanda agregada causadas en parte por expectativas de
inflación, la relación entre la tasa de inflación y el nivel de desempleo fue errática en la década de los setenta. Hay un
equilibrio transitorio entre inflación y desempleo, pero además
del desempleo, otros factores repercuten en la inflación.
CURVA DE OFERTA AGREGADA DE
LARGO PLAZO, PIB POTENCIAL Y TASA
NATURAL DE DESEMPLEO
13. Quienes creen que la curva OA es vertical a la larga, también
creen que la curva de Phillips es vertical en el largo plazo a la
tasa natural de desempleo. Por lo general, la tasa natural es la
suma de las tasas estructural y por fricción. Si la curva de Phillips
es vertical en el largo plazo, entonces hay un límite a cuánto
puede el gobierno abatir la tasa de desempleo sin activar la
inflación.
14. La teoría TINAD afirma que el nivel de precios se acelera cuando
la tasa de desempleo está debajo de la TINAD y se desacelera
cuando la tasa de desempleo está arriba de esa cifra.
