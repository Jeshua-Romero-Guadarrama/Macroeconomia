## 1. Introducción a la macroeconomía

### 1.1 Contenido

- Las raíces de la macroeconomía se ecneuntran en la Gran Depresión, que forma parte de la historia reciente de la macroeconomía.
- Los principales intereses de la macroeconomía son la inflación y deflación, el aumento de la producción de corto y largo plazos y el desempleo.
- El gobierno formula la política macroeconomía de una nación; esto es, crea la política fiscal, política monetaria y políticas de crecimiento que deben seguir la sociedad.
- Uno de los componentes escenciales de la macroeconomía es el diagrama del flujo circular, que refleja los tres ámbitos del mercado.
- La metodología de la macroeconomía se encuentra microfundamentada, llegando hasta la demanda y oferta agregadas.
- La economía mexicana desde 1900: ciclos y tendencias; esto es, la expansión y contracción de los ciclos comerciales y económicos.
- La macroeconomía mexicana desde 1970.

### 1.2 Introducción

La macroeconomía es parte de la vida diaria de cada país; cuando marcha bien, son pocos quienes no tienen trabajo, los ingresos de las personas aumentan y las utilidades de las empresas son altas. En dicho contexto, dominado por una expansión del sistema económico en su conjunto, a los recién llegados al mercado laboral (como los estudiantes que acaban de titularse) les resulta fácil encontrar un trabajo bien pagado. 

Por el contrario, si el sistema macroeconómico se ecuentra en declive, es difícil encontrar un empleo nuevo, los ingresos no
aumentan lo suficiente y las utilidades son marginales. Dados los grandes efectos que llega a tener la macroeconomía en la vida de cientos de miles de personas, es importante que la sociedad, en general, y los economistas, en particular, entiendan el fucnionamiento del sistema económico a gran escala.

### 1.3 Diferencias entre microeconomía y macroeconomía.

La microeconomía examina el funcionamiento de las empresas e industrias de forma individual, así como el comportamiento de las unidades de toma de decisiones en el sistema económico (las empresas y los hogares) con algunas premisas clave, ya que a partir del comportamiento racional de las unidades [1] (las empresas maximizan las utilidades y los hogares buscan aumentar el excedente del consumidor), se puede llegar a conclusiones útiles sobre la operación de los mercados, cómo se asignan los recursos, entre otros.

La macroeconomía, en lugar de centrarse en los factores que influyen en la producción de bienes
particulares y el comportamiento de industrias individuales, se enfoca en los determinantes del
producto nacional en su conjunto
(. La macroeconomía estudia no el ingreso de los hogares, sino el
ingreso nacional; no los precios por separado, sino el nivel general que éstos guardan. No analiza la
demanda de trabajadores en la industria automovilística, sino el empleo total en la economía.
La microeconomía y la macroeconomía se interesan en las decisiones de los hogares y las empresas. La microeconomía se ocupa de las decisiones individuales; la macroeconomía, de la suma de
éstas. Acumulado, total o conjunto son los términos que se usan en macroeconomía para referirse a
las sumas. Cuando hablamos de comportamiento conjunto, queremos decir el proceder de todos los
hogares y las empresas tomadas juntas. También hablamos de consumo acumulado e inversión total,
que se refieren a la totalidad del consumo y la inversión en la economía.
Como los expertos en microeconomía y macroeconomía estudian la economía desde ángulos
distintos, uno esperaría que lleguen a conclusiones un tanto diversas sobre el comportamiento de
ésta, lo cual es verdad hasta cierto punto. En general, los expertos en microeconomía concluyen
que el mercado funciona bien. Consideran que los precios son flexibles, que se ajustan para mantener la correspondencia entre las cantidades ofrecidas y las demandadas. En cambio, los expertos
