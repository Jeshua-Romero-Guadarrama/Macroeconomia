## 1. Introducción a la macroeconomía

### 1.1 Contenido

- Las raíces de la macroeconomía se ecneuntran en la Gran Depresión, que forma parte de la historia reciente de la macroeconomía.
- Los principales intereses de la macroeconomía son la inflación y deflación, el aumento de la producción de corto y largo plazos y el desempleo.
- El gobierno formula la política macroeconomía de una nación; esto es, crea la política fiscal, política monetaria y políticas de crecimiento que deben seguir la sociedad.
- Uno de los componentes escenciales de la macroeconomía es el diagrama del flujo circular, que refleja los tres ámbitos del mercado.
- La metodología de la macroeconomía se encuentra microfundamentada, llegando hasta la demanda y oferta agregadas.
- La economía mexicana desde 1900: ciclos y tendencias; esto es, la expansión y contracción de los ciclos comerciales y económicos.
- La macroeconomía mexicana desde 1970.

### 1.2 Introducción

La macroeconomía es parte de la vida diaria de cada país; cuando marcha bien, son pocos quienes no tienen trabajo, los ingresos de las personas aumentan y las utilidades de las empresas son altas. En dicho contexto, dominado por una expansión del sistema económico en su conjunto, a los recién llegados al mercado laboral (como los estudiantes que acaban de titularse) les resulta fácil encontrar un trabajo bien pagado. 

Por el contrario, si el sistema macroeconómico se ecuentra en declive, es difícil encontrar un empleo nuevo, los ingresos no
aumentan lo suficiente y las utilidades son marginales. Dados los grandes efectos que llega a tener la macroeconomía en la vida de cientos de miles de personas, es importante que la sociedad, en general, y los economistas, en particular, entiendan el fucnionamiento del sistema económico a gran escala.

LAS RAÍCES DE LA MACROECONOMÍA
LA GRAN DEPRESIÓN
Los sucesos económicos de los años treinta, el decenio de la Gran Depresión, acicatearon las reflexiones sobre temas de macroeconomía, sobre todo el desempleo. La década de los veinte fue una
etapa próspera para la economía estadounidense. De hecho, todos los que querían trabajo encontraban uno, el ingreso aumentaba de manera sustancial y los precios eran estables. Pero a finales de
1929, de repente, las cosas empeoraron. Ese año hubo 1.5 millones de desempleados. En 1933 habían aumentado a 13 millones, en una fuerza laboral de 51 millones. En 1933, Estados Unidos produjo casi 27% menos bienes y servicios que en 1929. En octubre de 1929, cuando los precios de las
acciones se desplomaron en Wall Street, se perdieron miles de millones de dólares de riquezas personales. El desempleo se mantuvo en más de 14% de la fuerza de trabajo hasta 1940.
Modelos clásicos Antes de la Gran Depresión, los economistas aplicaban a los problemas
generales de la economía modelos microeconómicos, llamados “clásicos” o de “limpieza de mercado”.
Por ejemplo, en el análisis clásico de la oferta y la demanda se supone que un exceso de mano de obra
hace bajar los salarios hasta un nuevo nivel de equilibrio; por lo tanto, el desempleo no persistiría.
En otras palabras, los economistas de cuño clásico pensaban que las recesiones (bajas de la economía) se corregían solas. Cuando la producción cae y la demanda de mano de obra se desplaza a la
izquierda —continuaban—, la tasa salarial se reduce, lo que aumenta la cantidad de mano de obra
que demandan las empresas, las cuales quieren contratar más trabajadores con la nueva y más baja
tasa salarial. Sin embargo, durante la Gran Depresión los niveles de desempleo se mantuvieron muy
altos durante casi 10 años. En gran medida, la incapacidad de los modelos clásicos simples de explicar la prolongación del elevado desempleo dio el impulso para el desarrollo de la macroeconomía.
No es de sorprender que lo que hoy llamamos macroeconomía haya nacido en los años treinta.
La revolución keynesiana Una de las obras más importantes de la historia de la economía es la Teoría general de la ocupación, el interés y el dinero, de John Maynard Keynes, publicada en
1936. Keynes se basó en lo que ya se sabía sobre los mercados y su comportamiento y, de esta manera, formuló una teoría para explicar los confusos acontecimientos económicos de su tiempo.
Buena parte de la macroeconomía tiene sus raíces en la obra de Keynes. Según él, no son los
precios y salarios los que determinan el nivel del empleo, como indicaban los modelos clásicos,
sino el nivel de la demanda agregada de bienes y servicios. Keynes creía que los gobiernos debían
intervenir en la economía e influir en el nivel de la producción y el empleo. La función del gobierno en las épocas de poca demanda de los particulares —afirmaba Keynes— es estimular la
demanda agregada y, con ello, sacar la economía de la recesión (véase el apartado Para profundizar, “La Gran Depresión y John Maynard Keynes”).
HISTORIA RECIENTE DE LA MACROECONOMÍA
Después de la Segunda Guerra Mundial, más que nada en los años cincuenta, las ideas de Keynes
comenzaron a ganar influencia creciente entre los economistas profesionales y las autoridades.
Los gobiernos pensaron que podían intervenir en la economía para alcanzar determinadas metas
de empleo y producción. Comenzaron a aprovechar sus poderes de gravar y gastar, así como su ca

pacidad de incidir en las tasas de interés y el circulante, con la finalidad explícita de controlar los altibajos de la economía. Esta noción de las políticas gubernamentales se estableció con firmeza en Estados Unidos con la aprobación de la Ley del Empleo de 1946, que instituyó el Consejo de Asesores Económicos del Presidente, un grupo de economistas que aconsejaban al jefe del ejecutivo en materia de
economía. También comprometió la participación del gobierno federal en la economía para evitar caídas grandes de la producción y el empleo.

Ajuste fino (fine tuning) en la década de 1960 La noción de que el gobierno puede y debe actuar para estabilizar la macroeconomía llegó a la cúspide de su popularidad en la década de los sesenta. En esos años, Walter Heller, presidente del Consejo de Asesores Económicos de los
presidentes Kennedy y Johnson, se refirió al ajuste fino como el cometido del gobierno para regular
la inflación y el desempleo. En los años sesenta, muchos economistas creían que el gobierno podía
usar sus herramientas para manipular con bastante precisión los niveles de inflación y desempleo.
Desilusión en los años setenta y comienzos de los ochenta En la década de
los setenta y comienzos de los ochenta, la economía estadounidense tuvo amplias fluctuaciones
de empleo, producción e inflación. De 1974 a 1975 y de nuevo entre 1980 y 1982, Estados Unidos
sufrió dos graves recesiones. No fueron tan catastróficas como la Gran Depresión, pero estas dos
recesiones dejaron sin trabajo a millones de personas y se perdieron miles de millones de dólares
de producción e ingresos. De 1974 a 1975 y otra vez entre 1979 y 1981, Estados Unidos sufrió tasas de inflación muy elevadas.
Además, en los setenta apareció la estanflación (estancamiento más inflación). La estanflación
se presenta cuando el nivel general de precios aumenta a gran velocidad (inflación) en periodos de
recesión o de desempleo elevado y persistente (estancamiento). Hasta 1970, sólo se había visto
un incremento acelerado de los precios en periodos en que la economía prosperaba y el desempleo
era bajo (o por lo menos menguante). El problema de la estanflación preocupaba a los teóricos de
la macroeconomía y a los políticos interesados en la salud de la economía.
Para 1975 había quedado claro que la macroeconomía era más difícil de controlar, no obstante
la confianza que los economistas tenían en la declaración de Heller y la teoría académica. Los acontecimientos de la década de los setenta y comienzos de los años ochenta ejercieron una influencia
importante en la teoría macroeconómica. Se debilitó la fe en el modelo keynesiano simple y en el
“sentido común” de los sesenta.
Buen tiempo en los años noventa, pausa en 2000-2001 y recuperación en
2002-2005 La economía creció bien en los años ochenta después de la recesión de 1980 a
1982. Hubo una recesión menor entre 1990 y 1991 y luego la economía creció el resto de la década de los noventa. El crecimiento en el lapso 1996-1999 fue particularmente sólido, impulsado en
parte por el auge de las bolsas de valores que comenzó en 1995. Notablemente, la inflación no fue
un problema en toda la década de los noventa. Luego la economía se desaceleró a mediados del
2000, antes de los atentados terroristas del 11 de septiembre de 2001. A pesar de los ataques, la
economía remontó a finales de 2001. El ritmo del crecimiento fue moderado en 2002 y 2003 y no
bastó para evitar que aumentara la tasa del desempleo. Entre 2000 y 2001 la tasa de desempleo
aumentó de casi 4% a 5.5%; después, en el tercer trimestre de 2003, volvió a subir a 6.1%. La economía creció bastante bien en 2004 y 2005 y a mediados de 2005 la tasa de desempleo retrocedió
a 5%. La inflación no presentó ningún problema hasta el 2005.
La fuerte economía de los noventa y la recuperación de 2002 a 2005 no hicieron converger las
opiniones de los expertos sobre el funcionamiento de la macroeconomía. La disciplina de la macroeconomía se encuentra todavía en estado de cambio y quedan por resolver muchos temas importantes. Por eso es difícil enseñarla, aunque es emocionante estudiarla.
INTERESES DE LA MACROECONOMÍA
Tres intereses fundamentales de la macroeconomía son:
■ Inflación
■ Aumento de la producción
■ Desempleo
Las autoridades gubernamentales quisieran que la inflación fuera mínima, grande el aumento de
la producción y bajo el desempleo. Quizá no tengan éxito, pero como metas son claras.
INFLACIÓN Y DEFLACIÓN
La inflación es un aumento en el nivel general de precios. Desde hace mucho, la meta de las políticas gubernamentales es mantener baja la inflación. Más problemáticas son las hiperinflaciones,
periodos de incrementos muy rápidos del nivel general de precios.

Casi ningún estadounidense está consciente de cómo es la vida con una inflación muy alta.
En algunos países y en ciertas épocas, la gente se ha acostumbrado a que los precios aumenten
cada día, cada hora e incluso cada minuto. En la hiperinflación de Bolivia (en 1984 y 1985) el
precio de un huevo pasó de 3,000 a 10,000 pesos en una semana. En 1985, tres frascos de aspirina
costaban lo mismo que un auto de lujo en 1982. Asimismo, el problema de manejar el dinero se
volvió una carga. Los bancos dejaron de contabilizar los depósitos. Un depósito de 500 dólares
equivalía a 32 millones de pesos y no tenía sentido ponerse a contar un costal lleno de billetes. La
moneda de Bolivia, que se imprimía en Alemania Occidental e Inglaterra, fue la tercera importación
más grande del país en 1984, superada únicamente por el trigo y el equipo de extracción minera.
El aumento estratosférico de los precios en Bolivia es una parte pequeña de la historia.
Cuando la inflación se acerca a tasas anuales de 2,000%, la economía y toda la organización de
un país comienzan a desgajarse. Los trabajadores recurren a la huelga para exigir aumentos salariales que correspondan a la tasa alta de inflación y las empresas tienen dificultades para conseguir
créditos.
Las hiperinflaciones son raras. No obstante, los economistas han dedicado muchos esfuerzos
a detectar los costos y las consecuencias incluso de una inflación moderada. ¿Alguien gana con la
inflación? ¿Quién pierde? ¿Qué costos impone la inflación a la sociedad? ¿Qué tan onerosos son?
¿Cuáles son las causas de la inflación? ¿Cuál es la mejor manera de detenerla? Tales son algunas
preocupaciones importantes de los especialistas en macroeconomía.
Una reducción en el nivel general de precios se llama deflación. En algunas épocas de la historia estadounidense y hace poco en Japón ha ocurrido una prolongada deflación. El objetivo de
las autoridades es evitar periodos largos de deflación, lo mismo que de inflación, para alcanzar la
meta macroeconómica de estabilidad.
AUMENTO DE LA PRODUCCIÓN DE CORTO Y LARGO PLAZOS
En lugar de crecer a un ritmo uniforme todo el tiempo, las economías tienen un rendimiento con
altas y bajas de corto plazo. El nombre técnico de estas altas y bajas es ciclo comercial. La principal medida del desempeño de una economía es la producción total (agregada), que es el volumen
total de bienes y servicios producidos por una economía en determinado periodo. Cuando se
produce menos (en otras palabras, cuando la producción total disminuye) hay menos bienes y
servicios para todos y el promedio de la calidad de vida baja. Cuando las empresas reducen la
producción, también despiden trabajadores, lo cual aumenta la tasa de desempleo.
Las recesiones son periodos en los que mengua la producción total. Se ha vuelto convencional clasificar una baja económica como “recesión” cuando la producción total disminuye en
dos trimestres consecutivos. Una recesión prolongada y honda se llama depresión, aunque los
economistas no se ponen de acuerdo sobre cuándo una recesión se convierte en depresión. Desde
comienzos del siglo XX, Estados Unidos ha sufrido una depresión (en los años treinta), tres recesiones graves (1946, 1974-1975 y 1980-1982) y varias recesiones menos graves y más breves
(1954, 1958, 1990-1991 y 2001). Otros países también pasaron por recesiones en el siglo pasado,
algunas que coincidieron aproximadamente con las de Estados Unidos y otras que no lo hicieron.
En la producción incide algo más que los movimientos de alza y baja de los ciclos comerciales.
La magnitud del ritmo de crecimiento de la producción en un periodo prolongado (mayor, digamos, que la duración habitual de un ciclo comercial) también preocupa a los especialistas en
macroeconomía y a los políticos. Si el ritmo de crecimiento de la producción es mayor que el incremento demográfico, hay más bienes y servicios por persona. Así, en promedio, cada quien está
en mejor situación. Por consiguiente, los políticos se interesan no sólo en allanar las fluctuaciones
de la producción durante un ciclo comercial, sino también en las medidas que a la larga incrementen el ritmo del crecimiento.
DESEMPLEO
Uno no puede escuchar los noticiarios ni leer los periódicos sin advertir que cada mes se publican
los datos sobre la tasa de desempleo. La tasa de desempleo (el porcentaje de la fuerza laboral que
no tiene trabajo) es un indicador clave de la salud de la economía. Como la tasa de desempleo
guarda una relación estrecha con la producción total (agregada) de la economía, los anuncios de
la cifra del mes son seguidos con gran interés por economistas, políticos y autoridades.

Aunque los expertos en macroeconomía están interesados en saber por qué la tasa de desempleo aumenta o baja en un periodo, también tratan de responder una pregunta básica: ¿por qué
hay desempleo? No esperamos que no haya desempleo. En todo momento, algunas empresas se
van a la quiebra por la competencia de los rivales, mala administración o mala suerte. Los antiguos empleados de esas compañías no encuentran un trabajo nuevo de inmediato y, mientras lo
consiguen, están desempleados. De la misma manera, todos los trabajadores que entran en el
mercado laboral por primera vez pasan semanas y hasta meses antes de colocarse.
Si basamos nuestro análisis en la oferta y la demanda, esperaríamos que las condiciones
cambiaran en respuesta a la existencia de trabajadores desempleados. En particular, cuando el
desempleo rebasa cierto límite, hay una oferta excesiva de trabajadores; con las tasas salariales
actuales, hay personas que quieren trabajar pero no encuentran colocación. En la teoría microeconómica, la respuesta a la oferta excesiva es una disminución en el precio de la mercancía de
que se trate y, por lo tanto, un aumento en la cantidad demandada, una reducción de la cantidad
ofrecida y un restablecimiento del equilibrio. Cuando la cantidad ofrecida es igual a la demandada, el mercado queda compensado.
La existencia del desempleo parecería implicar que el mercado total de mano de obra no está en equilibrio, que algo impide que la cantidad ofrecida y la cantidad demandada sean iguales.
¿Por qué los mercados laborales no se equilibran cuando otros mercados lo hacen o es que los
primeros están equilibrados y el desempleo significa otra cosa? Es otra de las principales inquietudes de los expertos en macroeconomía.
EL GOBIERNO EN LA MACROECONOMÍA
Buena parte de nuestra exposición de la macroeconomía se refiere a la función posible del gobierno como influencia en la economía. Hay tres clases de políticas que los gobiernos aplican para influir en la economía.
1. Política fiscal
2. Política monetaria
3. Políticas de crecimiento o de fomento de la oferta
POLÍTICA FISCAL
Una manera en que el gobierno federal afecta la economía es con decisiones fiscales y de gasto, es
decir, con políticas fiscales. El gobierno federal recauda impuestos de hogares y empresas para
luego gastarlos en rubros que van de misiles a parques, pagos del seguro social o carreteras interestatales. Tanto la magnitud como la composición de estos impuestos y gastos tienen un efecto
importante en la economía.
Una de las principales ideas de Keynes en los años treinta era que la política fiscal podía y debía usarse para estabilizar el nivel de producción y empleo. En concreto, Keynes pensaba que el
gobierno debería reducir los impuestos y/o aumentar el gasto (las llamadas políticas fiscales de expansión) para rescatar a la economía de una crisis. Por el contrario, sostenía que el gobierno debería elevar los precios y/o reducir el gasto (las llamadas políticas fiscales de contracción) para sacar a la economía de una inflación.
POLÍTICA MONETARIA
Impuestos y gastos no son las únicas variables que controla el gobierno. La Reserva Federal, el
banco central de Estados Unidos, determina el monto del circulante en la economía de ese país.
Los efectos y la función conveniente de la política monetaria se cuentan entre los temas más debatidos de la macroeconomía. Casi todos los economistas están de acuerdo en que la oferta de
circulante repercute en el nivel general de precios, las tasas de interés, el tipo de cambio, la tasa
de desempleo y el nivel de la producción. Las principales controversias se refieren a cómo se manifiesta la política económica y cuál es la magnitud de sus efectos.
POLÍTICAS DE CRECIMIENTO
Muchos economistas se sienten escépticos sobre la capacidad del gobierno para regular, con cierta
precisión, el ciclo comercial mediante políticas monetarias y fiscales. Opinan que el objetivo de
las políticas gubernamentales debe ser estimular la oferta agregada, es decir, alentar el crecimiento
potencial de la producción y el ingreso totales. Se ha dirigido un cúmulo de medidas a incrementar
el ritmo del crecimiento. Muchas de estas acciones están pensadas para mercados particulares y en
macroeconomía se discuten mucho. Una preocupación importante de los expertos es que el gobierno pida prestado para financiar un gasto excesivo con respecto a la recaudación de impuestos (el
“déficit”) y absorba ahorros que de manera corriente se canalizarían a las empresas para que invirtieran en capital. Otra finalidad de las medidas gubernamentales de fomento del crecimiento ha
sido el sistema impositivo. Una meta importante de las reformas fiscales de 1981 y 1986 fue reducir
las tasas de impuestos para aumentar los incentivos para trabajar, ahorrar e invertir. Además, la Ley
de Ayuda a los Contribuyentes de 1997 incluyó varias medidas que favorecen el crecimiento. Todas
estas acciones se conocen como políticas de la oferta.

COMPONENTES DE LA MACROECONOMÍA
La macroeconomía se centra en cuatro grupos: 1) hogares y 2) empresas, que en conjunto forman el
sector privado; 3) el gobierno (sector público) y 4) el resto del mundo (sector internacional). Estos
cuatro grupos interactúan de varias maneras y muchas incluyen la recepción o el pago de ingresos.
DIAGRAMA DEL FLUJO CIRCULAR
Una manera útil de ver las interacciones económicas de los cuatro sectores de la economía es un
diagrama de flujo circular, que muestra los ingresos recibidos y los pagos hechos por cada uno. En
la figura 5.1 se ilustra un diagrama de flujo circular simple.
Recorramos poco a poco el flujo circular. Los hogares trabajan para empresas y el gobierno,
y a cambio de su trabajo reciben un salario. En nuestro diagrama se muestra un flujo de salarios
al sector de los hogares como pago por esos servicios. Los hogares también reciben intereses por
bonos de corporaciones y el gobierno más dividendos de las empresas. Muchos hogares reciben
otros pagos del gobierno, como prestaciones del seguro social, beneficios de veteranos de guerra y pagos de asistencia. Los economistas llaman pagos de transferencia a estas retribuciones del
gobierno (por los que los receptores no proporcionan bienes, servicios ni trabajo). En conjunto,
todas estas entradas comprenden el ingreso total recibido por los hogares.
Los hogares gastan cuando compran bienes y servicios a las empresas y cuando pagan impuestos al gobierno. Estos rubros comprenden la suma total pagada por los hogares. La diferencia
entre las entradas absolutas y los pagos totales de las familias es el monto que éstas ahorran o lo
que pierden de su riqueza. Si los hogares reciben más de lo que gastan, ahorran en el periodo. Si
reciben menos de lo que gastan, pierden riqueza. Una familia pierde riqueza si toma parte de sus
ahorros anteriores o si pide un préstamo. En el diagrama de flujo circular, el gasto de los hogares se
representa como una salida de ese sector. También se dice que el ahorro de los hogares es una
“fuga” del flujo circular porque se retiran ingresos del sistema (o sea, poder de compra actual).
Las empresas venden bienes y servicios a hogares y el gobierno. Estas ventas representan ingresos, que se muestran en el diagrama como un flujo de entrada al sector de las empresas. Éstas
pagan salarios, intereses y dividendos a los hogares a impuestos al gobierno. Dichos pagos se muestran como un flujo de salida del sector de las empresas.
El gobierno recauda impuestos de los hogares y las empresas. El gobierno también realiza
pagos. Compra bienes y servicios a las empresas, paga salarios e intereses a los hogares y les hace pagos de transferencia. Si el ingreso del gobierno es menor que sus pagos, pierde riqueza.
Por último, los hogares gastan parte de sus ingresos en importaciones, que son bienes y servicios
producidos en el resto del mundo. Del mismo modo, gente de otros países compra exportaciones,
bienes y servicios generados por empresas nacionales y vendidos a otros países.
Una lección del diagrama de flujo circular es que el gasto de uno es la entrada de otro. Si
uno compra una computadora personal Dell, se hace un pago y la compañía recibe un ingreso.
Si Dell entera impuestos al gobierno, hace un pago y éste recibe un ingreso.
Los gastos de todos van a alguna parte. Es imposible vender algo sin que haya un comprador, así como hacer un pago sin un destinatario. Toda transacción ha de tener dos lados.
LOS TRES ÁMBITOS DEL MERCADO
Otra manera de considerar las relaciones recíprocas de hogares, empresas, gobierno y el resto del
mundo es examinar los mercados en que interactúan.
Los tres ámbitos del mercado son:
1. El mercado de bienes y servicios
2. El mercado de mano de obra
3. El mercado de dinero (mercado financiero)
Mercado de bienes y servicios Hogares y gobierno compran bienes y servicios a las
empresas en el mercado de bienes y servicios. En éste, las firmas intercambian bienes y servicios.
Por ejemplo, Levi Strauss adquiere de otras compañías la mezclilla para hacer sus pantalones.
Además, algunas empresas compran bienes de capital a otras. Si General Motors necesita nuevos
robots para sus líneas de montaje, en lugar de fabricarlos, los consigue de otra empresa.
Las empresas ofrecen en el mercado de bienes y servicios. Los hogares, el gobierno y las firmas demandan en este mercado. Por último, el resto del mundo compra y vende en el mercado
de bienes y servicios. Estados Unidos importa cientos de miles de millones de dólares en automóviles, DVD, petróleo y otros bienes. Asimismo, ese país exporta cientos de miles de millones
de dólares en computadoras, aviones y productos agrícolas.
Mercado de mano de obra Las relaciones en el mercado de mano de obra tienen lugar cuando las empresas y el gobierno compran trabajo a los hogares. En este mercado, los hogares
ofrecen mano de obra y las empresas y el gobierno la demandan. En la economía estadounidense,
las empresas son las principales solicitantes de mano de obra, aunque el gobierno también es un
patrón importante. La oferta total de mano de obra en la economía depende de la suma de decisiones que toman los hogares. Los individuos tienen que decidir si entran en la fuerza laboral (si
buscan trabajo o no) y cuántas horas van a trabajar.

La mano de obra también se ofrece y se demanda en el resto del mundo. En los últimos años,
el mercado de mano de obra se ha vuelto internacional. Por ejemplo, los productores de frutas y
verduras de California tendrían grandes dificultades para llevar sus productos al mercado si no
fuera por el trabajo de los jornaleros mexicanos. Durante años, Turquía ha proporcionado a Alemania “trabajadores invitados” que están dispuestos a aceptar las ocupaciones peor pagadas de
las que se abstienen los prósperos trabajadores alemanes.
Mercado de dinero En el mercado de dinero (también llamado mercado financiero) los
hogares compran acciones y bonos de las empresas. Asimismo, ofrecen fondos con la esperanza
de ganar ingresos en la forma de dividendos sobre las acciones e intereses sobre los bonos. Los
hogares también demandan (piden prestado) fondos de este mercado para financiar diversas
compras. Las empresas toman préstamos para construir nuevas instalaciones con el fin de ganar
más en el futuro. El gobierno pide prestado expidiendo bonos. El resto del mundo pide y presta
en el mercado de dinero. Todas las mañanas se dan informes en radio y televisión sobre las bolsas
de valores de Japón e Inglaterra. Buena parte de los empréstitos que piden y otorgan los hogares,
las empresas, el gobierno y el sector internacional es coordinada por instituciones financieras:
bancos comerciales, asociaciones de ahorro y préstamo, aseguradoras, etc. Estas organizaciones
toman los depósitos de un grupo y los prestan a otros.
Cuando una empresa, un hogar o el gobierno toman un préstamo para financiar una compra, tienen la obligación de pagarlo, generalmente en una fecha futura especificada. La mayoría
de los empréstitos incluyen el pago de intereses como una tasa por el uso de los fondos prestados.
Cuando se concede un préstamo, el prestatario casi siempre firma un “compromiso de pago”, o
nota de promesa de pago, y la entrega al prestamista. Cuando el gobierno federal toma un préstamo,
emite “promesas” llamadas bonos, notas o certificados de la Tesorería a cambio del dinero. Las
corporaciones expiden bonos corporativos. Un bono corporativo diría, por ejemplo, “General
Electric Corporation accede a pagar 5,000 dólares al tenedor de este bono el 1 de enero de 2006 y
un interés sobre éste de 5.1% anual hasta esa fecha”.
En lugar de expedir bonos para reunir fondos, las empresas también pueden emitir títulos de
acciones. Un título de acciones es un instrumento financiero que da al tenedor una participación
en la propiedad de la empresa y, por consiguiente, el derecho a participar en las utilidades de ésta.
Si la compañía prospera, el valor de la acción aumenta y el accionista recibe una ganancia de capital1 sobre la compra inicial. Además, la acción puede pagar dividendos, es decir, la empresa entrega parte de sus utilidades directamente a sus accionistas, en lugar de retenerlas para comprar
capital. Si la empresa no tiene un buen desempeño, el accionista no tiene rendimientos. El valor
de capital de la acción puede bajar y no se pagan dividendos.
Acciones y bonos no son más que contratos o acuerdos entre las partes. Acepto prestarle a
usted cierta suma y usted acepta pagarme esa suma más algo adicional en una fecha determinada; o bien, accedo a comprar parte de su propiedad de la empresa y usted me cede una participación en las utilidades futuras de la empresa.
Una variable crucial del mercado de dinero es la tasa de interés. Aunque a veces hablamos como si sólo hubiera una tasa de interés, en ningún momento hay una sola. Por el contrario, la tasa
de interés sobre cierto préstamo refleja el plazo de éste y el riesgo que percibe el prestamista. Una
empresa que acaba de fundarse tendrá que pagar una tasa más alta que General Motors. Una hipoteca a 30 años tiene una tasa de interés distinta que un préstamo a 90 días. Sin embargo, las tasas de
interés suben y bajan juntas y su movimiento es una manifestación de las condiciones generales
del mercado financiero.
METODOLOGÍA DE LA MACROECONOMÍA
Los especialistas en macroeconomía elaboran modelos sobre la base de teorías y verifican los modelos con datos. En este sentido, la metodología de la macroeconomía es semejante a la que se
aplica en la microeconomía.

CONEXIONES CON LA MICROECONOMÍA
¿Cómo explican el comportamiento conjunto los especialistas en macroeconomía? En una versión se asume que los mismos factores que afectan la conducta de individuos influyen en el comportamiento colectivo. Por ejemplo, gracias a la microeconomía sabemos que la tasa salarial de
un individuo incide en sus hábitos de consumo y en la cantidad de trabajo que está dispuesto a
realizar. Si aplicamos esta hipótesis microeconómica a datos acumulados, diríamos que la tasa salarial promedio de la economía debe afectar al consumo total y a la oferta total de trabajo (lo que
parece ser cierto).
El motivo para recurrir a la microeconomía en busca de explicaciones para los acontecimientos macroeconómicos es simple:
El comportamiento macroeconómico es la suma de todas las decisiones macroeconómicas
hechas por los hogares y las empresas. Si los movimientos de los macroeconómicos agregados, como la producción o el empleo totales, reflejan decisiones tomadas en lo individual
por las empresas y los hogares, no se entienden aquéllos sin algún conocimiento de los factores que influyen en éstas.
Pensemos en el desempleo. La tasa de desempleo es el número de personas sin trabajo como
proporción de la fuerza laboral. Para ser considerada “parte de la fuerza laboral”, una persona debe tener trabajo o buscarlo activamente. Para comprender el desempleo total, necesitamos entender el comportamiento individual de los hogares en el mercado de trabajo. ¿Por qué las personas
deciden entrar en la fuerza laboral? ¿En qué circunstancias la abandonan? ¿Por qué hay desempleo aunque la economía se desenvuelva muy bien? Un conocimiento del comportamiento macroeconómico es un punto de partida lógico para el análisis macroeconómico.
DEMANDA Y OFERTA AGREGADAS
Un tema de gran interés en los capítulos siguientes es el comportamiento de la demanda y la oferta
agregadas. La demanda agregada es la demanda total de bienes y servicios. La oferta agregada es la
oferta total de bienes y servicios.
En la figura 5.2 se muestran las curvas de la demanda agregada y la oferta agregada. Sobre el
eje horizontal se mide la producción total; sobre el eje vertical, el índice general de precios, no el
precio de un bien o servicio en concreto. La economía está en equilibrio en el punto donde las
curvas se cruzan.

Veremos que las curvas de la oferta y la demanda agregadas son mucho más complicadas que
las curvas simples que describimos en los capítulos 3 y 4. La lógica elemental de la oferta, demanda y
equilibrio en mercados individuales no explica lo que se retrata en la figura 5.2. Dedicaremos la
totalidad del capítulo siguiente a explicar el significado de “producción total” y “nivel general de
precios”. Además, aunque estudiaremos el comportamiento de hogares y empresas en mercados
individuales en busca de claves sobre cómo analizar el comportamiento conjunto, hay diferencias
importantes cuando pasamos del nivel individual al acumulado.
Tomemos por ejemplo la demanda, uno de los conceptos más importantes de la economía.
Cuando el precio de un bien aumenta, quizá el determinante más significativo de la respuesta del
consumidor sea la disponibilidad de otros bienes que sustituyan al bien que aumentó de precio.
Parte de la razón de que un aumento en el precio de los boletos de avión cause una disminución en
la cantidad de boletos demandados es que un precio más alto en relación con otros bienes significa
que el costo de oportunidad de comprar un boleto es mayor: el sacrificio requerido en términos de
otros bienes y servicios aumentó. Sin embargo, cuando el nivel general de precios cambia, no siempre hay variaciones en el costo de oportunidad de los bienes. Por ejemplo, un boleto de avión más
caro, comparado con otros bienes de precio también mayor, podría requerir el mismo balance que
antes del incremento de precios.
La microeconomía nos enseña que,ceteris paribus, la cantidad demandada de un bien cae cuando el precio se incrementa y aumenta cuando su precio se abate (es la ley microeconómica de la
demanda). En otras palabras, las curvas de la demanda individual y la demanda del mercado se inclinan abajo a la derecha. La razón de que la curva de la demanda agregada de la figura 5.2 se incline
abajo a la derecha es compleja. Como veremos luego, la pendiente debajo de la curva de la demanda
agregada se relaciona con lo que sucede en el mercado de dinero (financiero).
La curva de la oferta agregada es muy distinta que la curva de la oferta de una empresa o mercado individual. La curva de la oferta de una empresa se deriva de la premisa de que todos los precios
de sus insumos son fijos. En otras palabras, se supone que los precios de los insumos de la empresa no
cambian cuando cambia el precio del producto de esa empresa. Cuando en el capítulo 3 derivamos el
esquema de la oferta de la soya de Clarence Brown, consideramos fijos los precios de los insumos. Un
cambio en el precio de un insumo produce un desplazamiento en la curva de la oferta de Brown, no
un movimiento sobre la curva. Ahora bien, si examinamos los cambios en el nivel general de precios,
todos los precios cambian (incluyendo los correspondientes a los insumos), así que la curva de la
oferta agregada no puede basarse en la suposición de que los precios de insumos son fijos. Veremos
que la curva de la oferta agregada es otra fuente de controversia en la macroeconomía.
LA ECONOMÍA ESTADOUNIDENSE DESDE 1900:
CICLOS Y TENDENCIAS
Como dijimos, casi todas las variables macroeconómicas pasan por altas y bajas con el tiempo y la
economía en su conjunto experimenta periodos de prosperidad y recesión. No obstante, la tendencia
de la economía estadounidense ha sido hacia la prosperidad. Una medida de la prosperidad de una
economía es la cantidad de bienes y servicios que produce en el año; es decir, su producto interno bruto (PIB), que es el tema del capítulo siguiente. Se dice que una economía crece de un año al siguiente
si el PIB es mayor en el segundo año. Entre 1900 y 2004 la economía estadounidense creció a un ritmo
promedio de 3.4% anual. Esto significa que durante ese periodo la economía fue, en promedio, 3.4%
más rica cada año que el anterior. Es la tasa de crecimiento de largo plazo de la economía.
Recuerde que lo que estudiamos aquí es el ritmo promedio de crecimiento. La economía no
creció realmente 3.4% cada año. En algunos años el crecimiento fue inferior a 3% y en otros fue negativo (el PIB bajó). En otros años el crecimiento fue mayor a 3.4%. Entonces, necesitamos distinguir
entre las tendencias duraderas o seculares del desempeño económico y las variaciones de corto plazo
o cíclicas.
EXPANSIÓN Y CONTRACCIÓN: EL CICLO COMERCIAL
La macroeconomía se ocupa de las tendencias prolongadas (¿qué ha hecho la economía estadounidense en el último siglo?) y de las fluctuaciones de corto plazo del rendimiento económico
(¿por qué el mundo sufrió una recesión grave a comienzos de los años ochenta y una recesión ligera en 2001?). El término ciclo comercial se refiere a las fluctuaciones efímeras de una economía.
En la figura 5.3 se ilustra un ciclo comercial característico.

Como la economía estadounidense crece, en promedio, al paso del tiempo, el ciclo comercial
de la figura 5.3 muestra una tendencia positiva: la cresta (el punto más alto) de un nuevo ciclo está
más arriba que la cresta del ciclo anterior. El periodo que va de un valle, el punto bajo de un ciclo,
a una cresta se llama expansión o auge. Durante una expansión, la producción y el empleo crecen.
El periodo que va de una cresta a un valle se llama contracción, recesión o crisis, cuando la producción y el empleo caen.
Al deliberar si una economía se expande o se contrae, observe la diferencia entre el nivel de
actividad económica y su tasa de cambio. Si la economía acaba de dejar un valle (el punto A de la
figura 5.3), crece (su tasa de crecimiento es positiva), pero su producción todavía es baja. Si la economía acaba de empezar a declinar después de una cresta (punto B), se contrae (la tasa de crecimiento es negativa), pero su producción todavía es alta.
El ciclo comercial de la figura 5.3 es simétrico, lo cual significa que la duración de una expansión es igual a la duración de una contracción. Sin embargo, la mayor parte de los ciclos comerciales
son asimétricos. Por ejemplo, es posible que la fase de expansión dure más que la de contracción.
Cuando viene una contracción, puede ser rápida y aguda, mientras que la expansión es lenta y
gradual. Además, la economía no es ni por mucho tan regular como indica el ciclo comercial de
la figura 5.3. Las altas y bajas de la economía son erráticas.
En la figura 5.4 se muestran ciclos comerciales reales de Estados Unidos entre 1900 y 2004.
En un siglo han ocurrido muchos ciclos, pero cada uno ha sido único. La economía no es tan simple como para tener ciclos regulares.
Las épocas de la Gran Depresión y las dos guerras mundiales muestran las fluctuaciones más
grandes de la figura 5.4, aunque se han verificado otras contracciones y expansiones grandes. Observe la expansión en la década de los sesenta y las recesiones de comienzos de los años ochenta y
noventa. Algunos ciclos han sido dilatados; otros, muy breves. Observe también que el PIB de hecho aumentó entre 1933 y 1937, aunque en 1937 todavía era muy bajo. La economía no salió de
la Gran Depresión sino hasta la acumulación de armamentos antes de la Segunda Guerra Mundial. Asimismo, tome nota de que los ciclos comerciales fueron más acusados antes de la Segunda
Guerra Mundial que después.

## Resumen

1. La microeconomía examina el funcionamiento de las industrias en
lo particular y el comportamiento de las unidades separadas de la
toma de decisiones. La macroeconomía se ocupa de la suma, acumulado o total de estas decisiones individuales: el consumo de todos los hogares de la economía, el monto del trabajo que ofrecen
y demandan todas las personas y empresas, y el total de todos los
bienes y servicios producidos.
LAS RAÍCES DE LA MACROECONOMÍA
2. La macroeconomía surgió del esfuerzo por explicar la Gran Depresión de los años treinta. Desde entonces, la disciplina ha evolucionado y se ha ocupado de nuevos temas, a medida que han
cambiado los problemas de la economía. Hasta finales de los sesenta se pensaba que el gobierno podía “pulir” la economía para
que marchara siempre en equilibrio. Sin embargo, el mal rendimiento de la economía en los setenta mostró que el pulimento
no siempre funciona.
INTERESES DE LA MACROECONOMÍA
3. Los tres temas de mayor interés para los estudiosos de la macroeconomía son los aumentos del nivel general de precios, o sea, la
inflación, el ritmo de crecimiento de la producción total y el nivel
de desempleo.
EL GOBIERNO EN LA MACROECONOMÍA
4. Entre las herramientas que tienen los gobiernos para influir en la
macroeconomía están la política fiscal (decisiones sobre impuestos y gasto gubernamental), política monetaria (control del circulante) y políticas de crecimiento o de la oferta (políticas que se centran en aumentar el ritmo de crecimiento de largo plazo).
COMPONENTES DE LA MACROECONOMÍA
5. El diagrama del flujo circular muestra el flujo del ingreso recibido
y los pagos hechos por los tres sectores de la economía: privado,
público e internacional. Los gastos de uno son los ingresos de
otro; toda transacción tiene dos lados.
6. Otra manera de estudiar las relaciones entre hogares, empresas,
gobierno y sector internacional es examinar los mercados donde
participan: el mercado de bienes y servicios, de mano de obra y
de dinero (mercado financiero).
METODOLOGÍA DE LA MACROECONOMÍA
7. Como el comportamiento macroeconómico es la suma de todas
las decisiones macroeconómicas tomadas por hogares y empresas, no es posible entender el primero sin algún conocimiento de
los factores que influyen en las segundas. Los movimientos de los
acumulados macroeconómicos reflejan las decisiones tomadas
por las empresas y hogares en lo individual.
8. Un tema importante de la macroeconomía es la comprensión del

Las cuentas del ingreso nacional
y del producto son más que un comunicado de los datos sobre el rendimiento de la economía. También proporcionan un marco conceptual que

PRODUCTO INTERNO BRUTO
El concepto clave de las cuentas del ingreso nacional y del producto es el producto interno bruto (PIB).
El PIB es el valor total en el mercado de la producción de un país. Es el valor de mercado de
todos los bienes y servicios finales generados en determinado periodo por factores de producción localizados en el país.
El PIB de Estados Unidos en 2004 (el valor de todo lo producido por los factores de producción de
ese país en 2004) fue de 11 billones 734,300 millones de dólares.
El carácter central del PIB como concepto de trabajo no puede ser exagerado. Así como una empresa necesita valorar el éxito o el fracaso de sus operaciones cada año, la economía tiene que evaluarse en su conjunto. El PIB, como medida de la producción total de una economía, nos proporciona un
boletín informativo sobre la economía nacional. Como el PIB es un concepto tan importante, tenemos que dedicar algún tiempo a explicar qué significa su definición.
BIENES Y SERVICIOS FINALES
En primer lugar, observe que la definición se refiere a bienes y servicios finales. Muchos bienes que
se producen en la economía no se clasifican como finales, sino como intermedios. Los bienes intermedios son los que produce una empresa y que otra usa en un proceso posterior. Por ejemplo, las llantas que se venden a los fabricantes de automóviles son bienes intermedios. El valor de estos bienes no
se cuenta en el PIB.
¿Por qué los bienes intermedios no se cuentan en el PIB? Supongamos que para producir un
automóvil, General Motors (GM) paga a Goodyear 200 dólares por las llantas. GM usa esas ruedas
(entre otros componentes) para armar un vehículo que vende en 24,000 dólares. El valor de éste
(incluyendo los neumáticos) es de 24,000 dólares, no de 24,000 + 200. El precio final ya refleja el
valor de todos sus componentes. Si se cuenta en el PIB tanto el valor de las llantas vendidas a los
fabricantes de automotores como el valor de los automóviles vendidos a los consumidores, se contabilizaría por duplicado.
La doble contabilidad también se puede evitar si se considera únicamente el valor agregado por
cada empresa a un producto en el proceso de su elaboración. El valor agregado en una etapa de producción es la diferencia entre el valor de los bienes cuando salen de esa fase y el costo de los bienes
cuando entraron. En la tabla 6.1 se ejemplifica el valor agregado. Las cuatro etapas de la producción de
un litro de gasolina son 1) extracción del petróleo, 2) refinación, 3) embarque y 4) venta al público. En
la primera etapa, el valor agregado es el valor de venta. En la segunda, la refinería compra el petróleo
a la compañía extractora, lo convierte en gasolina y lo vende al distribuidor. La refinería paga a la extractora un dólar por galón y cobra al distribuidor 1.30. Así, el valor agregado por la refinería es de
0.30 por galón. El distribuidor vende la gasolina a las gasolineras en 1.60. El valor agregado en esta etapa de la producción es de 0.30. Por último, las gasolineras venden la gasolina a los consumidores en
dos dólares. El valor agregado en la cuarta etapa es de 0.40 y el valor total agregado en el proceso de
producción es de 2.00, el mismo que el valor de venta al púbico. Si se sumaran las cifras totales de ventas de cada etapa (1.00 + 1.30 + 1.60 + 2.00 = 5.90), el valor del galón de gasolina sería exagerado.
Al calcular el PIB, se suma el valor agregado en cada etapa de producción o se toma el valor
de las ventas finales. No se toma el valor de todas las ventas de la economía para medir
cuánto se ha producido.
EXCLUSIÓN DE BIENES USADOS Y TRANSACCIONES EN PAPEL
El PIB abarca sólo la producción nueva o actual. La producción vieja no se cuenta en el PIB actual
porque ya se contó en su momento. Sería una duplicación contar en el PIB las ventas de bienes
usados. Si alguien le vende un automóvil usado, la transacción ya no cuenta en el PIB porque no
ha tenido lugar una producción nueva. Del mismo modo, una casa se cuenta en el PIB sólo en el
momento de construirla, no cada vez que se vende. En resumen:
En el PIB se ignoran todas las transacciones en las que cambian de manos dinero o bienes,
pero que no producen nuevos bienes ni servicios.
Las ventas de acciones y bonos no se cuentan en el PIB. Estos intercambios son transferencias
de posesión de activos, ya sea por medios electrónicos o en documentos, y no corresponden a la
producción real. ¿Pero qué sucede si la acción o el bono se venden en más de lo que se pagó por ellos? Las utilidades del mercado de acciones o bonos no tienen nada que ver con la producción
actual, así que no se cuentan en el PIB. Sin embargo, si pago a un corredor para que le venda a alguien mi acción, sus comisiones cuentan en el PIB porque el corredor me presta un servicio. Este
servicio es parte de la producción actual. Tenga cuidado de distinguir entre el cambio de acciones
y bonos por dinero (o por otras acciones y bonos) que no comprenden producción actual, y los
pagos por realizar tales cambios, que sí son parte de la producción.
EXCLUSIÓN DE LA PRODUCCIÓN EN EL EXTRANJERO DE
FACTORES DE PRODUCCIÓN DE PROPIEDAD NACIONAL
El PIB es el valor de lo generado por los factores de producción situados en un país.
Los tres factores básicos de la producción son tierra, trabajo y capital. El trabajo de los ciudadanos estadounidenses cuenta como factor de producción interno de Estados Unidos. La producción
de los estadounidenses en el extranjero (por ejemplo, los que trabajan en una compañía foránea) no
se cuenta en el PIB de la Unión Americana porque la producción no tuvo lugar en ese país. Del
mismo modo, las utilidades ganadas en el extranjero por compañías estadounidenses no se cuentan
en el PIB de Estados Unidos. Sin embargo, la producción de los extranjeros que laboran en Estados
Unidos cuenta en el PIB porque se da dentro del país. Asimismo, las utilidades obtenidas en ese país
por compañías extranjeras se consideran en su PIB.
A veces es útil tener una medida de lo producido por los factores de producción que sean propiedad de los ciudadanos de un país, dondequiera que haya tenido lugar. Esta medida se llama producto
nacional bruto (PNB). En la mayoría de los países, incluyendo Estados Unidos, la diferencia entre el
PIB y el PNB es pequeña. En 2004, el PNB de Estados Unidos fue de 11 billones 788,000 millones de
dólares, que se aproxima al valor de 11 billones 734,300 millones de dólares del PIB de ese país.
La diferencia entre PIB y PNB puede ser tramposa. Pensemos en la planta de Honda en
Marysville, Ohio, que es propiedad de la Corporación Honda, de Japón, aunque la mayor parte de
los trabajadores de la fábrica son estadounidenses. Toda la producción de la planta se incluye en el
PIB de Estados Unidos, pero sólo una parte corresponde al PNB del país. Los salarios pagados a los
operarios estadounidenses son parte del PNB de Estados Unidos, lo cual no sucede con las utilidades
de la planta. Éstas se cuentan en el PNB de Japón porque la manufactura se realizó con factores de
producción que pertenecen a los japoneses (en este caso, capital japonés). Ahora bien, estas utilidades no se cuentan en el PIB de Japón porque no se ganaron en aquel país.
CÁLCULO DEL PIB
El PIB se calcula de dos maneras. Una consiste en sumar el total gastado en todos los bienes finales durante cierto periodo. Éste es el método del gasto para calcular el PIB. La otra es sumar el ingreso (sa larios, rentas, intereses y utilidades) que recibieron todos los factores de producción al manufacturar
los bienes finales. Es el método del ingreso para calcular el PIB. Los dos métodos llevan a la misma cifra del PIB por la razón que explicamos en el capítulo anterior: todo pago (gasto) de un comprador es,
al mismo tiempo, una entrada (ingreso) del vendedor. Podemos medir los ingresos recibidos o los gastos incurridos, y terminaremos con la misma producción total.
Supongamos que la economía está compuesta por una empresa única y que vende su producción anual total en un millón de dólares. Como el total gastado en la producción de ese año es
de un millón de dólares, el PIB del año es por la misma cantidad (recuerde que con el método del
gasto se calcula el PIB basándose en el monto total gastado en bienes y servicios finales de la economía). Pero cada dólar del millón del PIB se paga a alguien o se queda como utilidad para los
dueños de la compañía. Con el método del ingreso sumamos los salarios pagados a los empleados
de la empresa, los intereses pagados a quienes le prestaron dinero y las rentas pagadas a los que le
arrendaron tierra, instalaciones o equipo. Lo que queda es la utilidad, que, desde luego, es el ingreso de los propietarios de la compañía. Si sumamos los ingresos de todos los factores de producción, incluyendo las utilidades de los propietarios, tenemos un PIB de un millón.
EL MÉTODO DEL GASTO
Recordemos que en el capítulo anterior señalamos los cuatro grupos principales de la economía:
hogares, empresas, el gobierno y el resto del mundo. También hay cuatro categorías de gasto:
Categorías de gasto:
■ Gastos de consumo (C). Gasto de los hogares en bienes de consumo
■ Inversión interna privada bruta (I). Gasto de empresas y hogares en nuevo capital, es decir, plantas, equipo, existencias y nuevas estructuras residenciales
■ Consumo e inversión bruta del gobierno (G)
■ Exportaciones netas (EX – IM). Gasto neto del resto del mundo o exportaciones
(EX) menos importaciones (IM)
El método del gasto calcula el PIB al sumar esos cuatro componentes. Su ecuación es:
PIB = C + I + G + (EX − IM)
El PIB de Estados Unidos fue de 11 billones 734,300 millones de dólares en 2004. Los cuatro componentes del método del gasto se muestran en la tabla 6.2, junto con sus categorías.

Gastos de consumo (C) La mayor parte del PIB consiste en gastos de consumo (C).
En la tabla 6.2 se muestra que en 2004 los gastos de consumo sumaron 70% del PIB. Son gastos
hechos por los consumidores en bienes y servicios.
Hay tres categorías principales de gastos de consumo: imperecederos, perecederos y servicios. Los
imperecederos son bienes como automóviles, muebles y aparatos electrodomésticos que duran mucho tiempo. Los perecederos, como alimentos, ropa, gasolina y cigarros, se consumen muy pronto.
Los pagos de servicios (lo que compramos y que no requiere la producción de cosas materiales) incluyen gastos de médicos, abogados e instituciones educativas. Como se muestra en la tabla 6.2, en 2004
los gastos en imperecederos sumaron 8.4% del PIB, los perecederos, 20.2%, y los servicios, 41.4%.
Inversión interna privada bruta (I) Inversión, como usamos el término en economía, se
refiere a la compra de capital nuevo: viviendas, fábricas, equipo y existencias. El uso económico del
término es distinto del uso diario, en el que inversión se refiere a compras de acciones, bonos o
fondos mutuos.
La inversión total en capital del sector privado se llama inversión interna privada bruta (I).
Los gastos de empresas para máquinas, herramientas, fábricas, etc., componen la inversión no residencial.1 Como se trata de bienes que las empresas compran para uso final, son parte de las
“ventas finales” y se cuentan en el PIB. Los gastos de nuevas casas y edificios de departamentos
constituyen la inversión residencial. El tercer componente de la inversión privada bruta es el cambio en inventarios, que es el monto en que varían las existencias de las empresas durante cierto
periodo. Estos inventarios pueden ser considerados como los bienes que las empresas producen
hoy, pero que pretenden vender después. En 2004, la inversión privada bruta sumó 16.4% del
PIB. De esta cifra, 10.2% fue inversión no residencial, 5.7%, inversión residencial, y 0.5%, cambio
en las existencias de las empresas.
Cambio en las existencias de las empresas A veces confunde a los estudiantes que las existencias
se cuenten como capital y que los cambios en ellas se consideren como parte de la inversión privada
interna bruta, pero en la teoría es lógico. Los inventarios que posee una empresa tienen un valor y le
cumplen un propósito o le prestan un servicio. Es evidente que tienen un valor. Piense en las existencias de una nueva distribuidora de automóviles o de una tienda de ropa, o en las computadoras
recién fabricadas que no se han vendido y que esperan ser embarcadas. Todo esto tiene valor.
Aparte, ¿qué servicio prestan las existencias? Las empresas mantienen existencias de reserva
por varias razones. Una es satisfacer una demanda imprevista. Las compañías nunca están seguras sobre cuánto van a vender en un periodo; las ventas suben y bajan. Para conservar la buena
voluntad de los clientes, las empresas deben responder a aumentos imprevistos de las ventas. La
única manera de hacerlo es con existencias.
Algunas compañías usan las existencias para ofrecerles servicios directos a sus clientes, que
es la principal función de una tienda al público. Una tienda de abarrotes ofrece un servicio: comodidad. La tienda como tal no produce ningún alimento, sino que reúne un surtido amplio de
artículos y los exhibe para que los consumidores, con gustos diversos, entren y compren en un
solo lugar lo que quieran. Lo mismo pasa con una tienda de ropa o de hardware. Para brindar
sus servicios, estas tiendas necesitan instalaciones de alumbrado, mostradores, cajas registradoras, establecimientos y muchas existencias. Los inventarios de capital están compuestos por fábricas, equipo y existencias; la acumulación de existencias es parte del cambio en inventarios o
inversiones de capital.
Recuerde que el PIB no es el valor en el mercado de las ventas finales totales de un periodo:
es el valor en el mercado de la producción total. La relación entre producción total y ventas totales es:
PIB = ventas finales + cambio en las existencias de las empresas
La producción total (PIB) es igual a las ventas finales de bienes nacionales más el cambio en las
existencias de las empresas. En 2004, la producción de Estados Unidos superó las ventas en 55,400
millones de dólares. Las existencias al final de 2004 fueron 55,400 millones de dólares mayores
que al final de 2003.

Inversión bruta o inversión neta En el proceso de producción se desgasta el capital producido
en los periodos anteriores, en particular equipo y maquinaria. El PIB no da una imagen verídica
de la producción de una economía. El PIB comprende los bienes de capital recién producidos, pero
no toma en cuenta los bienes de capital “consumidos” en el proceso de producción.
Los activos de capital pierden valor al paso del tiempo. El monto del valor perdido de un activo en cada periodo es su depreciación.
2 Se espera que una computadora personal que hoy
compra una empresa tenga una vida útil de cuatro años y que luego quede obsoleta o gastada.
En ese lapso, la computadora se deprecia constantemente.
¿Cuál es la relación entre inversión bruta (I) y depreciación? La inversión bruta es el valor total de todos los bienes de capital nuevos (fábricas, equipo, viviendas y existencias) generados en
cierto periodo. No toma en cuenta el hecho de que parte del capital se gasta y hay que reemplazarlo.

La inversión neta es igual a la inversión bruta menos la depreciación. La inversión neta es una
medida de cuánto cambia el inventario de capital durante un periodo. Una inversión neta positiva significa que el monto del nuevo capital producido supera el monto del gastado. Una inversión neta negativa significa que el monto del capital recién producido es menor que el monto
del gastado. Por consiguiente, si la inversión neta es positiva, el inventario de capital aumentó;
si es negativa, el inventario de capital se redujo. Dicho de otra manera, el inventario de capital
al final de un periodo es igual al inventario de capital al comienzo del periodo, más la inversión neta:
capitalfin del periodo = capitalcomienzo del periodo + inversión neta
Gasto de gobierno (G) El consumo e inversión bruta del gobierno (G) comprende los
gastos de los gobiernos federal, estatales y locales por bienes finales (bombas, lápices, escuelas) y
servicios (salarios del ejército, de los congresistas, de los maestros). Una parte de estos gastos se
cuenta como consumo del gobierno y otro segmento como inversión bruta gubernamental. Los
pagos de transferencia del gobierno (prestaciones del seguro social, estipendios por incapacidad
de veteranos de guerra, etc.) no se incluyen en G porque estas transferencias no son compras de
nada que se haya producido. Los pagos no se hacen a cambio de bienes y servicios. Como los pagos de intereses sobre la deuda del gobierno cuentan también como transferencias, igualmente
se excluyen del PIB, en tanto que no son pagos de bienes ni servicios reales.
Como se indica en la tabla 6.2, en 2004 el consumo y la inversión bruta sumaron 2 billones,
215,900 millones de dólares, o 18.9% del PIB de Estados Unidos. El consumo del gobierno federal y la inversión bruta de 2004 totalizaron 7.1% del PIB, y el consumo y la inversión bruta de los
gobiernos estatales y locales sumaron 11.8%.
Exportaciones netas (X – M) El valor de las exportaciones netas (X – M) es la diferencia entre las exportaciones (ventas al extranjero de productos y bienes nacionales) y las importaciones (compras en el país de bienes y servicios del extranjero). Esta cifra puede ser positiva
o negativa. En 2004, Estados Unidos exportó menos de lo que importó, así que su monto de exportaciones netas fue negativo (-624,000 millones de dólares). Antes de 1976, Estados Unidos era
un exportador neto: las exportaciones superaban a las importaciones, de modo que la cifra neta
de exportaciones era positiva.
El motivo para incluir las exportaciones netas en la definición del PIB es simple. El gasto de
consumo, en inversión y del gobierno (C, I y G) incluye los gastos en bienes producidos tanto en
el país como por extranjeros. Por consiguiente, C + I + G infla la producción interna, dado que
contiene gastos en bienes producidos en el extranjero, es decir, importaciones (IM) que hay que
restar al PIB para obtener la cifra correcta. Asimismo, C + I + G es menor que la producción nacional porque parte de lo que produce el país se vende en el extranjero y, por lo tanto, no se incluye en C, ni en I, ni en G: es preciso sumar las exportaciones (EX). Si una empresa estadounidense
produce computadoras y las vende en Alemania, las computadoras son parte de la producción de
Estados Unidos y deben contarse como parte del PIB de este último país.

EL MÉTODO DEL INGRESO
Ahora pasamos a calcular el PIB con el método del ingreso, en el que se estudia el PIB en términos de quién lo recibe como ingreso, en lugar de quién lo compra.
Comenzamos con el concepto de ingreso nacional, que se define en la tabla 6.3. El ingreso
nacional es la suma de ocho rubros de ingreso. El pago de los empleados, que es con mucho el
más grande de los ocho rubros, comprende sueldos y salarios pagados a los hogares por empresas
y el gobierno, así como varios complementos salariales, como las aportaciones de los patrones al
seguro social y los fondos privados de retiro. El ingreso de los propietarios corresponde a las empresas desincorporadas. El ingreso de rentas, que representa un rubro menor, es el que reciben
los dueños de propiedades en la forma de rentas. Las utilidades de las corporaciones, que son el
segundo rubro más grande de los ocho, es el ingreso que reciben las empresas corporativas. El interés neto es una obligación que paga una empresa. (Los intereses que pagan hogares y el gobierno no se cuentan en el PIB porque no se supone que procedan de la producción de bienes y
servicios.)
El sexto rubro, impuestos indirectos menos subsidios, abarca gravámenes sobre ventas, tarifas aduaneras y pagos de licencias, menos los subsidios que paga el gobierno y por los que no
recibe bienes ni servicios. (Los subsidios son como impuestos negativos.) Así, el valor de los impuestos indirectos menos subsidios es el ingreso neto que recibe el gobierno. Los pagos netos de
transferencias comerciales son los pagos de transferencia neta de unas empresas a otras y que,
por consiguiente, son el ingreso de las segundas. El último rubro es el superávit de las empresas
paraestatales, que es el ingreso de las empresas gubernamentales. En la tabla 6.3 se muestra que
este aspecto fue negativo en 2004: las paraestatales operaron con pérdidas.
El ingreso nacional es el ingreso total del país, aunque no es todo el PIB. En la tabla 6.4 se
indica lo que se requiere para pasar del ingreso nacional al PIB. Primero, en la tabla 6.4 se muestra que al pasar del producto interno bruto (PIB) al producto nacional bruto (PNB) tenemos
que añadir entradas de ingreso de factores que vienen del resto del mundo y se restan los pagos al resto del mundo por esos factores. El ingreso nacional es el ingreso de los ciudadanos de un
país, no el ingreso de los residentes del país, así que primero tenemos que pasar del PIB al PNB.
Como vimos arriba, el ajuste es mínimo.
A continuación tenemos que restar la depreciación del PIB, que es un ajuste importante. El
PIB menos la depreciación se llama producto nacional neto (PNN). ¿Por qué se resta la depreciación? Para ver por qué, volvamos al ejemplo anterior en el que una economía estaba compuesta
por una sola empresa, con una producción total (PIB) de un millón de dólares ese año. Supongamos que después de que esa compañía paga salarios, intereses y rentas, le quedan 100,000 dólares.
Digamos también que su inventario de capital se depreció 40,000 dólares durante el año. El ingreso nacional incluye las utilidades de la corporación (véase la tabla 6.3) y al calcularlas, esta
depreciación de 40,000 se resta de los 100,000 dólares, lo que arroja utilidades de 60,000 dólares.
Cuando calculamos el PIB con el método del gasto no se resta la depreciación. Tan sólo sumamos
consumo, inversión, gasto del gobierno y exportaciones netas. En nuestro sencillo ejemplo, esto
es exactamente un millón. Así, para obtener el ingreso nacional tenemos que restar la depreciación del PIB (de hecho, el PNB cuando hay un sector internacional).
En la tabla 6.4 se aprecia que el producto nacional neto y el ingreso nacional son iguales salvo
por una discrepancia estadística. Si el gobierno recopilara los datos con total exactitud, la discrepancia estadística sería de cero. Pero el acopio de los datos no es perfecto y la discrepancia estadística es el error de medición de cada periodo. En la tabla 6.4 se muestra que en 2004 este error fue
de –76,900 millones de dólares, que es pequeño en comparación con el ingreso nacional de 10
billones, 275,900 millones de dólares.
Hasta aquí hemos visto, por la tabla 6.3, los rubros de ingreso que componen el ingreso nacional total y, por la tabla 6.4, cuál es la relación entre el PIB y el ingreso nacional. Una manera
provechosa de pensar en el ingreso nacional es considerar cuánto va a los hogares. El ingreso total
de los hogares se llama ingreso personal y resulta de que casi todo el ingreso nacional es personal.
En la tabla 6.5 se indica que de la cifra de 10 billones, 275,900 millones de dólares del ingreso nacional —en 2004—, nueve billones, 713,300 millones correspondieron al ingreso personal. Aunque
no se ve en la tabla 6.5, una de las diferencias entre los ingresos nacional y personal son las utilidades de las corporaciones que no se pagaron a los hogares en la forma de dividendos, las llamadas
ganancias retenidas por las empresas. Se trata de un ingreso que va a las corporaciones pero no a
los hogares y, por lo tanto, es parte del ingreso nacional mas no del personal.
El ingreso personal es el que reciben los hogares, antes de pagar los impuestos individuales.
La proporción del ingreso que los hogares tienen para gastar o ahorrar se llama ingreso personal
disponible o ingreso después de impuestos. Es igual al ingreso por individuo menos los impuestos
personales, como se ve en la tabla 6.5.
Como el ingreso personal disponible es aquel que los hogares pueden gastar o ahorrar, es
un importante concepto del ingreso. En la tabla 6.5 se muestra que hay tres categorías de gasto:
1) gastos de consumo, 2) pagos de intereses personales y 3) pagos de transferencias hechos por los
hogares. El monto del ingreso personal disponible que queda después del gasto personal es el ahorro
personal. Si su ingreso mensual disponible es de 500 dólares y gastó 450, al final del mes le quedan
50. Su ahorro personal del mes es de 50 dólares. El ahorro personal puede ser negativo: si ganó 500 y gastó 600 en el mes, perdió 100 dólares de su riqueza. Para gastar 100 dólares más de lo que ganó tendrá que pedirle 100 dólares a alguien, tomarlos de su cuenta de ahorros o vender alguna
propiedad suya.
La tasa de ahorro personal es el porcentaje ahorrado del ingreso personal disponible, un
importante indicador del comportamiento de los hogares. Una tasa de ahorro baja significa que
los hogares gastan buena parte de sus ingresos. Una tasa elevada indica que los hogares son cuidadosos con sus gastos. Como se ve en la tabla 6.5, la tasa de ahorro personal de Estados Unidos
en 2004 fue de 1.8%. Las tasas de ahorro aumentan en los periodos de recesión, cuando los consumidores se sienten inquietos por el futuro, y bajan en las épocas de auge, cuando se libera la
demanda contenida del gasto.
PIB NOMINAL Y REAL
Hasta aquí hemos visto el PIB medido en dinero actual, esto es, a los precios actuales que pagamos
por las cosas. Cuando una variable se mide en dinero actual, se describe en términos nominales.
El PIB nominal es el PIB medido en dinero actual, con todos los componentes del PIB valuados
en sus precios actuales.
En muchas aplicaciones de macroeconomía, el PIB nominal no es una medida muy deseable
de la producción. ¿Por qué? Supongamos que nada más hay un bien; por ejemplo, pizza. En los
años 1 y 2 se produjeron 100 unidades (rebanadas) de pizza. Así, la producción fue la misma en
los años 1 y 2. Supongamos que el precio de la pizza aumentó de un dólar por rebanada —en el
año 1— a 1.10, en el año 2. El PIB nominal del año 1 es de 100 dólares (100 unidades × 1.00 dólar) y el PIB nominal del año 2 es de 110 (100 unidades × 1.10 dólares). El PIB nominal aumentó
10 dólares, aunque no se produjeron más rebanadas de pizza. Si tomamos el PIB nominal para
medir el crecimiento, podemos equivocarnos y pensar que la producción se incrementó, cuando
en verdad todo lo que pasó fue que se elevó el precio (inflación).
Si sólo hubiera un bien en la economía (como la pizza), sería fácil medir la producción y
comparar el valor de un año con otro. Sumaríamos todas las rebanadas de pizza producidas cada año. En el ejemplo, la producción es de 100 en los dos años. Si el número de pizzas hubiera
ascendido a 105 en el año 2, diríamos que la producción aumentó cinco rebanadas en los dos
años, lo cual equivale a un incremento de 5%. No obstante, la economía tiene mucho más que
un solo bien.
Lo que sigue es una exposición de cómo la Oficina de Análisis Económico ajusta el PIB nominal para dar cuenta de los cambios de precios. Al leer la exposición tenga presente que no es un
ajuste fácil. Ni siquiera en una economía de manzanas y naranjas sería patente cómo sumarlas para
tener una medida general de la producción. La tarea de la OAE es sumar miles de bienes que cambian de precio al paso del tiempo.
En lo que sigue nos valdremos del concepto de peso, ya sea de precios o de cantidades. ¿Qué
es un peso? Es más fácil definirlo con un ejemplo. Supongamos que en su clase de economía hay
un examen final y otras dos pruebas. Si el examen final cuenta por media calificación y las otras
dos pruebas un cuarto cada una, los “pesos” son un medio, un cuarto y un cuarto. Si por el contrario el examen final contara 80% de la calificación y las otras dos pruebas 10% cada una, los
pesos serían 0.8, 0.1 y 0.1. Cuanto más importante es un rubro de un grupo, más peso tiene.
CÁLCULO DEL PIB REAL
El PIB ajustado a los cambios de precios se llama PIB real. Los principales temas relacionados con
el cálculo del PIB real se pueden analizar en una economía simple de tres bienes y dos años. En la
tabla 6.6 se dan todos los datos que necesitaremos. Allí, se presentan datos de precios y cantidades
para dos años y tres bienes. Éstos se designan como A, B y C, y los años, 1 y 2. P denota el precio y
Q la cantidad.
Lo primero que hay que observar en la tabla 6.6 es que la producción nominal (en dinero corriente) del producto A en el año 1 es el precio de ese producto A en el año 1 (0.50), multiplicado
por el número de unidades del bien A producidas en el año 1 (6), lo que da 3.00. Del mismo modo, la producción nominal en el año 1 es de 7 × 0.30 = 2.10 para el producto B y 10 × 0.70 = 7.00
para el bien C. La suma de las tres cantidades, 12.10 dólares en la columna 5, es el PIB nominal
del año 1 en esta economía simple. El PIB nominal del año 2 (que se calcula con las cantidades y
los precios del año 2) es de 19.20 dólares (columna 8). El PIB nominal aumentó de $12.10 en el
año 1 a $19.20 en el año 2, un incremento de 58.7%.3
Se observa que el precio de cada bien cambió entre los años 1 y 2: el precio del bien A bajó (de
$0.50 a $0.40) y los precios de los bienes B y C subieron (B de $0.30 a $1.00 y C de $0.70 a $0.90).
Parte del cambio del PIB nominal entre los años 1 y 2 se debió a variaciones de precios pero no de
producción. ¿Cuánto atribuimos al cambio de precios y cuánto a alteraciones de producción? Aquí
las cosas se vuelven sinuosas. El procedimiento que la Oficina de Análisis Económicos usaba antes
de 1996 consistía en escoger un año base y tomar los precios de ese año como pesos para calcular
el PIB real. Es un procedimiento de pesos fijos, porque los pesos que se usan, que son los precios,
son iguales para todos los años; a saber, los precios que prevalecían en el año base.
Sigamos el procedimiento de los pesos fijos, tomando el año 1 como año base, lo cual significa que los precios del año 1 son los pesos. En la tabla 6.6, el PIB real del año 1 es de $12.10 (columna 5) y el PIB real del año 2, de $15.10 (columna 6). Observe que las dos columnas usan los
precios del año 1 y que el PIB nominal y el real son los mismos en el año 1, porque es el año base.
El PIB real aumentó de $12.10 a $15.10, un incremento de 24.8%.
Ahora repitamos el procedimiento de los pesos fijos, tomando el año 2 como año base, lo
cual significa que los precios del año 2 son los pesos. En la tabla 6.6, el PIB real del año 1 es de
$18.40 (columna 7) y el PIB real del año 2, de $19.20 (columna 8). Vea que en ambas columnas
se toman los precios del año 2 y que el PIB nominal y el real son los mismos en el año 2, que es
el año base. El PIB real aumentó de $18.40 a $19.20, un incremento de 4.3%.
En este ejemplo se muestra que las tasas de interés pueden ser sensibles a la elección del
año base: 24.8% con los precios del año 1 y 4.3% con los precios del año 2 como pesos. Con el
procedimiento antiguo de la Oficina de Análisis Económicos, se tomaba un año como cero y se
hacían todos los cálculos usando como pesos los precios de ese año. El nuevo procedimiento
realiza dos cambios importantes. El primero (si retomamos el ejemplo actual) es “dividir la diferencia” que hay entre 24.8% y 4.3%. ¿Qué significa “dividir la diferencia”? Una manera sería
tomar el promedio de las dos cifras, que es 14.55%. Lo que hace la Oficina de Análisis Económico es tomar el promedio geométrico, que en el ejemplo actual es 14.09%.4 Estos dos promedios (14.55% y 14.09%) son muy próximos y cualquiera daría resultados similares. Aquí lo que
importa no es que se use el promedio geométrico, sino que el primer cambio consiste en dividir la diferencia mediante algún promedio. Observe que este procedimiento nuevo requiere dos
años “cero”, porque 24.8% se calculó con los precios del año 1 y 4.3% con los precios del año 2
como pesos.
El segundo cambio de la OAE es tomar los años 1 y 2 como cero al calcular el cambio porcentual entre los años 1 y 2, luego usar los años 2 y 3 como base para calcular el cambio porcentual entre los años 2 y 3, etc. Los dos años cero cambian conforme los cálculos pasan por el tiempo. La serie de cambios porcentuales computados de esta manera se toma como la serie de tasas de
crecimiento del PIB real y así se ajusta el PIB nominal a los cambios de precios. Para estar seguros de que compendió lo anterior, revise los cálculos de la tabla 6.6, que tiene todos los datos que
necesita para ver lo que pasa.
CÁLCULO DEL DEFLACTOR DEL PIB
Ahora pasamos del PIB real, una medida cuantitativa, al deflactor del PIB, una medida de precios.
Uno de los objetivos de las autoridades económicas es minimizar los cambios del nivel general de
precios. Por este motivo, no sólo necesitan buenas medidas de los cambios de la producción real,
sino también buenas medidas del nivel general de precios. El deflactor del PIB es una medida del
índice general de precios. Podemos usar los datos de la tabla 6.6 para mostrar cómo la OAE calcula el deflactor del PIB.
En la tabla 6.6, el precio del bien A bajó de 0.50 en el año 1 a 0.40 en el año 2; el precio del
bien B aumentó de $0.30 a $1.00, y el precio del bien C aumentó de $0.70 a $0.90. Si únicamente
nos interesara cómo cambian los precios por separado, ésta sería toda la información que nos hiciera falta. Sin embargo, si estamos interesados en cómo cambia el nivel general de precios, de alguna
manera debemos ponderar los precios individuales. Los pesos obvios que hay que usar son las
cantidades producidas, pero ¿qué cantidades, las del año 1 o las del 2? Aquí se plantean los mismos
problemas para los pesos de cantidad que para los pesos de precios al calcular el PIB real.
Primero apliquemos el procedimiento de pesos fijos tomando el año 1 como año base, lo
cual significa que usamos como pesos las cantidades del año 1. Luego, en la tabla 6.6 el precio
“conjunto” del año 1 es de $12.10 (columna 5) y el precio conjunto del año 2, de $18.40 (columna 7). Las dos columnas toman cantidades del año 1. El precio conjunto aumentó de $12.10 a
$18.40, un incremento de 52.1%.
A continuación empleamos el procedimiento de los pesos fijos tomando el año 2 como año
base, lo cual equivale a tomar como pesos las cantidades del año 2. Así, el precio conjunto del
año 1 es de $15.10 (columna 6) y el precio conjunto en el año 2, de $19.20 (columna 8). Ambas
columnas usan cantidades del año 2. El precio conjunto aumentó de $15.10 a $19.20, un incremento de 27.2%.
En este ejemplo se muestra que los incrementos del nivel general de precios pueden ser sensibles a la elección del año cero: 52.1% con cantidades del año 1 como pesos y 27.2% con cantidades del año 2 como pesos. De nueva cuenta, en el procedimiento anterior de la OAE se tomaba
un año como base y se hacían todos los cálculos tomando las cantidades de ese año como pesos.
En el nuevo procedimiento, primero se divide la diferencia entre 52.1% y 27.2% mediante el
promedio geométrico, que es de 39.1%. En segundo lugar, se toman los años 1 y 2 como cero para
calcular el cambio porcentual entre los años 1 y 2, los años 2 y 3 son la base para calcular el cambio
porcentual entre los años 2 y 3, etc. La serie de cambios porcentuales calculada de esta manera se
toma como la serie de cambios porcentuales del deflactor del PIB, es decir, como una serie de tasa
de inflación del nivel general de precios.
LOS PROBLEMAS DE LOS PESOS FIJOS
Para entender por qué la Oficina de Análisis Económicos pasó al nuevo procedimiento, consideremos varios problemas de usar pesos fijos para calcular el PIB real. En primer lugar, los pesos de
los precios de 1987, los últimos que usó la oficina antes de cambiar de procedimientos, no son
muy exactos para, digamos, los años cincuenta. Muchos cambios estructurales han tenido lugar
en la economía estadounidense en los últimos 40 o 50 años, y no es de creer que los precios de
1987 sean buenos pesos para usarlos en los cincuenta.
Otro problema es que el uso de los pesos fijos no toma en cuenta las respuestas de la economía a los cambios de la oferta. Digamos que el mal tiempo es causa de una producción magra de
naranjas en el año 2. En un diagrama simple de oferta y demanda de naranjas, esto corresponde
a un desplazamiento de la curva de la oferta a la izquierda, lo cual causa un aumento en el precio de las naranjas y una reducción de la cantidad demandada. A medida que los consumidores
ascienden por la curva de la demanda, cambian naranjas por otro producto. Si los adelantos técnicos del año 2 dan por resultado métodos más baratos para producir computadoras, el resultado
es un desplazamiento a la derecha de la curva de oferta de computadoras, lo cual lleva a rebajas
del precio de estas máquinas y un incremento de la cantidad demandada. Los consumidores optan
por comprar computadoras. (Usted debe ser capaz de dibujar los diagramas de oferta y demanda
en ambos casos.) En la tabla 6.6 se indica esta tendencia. La cantidad del bien A creció entre los
años 1 y 2 y el precio se redujo (el caso de las computadoras), mientras que la cantidad del producto B bajó y el precio se elevó (el caso de las naranjas). La curva de la oferta de computadoras se
ha desplazado a la derecha con el paso del tiempo, debido sobre todo a los adelantos técnicos. De
ahí resultaron grandes reducciones en el precio de las computadoras y sus correspondientes incrementos en la cantidad demandada.
Para entender por qué estas reacciones plantean un problema para el uso de los pesos fijos,
tomemos los datos de la tabla 6.6. Como el precio del bien A fue mayor en el año 1, el aumento
de la producción del bien A tiene más peso si tomamos el año 1 como base que si tomamos el
año 2. Asimismo, como el precio del bien B fue menor en el año 1, la disminución de la producción del bien B se pondera menos si usamos el año 1 como año base. Estos efectos hacen que el
cambio general del PIB real sea mayor si tomamos los pesos de los precios del año 1, que si recurrimos a los del año 2. Tomar los pesos de los precios en el año 1 omite las respuestas de sustitución que indicamos en el párrafo anterior y lleva a lo que, en opinión de muchos, son cálculos
demasiado grandes de los cambios en el PIB real. Antes, la Oficina de Análisis Económicos adelantaba el año base aproximadamente cada cinco años, lo cual daba por resultado que los cálculos del crecimiento anterior del PIB real tenían que ser ajustados a la baja. Es indeseable que los
cálculos del crecimiento pasado cambien tan sólo porque se cambia el año base. En el nuevo
procedimiento de la Oficina de Análisis Económicos se evitan muchos de estos problemas de los
pesos fijos.
Surgen problemas semejantes cuando se usan pesos de cantidad fija para calcular los índices de precios. Por ejemplo, en el procedimiento de los pesos fijos se ignoran las sustituciones de
bienes cuyo precio aumenta, lo cual favorece a los bienes cuyos precios bajan o se incrementan
menos deprisa. El procedimiento exagera el incremento en el nivel general de precios. Como veremos en el capítulo siguiente, todavía hay varios índices de precios que se calculan con pesos fijos.
El deflactor del PIB difiere porque no usa pesos fijos. También es un índice de precios de todos
los bienes y servicios producidos; además, incluye algunos bienes y servicios importados (generados en el extranjero).
Por último, debe recalcarse que no hay un método “correcto” de calcular el PIB real. La economía consta de muchos bienes, cada uno con su propio precio y, por lo tanto, no hay un método
exacto de sumar toda la producción de los bienes. Podemos decir que el nuevo procedimiento de
la OAE para calcular el PIB evita los problemas del uso de pesos fijos y parece ser mejor que el
método antiguo. No obstante, en el capítulo siguiente veremos que el índice de precios al consumidor (IPC), que es muy usado, todavía se calcula con pesos fijos.
LIMITACIONES DEL CONCEPTO DE PIB
En general, pensamos que es bueno que aumente el PIB. Se considera que aumentar el PIB (o
impedir que disminuya) es una de las principales metas de las políticas macroeconómicas de los
gobiernos. Como se presentan problemas graves si queremos tomar al PIB como medida de la
felicidad, o el bienestar, enseguida apuntaremos algunas limitaciones del concepto del PIB como
medida de riqueza.
PIB Y BIENESTAR SOCIAL
Si los índices de delincuencia bajaran, la sociedad estaría mejor; pero una reducción de la delincuencia no es un acrecentamiento de la producción ni aparece en el PIB. Tampoco lo es un aumento del ocio. Sin embargo, en la medida en que los hogares quieren más tiempo libre (a diferencia de ser obligados a parar por falta de fuentes de empleo en la economía), un incremento del
ocio es también una mejora del bienestar social. Además, algo de esto último se relaciona con una
disminución del PIB. Por ejemplo, un aumento del tiempo libre en una época de pleno empleo se
traduce en una reducción del PIB, porque se dedica menos tiempo a la producción.
La mayor parte de las actividades domésticas y no remunerativas, como las faenas de la casa y
el cuidado de los niños, no se cuentan en el PIB aunque contribuyen a la producción real. No
obstante, si decido enviar a mis hijos a la guardería o contratar a alguien que limpie mi casa o
que maneje mi coche, el PIB aumenta. Los salarios del personal de la guardería, afanadores y
choferes se cuentan en el PIB, pero el tiempo que dedico a hacer esas mismas cosas no se incluye.

Un simple cambio de organización o institución, aunque no se produzca más, se muestra como
un cambio en el PIB.
Por otra parte, el PIB casi nunca refleja pérdidas ni males sociales. Las reglas contables del
PIB no prevén la contaminación que daña el ambiente natural. Cuanta más producción haya,
mayor es el PIB, sin que importe qué tanta contaminación se genere.
El PIB tampoco dice nada sobre la distribución de la producción entre los miembros de la
sociedad. Por ejemplo, no distingue entre el caso en que la mayor parte de la producción va a pocas personas y el caso en que se divide con justicia entre todos. No podemos valernos del PIB para
medir los efectos de las políticas distributivas (que toman del ingreso de unas personas para darles a otras). Estas políticas no tienen un efecto directo en el PIB. También el PIB es neutro en
cuanto a los bienes que genera la economía. Los conciertos de las sinfónicas, pistolas, cigarros,
partidos de futbol, ejemplares de la Biblia, refrescos, leche, libros de economía e historietas cuentan de la misma manera, sin atender a su diferente valor para la sociedad.
A pesar de estas limitaciones, el PIB es muy útil para medir la actividad económica y el bienestar. Si usted duda de esta afirmación, responda la siguiente pregunta: ¿le hubiera gustado vivir
en Estados Unidos, hace 200 años, cuando los ríos no estaban contaminados y la delincuencia era
mínima, o en la actualidad? La mayoría ha dicho que prefiere el presente. No obstante todos los
“inconvenientes”, el PIB por persona y los estándares de vida promedio de hoy son mucho más altos que los de hace dos centurias.
LA ECONOMÍA SUBTERRÁNEA
En los cálculos del PIB se omiten muchas transacciones, aunque en principio deberían incluirse. En
general, faltan las operaciones ilegales, salvo que se “laven” dentro de negocios legítimos. El ingreso
que se gana pero no se declara para fines fiscales también queda fuera, aunque se hacen ciertos
ajustes al PIB para tomar en cuenta el ingreso mal declarado. La parte de la economía que debería
integrarse al PIB pero que no se hace de esa forma se llama economía subterránea.
Se piensa que la evasión fiscal es la principal motivación de la gente para participar en la economía subterránea. En estudios se ha calculado que el tamaño de la economía subterránea de Estados Unidos, que va de 5% a 30% del PIB,5 es equivalente a la economía subterránea de la mayoría de
los países de Europa y quizá mucho menor que la de los países de Europa oriental. Se calcula que la
economía subterránea de Italia va de 10% a 35% del PIB de ese país. En el otro extremo de la escala, los cálculos de Suiza señalan un margen de 3% a 5%.
¿Por qué nos preocupa la economía subterránea? En la proporción en que el PIB refleja sólo
una parte de la actividad económica y no es una medida completa de lo que produce la economía,
se presta a equivocaciones. Por ejemplo, las tasas de desempleo serían menores de lo que reflejan
las cifras oficiales si el gobierno estuviera informado acerca de la gente que trabaja en la economía
subterránea. También, como en efecto el tamaño de la economía subterránea varía entre los países,
cabe equivocarse al comparar sus respectivos PIB. Por ejemplo, el PIB de Italia sería mucho mayor
si consideráramos al sector subterráneo como parte de su economía, mientras que el PIB de Suiza
cambiaría poco.
INGRESO NACIONAL BRUTO PER CÁPITA
Es difícil hacer comparaciones entre países porque hay que efectuarlas con una sola moneda, por
lo regular el dólar estadounidense. Convertir las cifras del PIB de Japón en dólares requiere cambiar yenes por dólares. Como los tipos de cambio pueden variar de manera acelerada en lapsos
breves, estas conversiones son engañosas. Hace poco, el Banco Mundial adoptó un nuevo sistema
de medición para comparaciones internacionales. El concepto del ingreso nacional bruto (INB) es
el PIB convertido en dólares usando un tipo de cambio promedio de siete años, ajustado a la inflación. En la tabla 6.7 se anota el ingreso nacional bruto per cápita (el INB dividido entre la población) de varios países en 2004. Noruega tiene el INB per cápita más grande, seguido por Suiza, Estados Unidos y Dinamarca. Se calculó que, en 2004, Etiopía tuvo un INB per cápita de apenas
110 dólares.

1. Una fuente de información sobre las variables fundamentales de la
macroeconomía son las cuentas del ingreso nacional y del producto.
Estas cuentas ofrecen un marco conceptual para que los expertos
en macroeconomía analicen la correspondencia entre las partes
de la economía
PRODUCTO INTERNO BRUTO
2. El producto interno bruto (PIB) es el concepto clave de la
contabilidad del ingreso nacional. El PIB es el valor total en
el mercado de todos los bienes y servicios finales producidos
en determinado periodo por factores de producción situados
dentro del país. El PIB excluye bienes intermedios. Incluir bienes
cuando se compran como insumos y cuando se venden como
productos finales sería duplicar la contabilidad y exagerar el
valor de la producción.
3. El PIB excluye todas las transacciones de dinero o bienes que no
producen nuevos productos ni servicios. El PIB incluye el ingreso
de los extranjeros que trabajan en el país y las utilidades que hacen
las compañías foráneas. El PIB descarta el ingreso de los ciudadanos
que trabajan en el extranjero y las utilidades ganadas por las
compañías locales en otros países.
4. El producto nacional bruto (PNB) es el valor en el mercado de
todos los bienes y productos finales generados en cierto periodo
por los factores de producción que son propiedad de los ciudadanos de un país.
CÁLCULO DEL PIB
5. El método del gasto para calcular el PIB suma los montos gastados
en todos los bienes y servicios finales durante cierto periodo. Las
cuatro principales categorías de gastos son gastos de consumo (C),
inversión interna privada bruta (I), consumo e inversión bruta del
gobierno (G) y exportaciones netas (X - M). La suma de estos rubros es igual al PIB.
6. Los tres principales componentes de los gastos de consumo (C)
son imperecederos, perecederos y servicios.
7. La inversión interna privada bruta (I) es la inversión total hecha
por el sector privado en determinado periodo. Hay tres clases de inversión: inversión no residencial, inversión residencial y cambios
en las existencias de las empresas. La inversión bruta no toma en
cuenta la depreciación, que es la pérdida de valor de los activos. La
inversión neta es igual a la inversión bruta menos la depreciación.
8. El consumo y la inversión neta del gobierno (G) abarca los gastos de
los gobiernos federal, estatales y locales en bienes y servicios finales. El valor de las exportaciones netas (EX - IM) es igual a la
diferencia entre las exportaciones (las ventas a extranjeros de
productos y servicios nacionales) y las importaciones (las compras
de bienes y servicios de fuera).
9. Como todo pago (gasto) de un comprador es una entrada (ingreso) de un vendedor, el PIB puede calcularse según quién lo recibe
como ingreso: es el método del ingreso para calcular el producto
interno bruto.
10. El PIB menos la depreciación es el producto nacional neto (PNN).
El ingreso nacional es el monto total ganado por los factores de
producción de la economía; es igual al PNN salvo por una discrepancia estadística. El ingreso personal es el ingreso total de los hogares. El ingreso personal disponible es lo que los hogares tienen
para gastar o ahorrar después de pagar sus impuestos. La tasa de
ahorro personal es el porcentaje de ingreso personal disponible
que se ahorra en lugar de gastarlo.
PIB NOMINAL Y REAL
11. El PIB medido en dinero actual (los precios actuales que se pagan
por los bienes) es el PIB nominal. Si tomamos el PIB nominal para medir el crecimiento, podemos cometer el error de pensar que
aumentó la producción, cuando todo lo que pasó fue que se incrementó el nivel de precios, hubo inflación. Una mejor medida
de la producción es el PIB real, que es el PIB nominal ajustado a
los cambios de precios.
12. El deflactor del PIB es una medida del índice general de precios.
LIMITACIONES DEL CONCEPTO DE PIB
13. En general, pensamos que los aumentos del PIB son buenos, pero
surgen problemas cuando queremos usar al PIB como medida de
la felicidad o el bienestar. Las peculiaridades de la contabilidad
del PIB significan que los cambios institucionales pueden modificar su valor, aun si la producción real no ha cambiado. El PIB ignora los males sociales, como la contaminación. Además, el PIB
no dice nada sobre qué bienes se producen ni cómo se distribuye
el ingreso entre la población. El PIB también excluye muchas
transacciones de la economía subterránea.
14. El concepto de ingreso nacional bruto (INB) es el PIB traducido
en dólares con un promedio de los tipos de cambio de siete años,
ajustado a la inflación.

## 3. Preocupaciones de corto y largo plazos: crecimiento, productividad, desempleo e inflación

La economía ideal es aquella
donde aumenta muy pronto
la productividad por trabajador,
hay poco desempleo y la inflación
es mínima. En esta situación, las autoridades encargadas de las políticas económicas pueden dormir
tranquilas. Pero, ay, la economía no
siempre se encuentra en ese estado
ideal. Puede haber épocas de crecimiento lento, mucho desempleo e
inflación elevada. Una parte esencial
de la macroeconomía es considerar
qué determina la producción, el desempleo y la inflación. ¿Por qué el
crecimiento de la productividad
unas veces es mayor que otras? ¿Por
qué en ocasiones se incrementan el
desempleo y la inflación? En el siguiente capítulo comenzaremos la tarea de explicar el funcionamiento de la macroeconomía. No obstante, antes de sumergirnos en el análisis, sería útil que
en este capítulo dedicáramos algo más de tiempo a las descripciones.
En primer lugar expondremos los problemas de largo plazo, a saber, el ritmo de crecimiento
de la producción y la productividad por trabajador durante un periodo largo. Luego pasamos a
las cuestiones de corto plazo del desempleo y la inflación.
Antes de estudiar la producción y el incremento de la productividad, sería conveniente repasar algunos conceptos. El crecimiento de la producción es el ritmo de crecimiento de la producción
de toda la economía. El crecimiento per cápita de la producción es el ritmo de crecimiento de la
producción por persona en una economía. Si la población de un país crece a la misma tasa que
la producción, la producción per cápita no aumenta, sino que tan sólo se mantiene al parejo del
crecimiento demográfico. No todos los habitantes de un país trabajan y por eso la producción
por trabajador no es lo mismo que la producción por persona. La producción por trabajador es
mayor que la producción por persona y se llama productividad. Así, el incremento de la productividad es el ritmo de crecimiento de la producción por trabajador.
Si tomamos la producción per cápita como medida del bienestar económico de un país,
una tasa acelerada de crecimiento de la producción no necesariamente mejora el bienestar de
los individuos. Si la población crece tan deprisa como la producción, la producción per cápita
no aumenta y el bienestar económico del promedio de las personas no se acrecienta. Si el segmento de la población que trabaja no aumenta y la productividad se expande, la producción
per cápita crecerá porque cada individuo produce más. La producción per cápita también aumenta si la productividad no cambia, aunque la parte de la población que trabaja se multiplica. En este caso, también hay más producción por persona, si bien se debe a que más personas
trabajan. PRODUCCIÓN DE LARGO PLAZO E INCREMENTO DE
LA PRODUCTIVIDAD
En la figura 5.4 del capítulo 5 se muestra que el crecimiento promedio de la producción de la
economía estadounidense ha sido de alrededor de 3.4% al año, desde 1900. Algunos años han
sido mejores que otros, pero en promedio, la tasa es de 3.4%. Una rama de la economía llamada teoría del crecimiento se ocupa de investigar qué determina esta tasa. ¿Por qué 3.4% y no 2 o
4%? Retomaremos esta cuestión en el capítulo 18, pero es prudente hacer algunos señalamientos aquí.
Antes de iniciar nuestra exposición del crecimiento económico, es importante revisar lo
que se entiende por capital. El capital es todo lo que se produce y que a continuación se usa como
insumo para generar otros bienes y servicios. El capital puede ser tangible, como edificios y
equipo, o intangible. Los conocimientos y habilidades que se adquieren por educación y capacitación pueden considerarse capital humano intangible. El capital es privado o público. Los caminos y puentes por los que pasamos son parte del capital social público. Así, el capital adopta
muchas formas. Sin embargo, para simplificar la exposición a veces nos referiremos al capital
nada más como “máquinas”.
En una economía simplificada se necesitan máquinas (capital) y trabajadores (mano de
obra) que realicen la producción. Supongamos que una economía consta de seis máquinas y 60
trabajadores. En cada máquina se ocupan 10 operarios. La semana laboral tiene 40 horas y
arroja 50 unidades de producción por mes y por máquina. Entonces, en esta economía simple
la producción total (PIB) es de 300 unidades (seis máquinas por 50 unidades por máquina).
¿Cómo se puede incrementar la producción en esta economía? Hay varias formas. Una es
traer más trabajadores. Digamos, si se suman 12 operarios, dos más por máquina, aumenta la
producción por máquina y por hora, puesto que hay más obreros afanados en cada máquina.
Otro método consiste en agregar máquinas. Por ejemplo, si se añaden cuatro máquinas, los 60
empleados tienen 10 máquinas para laborar, en lugar de seis, y hay más producción por trabajador y por hora trabajada. La tercera posibilidad es acrecentar la duración de la semana laboral
(por ejemplo, de 40 a 45 horas). Como los obreros y las máquinas operan más horas, arrojan
más producción. Así, ésta se acrecienta si se incrementa la mano de obra o el capital, o si aumenta el tiempo de operación por semana de los trabajadores o el capital.
Otra manera de agrandar la producción de nuestra economía es mejorar la calidad de los
trabajadores. Por ejemplo, si se incentiva la educación de los trabajadores, éstos aprenden destrezas y refuerzan su capacidad de trabajar en las máquinas. Entonces la producción de las
máquinas podría pasar de 50 unidades al mes a una cifra mensual mayor. Además, si los trabajadores mejoran su condición física porque hacen más ejercicio y comen menos grasas y más
cereales enteros, frutas y verduras frescas, es posible que su producción en las máquinas crezca. Se dice que la gente aumenta su capital humano cuando mejora sus facultades mentales o
físicas.
También la calidad de las máquinas puede superarse. En particular, los aparatos nuevos
que reemplacen a los viejos aumentarían la producción por hora con el mismo número de trabajadores. En nuestro ejemplo, quizá con cada máquina se generen 55 unidades mensuales, en
lugar de 50, con los mismos operarios y 40 horas hábiles por semana. Un caso obvio es el cambio de equipo de cómputo viejo por uno nuevo y más veloz que permita hacer más trabajo por
minuto.
Para resumir, la producción aumenta si hay más trabajadores, más capacidades por operario, más máquinas, aparatos más veloces o semanas laborales más largas.

Ahora pasemos del estudio de las tendencias de largo plazo a considerar sus desviaciones en la
economía. En otras palabras, vamos a considerar los ciclos comerciales, las altas y bajas periódicas de la economía.
Recuerde que en el capítulo 5 dijimos que una recesión es, en términos generales, una
época en que el PIB real se reduce por lo menos dos trimestres consecutivos. También, que el
PIB real es una medida de la producción concreta de bienes y servicios de una economía en
cierto periodo. Cuando el PIB real baja, es porque se produce menos. Cuando esto ocurre, se
usan menos insumos, los trabajos escasean, la tasa de desempleo aumenta y se aprovecha una
proporción menor del inventario de capital disponible (más fábricas y equipos operan por debajo de su capacidad total). Cuando el producto real disminuye, el ingreso real mengua.
Una depresión es una recesión prolongada y profunda, aunque hay desacuerdos sobre qué
tan grave o prolongada debe ser una recesión para llamarla depresión. Casi todos están de
acuerdo en que la economía estadounidense sufrió una depresión entre 1929 y finales de los
años treinta. Las recesiones más graves desde esa década tuvieron lugar entre 1980 y 1982.

Tasas de desempleo por estados y regiones Las tasas de desempleo varían según
la ubicación geográfica. Por diversas razones, no todos los estados y regiones tienen el mismo nivel
de desempleo. Estados y regiones tienen diversas combinaciones de industrias, que no crecen ni
menguan al mismo tiempo ni con la misma velocidad. Asimismo, la fuerza laboral no es del todo
móvil: los trabajadores no siempre pueden ni quieren empacar y mudarse para aprovechar las
oportunidades laborales que se ofrecen en otras partes del país.
En los últimos 30 años han ocurrido cambios notables en la prosperidad relativa de las regiones estadounidenses, sobre todo en el noreste y la zona petrolera del sureste. A comienzos de
los años setenta, el noreste (en particular Nueva Inglaterra) fue golpeado por una grave reducción de la base industrial. Plantas de textiles, empresas de artículos de cuero y fábricas de muebles cerraron por la competencia extranjera o se mudaron a estados sureños con salarios más
bajos. En la recesión de 1975, Massachusetts y Michigan tuvieron tasas de desempleo muy elevadas (11.2 y 12.5%). Sobre la ola de aumentos en los precios del petróleo, Texas tuvo una de las tasas de desempleo más bajas de la época, de 5.6% (tabla 7.4).
En la recesión de 1982, Texas también salió bien librado y Massachusetts tuvo una mejora
completa. La tasa de desempleo en Massachusetts era superior en casi tres puntos al promedio
nacional en la recesión de 1975, mientras que en la de 1982 se ubicaba casi dos puntos por abajo
del índice general.
Para 1987 las cosas habían cambiado. Aunque no se muestra en la tabla 7.4, Massachusetts
tuvo una de las tasas de desempleo más bajas del país en ese año (un sorprendente 2.8%) y Texas, con 8.5%, uno de los más elevados. En Massachusetts, las empresas de alta tecnología, como
Wang Laboratories y Digital Equipment, dos empresas que dieron trabajo a más de 100,000 personas, habían crecido notablemente. En contraste, la caída de los precios del petróleo, de más de
30 dólares por barril a menos de 15 a comienzos de los ochenta, sumió a la economía petrolizada
de Texas en una recesión profunda y prolongada. Luego, en 1991 Massachusetts experimentó otro
revés con una tasa de desempleo de 9%.
La economía de Michigan está muy ligada a la suerte de la industria automotriz. En la recesión
de 1982, Michigan tuvo la tasa de desempleo más alta del país, de 15.5%. El sector automovilístico no sólo sufrió por la baja de la economía estadounidense, sino que también enfrentó una
fuerte competencia extranjera, sobre todo de Japón. Michigan también tuvo problemas en 1991,
con una tasa de desempleo de 9.2%.
La recesión de 2001 fue ligera en comparación, y el efecto fue mucho más uniforme en todo
el país. Para el segundo trimestre de 2003, la tasa nacional de desempleo había aumentado a
6.1%, de 3.9% en el cuarto trimestre de 2000. Para los estados anotados en la tabla 7.4, la tasa de
desempleo en 2003 varió de 5.1% en Florida a 7.3% en Michigan.
La tasa nacional de desempleo no da toda la explicación. Una tasa baja no significa que todo el país crece y produce al mismo ritmo.
Efectos del trabajador desmotivado Recuerde que las personas que dejaron de buscar trabajo no se clasifican como desempleadas, sino que se considera que abandonaron la fuerza
de trabajo. En las recesiones, muchos se desalientan en la búsqueda de trabajo y desisten de ese
propósito. Esto reduce la tasa de desempleo porque quienes ya no buscan dejan de contar como
desempleados.
Para demostrar que este efecto del trabajador desmotivado reduce la tasa de desempleo, supongamos que hay 10 millones de desempleados en una fuerza laboral de 100 millones. Esto
significa una tasa de desempleo de 10/100 = 0.10, o sea 10%. Si un millón de esos 10 millones de
desempleados dejan de buscar trabajo y abandonan el mercado laboral, habría nueve millones
de desempleados en una fuerza laboral de 99 millones. Entonces, la tasa de desempleo sería de
9/99 = 0.91, o 9.1%.
La encuesta de la Oficina de Estadísticas Laborales arroja pruebas de la magnitud del efecto
del trabajador desmotivado. A los entrevistados que indicaron que habían dejado de pedir trabajo
se les preguntó por qué. Si alguno cita la imposibilidad de encontrarlo como única razón para
abandonar la búsqueda, se le clasifica como trabajador desmotivado.
El número de trabajadores desmotivados parece quedar cerca de 1% de la fuerza laboral de
tiempos normales. En la recesión del periodo 1980-1982, el número de trabajadores desmotivados tuvo un aumento constante hasta un máximo de 1.5%. Al terminar el primer trimestre de
1991, la recesión de 1990 a 1991 había producido 997,000 trabajadores desmotivados. Algunos economistas afirman que sumar la cifra de trabajadores desmotivados al número de los que ahora
se clasifican como desempleados da una mejor imagen de la situación del desempleo.

promedio.

semanas. El lento crecimiento que siguió a la recesión del lapso 1990-1991 dio por resultado un
incremento en la duración del desempleo de 17.7 semanas en 1992 a 18.8 semanas en 1994. En
el año 2000, la duración promedio bajó a 12.6 semanas, que luego volvió a incrementarse a 19.6
semanas en 2004.
LOS COSTOS DEL DESEMPLEO
En la Ley del Empleo de 1946, el Congreso estadounidense declaró que era
política permanente y responsabilidad del gobierno federal aplicar todos los medios viables [...] para fomentar el máximo empleo, producción y poder de compra.
En 1978 el Congreso aprobó la Ley de Pleno Empleo y Crecimiento Equilibrado, conocida como
Ley Humphrey-Hawkins, que estableció formalmente una meta concreta de desempleo de 4%.
¿Por qué el pleno empleo tendría que ser un objetivo del gobierno federal? ¿Qué costos impone el
desempleo a la sociedad?
Algún desempleo es inevitable Antes de estudiar los costos del desempleo, tenemos
que darnos cuenta de que algo de desempleo es parte de la operación natural del mercado laboral. Recuerde que para ser clasificada como desempleada, una persona debe estar en busca de
colocación. Todos los años, miles de individuos se integran por primera vez a la fuerza de trabajo.
Algunos abandonaron la preparatoria, otros la terminaron o completaron la licenciatura y unos
más obtuvieron un posgrado. Asimismo, se fundan empresas nuevas y otras se expanden y
abren nuevas fuentes de trabajo, mientras que algunas más se contraen o quiebran.
En todo momento hay un grupo que busca trabajo y una oferta de plazas laborales, entonces,
hay que correlacionar el primero con la segunda. Es importante que los individuos correctos queden en los puestos adecuados. El puesto apropiado para cada quien depende de sus capacidades,
preferencias de entorno laboral (empresas grandes o pequeñas, formales o informales), lugar de
residencia y disposición a cambiar de domicilio. De este modo, las empresas quieren trabajadores
que cumplan los requisitos del puesto y crezcan con ellas.
Para poder ocupar un puesto, los trabajadores deben conseguir información sobre su disponibilidad, tasas salariales, ubicación y entorno laboral. Las empresas tienen que recabar información sobre la disponibilidad y las capacidades de los trabajadores. Acopiar información
requiere tiempo y recursos. La búsqueda impone viajes, entrevistas, elaboración de currículos,
llamadas telefónicas y horas dedicadas a revisar los anuncios clasificados. En la medida en que
estos esfuerzos sirvan para hacer corresponder mejor a los trabajadores con los puestos, son
una buena inversión. El resultado es eficiente si las ganancias para las empresas y los trabajadores exceden los costos de la búsqueda.
Cuando consideramos los costos del desempleo, es útil clasificarlo en tres tipos:
■ Desempleo friccional (por fricción)
■ Desempleo estructural
■ Desempleo cíclico
Desempleo friccional y estructural Cuando la Oficina de Estadísticas Laborales
hace su estudio de la actividad laboral en la semana del día 12 de cada mes, entrevista a muchas personas que se dedican a la búsqueda ordinaria de ocupación. Algunas se integran
apenas a la fuerza de trabajo o cambian de empleo. Este desempleo es natural y benéfico para
la economía.
La proporción del desempleo debida al funcionamiento normal del mercado laboral se llama
desempleo friccional. La tasa del desempleo friccional nunca puede ser de cero, aunque cambia
con el tiempo. A medida que los empleos se vuelven cada vez más diferenciados y aumentan las
capacidades requeridas, se complica la correspondencia entre habilidades y ocupaciones, con lo
cual aumenta el desempleo friccional.
El concepto de desempleo friccional es un tanto abstracto porque es difícil saber qué significa el “funcionamiento normal del mercado laboral”. La estructura industrial de la economía
estadounidense cambia de manera continua. Por ejemplo, la manufactura ha rendido un porcentaje de su participación en el empleo total a los servicios y las finanzas, seguros y bienes raíces. En
el sector fabril, las industrias del acero y textiles se han contraído notablemente, mientras que se
han expandido los sectores de alta tecnología, como los componentes electrónicos.
El desempleo que se produce por estos cambios estructurales podría clasificarse como de
fricción, pero más bien se llama desempleo estructural. El término desempleo friccional denota
problemas pasajeros de correspondencia entre puestos y habilidades que duran pocas semanas.
El desempleo estructural denota problemas de desajustes prolongados que tienden a durar años.
Aunque en una economía dinámica se espera desempleo estructural, es doloroso para los trabajadores que lo sufren. En cierto sentido, los que pierden su trabajo porque sus capacidades se
hicieron obsoletas son los que más lo padecen. El hecho de que el desempleo estructural es natural e inevitable no significa que no le cueste nada a la sociedad.
A veces, los economistas se refieren a una tasa natural de desempleo, como la desocupación
que ocurre como parte normal del funcionamiento de la economía. Este concepto también es un
tanto vago, porque natural no es una palabra precisa. Quizá sea mejor pensar que la tasa natural
es la suma de las tasas friccional y estructural. Los cálculos de la tasa natural van de 4 a 6%.
Desempleo cíclico y pérdida de producción Si bien algo de desempleo es natural,
hay ocasiones en que la cifra parece superar esa tasa natural. En 1979 la tasa de desempleo era de
5.8%, pero no volvió a bajar de 6% sino hasta 1987, ocho años después. Entre tanto, Estados
Unidos pasó por una recesión grave en la que la tasa de desempleo se incrementó sustancialmente. El aumento del desempleo que ocurre en las recesiones y depresiones se llama desempleo
cíclico.
En cierto sentido, un aumento del desempleo en una recesión es una pura manifestación de
un problema más fundamental. Éste es que las empresas producen menos. Recuerde que una recesión entraña una reducción del PIB real, o sea, de la producción real. Cuando las empresas se
achican y producen menos, emplean menos trabajadores y menos capital. Así, el primer costo y
el más directo de una recesión es la pérdida de bienes y servicios reales que se habrían producido en otras circunstancias.
La pérdida de producción nunca fue más drástica que durante la Gran Depresión. En la
tabla 7.1 vimos que la producción real cayó casi 27% entre 1929 y 1933. Desde luego, lo que
más importa es la producción real de la economía: lo que comemos, la atención médica que

recibimos, el automóvil que manejamos, las películas que vemos, las casas nuevas que se construyen, las ollas con que cocinamos y la educación que recibimos. Cuando la producción baja
27%, la vida de mucha gente cambia.
En la recesión de 1980 a 1982, la tasa de crecimiento del PIB real fue, en promedio, de casi
cero. En la tabla 7.1, si el PIB real hubiera crecido 3% anual desde 1979, el crecimiento total de
la producción en el periodo 1980-1982 habría sido de alrededor de 9%, en lugar de 0.1%, como
ocurrió en efecto. Se trató de una pérdida sustancial de la producción y, por consiguiente, del
ingreso.
Consecuencias sociales Los costos de recesiones y depresiones no se distribuyen equitativamente entre la población ni se dejan cuantificar con facilidad. Acaso las consecuencias sociales de la depresión de los años treinta sean las más difíciles de comprender. La mayoría de las
personas que viven actualmente no pasaron por la Gran Depresión y sólo pueden leer al respecto
en libros o escuchar lo que cuentan padres y abuelos. Pocos salieron indemnes de esa etapa. Al último quedaron los pobres y los más desocupados, alrededor de 25% de la fuerza laboral. Incluso
quienes conservaron su puesto trabajaban medio tiempo. Muchas personas perdieron parte o todos sus ahorros porque la bolsa de valores se derrumbó y miles de bancos quebraron.
Comisiones del Congreso estadounidense oyeron un caso tras otro. En Cincinnati, donde la
fuerza laboral sumaba alrededor de 200,000 trabajadores, casi 48,000 estaban desempleados,
40,000 más habían pasado a medio tiempo y los pagos de ayuda a los necesitados iban de siete a
ocho dólares por semana:
Se entrega la ayuda a una familia una semana y luego se suspende otra semana, con la
esperanza de que de un modo u otro el sostén de la casa encuentre algún trabajo [. . .]
No pagamos renta. Desde luego, es un problema muy difícil porque enfrentamos desahucios constantes y vienen trabajadoras sociales [. . .] Es difícil encontrar acomodo para
gente que tiene sus muebles en la calle.1
De Birmingham, Alabama, en 1932:
. . . en mi distrito tenemos aproximadamente 108,000 asalariados. De esa cifra, creo que
no pasan de 8,000 los que reciben sus ingresos normales. Por lo menos 25,000 hombres
no tienen ningún trabajo. Algunos no han tenido la mínima ocupación en más de 12 meses. Quizá de 60,000 a 70,000 trabajan entre uno y cinco días por semana y, de hecho,
todos han sufrido fuertes recortes de salario y muchos no promedian más de 1.50 dólares al día.2
Las dificultades económicas también acompañaron a la recesión del lapso 1980-1982. Entre
1979 y 1983 el número de estadounidenses clasificados oficialmente en la pobreza pasó de 26.1
millones (11.7% de la población) a 35.5 millones (15.3%). Además de las dificultades económicas,
el desempleo prolongado también trae males sociales y personales: ansiedad, depresión, deterioro
de la salud física y psicológica, consumo de drogas (incluyendo alcoholismo) y suicidio.
LOS BENEFICIOS DE LAS RECESIONES
¿Las recesiones tienen beneficios? Sí: es probable que disminuyan la inflación. En la figura 5.7 vimos dos periodos de inflación grave desde 1970: 1974-1975 y 1979-1981. Cada uno fue seguido
por una recesión en la que se redujo la inflación. Como se observa en la tabla 7.6, la inflación bajó
de 11.0% en 1974 a 5.8% en 1976. En 1983 la inflación también bajó, a 3.2%, de una tasa en 1980 de
13.5. Al parecer, las recesiones sirven para contrarrestar la inflación, pero hacen falta más análisis
para entender el porqué (véase el capítulo 13). El punto aquí es que:
Las recesiones pueden ayudar a reducir la inflación.
Algunos dicen que las recesiones aumentan la eficiencia porque hacen quebrar a las empresas menos capaces en la economía y obligan a las sobrevivientes a reducir desperdicios y manejar
mejor sus recursos. Como veremos en el capítulo 21, una recesión hace disminuir la demanda de
importaciones, lo cual mejora la balanza de pagos de la nación, es decir, el estado de su comercio
con otros países.
INFLACIÓN
Muchas veces, aunque no siempre, las altas de los ciclos comerciales alientan la inflación. En la tabla 7.7 se indica la tasa de inflación de los tres últimos periodos de expansión. La inflación aumentó
de 3.2 a 11.0% de 1972 a 1974. La tendencia de la inflación también fue ascendente de 1976 a 1980.
En cambio, el crecimiento sostenido que empezó en 1983 no trajo una inflación acelerada.
¿Por qué es problemática la inflación? Si entiende que sueldos y salarios, así como otras formas de ingresos, aumentan junto con los precios en los periodos de inflación, usted verá que la
pregunta es más sutil de lo que pudiera pensar. Si mi ingreso se duplica, lo mismo que el precio de
las cosas que compro, ¿mi situación es peor? Compro exactamente las mismas cosas que ayer, así
que en la medida en que mi bienestar dependa de lo que puedo comprar, la respuesta es negativa.
No obstante, en las inflaciones los precios y los ingresos no se incrementan al mismo tiempo.
Para algunas personas, el ingreso se acrecienta de forma más apresurada que la inflación; para
otras, los precios crecen más deprisa. Algunos individuos se benefician de las inflaciones, otros salen perjudicados.
En el resto del capítulo nos enfocamos en el problema de la inflación: su medición, sus costos y las ganancias y pérdidas sufridas en periodos inflacionarios.
DEFINICIÓN DE INFLACIÓN
¿Qué es la inflación? No todos los incrementos de precios constituyen una inflación. Los precios
de bienes y servicios se determinan de muchas maneras. En los mercados competitivos, la interacción de muchos compradores y vendedores (el juego de la oferta y la demanda) determina los
precios. En los mercados imperfectamente competitivos, los precios los determinan las decisiones
de los productores (y éste es el meollo de la teoría microeconómica).
En toda economía, los precios cambian sin parar conforme los mercados se ajustan a las modificaciones de la situación. La falta de lluvias puede secar los cultivos de maíz y trigo, lo cual reduce la
oferta y hace subir los precios de los productos agrícolas. Asimismo, la extracción excesiva de los
productores de petróleo abate el precio del crudo y los productos derivados. También es posible que
el sindicato nacional de trabajadores de la industria automovilística negocie un contrato con la
Ford Motor Company en el que se eleven (o reduzcan) los niveles salariales.
Cuando el precio de un bien aumenta, ese acontecimiento puede ser o no parte de una inflación más extensa. Recuerde que la inflación es un incremento del nivel general de precios. Ocurre
cuando muchos precios suben al mismo tiempo. Para medir la inflación, se toman muchos bienes
y servicios y se calculan las alzas promedio en cierto periodo. La deflación es una reducción del
nivel general de precios. Ocurre cuando muchos precios bajan a un tiempo.
Es útil distinguir entre un incremento único del nivel general de precios y un aumento de este
nivel que se prolonga al paso del tiempo. Por ejemplo, el nivel general de precios podría subir 10%
en un mes y dejar de hacerlo, o bien, podría estar subiendo durante varios años. Los economistas
usan el término inflación para referirse sólo a incrementos del nivel de precios que se extienden durante un tiempo significativo. Nos referiremos a este tiempo como periodo de inflación sostenida.
ÍNDICES DE PRECIOS
Los índices de precios miden el nivel general de precios. En el capítulo anterior vimos cómo se elaboran estos indicadores. El índice de precios que corresponde a todos los bienes y servicios de la
economía es el deflactor del PIB. Como vimos en el capítulo 6, la Oficina de Análisis Económicos
no usa pesos fijos para preparar el deflactor del PIB. Muchos otros índices de precios se preparan
con pesos fijos.
El índice de precios con pesos fijos más conocido es el índice de precios al consumidor
(IPC). El IPC se creó durante la Primera Guerra Mundial como la base para ajustar los salarios de
los constructores de buques, que entonces controlaba el gobierno. Hoy, el IPC lo calcula la Oficina
de Estadísticas Laborales todos los meses con una selección de bienes que son representativos de la
“canasta” que adquiere el consumidor urbano común cada mes. Las cantidades de cada artículo
de la selección que se usan para los pesos se basan en amplias encuestas de consumidores. De hecho, la Oficina de Estadísticas Laborales reúne cada mes los precios de alrededor de 71,000 bienes
y servicios en 22,000 puntos de venta en 44 zonas geográficas. Por ejemplo, el costo de la vivienda
se incluye en la recopilación de datos luego de encuestar a casi 5,000 inquilinos y 1,000 casatenientes por mes. En la figura 7.3 se muestra la canasta del IPC de finales de 2002.
En la tabla 7.8 se muestran los valores del IPC desde 1950. Los cambios porcentuales de la
tabla se calculan con el índice (el periodo de base del índice es de tres años, de 1982 a 1984, en
lugar del más usual de un año). Por ejemplo, de 1970 a 1971 el IPC aumentó de 38.8 a 40.5%.
El cambio porcentual es de [(40.5 − 38.8)/38.8] × 100, que da 4.4%.
Recuerde que en el capítulo anterior vimos que los índices de precios con pesos fijos, como el
IPC, no toman en cuenta que los consumidores sustituyen los productos caros con otros. Para
más información, véase el apartado Para profundizar, “La economía mundial en transición: desviación del índice estadounidense de precios al consumidor; por qué podría ser importante”.
Los cambios del IPC exageran en parte los cambios del costo de la vida.
En respuesta al problema de los pesos fijos, en agosto de 2002 la Oficina de Estadísticas Laborales inició la publicación de una versión del IPC llamado índice de precios al consumidor
concatenado, que usa precios variables. Aunque esta versión no es la principal, puede ser que
dentro de unos años la oficina abandone por completo la versión de pesos fijos del IPC. Pero recuerde que aun si esto sucede, el IPC va a diferir de manera relevante del deflactor del PIB. El IPC
abarca sólo los bienes y servicios de consumo, mientras que el deflactor del PIB cubre todos los
bienes y servicios producidos por la economía. Asimismo, el IPC incluye los precios de los bienes
importados, que no están en el deflactor del PIB. En la figura 7.3 se muestra la canasta de bienes que
se usó en 2002 para calcular el IPC.
Otros índices de precios conocidos son los índices de precios al productor (IPP), que antes
se llamaban índices de precio de mayoreo. Son índices de los precios que cobran los productores
por la manufactura en todas las etapas del proceso de producción, no en la etapa final. Se calculan por separado para las diversas etapas de la producción. Las tres principales categorías son
productos terminados, materiales intermediarios y materias primas, aunque hay divisiones en cada categoría.
Una de las ventajas de algunos IPP es que detectan pronto los incrementos de precios en el
proceso de producción. Como estos movimientos anticipan cambios futuros en los precios al
consumidor, se consideran indicadores confiables de cuáles serán los precios de venta.
COSTOS DE LA INFLACIÓN
Si uno le pregunta a mucha gente por qué es mala la inflación, la mayoría respondería que reduce el nivel de vida porque los bienes y servicios son más caros. Es decir, se mella el poder de
compra de las personas. A la gente le gusta recordar los días en que una botella de Coca-Cola
costaba centavos y una hamburguesa poco dinero. Imagínese lo que podríamos comprar hoy si
no hubieran subido los precios. Pero en general, la gente no piensa en lo que ganaba en “los
buenos tiempos”. El hecho de que un refresco haya subido de 10 centavos a un dólar no significa
nada en términos reales, si la gente que antes ganaba 5,000 dólares ahora gana 50,000. ¿Por qué?
La razón es simple:
El ingreso de las personas por sueldos y salarios, utilidades, intereses y rentas aumenta con
las inflaciones. La tasa salarial es el precio de la mano de obra, la renta es el precio de las tierras, etc. En las inflaciones casi todos los precios suben juntos, incluyendo los correspondientes a los insumos, que son los que determinan los ingresos de los trabajadores y de los
dueños del capital y la tierra.
La inflación cambia la distribución del ingreso Que uno gane o pierda en un periodo de inflación depende de que su ingreso aumente más deprisa o más despacio que los precios de los bienes que uno compra. El grupo más citado cuando se habla del efecto de la inflación
es el de quienes viven de ingresos fijos. Si usted recibe un ingreso fijo y los precios suben, su capacidad de comprar bienes y servicios baja de manera proporcional. ¿Quiénes ganan ingresos fijos?
La mayoría de las personas piensan en los ancianos. Muchos jubilados que viven de pensiones privadas reciben pagos mensuales que no aumentan jamás. Sin embargo, muchos planes de
pensiones entregan pagos indexados a la inflación. Los beneficios que otorgan estos planes se incrementan automáticamente cuando sube el nivel general de precios. Si los precios suben 10%,
los beneficios también crecen 10%. La mayor fuente de ingreso para los ancianos es el seguro social. Estas prestaciones están totalmente indexadas; cuando los precios (es decir, el IPC) suben
5%, las prestaciones del seguro social también aumentan 5%.
Efecto en deudores y acreedores También se cree que los deudores se benefician a
expensas de los acreedores en una inflación. En verdad, si le presto a usted 100 dólares para que
me los pague en un año y los precios aumentan 10% entre tanto, en términos reales recibo 10%
menos de lo que le presté.
Supongamos que usted y yo anticipamos que los precios aumentarían 10%. Yo tendría que
tomar esto en cuenta en mi trato con usted. Le cobraría una tasa de interés bastante alta para
cubrir la pérdida de valor debida a la inflación prevista. Si nos ponemos de acuerdo en una tasa de interés de 15%, usted debe pagarme 115 dólares al término del año. La diferencia entre la
tasa de interés sobre un préstamo y la tasa de inflación es la tasa de interés real. En nuestro
trato, ganaré una tasa de interés real de 5%. Al cobrar una tasa de interés de 15%, he tomado
en cuenta la inflación prevista de 10%. En este sentido, no me perjudica la inflación pese al hecho de ser un acreedor, porque sigo el paso de la inflación y también obtengo una utilidad por
mi dinero.
Por otro lado, una inflación imprevista, que tome a la gente por sorpresa, perjudica a los
acreedores. Si la inflación real en el periodo del préstamo que le hice fuera de 20%, saldría menoscabado en mi calidad de acreedor. Le cobré 15% de intereses con la esperanza de obtener
un rendimiento real de 5%, pero debí cobrarle 25% para recibir la misma utilidad real de 5%.
Como la inflación fue mayor de lo esperado, mi rendimiento real fue negativo, de 5%.
Una inflación mayor que la esperada beneficia a los deudores; una inflación menor, a los
acreedores.
Costos e ineficiencias administrativas Hay costos ya incluso por anticipar la inflación. Uno es el costo administrativo por sólo estar al corriente de los precios. En la inflación acelerada de Israel a comienzos de los años ochenta, se abrió una línea telefónica de urgencia que cada
hora daba el índice de precios. Los propietarios de las tiendas tenían que recalcular los precios y
reetiquetar a menudo, lo cual requería un tiempo que pudo haber sido aprovechado mejor.
Es necesario hacer transacciones bancarias más frecuentes. Por ejemplo, las tasas de interés
aumentan con la inflación anticipada. Cuando las tasas son elevadas, también son altos los costos de oportunidad de no tener el efectivo en el banco. Por eso, la gente conserva menos efectivo
y visita más el banco (en la siguiente parte del libro estudiaremos con mayor detalle este efecto).
Además, si la gente no está bien informada o si no entiende lo que pasa con los precios en general,
puede cometer errores en sus tratos de negocios. Estos descuidos son causa de que se asignen
mal los recursos.
Mayor riesgo y menor crecimiento económico Cuando se presenta una inflación
inesperada, aumenta el riesgo de invertir en la economía. Puesto que la incertidumbre es mayor,
los inversionistas son reacios a invertir en capital y a comprometerse a largo plazo. Como baja el
monto de las inversiones, aminoran las perspectivas de crecimiento económico de largo plazo.
INFLACIÓN: ¿ENEMIGO PÚBLICO NÚMERO UNO?
Durante décadas, los economistas han debatido la gravedad de los costos de la inflación. Algunos,
entre ellos Alan Blinder, dicen: “Como todo adolescente, la inflación está muy incomprendida y
esto infla su importancia política fuera de toda proporción con su relevancia económica.”3 Phillip Cagan y Robert Lipsey afirman: “Se pensaba que con el tiempo la economía haría los ajustes
necesarios [para la inflación], pero muchos de éstos han resultado muy difíciles [. . .] Para las
instituciones financieras y los mercados, los efectos de la inflación han sido extremadamente
problemáticos.”4
Cualesquiera que sean los costos de la inflación, a la gente no le gusta. Nos vuelve inquietos e
inconformes. En 1974, el presidente estadounidense Gerald Ford expresó esta insatisfacción
cuando dijo: “Si no extirpamos la inflación, el enemigo público número uno, destruirá nuestro
país, nuestros hogares, nuestras libertades, nuestras propiedades y, en fin, nuestro orgullo nacional
tan categóricamente como un enemigo bien armado en tiempos de guerra.”5 Con este convencimiento, nuestras autoridades electas han perseguido vigorosamente políticas orientadas a detener
la inflación. Y esto nos regresa adonde empezamos. Si, como dijimos arriba, las recesiones de 1974
a 1975 y de 1980 a 1982 fueron el precio que tuvimos que pagar para detener la inflación, sale caro pararla.

menor del inventario de capital. Cuando la producción real mengua, el ingreso real baja.
4. Una depresión es una recesión prolongada y profunda, aunque
hay desacuerdos sobre qué tan grave y dilatada tiene que ser una
recesión para ser llamada depresión.
5. La tasa de desempleo es la proporción entre el número de desempleados y el número de integrantes de la fuerza de trabajo. Para
que un individuo se considere desempleado y perteneciente a la
fuerza de trabajo, debe hallarse en la búsqueda de trabajo.
6. Hay grandes diferencias en las tasas de desempleo por grupos
demográficos, regiones y sectores económicos. Por ejemplo, los
afroamericanos tienen tasas de desempleo mucho mayores que
los blancos.
7. Se considera que una persona queda excluida de la fuerza
laboral cuando ya no busca trabajo y, por lo tanto, tampoco
es desempleada. Quienes dejan de buscar empleo porque se
sienten pesimistas en cuanto a colocarse se llaman trabajadores
desmotivados.
8. Algún desempleo es inevitable. Como constantemente entran
nuevos trabajadores a la fuerza de trabajo, empresas e industrias

9. El desempleo que obedece a problemas pasajeros de correspondencia entre puestos y habilidades se llama desempleo friccional.
El desempleo que se da por cambios estructurales duraderos de
la economía se llama desempleo estructural. La tasa natural de
desempleo es la suma de las tasas por fricción y estructural.
El aumento del desempleo que ocurre en recesiones y depresiones
se llama desempleo cíclico.
10. Los principales costos de recesiones y desempleo son la reducción
de la producción real, perjuicios causados a los desempleados y
pérdida de producción futura. Los beneficios de las recesiones
son que abaten la inflación, incrementan la eficiencia y mejoran
la balanza de pagos de un país.
INFLACIÓN
11. Una inflación es un aumento del nivel general de precios.
Ocurre cuando muchos precios suben al mismo tiempo. Para
medir la inflación se calcula el alza promedio de los precios de
muchos bienes durante cierto periodo. Una deflación es una
baja del nivel general de precios. Una inflación sustancial es un
aumento del nivel general de precios que se extiende durante
un tiempo significativo.
12. Se aplican varios índices para medir el nivel general de precios.
Entre ellos están el deflactor del PIB, índice de precios al consumidor
(IPC) e índices de precios al productor (IPP).
13. Que una persona gane o pierda en una época de inflación depende
de si su ingreso aumenta o no más deprisa que los precios de los
artículos que compra. Los ancianos están más protegidos de la
inflación de lo que se piensa, porque las prestaciones del seguro
social y muchas pensiones están indexadas a la inflación.
14. La inflación mayor de lo esperado beneficia a los deudores; la
inflación menor, a los acreedores.
