o baje en el siguiente periodo?
PRODUCCIÓN TOTAL E INGRESO AGREGADO (Y)
En cada periodo, las empresas producen cierta cantidad total de bienes y servicios, que llamamos
producción total (Y). En el capítulo 6 presentamos el producto interno bruto real como medida de la
cantidad generada por la economía, Y. La producción abarca servicios, bienes de consumo y bienes
de inversión. Es importante considerar que todos éstos son componentes de la producción “real”.
Ya vimos que el PIB (Y) puede calcularse por el ingreso o por el gasto. Como todo gasto lo
recibe alguien como ingreso, para calcular el PIB (Y) sumamos el gasto total en todos los bienes
finales del periodo o sumamos todos los ingresos (salarios, rentas, intereses y utilidades) que recibieron todos los factores de la producción.
Usaremos la variable Y para denotar a la vez la producción total y el ingreso agregado porque son lo mismo visto desde dos perspectivas. Cuando la producción aumenta, se genera ingreso adicional. Se contratan más trabajadores pagados, los empleados trabajan más horas y ganan
más, y los patrones obtienen más utilidades. Cuando se reduce la producción, el ingreso baja, los
trabajadores son despedidos o cubren jornadas reducidas (por lo que ganan menos) y las utilidades menguan.
En cualquier periodo hay una igualdad exacta entre la producción total y el ingreso
agregado. Debe recordar este hecho siempre que vea el término producción total
(ingreso).

La producción total también puede ser considerada la cantidad íntegra ofrecida, porque es el
monto que las empresas abastecen (producen) en el periodo. En la exposición que sigue usamos
la frase producción total (e ingreso), en lugar de cantidad total abastecida, pero tenga presente que
son equivalentes. Recuerde también que “producción total” significa “PIB real”.
Piense en términos reales Desde el comienzo tiene que pensar en “términos reales”.
Por ejemplo, cuando hablamos de producción (Y) queremos referirnos a la producción real, no a
la nominal. Ya dijimos en el capítulo 6 que el cálculo del PIB real es complicado, pero se pueden
pasar por alto estas complejidades en el análisis que sigue. Para facilitar la lectura, a menudo usaremos valores en dinero para Y, pero no confunda Y con la producción nominal. Lo principal es
pensar en Y en términos reales: las cantidades de bienes y servicios producidos y no el dinero que
circula en la economía.
INGRESO, CONSUMO Y AHORRO (Y, C y S)
Cada periodo (un mes o un trimestre) los hogares reciben un monto total de ingresos (Y). Comenzamos nuestro análisis en un mundo simple sin gobierno y con una economía “cerrada”, es
decir, sin importaciones ni exportaciones. En ese mundo, un hogar puede hacer dos, y sólo dos,

Recuerde que el ahorro no se refiere a los ahorros totales acumulados con el tiempo. Ahorro
(sin la s al final) se refiere a la parte del ingreso de un solo periodo que no se gastó en ese momento. El ahorro (S) es el monto sumado a los ahorros acumulados de cualquier periodo. El ahorro es

EXPLICACIÓN DE LOS HÁBITOS DE GASTO
Hasta aquí no hemos dicho nada sobre el comportamiento. No hemos descrito los hábitos de
consumo y ahorro de los hogares, ni hemos especulado sobre cuánto decidirán producir en total
las empresas en determinado periodo. Para nuestro trabajo sólo tenemos un marco teórico y un
conjunto de definiciones.
Recuerde que la macroeconomía es el estudio del comportamiento. Para comprender cómo
funciona la macroeconomía, debemos entender el proceder de hogares y empresas. En nuestra
economía simple que no tiene gobierno hay dos formas de gasto: el de los hogares (consumo) y el
de las empresas (inversión).
Consumo y ahorro de los hogares ¿Cómo deciden los hogares cuánto consumir? En
determinado periodo, el monto del consumo total en la economía depende de varios factores.
Entre los determinantes del consumo total están:
1. Ingreso de los hogares
2. Riqueza de los hogares
3. Tasas de interés
4. Expectativas de los hogares sobre el futuro
Estos cuatro factores se conjuntan para determinar los hábitos de gasto y ahorro de los hogares, tanto en lo individual como en total. No es de sorprender, pues los hogares con más ingresos
y riqueza gastan más que los hogares que ganan poco y tienen menos. Las tasas de interés bajas
reducen el costo de los préstamos, así que estimulan el gasto. (Las tasas elevadas encarecen los
préstamos y aminoran el gasto.) Por último, las buenas expectativas sobre el futuro aumentan el
gasto actual, en tanto que la incertidumbre sobre el porvenir lo disminuye.
Todos estos factores son importantes, pero por ahora vamos a centrarnos en la relación entre
ingreso y consumo.1 En la Teoría general, Keynes afirma que el monto del consumo realizado por
un hogar se relaciona en forma directa con su ingreso:
Cuanto mayor sea el ingreso, más probable es que el consumo sea elevado. Las personas con
más ingreso consumen más que las otras.

La propensión marginal a consumir (PMC) es la fracción de un aumento en el ingreso que
se consume (o la fracción de una baja del ingreso que se saca del consumo). La propensión
marginal a ahorrar (PMA) es la fracción de un aumento en el ingreso que se ahorra (o una
parte de la disminución del ingreso que se extrae del ahorro).
Como C es el consumo agregado y Y el ingreso total, se sigue que la PMC es la propensión marginal
de la sociedad a consumir del ingreso nacional y que la PMA es la propensión marginal de la sociedad
a ahorrar del ingreso nacional.

INVERSIÓN PLANEADA (I)
Como hemos visto, el consumo es el gasto de los hogares en bienes y servicios, ¿pero en qué gastos incurren las empresas? La respuesta es: inversión.
¿Qué es una inversión? Empecemos con un breve repaso de términos y conceptos. En
el habla ordinaria, usamos inversión para referirnos a lo que hacemos con nuestros ahorros: “Invertí en fondos mutuos y algunas acciones de AOL”. En cambio, en el lenguaje de la economía,
inversión se refiere siempre a la formación de inventarios de capital. Para un economista, una inversión es algo producido que se destina a crear valor en el futuro.

No debe confundir los dos usos del término. Cuando una empresa construye una fábrica o
compra máquinas nuevas para su equipo, invierte. El restaurantero que adquiere mesas, sillas,
equipo de cocina y cubertería, invierte. Cuando una universidad edifica un nuevo centro deportivo, invierte. Desde ahora, usaremos el término inversión exclusivamente para referirnos a las
compras que hacen las empresas de nuevas edificaciones, equipo e inventarios, todo lo cual acrecienta su capital social.
Recuerde que las existencias son parte del inventario de capital. Cuando las empresas acrecientan sus existencias, invierten: compran algo que crea valor en el futuro. La mayor parte del
inventario de capital de una tienda de ropa consiste en existencias de prendas no vendidas en sus

almacenes, estantes y exhibidores. El servicio prestado por un supermercado o una tienda de departamentos es la comodidad de tener un gran surtido de mercancía en inventario, lista para
comprarla en un solo lugar.
En general, las empresas de manufactura tienen dos clases de inventarios: insumos y productos
finales. General Motors (GM) tiene existencias de llantas, planchas de acero, bloques de motores,
tapas de válvulas y miles de otras cosas, listas para ser usadas en la fabricación de nuevos automóviles. Además, GM tiene existencias de vehículos terminados en espera de su envío.
La inversión es una variable de flujo, pues representa las adiciones al inventario de capital en
cierto periodo. La decisión de una empresa sobre cuánto invertir cada periodo está determinada
por muchos factores. Por ahora, nos enfocaremos sólo en los efectos que ciertos niveles de inversión tienen en el resto de la economía.
Inversión real y planeada Uno de los datos más importantes de la macroeconomía es engañosamente simple: una empresa no siempre invierte al final la suma exacta que había planeado.
La razón es que no tiene el control total sobre sus decisiones de inversión, ya que también deciden
otros participantes de la economía. (Sin embargo, lo anterior no es verdad tratándose del consumo. Como suponemos que los hogares tienen el control absoluto sobre su consumo, el consumo
planeado siempre es igual al consumo real.)
Por lo común, las empresas pueden escoger cuánto comprar de plantas y equipo nuevo en
determinado periodo. Si GM quiere adquirir un robot nuevo para estampar las defensas o si
McDonald’s decide comprar otra freidora para papas a la francesa, lo hacen sin problemas. No
obstante, hay otro componente de la inversión sobre el que las compañías tienen menos control: la inversión en inventario.
Supongamos que GM espera vender un millón de automóviles este trimestre y tiene las existencias que considera apropiadas. Si la compañía produce y vende un millón de vehículos, tendrá
sus existencias en donde se encuentran ahora (en el nivel deseado). Ahora imaginemos que GM
produce un millón de automotores, pero debido a un cambio repentino del interés de sus clientes
vende sólo 900,000. Por definición, las existencias de GM tienen que ascender a 100,000 automóviles. El cambio de inventario de la empresa es igual a la producción menos las ventas. El punto es:
Un componente del inventario (el cambio de existencias) está determinado en parte
por cuánto deciden comprar los hogares, lo cual no está bajo el control estricto de las
empresas. Si los hogares no adquieren todo lo que desean las compañías, los inventarios serán mayores que lo esperado y las empresas tendrán que hacer una inversión
en inventario que no habían planeado.
Como los ajustes involuntarios al inventario no son deseados ni planeados, tenemos que distinguir entre inversión real e inversión deseada o planeada. Denotamos con I exclusivamente la
inversión deseada o planeada. En otras palabras, I se referirá a las compras planeadas de plantas y
equipos, y a los cambios de existencias planeados. Por el contrario, la inversión real es el monto real
de la inversión. Si la inversión real en existencias resulta ser mayor de lo que proyectaba la empresa, entonces la inversión real es mayor que I, la inversión planeada.
Para los fines de este capítulo, tomaremos como fijo en cierto nivel el monto de la inversión
que las empresas, en su conjunto, planean hacer en cada periodo (I). Suponemos que este nivel no
varía con el ingreso. En el ejemplo que sigue, asumiremos que I  25,000 millones de dólares,

Hasta aquí hemos descrito el comportamiento de empresas y hogares. Ahora vamos a estudiar la
naturaleza del equilibrio y a explicar cómo lo alcanza la economía.

estado de equilibrio no hay tendencias de cambio. En microeconomía se dice que un mercado
está en equilibrio (por ejemplo, el mercado de plátanos) con relación al precio cuando la cantidad demandada es igual a la cantidad ofrecida. En ese punto, proveedores y compradores están
satisfechos. El precio de equilibrio de un bien es el precio al que los proveedores quieren suministrar lo que demandan los compradores.
En macroeconomía, definimos equilibrio del mercado de bienes como el punto donde el gasto planeado total es igual a la producción total.

Así, sólo cuando la inversión planeada es igual al ahorro se establece el equilibrio.
Este método de ahorro e inversión para el equilibrio es de sentido común si recordamos dos
cosas: 1) la producción y el ingreso son iguales, y 2) el ahorro es ingreso que no se gastó. Como
no se usó, el ahorro es como una fuga en la corriente del gasto. Sólo si la fuga se compensa con
otro componente del gasto planeado, el gasto total planeado resultante equipara a la producción
total. Ese otro componente es la inversión planeada (I).

planeada de las empresas es igual al ahorro de los hogares, el gasto total planeado (GT ≡ C + I)

AJUSTES AL EQUILIBRIO
Hemos definido equilibrio y aprendimos a encontrarlo, pero no hemos dicho nada sobre cómo
reaccionan las empresas al desequilibrio. Consideremos las acciones de las empresas cuando el
gasto total planeado excede a la producción total (ingreso).
Ya sabemos que la única manera en que las empresas pueden vender más de lo que producen, consiste en vender parte de sus existencias. Esto significa que cuando el gasto total planeado


es superior a la producción total, ha habido reducciones imprevistas de existencias. Parece razonable dar por sentado que las empresas van a responder a estas disminuciones imprevistas
aumentando la producción. Si acrecientan la producción, el ingreso también debe incrementarse (la producción y el ingreso son dos maneras de medir la misma cosa). Cuando GM fabrica
más automóviles, contrata más operarios (o paga a su planta laboral por trabajar horas extras),
compra más acero, usa más electricidad, etc. Estas compras de GM representan un ingreso para
la mano de obra, los productores de acero, electricidad, etc. Si GM (y todas las demás empresas)
aumentan su producción para mantener intactas sus existencias, van a generar más ingreso en la
totalidad de la economía, lo cual fomentará el consumo. Recuerde que cuando el ingreso crece,
el consumo aumenta.
El proceso de ajuste continúa siempre que la producción (ingreso) esté por debajo del gasto total planeado. Si las empresas reaccionan a las reducciones imprevistas de las existencias
aumentando la producción, una economía con un gasto planeado mayor que la producción
se ajustará para equilibrarse y Y será mayor que antes. Si el gasto planeado es menor que la
producción, habrá incrementos inesperados en las existencias. En ese caso, la reacción de
las empresas consiste en reducir la producción. Cuando ésta baja, el ingreso disminuye, el
consumo aminora, etc., hasta que se restituye el equilibrio con Y menor que antes.
Como se aprecia en la figura 8.8, en cualquier nivel de producción de más de Y  500,000 millones, como Y  800,000, la producción bajará hasta que alcance el equilibrio en Y  500,000. En
cualquier nivel de producción inferior a Y  500,000 millones, como Y  200,000, la producción
aumentará hasta equilibrarse en Y  500,000.5
EL MULTIPLICADOR
Ahora que sabemos cómo se determina el valor de equilibrio del ingreso, nos preguntamos cómo cambia el nivel de equilibrio de la producción cuando la inversión planeada varía. Si hay un
cambio súbito en la inversión planeada, ¿reaccionará la producción de alguna manera? Como
veremos, el cambio en la producción de equilibrio es mayor que en la inversión planeada. La
producción varía en múltiplos del cambio en la inversión planeada. Por lo tanto, el múltiplo correspondiente se llama multiplicador.

El multiplicador se define como la proporción de cambio del nivel de equilibrio de la producción en relación con alguna variable autónoma. Una variable autónoma es la que se supone que
no depende del estado de la economía; es decir, si no cambia como respuesta a cambios en la economía. En este capítulo consideramos que la inversión planeada es autónoma. Esto simplifica
nuestro análisis y aporta las bases para investigaciones posteriores.
Como la inversión planeada es autónoma, podemos preguntarnos cuánto varía el nivel de
equilibrio de la producción cuando cambia aquélla. Recuerde que aquí no queremos explicar
por qué cambia la inversión planeada; sólo nos preguntamos cuánto cambia el nivel de equilibrio
de la producción cuando (por el motivo que sea) cambia la inversión planeada. (A partir del capítulo 12 no tomaremos como dada la inversión planeada y explicaremos cómo se determina.)
Consideremos un aumento sustancial en la inversión planeada de 25,000 millones de dólares; es decir, supongamos que I aumenta de 25,000 millones a 50,000 millones y se sostiene en esa
cifra. Si había un equilibrio con I  25,000 millones de dólares, un aumento de 25,000 millones
en la inversión planeada causará un desequilibrio en el que el gasto total planeado será 25,000 millones mayor que la producción total. Las empresas ven de inmediato reducciones imprevistas de
sus existencias y, como resultado, comienzan a aumentar la producción.
Digamos que el aumento de la inversión planeada viene de un incremento anticipado en los
viajes, que lleva a una aerolínea a comprar más aviones, a las compañías de renta de automóviles
a adquirir más vehículos y a las de autobuses a obtener más unidades (todos éstos son bienes de
capital). Las empresas que pasen por reducciones imprevistas de existencias serán los fabricantes
de automóviles, autobuses y aviones: GM, Ford, Boeing, etc. En respuesta a la reducción de sus inventarios, estas firmas aumentarán la producción.
Ahora supongamos que estas empresas acrecientan su producción hasta completar el incremento de 25,000 millones de dólares en inversión planeada. ¿Con esto se restituye el equilibrio?
No, porque cuando crece la producción, la gente obtiene más ingresos y gasta una parte, lo que
eleva todavía más el gasto planeado. En otras palabras, un incremento en I lleva a un aumento indirecto en C. Para construir más aviones, Boeing tiene que contratar más trabajadores o pedir a sus
empleados actuales que trabajen más horas. También tiene que comprar más motores de General
Electric, más llantas de Goodyear, etc. Los dueños de estas empresas ganarán más utilidades, producirán más, contratarán más trabajadores y pagarán más sueldos y salarios.
Este ingreso agregado no se desvanece en el aire, sino que se paga a los hogares, los cuales
gastan una parte y ahorran el resto. La producción adicional genera más ingreso, que trae
más gasto en consumo.
Si la inversión planeada (I) sube inicialmente 25,000 millones de dólares y se sostiene en este
nivel más alto, un aumento de la producción de 25,000 millones no restaura el equilibrio porque
genera más gasto de consumo (C). Las personas compran más bienes de consumo. Hay reducciones
inesperadas de las existencias de artículos de consumo básico (lavadoras, comida, ropa, etc.) y eso
incita a otras empresas a incrementar la producción. El ciclo se repite.6
La producción y el ingreso pueden crecer mucho más que el aumento inicial de la inversión
planeada, ¿pero de qué magnitud es el multiplicador? La gráfica de la figura 8.11 es la respuesta.
Supongamos que la economía está en equilibrio en el punto A, donde la producción de equilibrio
es de 500,000 millones de dólares. El aumento de I de 25,000 desplaza la curva GT ≡ C + I en
25,000, porque I es mayor por 25,000 en todos los niveles de ingreso. El nuevo equilibrio se produce
en el punto B, donde la producción de equilibrio es de 600,000. En la misma forma que el punto A,

el punto B está sobre la línea de 45° y tiene un valor de equilibrio. Entre el punto A y el B, la producción (Y) aumentó 100,000 (600,000 – 500,000), o sea, cuatro veces el aumento inicial de la inversión
planeada de 25,000. En este ejemplo el multiplicador es 4. En el punto B, el gasto total también es
100,000 mayor. Si 25,000 de estos 100,000 adicionales son inversión (I) —algo que ya sabemos—,
los restantes 75,000 son consumo agregado (C). Del punto A al punto B, entonces, ΔY  100,000,
ΔI  25,000 y ΔC  75,000.
¿Por qué el proceso del multiplicador no se extiende para siempre? La respuesta es que sólo
una fracción del aumento del ingreso se consume en cada ronda. Los incrementos sucesivos del
ingreso son menores en cada ronda del proceso de multiplicador, debido a fugas en forma de
ahorro, hasta que el equilibrio se restituye.
La magnitud del multiplicador depende de la pendiente de la recta del gasto total planeado.
Cuanto más inclinada es la pendiente, mayor es el giro en la producción dado un cambio de inversión. Cuando la inversión planeada es fija, como en nuestro ejemplo, la pendiente de la recta
GT ≡ C + I es la propensión marginal a consumir (ΔC/ΔY). Cuanto mayor es la PMC, mayor es el
multiplicador. Esto no debe causar sorpresa. Una PMC grande significa que el consumo se incrementa mucho cuando se acrecienta el ingreso. Cuanto más varía el consumo, más tiene que cambiar la producción para alcanzar el equilibrio.

Supongamos que el mercado está en equilibrio sobre un nivel de ingreso de Y  500. Ahora

EL MULTIPLICADOR EN ACCIÓN: LA RECUPERACIÓN
DE LA GRAN DEPRESIÓN
La Gran Depresión comenzó en 1930 y duró casi una década. La producción real de 1938 fue
menor que la de 1929. La tasa de desempleo nunca fue menor a 14% de la fuerza laboral entre
1930 y 1940. ¿Cómo se atascó la economía en un nivel tan bajo de ingreso y tan alto de desempleo? El modelo que analizamos en el capítulo nos servirá para dar con la respuesta.
Si las empresas no quieren hacer muchas inversiones (I es baja) o si los consumidores deciden aumentar su ahorro y reducir su consumo, el gasto planeado será escaso. Las empresas no
quieren producir más porque, con tantos trabajadores desempleados, las familias no tienen el
ingreso para comprar la producción adicional de aquéllas. Los hogares, que comprarían más si
tuvieran más ingresos, no encuentran trabajos en los que ganen más. La economía queda atrapada en un círculo vicioso.
¿Cómo se rompe el círculo? Una manera es incrementar la producción total mediante el
efecto del multiplicador para aumentar el gasto total planeado. Este acrecentamiento de GT
ocurre en forma natural o lo causa un cambio en las políticas gubernamentales.
Por ejemplo, a finales de los años treinta la economía experimentó un alza de la inversión
residencial y no residencial. Entre 1935 y 1940, el gasto total en la inversión (en términos reales)
aumentó 64% y la inversión residencial creció más del doble. No hay duda de que este incremento
de la inversión tuvo un efecto multiplicador. En apenas cinco años, el empleo en la industria de la
construcción se elevó por arriba de 400,000 puestos y en el sector manufacturero saltó más de un
millón. El empleo total ascendió a más de cinco millones. Como más trabajadores tenían ocupación, se generó más ingreso y una parte de este ingreso adicional se gastó en bienes de consumo. Las existencias bajaron y las empresas comenzaron a expandir la producción. Entre 1935 y
1940 la producción real (ingreso) aumentó más de un tercio y la tasa de desempleo bajó de 20.3
a 14.6%.
No obstante, 14.6% es una tasa muy alta de desempleo. La Gran Depresión no había terminado. Entre 1940 y 1943 ésta terminó; la tasa de desempleo cayó a 1.9% en 1943. Esta recuperación
fue iniciada por la movilización de la Segunda Guerra Mundial y el significativo incremento de

las compras gubernamentales de bienes y servicios, que pasó de 14,000 millones de dólares en
1940 a 88,600 millones en 1943. En el capítulo siguiente vamos a explorar este multiplicador del
gasto gubernamental y veremos cómo el gobierno puede estimular la economía si aumenta su gasto.

### Resumen

PRODUCCIÓN TOTAL E INGRESO ACUMULADO (Y)
1. En cada periodo, las empresas producen una cantidad total de
bienes y servicios llamada producción total (Y). Como todo gasto
lo recibe alguien como ingreso, la producción total y el ingreso
total son lo mismo.
2. El monto del consumo total que tiene lugar en cualquier periodo
depende de factores como el ingreso y la riqueza de los hogares,
las tasas de interés y lo que esperan los hogares sobre el futuro.
3. Si los impuestos son de cero, los hogares hacen sólo dos cosas
con su ingreso: lo gastan en consumo o lo ahorran. C representa
al consumo total de los hogares y S, su ahorro total. Por definición, el ahorro es igual al ingreso menos el consumo: S ≡ Y – C.
4. Cuanto mayor es el ingreso de las personas, más alto tiende a
ser su consumo. Esto también es verdad para la economía en su
conjunto: hay una relación positiva entre el consumo total (C)
y el ingreso total (Y).
5. La propensión marginal a consumir (PMC) es la fracción de un
cambio en el ingreso que se consume o se gasta. La propensión
marginal a ahorrar (PMA) es la fracción de un cambio en el
ingreso que se ahorra. Como todo el ingreso se ahorra o se
gasta, PMA + PMC ≡ 1.
6. La principal forma del gasto que realizan las empresas es la
inversión. En términos precisos, la inversión se refiere a la compra que hacen las firmas de nuevas edificaciones y equipo y la
adición de existencias. Todo esto forma el inventario de capital
de las empresas.
7. La inversión real se distingue de la inversión deseada o planeada
porque los cambios en las existencias de las empresas son parte
de la inversión real y no están bajo el control total de éstas. Los
cambios de existencias están determinados, en parte, por cuánto
deciden comprar las compañías. I denota nada más la inversión
planeada.
PRODUCCIÓN TOTAL DE EQUILIBRIO (INGRESO)
8. En una economía en la que el gasto del gobierno y las exportaciones netas son cero, el gasto total planeado (GT) es igual al consumo más la inversión planeada: GT ≡ C + I. Se alcanza el equilibrio
del mercado de bienes cuando el gasto total planeado es igual a la
producción total: C + I  Y. Esto ocurre si, y sólo si, la inversión
planeada y la real son iguales.
9. Como el ingreso total debe ahorrarse o gastarse, la condición de
equilibrio Y  C + I puede reescribirse como C + S  C + I, o
bien S  I. Sólo si la inversión planeada es igual al ahorro habrá
equilibrio. Éste es el método de ahorro e inversión para el equilibrio
o el método de fugas e inyecciones.
10. Cuando el gasto total planeado excede a la producción total
(ingreso) hay una baja imprevista en las existencias, entonces las
empresas acrecientan la producción. Ésta favorece el ingreso y
refuerza el consumo. El proceso continúa siempre que la producción (ingreso) sea menor que el gasto total planeado. Si las
empresas reaccionan a las reducciones inesperadas de los inventarios aumentando la producción, una economía con un gasto
planeado mayor que la producción se ajustará al nuevo
equilibrio y Y será mayor que antes.
EL MULTIPLICADOR
11. La producción de equilibrio cambia por un múltiplo del cambio
de la inversión planeada o de cualquier otra variable autónoma. El
multiplicador es 1/PMA.
12. Cuando los hogares aumentan su ahorro planeado, el ingreso se reduce y el ahorro no cambia. El ahorro no se incrementa porque, en
equilibrio, debe ser igual a la inversión planeada, que es fija. Si la inversión planeada también aumenta, se evita esta paradoja del ahorro
y se alcanza un nuevo equilibrio en un nivel superior de ahorro e ingreso. Este resultado depende de la existencia de un canal por el que
el ahorro adicional de los hogares financie nuevas inversiones.

## El gobierno y la política fiscal

Ningún tema macroeconómico ni microeconómico suscita tantas polémicas como la participación del gobierno en la economía.
En microeconomía, la intervención activa del gobierno para regular
la competencia, construir caminos,
proporcionar educación y redistribuir el ingreso es aplaudida por
quienes creen que el libre mercado
no funciona bien si se deja a sus propios medios. Quienes se oponen a la
participación del gobierno dicen que
es éste, y no el mercado, el que funciona mal. Aseguran que la burocracia y
la ineficiencia se eliminarían o se reducirían si el gobierno interviniera
menos en la economía.
En macroeconomía, la polémica sobre lo que el gobierno puede y debe hacer ha seguido en
el mismo tenor, aunque los problemas son diferentes. De un lado se encuentran los keynesianos
y sus descendientes intelectuales, quienes piensan que la macroeconomía tendría muchas fluctuaciones si se librara a sus fuerzas y que el gobierno tiene que paliar las fluctuaciones del ciclo
comercial. Estas ideas se remontan al análisis de Keynes de la Teoría general, la cual indica que en
depresiones y recesiones los gobiernos pueden valerse de sus facultades de gasto y recaudación
de impuestos para aumentar el gasto total (a fin de estimular la producción total). En el otro
bando se encuentran quienes aseveran que el gasto del gobierno es incapaz de estabilizar la economía y que, peor aún, es desestabilizador y perjudicial.
Quizá lo único en lo que esté de acuerdo la mayoría es que, quiérase que no, los gobiernos
son actores importantes en la economía de casi todos los países. Sólo por este motivo vale la pena
dedicar tiempo a analizar el influjo del gobierno en el funcionamiento de la macroeconomía.
El gobierno cuenta con varias facultades: regular la entrada y salida de las empresas a los sectores económicos, declarar las normas de calidad de los productos, fijar salarios mínimos y normar la publicación de informes; pero en macroeconomía lo estudiamos como si tuviera poderes
generales limitados. En particular, el gobierno afecta la macroeconomía a través de dos paquetes
de políticas: fiscales y monetarias. La política fiscal, que es la materia de este capítulo, se refiere
al gasto y los esquemas impositivos del gobierno; en otras palabras, sus normas presupuestales.
(La palabra fiscal viene de la raíz fisc-, que se refiere al “tesoro” del gobierno.) En general, la política fiscal se divide en tres categorías: 1) políticas relativas a las compras gubernamentales de
bienes y servicios, 2) políticas relativas a los impuestos, y 3) políticas concernientes a los pagos
de transferencia (como indemnización por desempleo, prestaciones del seguro social, pagos de
asistencia y prestaciones a los veteranos de guerra) a los hogares. La política monetaria, que estudiaremos en los siguientes dos capítulos, se refiere a las líneas del banco central (en Estados
Unidos, la Reserva Federal) sobre el circulante de la nación.

EL GOBIERNO EN LA ECONOMÍA
Dada la extensión y el poder de los gobiernos locales, estatales y federal, hay aspectos sobre los
que ejercen mucho control y otros que quedan fuera de su alcance. Tenemos que distinguir entre
las variables donde el gobierno interviene de manera directa y las que son consecuencia de sus decisiones, combinadas con el estado de la economía.
Por ejemplo, el gobierno regula las tasas fiscales. Por ley, el Congreso tiene la autoridad para decidir quién y qué debe ser gravado y con qué tasa. Por otro lado, los ingresos gravables no
están sujetos al control total del gobierno. Los ingresos del sistema de impuestos para personas
físicas dependen de las cargas tributarias individuales (que establece el Congreso) y del ingreso
del sector de los hogares (que depende de muchos factores que no están bajo el control directo
del gobierno, como cuánto quieren trabajar las familias). Los ingresos por las utilidades de las
corporaciones dependen tanto de las tasas fiscales de esta categoría como de la magnitud de tales ganancias.
Los gastos del gobierno dependen también de sus decisiones y del estado de la economía. Por
ejemplo, en Estados Unidos el programa de seguro de desempleo les entrega beneficios a los desocupados. Cuando la economía entra en una recesión, aumenta el número de desempleados, lo
mismo que el monto gubernamental de las retribuciones por desempleo.
Como los impuestos y los gastos suben y bajan en respuesta a los cambios de la economía, y
no como resultado de decisiones deliberadas por parte de las autoridades, en ocasiones usaremos
el término política fiscal discrecional para referirnos a los cambios de impuestos o gastos que son
consecuencia de cambios premeditados de las políticas gubernamentales.

En el capítulo anterior exploramos el nivel de equilibrio de la producción nacional en una economía simple (sin impuestos, ni gasto gubernamental, ni exportaciones) para dar una idea general de
cómo opera la macroeconomía. Para ser más veraces, tenemos que considerar una economía en la
que el gobierno es un participante activo. Empezaremos por sumar el sector público al modelo

otra variable que vamos a considerar son las compras del gobierno de bienes y servicios (G).
En nuestra exposición anterior del consumo de los hogares no consideramos los impuestos.

Debido a que los gobiernos gastan en bienes y servicios, tenemos que ampliar nuestra definición de gasto total planeado (GT) como la suma del gasto de consumo de los hogares (C), la inversión planeada de las empresas (I) y las compras gubernamentales de bienes y servicios (G).1
GT ≡ C + I + G
El déficit presupuestario es la diferencia entre lo que gasta (G) el gobierno y los impuestos

déficit presupuestario ≡ G − T
Si G excede a T, el gobierno tiene que pedir prestado al público para financiar el déficit. Para ello

que recauda en impuestos, tiene un superávit. Un superávit presupuestario es un déficit presupuestario negativo.
Los impuestos en la función de consumo En el capítulo 8 examinamos los hábitos
de consumo de los hogares y observamos que el consumo total (C) depende del ingreso total (Y):
en general, cuanto mayor es el ingreso, más grande es el consumo total. A título de ilustración,
usemos una función lineal de consumo específica:
C = a + bY
donde a es el monto del consumo que tendría lugar si el ingreso nacional fuera de cero y b es la
propensión marginal a consumir.
Tenemos que modificar la función del consumo porque hemos sumado el gobierno a la economía. Con los impuestos como parte del paisaje, es lógico suponer que el ingreso disponible
(Yd) determina los hábitos de consumo y no el ingreso antes de impuestos (Y). Si usted gana un
millón de dólares, pero tiene que pagar 950,000 de impuestos, no dispone de más ingresos que
quien gana 50,000 dólares y no paga impuestos. Lo que uno tiene para gastar en consumo real es
su ingreso disponible, no el ingreso antes de impuestos.
Para modificar nuestra función del consumo total a modo que incorpore el ingreso disponible
en lugar del ingreso antes de impuestos, escribimos, en sustitución de C = a + bY, lo siguiente:
C = a + bYd
o bien
C = a + b(Y − T)
Ahora, nuestra función el consumo depende del ingreso disponible y no del ingreso antes de impuestos.
Inversión ¿Qué pasa con la inversión? El gobierno incide en las actividades de inversión mediante su tratamiento de la depreciación y otras políticas fiscales. La inversión también varía con
las condiciones económicas y las tasas de interés, como veremos luego. Para nuestros fines del
momento, suponemos todavía que la inversión planeada (I) es autónoma.
PRODUCCIÓN DE EQUILIBRIO: Y = C + I + G
El capítulo 8 nos dice que se alcanza el equilibrio cuando Y = GT; es decir, cuando la producción
total es igual al gasto total. Recuerde que el gasto total planeado de una economía con gobierno es
GT ≡ C + I + G, así que la condición de equilibrio es:
condición de equilibrio: Y = C + I + G
Aquí también se aplica el análisis del equilibrio que hicimos en el capítulo 8. Si la producción
(Y) excede al gasto planeado (C + I + G), habrá un incremento imprevisto de las existencias: la inversión real superará a la inversión planeada. Por el contrario, si C + I + G es mayor que Y, habrá
una reducción inesperada en las existencias.

Si Y = 1,300, entonces Yd = 1,200, C = 1,000 y el gasto total planeado es de 1,200. Aquí el

Supongamos que usted es el jefe de asesores en economía del presidente y que la economía se encuentra asentada en la producción de equilibrio de la figura 9.2. La producción y el ingreso se
generan a un ritmo de 900,000 millones de dólares por año y el gobierno compra 100,000 millones de dólares en bienes y servicios cada año, que financia con 100,000 millones de impuestos.
El presupuesto está equilibrado. Además, el sector privado invierte (produce bienes de capital) a
una tasa de 100,000 millones de dólares al año.
En la Oficina Oval, el presidente le dice: “El desempleo está muy elevado. Tenemos que aumentar la producción y el ingreso para reducirlo.” Después de algunos estudios, usted determina
que se alcanzaría una tasa de desempleo aceptable sólo si la producción total aumenta a 1.100
billones de dólares.
Usted tiene que determinar cómo puede el gobierno ejercer la política de impuestos y gastos
(la política fiscal) para aumentar el nivel de equilibrio de la producción nacional. Supongamos
que el presidente ha informado que los impuestos deben quedarse en los niveles actuales (porque
el Congreso acaba de aprobar un paquete importante de reforma fiscal), así que durante varios
años estará fuera de la menor consideración aumentar T. Entonces, sólo queda G. Su única opción es aumentar el gasto del gobierno, dejando constantes los impuestos.
Para aumentar el gasto sin elevar los impuestos (entradas de las autoridades para gastar), el
gobierno tiene que pedir prestado. Cuando G es mayor que T, el gobierno tiene un déficit y la diferencia entre G y T debe ser cubierta con un préstamo. Por el momento, vamos a ignorar el posible
efecto del déficit y nos centraremos sólo en el efecto de aumentar G y dejar a T constante.
Mientras tanto, el presidente espera su respuesta. ¿Qué aumento del gasto se necesitaría para
generar un incremento de 200,000 millones de dólares en el nivel de equilibrio, de modo que pase de 900,000 millones a 1.100 billones y que se reduzca el desempleo a un nivel aceptable para el
presidente?
Quizá se sienta tentado a decir que como tenemos que acrecentar el ingreso en 200 (1,100 −
900), debemos incrementar el gasto gubernamental en la misma cantidad, ¿pero qué sucedería?
El aumento en el gasto del gobierno desequilibrará la economía. Como G es un componente del
gasto total, el gasto total planeado aumentará 200. El gasto planeado será mayor que la producción,
las existencias serán menores que lo planeado y las empresas tendrán un incentivo para aumentar
la producción. Supongamos que la producción crece en la suma deseada de 200. Tal vez piense:
“Subimos el gasto 200 y la producción en 200, con lo que se restablece el equilibrio.”
Pero hay más que traer a colación. En cuanto la producción aumenta, la economía genera
más ingresos. Ése era el efecto deseado: la creación de más empleos. Los trabajadores con nuevas
ocupaciones son también consumidores y una parte de sus ingresos se gasta. Con más gasto de
consumo, el gasto planeado es mayor que la producción, las existencias bajan más de lo que se anticipaba y las compañías elevan la producción, lo cual vuelve a incrementar el ingreso. Esta vez las
empresas responden al nuevo gasto de consumo. Entonces, el ingreso total ya es superior a 1,100.
La historia debe sonarle conocida: se trata del multiplicador en acción. Ahora es el gasto gubernamental (G) el que cambió, en lugar de la inversión planeada (I), pero el efecto es el mismo
del multiplicador que estudiamos en el capítulo 8. Un aumento en el gasto del gobierno tiene la
misma repercusión en el nivel de equilibrio de la producción y el ingreso que un incremento en
la inversión planeada. Un dólar de gasto adicional en G o en I es idéntico por lo que respecta a
su efecto en la producción de equilibrio. La ecuación del multiplicador de gasto del gobierno es
la misma que la ecuación del multiplicador de un cambio en la inversión planeada:
multiplicador del gasto del gobierno ≡ 1
PMA
En el apéndice A de este capítulo derivamos de forma algebraica el multiplicador del gasto del
gobierno.
Formalmente, el multiplicador del gasto del gobierno se define como la proporción entre el
cambio del nivel de equilibrio de la producción y una variación en el gasto gubernamental. Es la
misma definición que usamos en el capítulo anterior, pero ahora la variable autónoma es el gasto
del gobierno en lugar de la inversión planeada.
Recuerde que pensábamos en aumentar el gasto gubernamental (G) en 200. Podemos aplicar
el análisis del multiplicador para ver cuál sería el nuevo nivel de equilibrio de Y con un incremento de G de 200. El multiplicador de nuestro ejemplo es 4 (como b, la PMC, es 0.75, la PMA debe
ser 1 - 0.75 = 0.25 y 1/0.25 = 4). Así, Y aumentará 800 (4 × 200). Como el nivel inicial de Y era
900, el nuevo nivel de equilibrio de Y es 900 + 800 = 1,700 cuando G aumenta 200.
El nivel de 1,700 es mucho mayor que 1,100 y lo hemos calculado como necesario para bajar
el desempleo a la cifra deseada. Regresemos, entonces. Si queremos que Y aumente 200 y si el
multiplicador es 4, necesitamos que G aumente nada más 200/4 = 50. Si G cambia 50, el nivel de
equilibrio de Y cambiará 200 y el nuevo valor de Y será de 1,100 (900 + 200), como se quería.
Si examinamos la tabla 9.2, verificamos nuestra respuesta para estar seguros de que se alcanzó un equilibrio. Veamos primero el equilibrio anterior de 900. Cuando las compras del gobierno (G) eran de 100, la producción total (ingreso) era igual al gasto total planeado (GT ≡ C +

baja imprevista de las existencias y la producción aumenta, ¿pero cuánto? El multiplicador indica que el ingreso de equilibrio crecerá cuatro veces el cambio de 50 en G. Así, Y debe aumentar

100 (el nivel original de G) + 50 (la cifra adicional de G que es resultado del cambio en la política fiscal) = 150, entonces C + I + G = 850 + 100 + 150 = 1,100. Y = GT y la economía está en
equilibrio.

MULTIPLICADOR DE LOS IMPUESTOS
Recuerde que la política fiscal comprende las medidas relativas al gasto del gobierno y también
la fijación de impuestos. Para ver el efecto en la economía de un cambio en las políticas fiscales,
imaginemos lo siguiente. Usted aún es jefe de los asesores económicos del presidente, pero ahora le piden que conciba un plan para reducir el desempleo a un nivel aceptable sin aumentar el
gasto gubernamental. En su plan, en lugar de incrementar el gasto del gobierno (G) usted decide reducir los impuestos y sostener el gasto actual. Una disminución de impuestos acrecienta el
ingreso disponible, lo cual acrecentará el gasto de consumo. (Recuerde nuestra regla general de
que al elevar el ingreso aumenta el consumo.) ¿La reducción de impuestos afectaría la producción total (ingreso) de la misma manera que un incremento en G?
Una reducción de los impuestos aumentaría el ingreso. El gobierno no gasta menos que antes de la reducción y los hogares tienen un ingreso disponible mayor, después de descontar los
impuestos. Esto refuerza el consumo. El gasto total planeado aumenta, de modo que hay menos
existencias de lo planeado y así se incrementa la producción. Cuando ésta aumenta, se contrata a
más trabajadores y se genera más ingreso, lo cual da lugar a otra ronda de fomento al consumo,
etc. Así, el ingreso crece en un múltiplo de la reducción de los impuestos, pero tiene su truco.
El multiplicador de un cambio en los impuestos no es el mismo que el multiplicador de un
cambio en el gasto del gobierno.
¿Por qué el multiplicador de los impuestos (la proporción entre el cambio del nivel de equilibrio
de la producción y el cambio en los impuestos) es distinto que el multiplicador del gasto? Para
dar la respuesta, tenemos que comparar cómo se difunden por la economía una reducción de impuestos y un incremento del gasto.
Veamos la figura 9.1. Cuando el gobierno aumenta el gasto, hay un efecto directo e inmediato
en el gasto total de la economía. Como G es un componente del gasto total planeado, un incremento de G lleva a un crecimiento idéntico en este último. Cuando se reducen los impuestos no
hay un efecto directo en el gasto. Los impuestos entran en el cuadro sólo porque tienen un efecto
en el ingreso disponible de los hogares, el cual influye en el consumo de las familias (que es parte
total del gasto). Como se aprecia en la figura 9.1, la rebaja de impuestos fluye por los hogares antes
de afectar el gasto total.
Supongamos que el gobierno decide reducir los impuestos un dólar. ¿Cuánto aumentaría el
gasto? Ya conocemos la respuesta. La propensión marginal a consumir (PMC) indica cuánto cambia el gasto de consumo cuando el ingreso disponible varía. En el ejemplo que hemos estudiado
en el capítulo, la propensión marginal a consumir a partir del ingreso disponible es de 0.75. Esto
significa que si el incremento del ingreso después de impuestos de los hogares aumenta un dólar,

MULTIPLICADOR DEL PRESUPUESTO EQUILIBRADO
Hemos estudiado 1) el cambio del gasto del gobierno sin variación en los impuestos, y 2) el cambio de
impuestos sin variación en el gasto gubernamental. ¿Qué pasa si el gasto del gobierno y los impuestos
aumentan la misma cantidad? Es decir, ¿qué pasa si el gobierno decide pagar su gasto adicional incrementando los impuestos en la misma medida? El déficit presupuestario del gobierno no cambiaría,
porque el crecimiento de los gastos estaría equiparado por un incremento en la recaudación.
Quizá piense que en este caso los aumentos iguales en el gasto del gobierno y los impuestos no
tienen ningún efecto en el ingreso de equilibrio. Después de todo, el gasto gubernamental adicional
corresponde al monto añadido de impuestos recaudados por el gobierno. Pero no ocurre así. Por
ejemplo, tomemos un incremento de 40,000 millones de dólares en el gasto gubernamental. Por el
análisis anterior sabemos que un aumento de 40 en G, con los impuestos (T) constantes, debe elevar
el nivel de equilibrio del ingreso en 40 × el multiplicador del gasto del gobierno. El multiplicador es
1/PMA, o sea, 1/0.25 = 4. El nivel de equilibrio del ingreso debe aumentar 160 o (40 × 4).
Ahora supongamos que en lugar de mantener constantes los impuestos, financiamos el alza
de 40 en el gasto del gobierno con un incremento semejante de impuestos, a fin de mantener equilibrado el presupuesto. ¿Qué pasa con el gasto total con un aumento en G y en T? Hay dos efectos
iniciales. El primero, que el gasto del gobierno crece 40. Este efecto es directo, inmediato y positivo.
El gobierno recauda 40 más de impuestos. El aumento de impuestos tiene un efecto negativo en el
gasto general de la economía, pero no compensa del todo el aumento del gasto gubernamental.
El efecto final de un incremento de impuestos en el gasto total depende de cómo reaccionan los
hogares. Hasta ahora, lo único que sabemos sobre su comportamiento es que gastan 75 por ciento
de su ingreso adicional y ahorran 25 por ciento. Sabemos que cuando el ingreso disponible baja, se
reducen el consumo y el ahorro. Un aumento de impuestos de 40 disminuye el ingreso disponible
en 40 y el consumo promedio desciende 40 × PMC. Como PMC = 0.75, el consumo baja 30, es decir, (40 × 0.75). Al principio, el resultado neto es que el gasto del gobierno aumenta 40 y el gasto de
consumo baja 30. El gasto total sube 10 después de incrementos simultáneos de presupuesto equilibrado en G y en T.
Por consiguiente, un aumento de presupuesto equilibrado en G y T aumenta la producción,
¿pero cuánto? ¿De qué magnitud es el multiplicador del presupuesto equilibrado? La respuesta lo
sorprenderá:
multiplicador del presupuesto equilibrado ≡ 1

Vamos a reunir lo que sabemos sobre el multiplicador de impuestos para explicar lo anterior.
Para averiguar el efecto final del aumento simultáneo en el gasto del gobierno y en los impuestos
netos, tenemos que sumarles a los dos los efectos del multiplicador. El multiplicador del gasto del
gobierno es 1/PMA. El multiplicador de los impuestos es −PMC/PMA. Su adición es (1/ PMA) +
(−PMC/PMA) ≡ (1 − PMC)/PMA. Dado que PMC + PMA ≡ 1, entonces 1 − PMC ≡ PMA. Esto
significa que (1 − PMC)/PMA ≡ PMA/PMA ≡ 1. (También en el apéndice A derivamos el multiplicador del presupuesto equilibrado.)
Para volver a nuestro ejemplo, recuerde que con el multiplicador del gasto del gobierno un
incremento de 40 en G daría un aumento en la producción de equilibrio de 160 (40 × el multiplicador del gasto gubernamental de 4). Al usar el multiplicador de impuestos, sabemos que un
alza de 40 reduce el nivel de equilibrio de la producción en 120 (40 × el multiplicador de impuestos, −3). El efecto neto es 160 menos 120, o sea 40. Por consiguiente, debe resultar claro que el
efecto en Y de equilibrio es igual al aumento equilibrado en G y T. En otras palabras, el incremento neto del nivel de equilibrio de Y que resulta de un cambio en G y en T es exactamente de
la magnitud del mismo cambio inicial en G y en T.
Si el presidente quiere aumentar 200 a Y sin aumentar el déficit, conseguiría un incremento
simultáneo de 200 en G y T. Para ver por qué, revise las cifras de la tabla 9.3. En la tabla 9.1 vimos
un nivel de equilibrio de la producción de 900. Si se suma 200 a G y T, el nuevo equilibrio es de
1,100: 200 más alto. En ningún otro nivel de Y encontramos que (C + I + G) = Y.
Un incremento en el gasto gubernamental tiene un efecto inicial directo en el gasto total planeado; un alza de impuestos no lo tiene. El efecto inicial del aumento de impuestos es que los
hogares reducen el consumo en PMC por el cambio en los impuestos. Este cambio del consumo es menor que el giro en los impuestos, porque la PMC es menor que 1. Por consiguiente,
el estímulo positivo del aumento del gasto del gobierno es mayor que el estímulo negativo de
subir los impuestos. El efecto neto es que el multiplicador del presupuesto equilibrado es 1.

EL PRESUPUESTO FEDERAL
Como la política fiscal consiste en la manipulación de temas del presupuesto federal, tenemos
que considerar los aspectos del presupuesto que sean relevantes para nuestro estudio de la macroeconomía. El presupuesto federal es un documento en extremo complicado, de cientos de
páginas cada año. Ahí se detallan todos los rubros en los que el gobierno proyecta gastar, además
de sus fuentes de ingreso para el año siguiente. Es el producto de complejas influencias recíprocas de fuerzas sociales, políticas y económicas.
En realidad no es un solo “presupuesto”, sino tres. En primer lugar, es un documento político en
el que se dispensan favores a ciertos grupos o regiones (los ancianos se benefician del seguro social;
los agricultores, de los precios agrícolas de garantía; los estudiantes, de los programas federales de
préstamos, etc.) y se aumenta la carga (impuestos) sobre otros. En segundo lugar, es un reflejo de las
metas que el gobierno quiere alcanzar. Por ejemplo, además de ayudar a los agricultores, los precios
de garantía tienen la intención de preservar la “granja familiar”. Las facilidades fiscales para corporaciones que realizan investigación y desarrollan nuevos productos tienen por objeto alentar las investigaciones. Por último, el presupuesto puede ser la materialización de ideas sobre si el gobierno debe
dirigir la macroeconomía. Los aspectos macroeconómicos del presupuesto son sólo una parte de una
materia más compleja, que estaría más en el campo de los politólogos que de los economistas.

EL SUPERÁVIT O DÉFICIT
La diferencia entre las entradas y los gastos del gobierno federal son el superávit (+) o el déficit (−)
federal, que es el ahorro neto del gobierno federal. En la tabla 9.5 se muestra que el gobierno federal
gastó más de lo que recibió en 2004, de lo que resultó un déficit de 406,500 millones de dólares.
El déficit federal fue grande desde comienzos de los años ochenta hasta mediados de los noventa. Esta tendencia se aprecia en la figura 9.4, donde se traza la gráfica del superávit o el déficit del
gobierno federal como porcentaje del producto interno bruto (PIB) para 1970 I-2005 II. Como se
muestra en la figura, el presupuesto tuvo un déficit hasta 1998 I, cuando pasó a superávit. Salvo por
un periodo a mediados de los años setenta, el déficit fue bastante pequeño hasta comienzos de la
década de los ochenta, cuando llegó a 5.4 por ciento del PIB en 1982 IV. El presupuesto tuvo una
mejora acelerada de 1992 a 2000, pero entonces comenzó a caer también deprisa. En el tercer trimestre de 2003 el déficit fue de 4.1 por ciento del PIB. Desde entonces, ha bajado lentamente.
¿Cómo se produjeron esos déficits tan grandes en los años ochenta? Hay varias razones. En
primer lugar, las compras del gobierno como porcentaje del PIB aumentaron a comienzos de esa
década y luego se mantuvieron altas. Este incremento fue, sobre todo, reflejo de la acumulación
de gastos de defensa en los años de Ronald Reagan. En segundo lugar, los pagos de intereses como
porcentaje del PIB se acrecentaron de manera sustancial. En tercero, las tasas de los impuestos a
las personas físicas bajaron como resultado de la Ley Fiscal de Recuperación Económica de 1981.
Como los gastos de defensa y los pagos de intereses aumentaron muy rápidamente y los impuestos
bajaron, no es de sorprender que el déficit haya aumentado de manera sustancial en los años
ochenta. El gobierno gastaba mucho más de lo que recaudaba en impuestos.
En las dos administraciones de William Clinton (1993-2000) el déficit se convirtió en superávit y en su punto más alto en el primer trimestre de 2000 llegó a 2.3 por ciento del PIB. Parte
de este cambio se debió a aumentos en la tasa impositiva que se aprobaron a comienzos del primer periodo gubernamental de Clinton. El paso rápido de superávit a déficit a comienzos de 2001
se debió, en parte, a varios recortes importantes de impuestos aprobados en el régimen de George
W. Bush y al aumento del gasto gubernamental en Irak.
LA DEUDA
Cuando el gobierno incurre en un déficit, tiene que pedir dinero para financiarlo. Para tomar
préstamos, el gobierno federal vende títulos gubernamentales al público. Expide documentos en
los que promete pagar cierta cantidad en el futuro, con intereses. A cambio, recibe fondos de los
compradores de los documentos y con el dinero paga sus cuentas. Estos préstamos aumentan la
deuda federal, que es el débito absoluto del gobierno. La deuda federal es la suma de los déficits
acumulados menos los superávits.

Algunos de los títulos que emite el gobierno acaban en sus propias manos, o sea, en la Reserva Federal o en fondos mutuos gubernamentales. El término deuda federal privada se refiere
sólo a la deuda del gobierno estadounidense que está en manos de particulares. A finales de julio
de 2005, la deuda federal era de 7.9 billones de dólares, de los cuales 4.6 billones pertenecían al
sector privado.
Dados los grandes déficits que tuvo el gobierno federal en los años ochenta, no debe causar sorpresa que la deuda federal tuviera un incremento desproporcionado en esos años. Se observa este
cambio en la figura 9.5, donde se da la gráfica de la deuda federal privada como porcentaje del PIB
para el periodo 1970 I-2005 II. La deuda aumentó muy rápido entre 1982 y 1993: de 19.0 por ciento
del PIB en 1981 IV a 47.5 por ciento en 1993 IV. De mediados de los noventa a 2000 la deuda bajó de
manera notable como porcentaje del PIB. Desde el año 2000 el porcentaje se ha elevado un poco.
INFLUENCIA DE LA ECONOMÍA
EN EL PRESUPUESTO DEL GOBIERNO
Algunas partes del presupuesto del gobierno dependen del estado de la economía, sobre la cual
no tiene un control directo. A continuación veremos algunas formas en que la economía afecta el
presupuesto.
LA RECAUDACIÓN FISCAL DEPENDE DEL ESTADO DE
LA ECONOMÍA
Consideremos el lado de los ingresos del presupuesto. El gobierno aprueba leyes que fijan las tasas
y las categorías fiscales, variables que el gobierno controla. Por otro lado, la recaudación fiscal depende del ingreso gravable y éste depende del estado de la economía, que el gobierno no controla.
A su vez, el gobierno puede fijar una tasa de impuesto a las personas físicas de 20 por ciento, pero
las entradas que generen los impuestos dependen del ingreso promedio ganado por los hogares.
El gobierno recauda más si el ingreso promedio es de 60,000 dólares que si es de 40,000.
ALGUNOS GASTOS GUBERNAMENTALES DEPENDEN DEL ESTADO
DE LA ECONOMÍA
Algunos rubros del gasto en el presupuesto gubernamental también dependen del estado de la
economía. Conforme la economía se expande, el desempleo disminuye y el resultado es una
mengua de las prestaciones por desempleo. Los pagos de asistencia social y la repartición de vales de comida también se reducen en alguna medida. Las personas que reciben estas prestaciones
en tiempos adversos consiguen trabajo cuando el estado de la economía mejora y empiezan a

ganar suficientes ingresos, de modo que pierden aquel derecho. Los pagos de transferencia bajan de
manera automática en una expansión. (En una baja de la economía, los pagos de transferencia
aumentan porque hay más personas desocupadas y más pobres en general.)
Otro motivo de que el gasto gubernamental no se controle por completo es que la inflación
tiende a elevarse cuando la economía se expande. Esto lleva al gobierno a gastar más de lo que había planeado. Supongamos que el gobierno encargó 20 aviones de dos millones de dólares cada
uno y que la inflación hace que el precio sea mayor que lo esperado. Si de todos modos el gobierno
decide comprar las naves, tendrá que aumentar su gasto. Por último, cualquier cambio de las tasas de interés modifica los pagos de intereses del gobierno. Un incremento de las tasas de interés
hará que el gobierno gaste más en estos pagos.
ESTABILIZADORES AUTOMÁTICOS
Cuando la economía se expande, aumentan las entradas del gobierno. Además, los pagos de
transferencia bajan, lo cual reduce los gastos del gobierno. Los rubros de entradas y gastos que se
modifican con los cambios de la actividad económica, de manera tal que moderen los cambios
del PIB, se llaman estabilizadores automáticos. Cuando la economía se expande o contrae, se verifican cambios “automáticos” en los ingresos y egresos del gobierno que reducen los cambios del
PIB o los estabilizan.
En una expansión, el hecho de que algunas entradas aumenten y algunos gastos bajen automáticamente significa que el superávit del gobierno es mayor —o que el déficit es menor— de lo
que sería en otras circunstancias. Supongamos que queremos evaluar si un gobierno aplica una
política dirigida a aumentar el gasto y el ingreso. Si sólo examinamos la magnitud del déficit
presupuestal del gobierno, podríamos cometer el error de pensar que éste trata de estimular a la
economía cuando, de hecho, la verdadera causa del déficit es una baja de la economía que hizo
que las entradas bajaran y los pagos de transferencia aumentaran.
LASTRE FISCAL
Si la economía prospera, el ingreso es alto, lo mismo que la recaudación. Y ésta aumenta con el
ingreso por dos razones. En primer lugar, si la gente gana más, hay más ingreso que gravar. Segundo, cuando la gente gana más, pasa a categorías superiores y acrecienta la tasa fiscal promedio. Este incremento de las tasas fiscales es un lastre fiscal porque el aumento de la
tasa promedio que ocurre cuando la gente pasa a categorías más elevadas actúa como “lastre”
de la economía. Cuando la economía se expande y el ingreso crece, se echa a andar el mecanismo
del incremento automático de los impuestos que está inserto en el sistema. Las tasas fiscales suben,
se reduce el salario después de impuestos y se desacelera la expansión.
Antes de 1982 la gente pasaba a categorías fiscales superiores nada más por la inflación.
Supongamos que mi ingreso aumentó 10 por ciento en 1981, pero que el nivel de precios también
subió 10 por ciento ese año. Mi ingreso no creció en términos reales, pero como las categorías fiscales no se legislaban en términos reales, al final pagaba más impuestos. Pero desde 1982 las categorías
fiscales están indexadas (es decir, se ajustan a la inflación) y este cambio ha reducido de manera
sustancial el lastre fiscal automático del sistema.
PRESUPUESTO DE PLENO EMPLEO
Como la condición de la economía afecta con severidad al déficit presupuestario, no podemos
juzgar con tino la intención ni el éxito de las políticas fiscales si sólo consideramos el superávit o
el déficit. En lugar de analizar la magnitud del superávit o el déficit, los economistas han creado
una alternativa para medir la eficacia real de una política fiscal. Si examinamos cuál sería el presupuesto si la economía produjera al nivel de pleno empleo (el llamado presupuesto de pleno empleo)
podríamos fijar un punto de referencia para valorar esa política fiscal.
La distinción entre presupuesto real y de pleno empleo es importante. Supongamos que la
economía está en crisis y el déficit es de 250,000 millones de dólares. Asimismo, imaginemos que
si el empleo fuera pleno, el déficit bajaría a 75,000 millones. El déficit de 75,000 millones que
permanecería incluso con pleno empleo se debería a la estructura de los programas fiscales y de
gasto, más que al estado de la economía. Este déficit (que queda con un pleno empleo) se llama

déficit estructural. La parte de 175,000 millones (250,000 − 75,000) del déficit causada por el
hecho de que la economía está en una crisis se conoce como déficit cíclico. La existencia del déficit cíclico depende del momento del ciclo comercial en que se encuentre la economía y deja de
existir cuando se alcanza el pleno empleo. Por definición, el déficit cíclico del presupuesto de
pleno empleo es cero.

### Resumen

1. El gobierno influye en la macroeconomía mediante tres paquetes de
políticas. La política fiscal se refiere al gasto y la fijación de impuestos. La política fiscal discrecional se refiere a variaciones de impuestos
o gastos que son resultado de cambios deliberados de las políticas
gubernamentales. La política monetaria se refiere al comportamiento
de la Reserva Federal con respecto al circulante de la nación.
EL GOBIERNO EN LA ECONOMÍA
2. El gobierno no tiene un control completo sobre la recaudación
fiscal y ciertos gastos que están determinados, en parte, por el estado de la economía.
3. Como participante de la economía, el gobierno compra bienes y
servicios (G), recauda impuestos y hace pagos de transferencia a
los hogares. Los impuestos netos (T) son iguales a los pagos de
gravámenes hechos por empresas y hogares al gobierno, menos
los pagos de transferencia efectuados por el gobierno a las familias.
4. El ingreso disponible o después de impuestos (Yd) es igual a la suma
del ingreso recibido por los hogares luego de descontar la parte
fiscal: Yd ≡ Y − T. El ingreso después de impuestos determina los
hábitos de consumo de los hogares.
5. El déficit presupuestario es igual a la diferencia entre lo que gasta
el gobierno y lo que recauda por impuestos: G − T. Cuando G excede de T, el gobierno debe pedir prestado al público para financiar su déficit.
6. En una economía donde el gobierno es un participante, el gasto
total planeado es igual al gasto de consumo de los hogares (C),
más el gasto de inversión planeada de las empresas (I), más el gasto
del gobierno en bienes y servicios (G): GT ≡ C + I + G. Como la
condición Y = GT es necesaria para que la economía esté en
equilibrio, se sigue que Y = C + I + G es la condición de equilibrio de la macroeconomía. La economía también está en equilibrio
cuando las fugas del sistema equivalen a las inyecciones. Lo anterior
ocurre cuando el ahorro y los impuestos netos (las fugas) son
iguales a la inversión planeada y los gastos gubernamentales (las
inyecciones): S + T = I + G.
LA POLÍTICA FISCAL EN LA PRÁCTICA: EFECTOS DE LOS
MULTIPLICADORES
7. La política fiscal tiene un efecto multiplicador en la economía. Un
cambio en el gasto gubernamental da lugar a un multiplicador
igual a 1/PMA. Una variación en los impuestos produce un multiplicador igual a −PMC/PMA. Un incremento o decremento
igual y simultáneo en el gasto del gobierno y los impuestos tiene
un efecto multiplicador de 1.
EL PRESUPUESTO FEDERAL
8. El déficit federal fue muy grande en la década de los
ochenta. Entre otras causas, se debió a la acumulación de
gastos de defensa en los años de Ronald Reagan, a las grandes
cantidades de intereses pagadas por la deuda anterior y a la
reducción de las tasas tributarias de las personas físicas. Como
los gastos de defensa y los pagos de intereses aumentaron muy
rápido y los impuestos bajaron, el gobierno gastaba más de lo
que recaudaba.
INFLUENCIA DE LA ECONOMÍA EN EL PRESUPUESTO
DEL GOBIERNO
9. Los estabilizadores automáticos son rubros de entradas y gastos
del presupuesto federal que cambian de manera automática con
el estado de la economía y estabilizan el PIB. Por ejemplo, durante
las expansiones el gobierno recibe de inmediato más ingresos
porque la gente gana más ingresos gravables. El aumento de los
ingresos y las categorías impositivas aminora los pagos de
transferencia.
10. El lastre fiscal es un efecto negativo en la economía que ocurre
cuando las tasas fiscales promedio aumentan porque, en una
expansión, los contribuyentes pasan a categorías impositivas más
altas. Estos impuestos más elevados reducen el ingreso disponible
y desaceleran la expansión. Desde 1982, las categorías fiscales
están indexadas a la inflación. Esta medida disminuye el lastre
fiscal del sistema impositivo.
11. El presupuesto de pleno empleo es una construcción hipotética
de los economistas sobre cuál sería el presupuesto federal si la
economía produjera con pleno empleo. El déficit estructural es

## Oferta de dinero y el sistema de la Reserva Federal

1. El gobierno influye en la macroeconomía mediante tres paquetes de
políticas. La política fiscal se refiere al gasto y la fijación de impuestos. La política fiscal discrecional se refiere a variaciones de impuestos
o gastos que son resultado de cambios deliberados de las políticas
gubernamentales. La política monetaria se refiere al comportamiento
de la Reserva Federal con respecto al circulante de la nación.
EL GOBIERNO EN LA ECONOMÍA
2. El gobierno no tiene un control completo sobre la recaudación
fiscal y ciertos gastos que están determinados, en parte, por el estado de la economía.
3. Como participante de la economía, el gobierno compra bienes y
servicios (G), recauda impuestos y hace pagos de transferencia a
los hogares. Los impuestos netos (T) son iguales a los pagos de
gravámenes hechos por empresas y hogares al gobierno, menos
los pagos de transferencia efectuados por el gobierno a las familias.
4. El ingreso disponible o después de impuestos (Yd) es igual a la suma
del ingreso recibido por los hogares luego de descontar la parte
fiscal: Yd ≡ Y − T. El ingreso después de impuestos determina los
hábitos de consumo de los hogares.
5. El déficit presupuestario es igual a la diferencia entre lo que gasta
el gobierno y lo que recauda por impuestos: G − T. Cuando G excede de T, el gobierno debe pedir prestado al público para financiar su déficit.
6. En una economía donde el gobierno es un participante, el gasto
total planeado es igual al gasto de consumo de los hogares (C),
más el gasto de inversión planeada de las empresas (I), más el gasto
del gobierno en bienes y servicios (G): GT ≡ C + I + G. Como la
condición Y = GT es necesaria para que la economía esté en
equilibrio, se sigue que Y = C + I + G es la condición de equilibrio de la macroeconomía. La economía también está en equilibrio
cuando las fugas del sistema equivalen a las inyecciones. Lo anterior
ocurre cuando el ahorro y los impuestos netos (las fugas) son
iguales a la inversión planeada y los gastos gubernamentales (las
inyecciones): S + T = I + G.
LA POLÍTICA FISCAL EN LA PRÁCTICA: EFECTOS DE LOS
MULTIPLICADORES
7. La política fiscal tiene un efecto multiplicador en la economía. Un
cambio en el gasto gubernamental da lugar a un multiplicador
igual a 1/PMA. Una variación en los impuestos produce un multiplicador igual a −PMC/PMA. Un incremento o decremento
igual y simultáneo en el gasto del gobierno y los impuestos tiene
un efecto multiplicador de 1.
EL PRESUPUESTO FEDERAL
8. El déficit federal fue muy grande en la década de los
ochenta. Entre otras causas, se debió a la acumulación de
gastos de defensa en los años de Ronald Reagan, a las grandes
cantidades de intereses pagadas por la deuda anterior y a la
reducción de las tasas tributarias de las personas físicas. Como
los gastos de defensa y los pagos de intereses aumentaron muy
rápido y los impuestos bajaron, el gobierno gastaba más de lo
que recaudaba.
INFLUENCIA DE LA ECONOMÍA EN EL PRESUPUESTO
DEL GOBIERNO
9. Los estabilizadores automáticos son rubros de entradas y gastos
del presupuesto federal que cambian de manera automática con
el estado de la economía y estabilizan el PIB. Por ejemplo, durante
las expansiones el gobierno recibe de inmediato más ingresos
porque la gente gana más ingresos gravables. El aumento de los
ingresos y las categorías impositivas aminora los pagos de
transferencia.
10. El lastre fiscal es un efecto negativo en la economía que ocurre
cuando las tasas fiscales promedio aumentan porque, en una
expansión, los contribuyentes pasan a categorías impositivas más
altas. Estos impuestos más elevados reducen el ingreso disponible
y desaceleran la expansión. Desde 1982, las categorías fiscales
están indexadas a la inflación. Esta medida disminuye el lastre
fiscal del sistema impositivo.
11. El presupuesto de pleno empleo es una construcción hipotética
de los economistas sobre cuál sería el presupuesto federal si la
economía produjera con pleno empleo. El déficit estructural es
el déficit federal que perdura incluso en situación de pleno
empleo. Un déficit cíclico ocurre cuando hay una baja en el
ciclo económico.

La mayoría de las personas da por hecha la capacidad de obtener y usar el dinero. Cuando el
sistema monetario funciona bien, como ocurre en Estados Unidos por lo general, el mecanismo
básico del sistema es prácticamente invisible. Las personas ven con naturalidad que pueden ir a
cualquier tienda, restaurante, boutique o gasolinera y comprar lo que deseen, siempre que tengan dinero.
La idea de que uno compre cosas con dinero es tan normal y obvia que parece absurdo
mencionarla. Pero deténgase y pregúntese: ¿por qué el dueño de la tienda está dispuesto a entregar un bistec y una hogaza de pan para que yo coma, a cambio de unos papeles que carecen de
valor intrínseco? ¿Y por qué hay lugares y ocasiones que requieren una carretilla llena de dinero
para comprar una docena de huevos? Las respuestas están en lo que es el dinero: un medio de
pago, un depósito de valor y una unidad de cuenta.
Medio de pago o de intercambio El dinero es crucial para el funcionamiento de una
economía de mercado. Imagine cómo sería la vida sin él. La alternativa de la economía monetaria es el trueque, por el cual las personas hacen intercambios directos de bienes y servicios por
otros, en lugar de realizarlos mediante el dinero.
¿Cómo funciona el sistema de trueque? Supongamos que usted quiere desayunar huevos
con tocino y jugo de naranja. En lugar de ir a la tienda a comprar estos artículos con dinero, tendría que encontrar a alguien que los tuviera y que deseara cambiarlos. Usted también debería de
tener algo que quisieran el marchante de tocinos, el proveedor de jugo de naranja y el vendedor
de huevos. Tener lápices para cambiar no sería de ningún provecho si esos comerciantes no los
quieren.
Un sistema de trueque requiere una doble coincidencia de deseos para que se lleve a cabo el
comercio. Es decir, para hacer un cambio, debo encontrar a alguien que posea lo que quiero y
que además acepte lo que tengo. Cuando el surtido de bienes es mínimo, como acontece en las
economías poco avanzadas, no es difícil encontrar a alguien con quien intercambiar y es común
recurrir al trueque. En una sociedad compleja, con muchos bienes, el trueque exige un esfuerzo
intolerable. Imagine que trata de encontrar a las personas que vendan todo lo que usted quiere
comprar en una visita cualquiera al supermercado y que estén dispuestas a aceptar los bienes
que les ofrezca a cambio de los suyos.
Algún medio de cambio (o de pago) pactado elimina de tajo el problema de la doble coincidencia. En un sistema monetario, el dinero se cambia por bienes y servicios cuando la gente
compra; a su vez, los bienes y servicios se cambian por dinero cuando las personas venden. Jamás,
nadie tiene que cambiar unos bienes por otros de manera directa. El dinero es un lubricante del
funcionamiento de una economía de mercado.
Depósito de valor Los economistas han señalado otras funciones del dinero, aparte de
su aplicación principal como medio de cambio. El dinero también se utiliza como depósito de
valor, o sea, un activo que sirve para trasladar poder adquisitivo de un periodo a otro. Si usted
cría pollos y al final del mes los vende en más de lo que quiere gastar y consumir de inmediato,
puede guardar parte de sus ganancias en la forma de dinero hasta el momento en que quiera
gastarlo.
Hay muchas otras reservas de valor aparte del dinero. Usted pudo haberse decidido a conservar sus ganancias “excedentes” comprando, digamos, cuadros antiguos, tarjetas de beisbol o
diamantes que luego podría vender cuando quisiera gastar sus utilidades. El dinero tiene varias
ventajas sobre estas reservas de valor. En primer lugar, viene en denominaciones convenientes y
es portátil. No tiene que preocuparse por cambiar una pintura de Renoir para adquirir litros de
gasolina. Segundo, como el dinero también es un medio de pago, se cambia con facilidad por
bienes en todo momento (un Renoir no se cambia con esa comodidad por otros bienes). Estos
dos factores componen la propiedad de liquidez del dinero. El dinero se gasta con simplicidad,
pasa por sus manos como un líquido. Los cuadros de Renoir y las antiguas esculturas aztecas no
son convenientes ni portátiles; tampoco se aceptan de buena gana como medios de pago.
La principal desventaja del dinero como depósito de valor es que éste baja cuando suben los
precios de bienes y servicios. Si el precio de las papas fritas aumenta de uno a dos dólares por
bolsa, el valor de un billete de un dólar, medido en papas fritas, baja de una a media bolsa. Cuando esto sucede, quizá valga más usar las papas fritas (o las antigüedades o bienes raíces) como
depósito de valor.Unidad de cuenta El dinero sirve también como unidad de cuenta: una manera constante
de cotizar los precios, los cuales siempre se cotizan en unidades monetarias. Un libro cuesta 90
dólares, no 150 plátanos ni cinco DVD, y un plátano cuesta 60 centavos de dólar, no 1.4 manzanas ni seis páginas de un libro. Como es obvio, una unidad estándar de cuenta es en extremo útil
para cotizar precios. Esta función del dinero le habrá pasado inadvertida: ¿en qué otra cosa se cotizan los precios, si no es en dinero?
MERCANCÍAS Y MONEDAS FIDUCIARIAS
Los libros de introducción a la economía están llenos de informes sobre los objetos que diversas
culturas han usado como dinero: dulces, cigarros (los prisioneros de los campos de concentración
en la Segunda Guerra Mundial), enormes ruedas de piedra labrada (en la isla de Yap, en el Pacífico
sur), conchas de cauríes (en África occidental), cuentas (entre los indígenas de Norteamérica), ganado (en el sur de África) y pequeños recortes de papel verde (hoy, en Estados Unidos). La lista
continúa. Estas formas del dinero se dividen en dos grupos: bienes de cambio y dinero simbólico.
Los bienes de cambio son artículos que se usan como dinero y que también tienen un valor
intrínseco en algún otro uso. Por ejemplo, los prisioneros de guerra hacían sus compras con cigarrillos, fijaban los precios en términos de cigarrillos y su riqueza tenía la forma de cigarrillos acumulados. Desde luego, también podían fumarlos: tenían otro uso, además de servir como dinero.
El oro representa otra forma de bien de cambio. Durante cientos de años el oro se empleó directamente para comprar cosas, pero también tiene otros usos, de la joyería a los empastes dentales.
En cambio, casi todo el dinero de Estados Unidos es fiduciario. La moneda fiduciaria, también llamada dinero simbólico, carece de valor intrínseco. El valor real de un billete de uno, 10 o
50 dólares es, de hecho, nada: ¿para qué más sirve un papelito impreso con tinta verde?
¿Por qué la gente acepta como dinero recortes despreciables de papel, en lugar de algo que
tenga valor, como oro, cigarrillos o ganado? Si usted responde que se debe a que el dinero está respaldado por oro o plata, se equivoca. Hubo un tiempo en que, de manera directa, los dólares se
cambiaban por oro. El gobierno respaldaba cada dólar en circulación con determinada cantidad
de oro en sus bóvedas. Por ejemplo, si el precio del oro era de 35 dólares por onza, el gobierno
aceptaba vender una onza de oro por billetes que sumaran 35 dólares. Pero ya los billetes no están
respaldados por ningún bien: oro, plata o lo que sea. Sólo se cambian por monedas de uno, cinco
o 10 centavos, por otros dólares, etcétera.
La gente acepta el papel moneda como medio de pago y depósito de valor porque el gobierno
ha tomado medidas para que así ocurra. El gobierno declara que el papel moneda es de curso legal;
es decir, que su dinero debe ser aceptado para finiquitar deudas. Lo hace por fíat (de ahí la expresión moneda fiduciaria). Asimismo, promulga leyes en las que define que ciertos papeles impresos
con diversas tintas y en determinadas planchas son de curso legal y eso es todo. En toda anotación
de la Reserva Federal de Estados Unidos se lee: “Esta nota es de curso legal para todas las deudas,
públicas y privadas”. Muchas veces, un gobierno empieza a ganar aceptación para su papel moneda
exigiendo que se use para pagar los impuestos (observe que no puede pagar sus impuestos con
pollos, tarjetas de beisbol ni cuadros de Renoir).
Aparte de declarar que su moneda es de curso legal, el gobierno hace otra cosa para asegurar
la aceptación del papel moneda: promete al público que no imprimirá billetes tan pronto como
para que pierda su valor. Aumentar el circulante en esa forma hace que pierda mucho de su valor y
ha sido un problema a lo largo de la historia. A esto se le conoce como degradación monetaria.
La degradación de la moneda ha sido una complicación especial de los gobiernos a los que les
falta fuerza para aplicar la impopular medida de incrementar los impuestos. Imprimir dinero
para los gastos gubernamentales en bienes y servicios es un sustituto de las alzas de impuestos.
Los gobiernos débiles han recurrido a las prensas para financiar sus gastos. Un ejemplo reciente,
de 2005, es Zimbabwe (véase el apartado de la página 199). En capítulos posteriores nos detendremos más en el dinero y la inflación.
MEDICIÓN DEL CIRCULANTE EN ESTADOS UNIDOS
Ahora pasamos a las variantes del dinero en Estados Unidos. Recuerde que el dinero se usa para
comprar cosas (como medio de pago), acumular riqueza (depósito de valor) y cotizar precios
(unidad de cuenta). Por desgracia, estas características corresponden a una gama extensa de activos de la economía estadounidense. Como veremos, no está del todo claro dónde hemos de trazar
la línea y decir que de un lado es dinero y del otro, algo más.
Para resolver el problema de muchos dineros, los economistas han dado nombres distintos a
las medidas del dinero. Las dos más comunes son el dinero de transacciones, también designado
como M1, y el dinero general, o M2.
M1: Dinero de transacciones ¿Qué debe contarse como dinero? Pues, monedas y billetes, así como denominaciones mayores, ya que cumplen con todos los requisitos. ¿Y las cuentas de
cheques? También con éstos se compran cosas y sirven como depósito de valor. De hecho, los
banqueros llaman a las cuentas de cheques depósitos de demanda porque los depositantes tienen
el derecho de ir al banco y liquidar (por demanda) sus cuentas de cheques en cualquier momento.
Por eso su saldo en la chequera es, de hecho, equivalente a tener billetes en la cartera y debe contarse como parte del dinero que posee.
Si tomamos el valor de todo el dinero (incluyendo monedas) que no está en las bóvedas de los
bancos y le sumamos el valor de todos los depósitos de demanda, cheques de viajero y otros depósitos en cheques, tenemos la definición del M1, o dinero de transacciones. Como indica el nombre,
se trata de dinero que se usa para efectuar transacciones directas, o sea, para comprar cosas.
M1 ≡ dinero fuera del banco + depósitos en demanda + cheques de viajero + otros
depósitos en cheques
Un depósito en cheques es cualquier cuenta de depósito en un banco u otra institución financiera contra la que pueda girarse un cheque. Estos depósitos abarcan los depósitos en demanda,cuentas
de órdenes de retiro negociables (ORN), que son cuentas de cheques que pagan intereses, y cuentas de
ahorro de transferencia automática (ATA), que trasladan de manera expedita fondos de ahorros a
cheques (o viceversa) cuando el saldo de una de las cuentas llega a cierto monto determinado.
El 18 de julio de 2005 el M1 en Estados Unidos era de un billón 327,300 millones de dólares.
El M1 es una medida de inventario que se realiza en un momento fijo. Es la suma de monedas y
divisas que no están en el banco y el total del dinero depositado en las cuentas de cheques un día
específico. Hasta ahora, hemos considerado la oferta como un flujo, una variable con una dimensión
temporal: la cantidad de trigo ofrecida por año, la cantidad de automóviles llevados al mercado por
año, etc. No obstante, el M1 es una variable de inventario.
M2: Dinero general Aunque M1 es la medida más usada del circulante, hay otras. ¿Las
cuentas de ahorro deben ser consideradas dinero? Muchas no pueden usarse en transacciones directas, pero es fácil convertirlas en efectivo o trasladar fondos de una cuenta de ahorros a una de
cheques. ¿A qué corresponden las cuentas del mercado de dinero (que sólo autorizan pocos cheques por mes, pero que pagan las tasas de interés determinadas por el mercado) y los fondos mutuos del mismo mercado (que venden acciones y usan las utilidades para comprar títulos de vencimiento a corto plazo)? Sirven para girar cheques y hacer compras, si bien sólo hasta cierta suma.
Si sumamos los dineros afines, que son sustitutos muy equivalentes al dinero de transacción,
el M1, obtenemos el M2, llamado dinero general porque abarca entidades que no son del todo
monetarias, como las cuentas de ahorro y del mercado de dinero, más otros dineros afines.
M2 ≡ M1 + cuentas de ahorros + cuentas de mercados de dinero + otros dineros afines
El 18 de julio de 2005, el M2 en Estados Unidos sumaba seis billones 532,200 millones, mucho
mayor que el total del M1 de un billón 327,300 millones. La principal ventaja de examinar el M2
en lugar del M1 estriba en que a veces es más estable. Por ejemplo, cuando a comienzos de los años
ochenta los bancos introdujeron nuevas formas de cuentas de cheques que dan intereses, el M1 se
disparó porque la gente pasó sus fondos de las cuentas de ahorro a las de cheques. Sin embargo, el
M2 se mantuvo muy constante porque la baja de los depósitos de las cuentas de ahorros y el aumento de los saldos de cuentas de cheques son parte del M2, con lo que se cancelaron sus efectos.
Más allá del M2 Debido a que muchos instrumentos financieros guardan algún parecido
con el dinero, algunos economistas han propuesto que se les incluya —a casi todos— como parte
del circulante. Por ejemplo, en los últimos años las tarjetas de crédito se han usado ampliamente
para efectuar transacciones. Quien tenga una de estas tarjetas cuenta con un crédito limitado:
sólo se le puede cargar cierta suma que a continuación hay que pagar. Es habitual que paguemos
las facturas de las tarjetas de crédito con un cheque. Una de las definiciones más amplias de dinero incluye la suma disponible en las tarjetas de crédito (el límite de crédito menos los cargos
no saldados) como un segmento del circulante.
No hay reglas para decidir qué es el dinero y qué no es. Esto plantea problemas a los economistas y a los responsables de la política económica. Sin embargo, para nuestros fines aquí, “dinero”
siempre se refiere al dinero de transacciones, el M1. Por simplicidad, diremos que el M1 es la suma
de dos categorías generales: el dinero en circulación y los depósitos. Pero recuerde que el M1 tiene
cuatro componentes específicos: dinero fuera del banco, depósitos de demanda, cheques de viajero y depósitos cambiables por cheques.
EL SISTEMA DE BANCA PRIVADA
En la actualidad, la mayor parte del dinero de Estados Unidos es, de una u otra forma, “dinero de
banco”. El M1 está formado sobre todo por saldos de cuentas de cheques en lugar de efectivo, y
el efectivo comprende una parte todavía menor del M2 y otras definiciones amplias de dinero.
Para entender el dinero es necesario conocer la estructura del sistema bancario privado.
Los bancos e instituciones semejantes toman dinero de individuos y empresas con exceso
de fondos y lo prestan a quienes necesitan financiamiento. Por ejemplo, los bancos comerciales
reciben fondos de varias formas, como los depósitos en cuentas de cheques y ahorros. Toman
estos fondos y los prestan para automóviles, hipotecas, créditos comerciales, etc. Los bancos y
dichas instituciones se llaman intermediarios financieros porque “median”, actúan como vínculos entre quienes tienen fondos para prestarlos y quienes necesitan los préstamos.
Los principales tipos de intermediarios financieros son los bancos comerciales, seguidos
por las sociedades de préstamo y ahorro, aseguradoras y fondos de pensiones. Desde 1970 se han
estrechado en forma considerable las distinciones legales entre los intermediarios financieros.
Por ejemplo, antes sólo los bancos comerciales ofrecían cuentas de cheques y no podían pagar
intereses por ellas. Las asociaciones de préstamo y ahorro tenían prohibido ofrecer ciertas clases
de depósitos y se restringían a hacer préstamos hipotecarios.
La Ley de Desregulación de Instituciones de Depósito y Control Monetario, promulgada
por el Congreso estadounidense en 1980, eliminó muchas de las limitaciones que pesaban sobre
las actividades de los organismos financieros. Ahora, instituciones de muchos tipos ofrecen
cuentas de cheques y se pagan intereses por muchas de sus variantes. Las asociaciones de préstamo y ahorro conceden empréstitos para muchas cosas, además de hipotecas.
CÓMO CREAN DINERO LOS BANCOS
Hasta aquí hemos descrito en lo general el funcionamiento de los bancos y la forma de medir el
circulante en Estados Unidos, pero ¿cuánto dinero hay en determinado momento? ¿Quién lo proporciona y cómo? Estamos listos para analizar con mucho cuidado estas preguntas. En particular,
queremos explorar un proceso que a muchos les parece misterioso: cómo crean dinero los bancos.
PANORAMA HISTÓRICO: LOS ORFEBRES
Para empezar a ver cómo crean dinero los bancos, consideremos los orígenes del sistema bancario moderno. En los siglos XV y XVI, ciudadanos de muchos lugares usaban el oro como
dinero, sobre todo para transacciones grandes. Como es engorroso cargar oro y se presta a que
lo roben, la gente empezó a depositarlo con orfebres para que lo guardaran. Cuando un orfebre
tomaba el oro, entregaba un recibo al depositante y le cobraba una pequeña suma por cuidarlo.
Al cabo del tiempo, estos recibos en sí, más que el oro que amparaban, comenzaron a cambiarse
por bienes. Se convirtieron en una forma de papel moneda que hizo innecesario ir con el orfebre
a retirar oro para alguna transacción.
En ese momento, los recibos expedidos por los orfebres estaban respaldados por oro en su
totalidad. Si un orfebre tenía 100 onzas de oro en su caja de caudales, entregaba un recibo por 100
onzas de oro y nada más. Los orfebres operaban como depósitos donde la gente guardaba oro a
buen recaudo. Pero los orfebres se percataron de que la gente no volvía pronto a retirar su oro.
Para qué, si algunos recibos de papel que podían convertirse en oro eran igual de buenos (de hecho, los recibos eran mejores: más portátiles, difíciles de robar, etc.). Por consiguiente, los orfebres tenían siempre a la mano grandes reservas de oro.
Como de hecho los orfebres tenían oro “sobrante” sin utilidad, poco a poco se dieron cuenta
de que podían prestar una parte sin temor a quedarse sin nada. ¿Por qué lo hacían? Porque en lugar de guardar el oro ocioso en sus bóvedas, ganaban intereses sobre los préstamos. En este momento ocurrió algo sutil, pero radical. Los orfebres pasaron de meros depositarios de oro a instituciones de tipo bancario que tenían el poder de crear dinero. Esta transformación ocurrió en

cuanto los orfebres comenzaron a conceder préstamos. Sin agregar nada de oro real al sistema, los
orfebres incrementaron la cantidad del dinero en circulación porque crearon más obligaciones
sobre el oro; es decir, recibos que autorizaban al portador a recabar tantas onzas de oro en cuanto
lo pidiera.1 Así, había más derechos que onzas de oro.
Un ejemplo detallado servirá para aclararlo. Supongamos que usted acude con un orfebre
que opera sólo como depositario, como reserva, y le pide un préstamo para comprar una parcela
que cuesta 20 onzas de oro. Digamos también que el orfebre tiene depositadas en su caja de caudales exactamente 100 onzas de oro de varias personas que se lo confiaron. Si el orfebre decide
que ya se cansó de su oficio y quiere convertirse en un banco de verdad, le prestará a usted algo
de oro. Desde luego, usted no quiere el oro como tal, sino un papel que representa 20 onzas del
metal amarillo. En esencia, el orfebre “crea” dinero para usted cuando le da un recibo de 20 onzas de
oro (aunque todo su suministro de oro pertenezca a otras personas).2 Entonces, habrá en circulación recibos por 120 onzas, en lugar de los de 100 onzas que había antes de su préstamo; el circulante habrá aumentado.
La gente cree que la creación de dinero es misteriosa. ¡Nada más alejado de la realidad! Tan
sólo obedece a un procedimiento contable, uno de los más ordinarios entre las actividades humanas. Acaso tenga la sospecha de que todo el mecanismo es ilógico o por lo menos dudoso en
lo fundamental. Después de todo, el sistema bancario nació cuando alguien expidió derechos
sobre un oro que pertenecía a otra persona (en esto último, usted estaría sobre terreno algo más
firme).
Los orfebres convertidos en banqueros tenían ciertos problemas. En cuanto empezaron a
conceder préstamos, sus recibos sin pagar (derechos sobre el oro) sumaban más que el oro que
tenían en sus bóvedas en cualquier momento. Si los dueños de los recibos por 120 onzas de oro
se presentaban al mismo tiempo y exigían su metal dorado, el orfebre estaría en aprietos. Como
sólo tenía 100 onzas, las personas no podrían llevarse su respectiva parte de una vez.
En tiempos normales, la gente se contentaba con tener los recibos en lugar del oro real y el
problema no surgía. Pero si comenzaban a preocuparse por la estabilidad financiera del orfebre,
quizá les surgirían dudas sobre si sus recibos fueran tan buenos como el oro. Si sabían que había
más recibos por pagar que onzas de oro en la bóveda del orfebre, podrían exigir el metal a cambio
de sus recibos.
La situación llevó a una paradoja. Es del todo lógico que las personas tengan los recibos de
papel (en lugar del oro) si saben que siempre pueden cambiarlos por el metal. En épocas normales los orfebres podrían sentirse por completo seguros para prestar más oro del que estaba en su
posesión. Pero cuando la gente comenzaba a dudar de la seguridad de un orfebre, era tonto no
exigir la devolución del oro.
Una corrida financiera de un orfebre (en nuestros días, de un banco) ocurre cuando mucha
gente se presenta a ejercer sus derechos al mismo tiempo. Estos retiros se avivan solos. Si veo que
usted va con el orfebre a retirar su oro, me pongo nervioso y decido también sacar el mío. Lo que da
lugar a una corrida financiera es el miedo a una corrida financiera. Diversas causas pueden disparar
los retiros multitudinarios de un banco: rumores de que una institución prestó a deudores insolventes, guerras, quiebra de instituciones que tomaron préstamos el banco, etc. Más adelante veremos
que los banqueros de hoy tienen una diferencia con los orfebres: los bancos actuales están obligados a tener un “porcentaje de reserva obligatoria”. Los orfebres no tenían requisitos de reserva
legal, aunque la suma que prestaban estaba sujeta a las restricciones impuestas por su propio
miedo a quedarse sin oro.

Los activos son propiedad de una empresa y tienen un valor. En el caso de un banco, entre sus
activos están el edificio que lo alberga, los muebles, su cartera de valores gubernamentales, efectivo
en bóvedas, bonos, acciones, etc. De los activos del banco son muy importantes, cuando menos
para nuestros fines, los préstamos. Un prestatario le firma al banco un pagaré, una promesa de
pagarle cierta suma en determinada fecha. Este documento es un activo del banco porque vale algo.
El banco puede (como en efecto hace a veces) vender el pagaré a otro banco a cambio de efectivo.
Otros activos del banco son el efectivo líquido (también llamado efectivo en bóveda) y depósitos en el banco central de Estados Unidos, el Banco de la Reserva Federal (Fed). Como veremos
más adelante, las normas federales para la operación de la banca exigen que los bancos mantengan cierta proporción de sus depósitos como efectivo en bóveda o que lo depositen en la Fed.
Los pasivos de una empresa son sus deudas, lo que debe. Los pasivos de un banco son las
promesas de pago, los pagarés que ha expedido. Los pasivos más importantes de un banco son
sus depósitos. Éstos son deudas con los depositantes, porque cuando alguien deposita dinero en
su cuenta, en esencia le hace un préstamo al banco.
La regla básica de la contabilidad dice que si sumamos los activos de una empresa y a continuación restamos el monto total que le debe a quienes le han entregado sus fondos, el resultado
es el valor neto de la empresa, el cual representa el valor de una empresa para sus accionistas o
propietarios. ¿Cuánto pagaría por una compañía que posee diamantes por un valor de 200,000
dólares y para pagarlos pidió al banco un préstamo de 150,000? La compañía vale 50,000 dólares:
la diferencia entre lo que posee y lo que debe. Si el precio de los diamantes bajara, de modo que
su valor fuera de 150,000, la firma no valdría nada.
Tomamos nota de la posición financiera de una empresa con un balance general simplificado
llamado cuenta T. Por convención, los activos de un banco se anotan a la izquierda de la cuenta T
y sus pasivos y valor neto, a la derecha. Por definición, el balance general siempre cuadra, de modo
que la suma de los rubros del lado izquierdo es igual a la adición de los rubros del derecho.
En la cuenta T de la figura 10.1 se muestra que un banco tiene 110 millones de dólares en
activos, de los cuales 20 millones son reservas —los depósitos que el banco ha hecho a la Fed—
y su efectivo disponible (monedas y divisas). Las reservas son un activo del banco porque éste
puede ir a la Fed y pedir el efectivo, así como uno va al banco y saca dinero de la cuenta de ahorros. El otro activo de nuestro banco son los préstamos, que valen 90 millones de dólares.
¿Por qué los bancos tienen reservas o depósitos en la Fed? Hay varias razones, pero acaso la
más importante sea la obligación legal de que mantengan en reserva cierto porcentaje de sus pasivos en depósito. El porcentaje de sus depósitos que un banco tiene que mantener como reserva
se llama proporción de reserva requerida. Si la proporción de reserva es de 20 por ciento, el
banco con depósitos de 100 millones de dólares debe tener reservas de 20 millones, ya sea en

efectivo o como depósitos en la Fed. Para simplificar, vamos a suponer que los bancos tienen todas
sus reservas como depósitos en la Fed.
Del lado de los pasivos de la cuenta T, el banco ha aceptado depósitos por 100 millones de dólares, así que debe esa cantidad a los depositantes. Esto significa que el banco representa para sus
propietarios un valor neto de 10 millones (110 millones en activos – 100 millones en pasivos = 10
millones de valor neto). El valor neto del banco es el saldo del balance general. Recuerde:
Cuando cambia una partida del balance general, debe haber por lo menos otro cambio en
otro asiento para mantener el equilibrio.
Si las reservas de un banco aumentan un dólar, uno de los siguientes debe ser verdad: 1) bajan
un dólar sus otros activos (por ejemplo, préstamos); 2) sus pasivos (depósitos) se incrementan un
dólar; o 3) su valor neto aumenta un dólar. También son posibles combinaciones fraccionarias de
estos enunciados.

Los bancos hacen préstamos hasta el punto en que ya no pueden realizarlos debido a la restricción del requisito de reserva.
La suma de reservas que se exige a un banco es igual a la proporción de reserva requerida por el
total de depósitos del banco. Si un banco tiene depósitos de 100 dólares y la proporción obligada
es de 20 por ciento, la suma requerida de las reservas es de 20 dólares. La diferencia entre las reservas reales y las reservas obligatorias de un banco son sus reservas excedentes:

Si los bancos hacen préstamos hasta el punto en que ya no pueden realizarlos debido a la restricción
del requisito de reservas, significa que otorgan empréstitos hasta que sus reservas excedentes son cero.
Para ver por qué, observe que cuando un banco tiene reservas excedentes, dispone de crédito
y puede conceder préstamos. En realidad, sólo puede prestar dinero si tiene excedentes. Cuando
confiere un empréstito, crea un depósito en demanda para el prestatario. Este hecho rebaja las
reservas excedentes del banco porque los depósitos adicionales creados por el préstamo consumen
una parte de esos excedentes. Un ejemplo servirá para demostrarlo.

tiene activos (reservas) de 100 dólares y pasivos (depósitos) de 100 dólares. Si la proporción de reservas obligatoria es de 20 por ciento, el banco tiene un excedente de 80 dólares.

¿Cuánto puede prestar el banco sin dejar de cumplir el requisito de reservas? Supongamos
por un momento que alguien que toma un préstamo deja todas las utilidades en el banco o le paga
a alguien que así lo hace. Nada se retira en efectivo. En este caso, el banco puede prestar 400 dólares sin infringir la norma de la reserva. En la sección 3 se da el balance general del banco después
de completar la suma máxima de préstamos que se le permite, con una proporción de reservas requeridas de 20 por ciento. Con 80 dólares como exceso de reservas, el banco puede tener hasta
400 dólares de depósitos adicionales. Los 100 dólares de reservas más los 400 de préstamos (que
se hacen como depósitos) son iguales a 500 dólares en depósitos. Con 500 dólares de depósitos y
una reserva obligatoria de 20 por ciento, el banco debe tener reservas de 100 dólares (20 por
ciento de 500), que es precisamente lo que tiene. El banco no puede prestar más de 400 porque
sus requisitos de reservas no deben exceder de 100 dólares. Cuando un banco no tiene reservas
excedentes y ya no puede prestar más, se dice que gastó sus préstamos.
Recuerde que el circulante (M1) es igual al efectivo que está en curso más los depósitos. Antes
del depósito inicial, el circulante era de 100 dólares (100 en efectivo y ningún depósito). Así, está
claro que cuando los préstamos se convierten en depósitos, el circulante cambia.

EL SISTEMA DE LA RESERVA FEDERAL
Hemos visto cómo el sistema de la banca privada crea dinero haciendo préstamos. Sin embargo, los
bancos privados no pueden crear dinero a voluntad. Su capacidad de generarlo depende del volumen
de las reservas del mercado en el sistema, que está controlado por la Fed. Por lo tanto, esta institución
tiene el control definitivo del circulante.Ahora vamos a examinar la estructura y el funcionamiento de
la Reserva Federal.
La Fed se fundó en 1913 por una ley del Congreso (a la que se hicieron reformas importantes en
los años treinta) y es el banco central de Estados Unidos. La Fed es un organismo complejo con muchas responsabilidades, entre ellas la regulación y supervisión de más de 8,000 bancos comerciales. En
la figura 10.4 se ilustra la organización de la Reserva Federal.
El consejo de gobernadores es el grupo más importante del sistema de la Reserva Federal. Consta de siete miembros, que el presidente de Estados Unidos nombra para periodos de 14 años. El presidente de la Fed, que es nombrado por el presidente del país y cuyo periodo se extiende a cuatro
años, domina todo el sistema de la Reserva Federal y se ha dicho que es la segunda persona más poderosa de Estados Unidos. La Fed es una entidad autónoma en el sentido de que no acepta órdenes
del presidente ni del Congreso.
Estados Unidos está dividido en 12 distritos de la Reserva Federal, cada uno con su propio banco
de la Reserva Federal. Estos distritos aparecen indicados en el mapa de la figura 10.4. Los bancos distritales son como sucursales de la Fed porque aplican las reglas, normas y funciones del sistema central en
sus distritos y rinden cuentas al consejo de gobernadores sobre las condiciones de la economía local.
La Comisión Federal de Mercados Abiertos (CFMA) fija la política monetaria estadounidense
(las actividades de la Fed concernientes al circulante). La CFMA está compuesta por siete miembros
del consejo de gobernadores de la Fed, el presidente del Banco de la Reserva Federal de Nueva York y
en periodos rotatorios, cuatro de los 11 presidentes de los otros bancos distritales. La CFMA fija las
metas sobre el circulante y las tasas de interés y dirige la Mesa de Mercados Abiertos del Banco de la
Reserva Federal de Nueva York para comprar y/o vender los títulos del gobierno. (Más adelante expondremos las particularidades de las operaciones en los mercados abiertos.)
FUNCIONES DE LA RESERVA FEDERAL
La Reserva Federal es el banco central de Estados Unidos. En ocasiones se dice que los bancos
centrales son los “bancos de los banqueros” porque sólo aceptan cuentas de bancos (y de vez en
cuando de gobiernos extranjeros). Como ciudadano particular, usted no puede ir a la sucursal
más cercana de la Fed y abrir una cuenta de cheques o solicitar un préstamo.
Desde el punto de vista macroeconómico, la función crucial de la Fed es controlar el circulante, pero también cumple otros cometidos importantes para los bancos, entre los que se cuentan
compensar los pagos entre bancos, regular el sistema bancario y ayudar a los bancos que están en
una posición financiera difícil. La Reserva Federal también es responsable de manejar los tipos de
cambio y las reservas de divisas de la nación.4 Además, participa en negociaciones entre países sobre temas de economía internacional.
Compensación de pagos entre bancos Supongamos que usted gira un cheque por
100 dólares contra su banco, el First Bank of Fresno (FBF), con el que paga tulipanes de Crockett
Importers de Miami, Florida. Dado que Crockett Importers no tiene cuenta en FBF, sino en el
Banco de Miami, ¿cómo pasa el pago de su banco al de Florida?
La respuesta: lo hace la Fed. El FBF y el Banco de Miami tienen cuentas en la Reserva Federal.
Cuando Crockett Importers reciba su cheque y lo deposite en el Banco de Miami, el banco presenta
el cheque en la Reserva Federal y le pide que cobre los fondos del FBF. La Fed entrega el cheque a FBF
y dispone que se tomen 100 dólares de su cuenta y se acrediten en la cuenta del Banco de Miami. Las
cuentas en la Fed son reservas, así que el FBF pierde 100 dólares de sus reservas para que el Banco de
Miami gane esa cantidad para sus reservas. De hecho, los bancos han cambiado la propiedad de sus
depósitos en la Fed. El volumen total de las reservas no ha cambiado; el circulante tampoco.
La función de compensar los pagos interbancarios permite a los bancos mover el dinero de manera casi instantánea. Todo lo que tienen que hacer es conectarse con la Fed y solicitar una transferencia: los fondos se mueven en las computadoras, a la velocidad de la luz, de una cuenta a otra.
Otros deberes de la Fed Aparte de facilitar la transferencia de fondos entre bancos, la
Fed cumple otros importantes deberes. Está a cargo de muchas normas y criterios que rigen las
prácticas bancarias. Por ejemplo, la Fed tiene la autoridad para controlar las fusiones de bancos y
es responsable de examinar a los bancos para verificar que se encuentran en buena situación financiera y que acatan todas las normas contables del gobierno. Como ya vimos, la Reserva Federal
también fija los requisitos de reservas de todas las instituciones financieras.
Una de las responsabilidades más importantes de la Fed es fungir como prestamista de último recurso del sistema bancario. Como se vio en nuestro estudio de los orfebres, los bancos enfrentan la posibilidad de sufrir retiros en masa de sus depósitos. En Estados Unidos, la mayoría de
los depósitos de menos de 100,000 dólares están asegurados por la Corporación Federal de Seguros de Depósitos (CFSD), una dependencia del gobierno estadounidense establecida en 1933 durante
la Gran Depresión. Asegurar los depósitos hace menos probable una oleada de pánico. Como los depositantes saben que siempre pueden recuperar su dinero, aun si quiebra el banco, se inclinan menos
a retirarlo. No todos los depósitos están asegurados, así que persiste la posibilidad de oleadas de
pánico. No obstante, la Fed está dispuesta a proporcionar fondos a un banco en problemas que no
encuentra otras formas de financiamiento.
La Fed es el prestatario ideal de último recurso por dos razones. En primer lugar, proporcionar
fondos a un banco que está en graves aprietos es arriesgado y quizá no sea muy rentable; es difícil
encontrar bancos u otras instituciones privadas que estén dispuestos a hacerlo. La Fed es una institución sin fines de lucro que tiene la función de velar por el bienestar general del público. Así, la
Fed estaría interesada en prevenir oleadas catastróficas de pánico bancario, como las que se sucedieron a finales de los años veinte y comienzos de la siguiente década.

En segundo lugar, la Fed tiene fondos de suyo ilimitados para sacar de apuros a bancos que
enfrentan la posibilidad de retiros en masa. La causa, como veremos, estriba en que puede crear
reservas a voluntad. Una promesa de la Reserva Federal de que respaldará a un banco es muy
convincente. A diferencia de los demás prestamistas, la Fed nunca se queda sin dinero. Por lo
tanto, el apoyo explícito o implícito de la Fed debe bastar para asegurar a los depositantes que no
corren el peligro de perder sus fondos.

pasivos (pagarés) de la Fed. Son pagarés más bien extraños porque sólo pueden cambiarse por
otros del mismo tipo. Comoquiera que sea, se clasifican como pasivos de la Fed.
El balance general muestra que, como un banco comercial ordinario, la Fed acepta depósitos,
que son pasivos. El grueso de los depósitos de la Reserva Federal proviene de bancos comerciales.
Recuerde que los bancos comerciales están obligados a mantener cierta parte de sus depósitos como
depósitos en la Fed. Los depósitos de un banco en la Fed (sus reservas) son un activo desde el
punto de vista del banco; esas mismas reservas tienen que ser un pasivo desde el punto de vista de
la Reserva Federal.
En la tabla 10.1 se muestra que la Fed ha aceptado un pequeño volumen de depósitos de la
Tesorería de Estados Unidos. En efecto, la Fed funge como el banco del gobierno estadounidense.
Cuando el gobierno tiene que pagar algo, como un nuevo avión, gira un cheque al proveedor de
la nave desde su “cuenta de cheques” en la Fed. Del mismo modo, cuando el gobierno recibe ingresos por la recaudación de impuestos, multas o ventas de sus activos, deposita esos fondos en su
cuenta de la Fed.

CÓMO CONTROLA LA RESERVA
FEDERAL EL CIRCULANTE
Para ver cómo controla la Fed el circulante en la economía estadounidense, tenemos que entender
la función de las reservas. Como dijimos, la proporción de reservas requeridas establece un vínculo
entre las reservas de los bancos comerciales y los depósitos (dinero) que éstos pueden crear. En
efecto, el requisito de la reserva determina cuánto tiene un banco para prestar. Si la proporción
obligatoria es de 20 por ciento, cada dólar de reserva puede respaldar cinco dólares en depósitos.
Un banco que tiene reservas por 100,000 dólares no puede tener más de 500,000 en depósitos. Si
los tuviera, desobedecería la proporción obligatoria.
Si recuerda que el circulante es igual a la suma de los depósitos en los bancos y el dinero en
circulación fuera de ellos, verá que las reservas son la palanca que la Fed necesita para controlar
dicho circulante.
Si la Fed desea aumentar el circulante, deberá crear más reservas; por lo tanto, los bancos
tendrán la libertad de crear depósitos adicionales mediante más préstamos. Si quiere disminuir el circulante, habrá de reducir las reservas.
La Fed tiene tres herramientas para alterar el circulante: 1) cambiar la proporción requerida de
reservas, 2) cambiar la tasa de descuento y 3) realizar operaciones en el mercado abierto. Aunque la
opción 3) es la que se usa de manera casi exclusiva para cambiar el circulante, un conocimiento
de cómo operan 1) y 2) es útil para entender el funcionamiento de 3). Así, comenzamos nuestra
exposición con las primeras dos herramientas.
LA PROPORCIÓN DE RESERVA EXIGIDA
Una manera en que la Reserva Federal altera el circulante consiste en cambiar la proporción de la
reserva requerida. Este proceso se ilustra en la tabla 10.2. Supongamos que la proporción obligatoria inicial es de 20 por ciento.
En la sección 1, en una versión simplificada del balance general de la Fed (en miles de millones de dólares) se indica que las reservas son de 100,000 millones y la moneda disponible también
de 100,000 millones. El valor total de los activos de la Fed es de 200,000 millones de dólares, que
vamos a suponer que son en su totalidad títulos del gobierno. Suponiendo que no hay reservas
excedentes (los bancos prestan a toda su capacidad), los 100,000 millones de las reservas respaldan
500,000 millones en depósitos en los bancos comerciales. (Recuerde que el multiplicador del dinero es igual a 1/proporción requerida de reservas = 1/0.20 = 5. Así, 100,000 millones de dólares de
reservas pueden respaldar 500,000 millones [100,000 millones × 5] en depósitos, cuando la proporción exigida es de 20 por ciento.) Por consiguiente, el circulante (M1, el dinero de transacciones)

es de 600,000 millones: 100,000 millones en moneda y 500,000 millones en depósitos (cuentas
de cheques) de los bancos comerciales.
Ahora supongamos que la Reserva Federal quiere aumentar el circulante a 900,000 millones
de dólares. Si baja la proporción de reservas requeridas e 20 a 12.5 por ciento (como en la sección 2
de la tabla 10.2), los mismos 100,000 millones de reservas pueden respaldar 800,000 millones en depósitos, en lugar de 500,000. En este caso, el multiplicador del dinero es 1/0.125 = 8. Con una
proporción de reservas requeridas de 12.5 por ciento, 100,000 millones en reservas respaldan
depósitos de 800,000 millones. El circulante total sería de 800,000 millones en depósitos más
100,000 billones en moneda: un total de 900,000 millones de dólares.7
Dicho de otra manera, con la nueva proporción inferior de reservas requeridas, los bancos
tienen excedentes de 37,500 millones de dólares. Con la proporción obligatoria de 20 por ciento,
necesitaban 100,000 millones en reservas para respaldar depósitos de 500,000 millones. Con la
proporción más baja de 12.5 por ciento, sólo necesitan 62,500 millones de reservas para respaldar esos depósitos de 500,000 millones, así que los 37,500 millones que sobran de los 100,000
millones de reservas son “extra”. Con esos excedentes de 37,500 millones, los bancos pueden prestar más dinero. Si suponemos que el sistema presta dinero y crea depósitos al máximo monto posible, los 37,500 millones de reservas van a respaldar otros 300,000 millones de depósitos (37,500
millones × el multiplicador de dinero de 8 = 300,000 millones). El cambio en la proporción de
reservas obligadas inyectó otros 300,000 millones al sistema bancario. En ese punto, los bancos
cumplen su capacidad total de préstamo y ya no pueden aumentar sus depósitos.
Las reducciones de la proporción de reservas requeridas permiten a los bancos tener más depósitos para el volumen de reservas existentes. Cuando los bancos crean más depósitos haciendo préstamos, aumenta el circulante (dinero + depósitos). También es cierto lo contrario: si la
Reserva Federal quiere restringir el circulante, puede aumentar la proporción requerida de las
reservas, de modo que los bancos no tienen suficientes reservas y deben restringir sus depósitos “retirando” algunos de sus préstamos.8 El resultado es una disminución del circulante.
Por varias razones, la Fed ha preferido no cambiar el requisito de las reservas para controlar
el circulante. En parte, este rechazo procede de una era en que sólo algunos bancos eran miembros de la Fed y estaban sujetos a este requisito. El organismo pensaba que si elevaba el requisito
de las reservas, los bancos preferirían dejar de ser miembros. (Dado que las reservas no pagan intereses, cuanto mayor es el requisito, más grande es la carga impuesta a los bancos por conservar
reservas.) El argumento ha dejado de ser válido. Desde la aprobación de la Ley de Desregulación
de las Instituciones de Depósito y Control Monetario de 1980, todas las instituciones de depósito
quedaron sujetas a las normas de la Fed.
También es verdad que cambiar la proporción de reservas obligadas es una herramienta tosca.
Por las demoras en los informes de los bancos a la Reserva Federal sobre el estado de sus reservas y
depósitos, un cambio hoy en los requisitos no afecta a los bancos en casi dos semanas. (Sin embargo, el hecho de que cambiar la proporción de reservas amplía o reduce el crédito en todos los
bancos del país la hace una herramienta poderosa cuando es empleada por la Fed.)
LA TASA DE DESCUENTO
Los bancos pueden tomar préstamos de la Reserva Federal. La tasa de interés que le pagan es la
tasa de descuento. Cuando los bancos aumentan sus préstamos, el circulante crece. Para ver que
esto es cierto, supongamos que sólo hay un banco en el país y que la proporción de reservas requeridas es de 20 por ciento. La posición inicial del banco y la Reserva Federal aparece en la sección 1
de la tabla 10.3, donde el circulante (dinero + depósitos) es de 480 dólares. En la sección 2, el banco
tomó de la Fed un préstamo de 20 dólares. Si usa esos 20 dólares como reserva, el banco puede
aumentar sus préstamos en 100 dólares, de 320 a 420. (Recuerde que una proporción de reserva
requerida de 20 por ciento da un multiplicador de dinero de 5; tener reservas excedentes de 20
dólares permite al banco crear un adicional de $20 × 5, o sea $100, en depósitos.) Así, el circulante aumentó de 480 a 580 dólares. Por consiguiente:
Cuando un banco pide un préstamo a la Fed, aumenta el circulante.
La Fed puede influir en los préstamos de los bancos y, con ello, en el circulante, a través de la
tasa de descuento:
Mientras más alta sea la tasa de descuento, mayor será el costo de pedir prestado y menos
querrán hacerlo los bancos.
Por ejemplo, si la Fed quiere acortar el crecimiento del circulante, eleva la tasa de descuento y desalienta a los bancos para que no tomen préstamos, con lo cual restringe el aumento de las reservas (y, en última instancia, de los depósitos).
Históricamente, la Fed no ha usado la tasa de descuento para controlar el circulante. Antes de
2003 fijaba una tasa de descuento inferior a la tasa que tenían que pagar los bancos si tomaban
préstamos del mercado privado. Como es obvio, esto significaba un incentivo para que los bancos
le pidieran prestado, pero la Fed los desalentaba presionándolos de varias maneras para que no
solicitaran préstamos. Esta presión se calificaba de disuasión moral. El 9 de enero de 2003, la Fed
anunció un nuevo procedimiento. En adelante, la tasa de descuento se fijaría arriba de la tasa que
pagan los bancos por tomar dinero del mercado privado y ya no recurriría a la disuasión moral.
Ahora los bancos pueden pedir prestado a la Fed si lo desean, pero es poco probable que lo hagan,
salvo en circunstancias inusitadas porque son más baratos los préstamos en el mercado privado.
Así queda claro que la Fed no usa la tasa de descuento como herramienta frecuente para cambiar
el circulante.
OPERACIONES EN EL MERCADO ABIERTO
Con mucho, las herramientas más importantes de la Fed para controlar el circulante son las
operaciones en el mercado abierto. El Congreso ha autorizado a la Reserva Federal para que
compre y venda títulos del gobierno en los mercados abiertos. Cuando la Fed compra un título,
paga con un cheque que, cuando se cambia, amplía la cantidad de las reservas del sistema y aumenta el circulante. Cuando vende un bono, los particulares o las instituciones lo pagan con un
cheque que, cuando se cambia, reduce la cantidad de reservas del sistema.
Para ver cómo funcionan las transacciones del mercado abierto y los controles de las reservas, tenemos que revisar varias ideas esenciales.

Dos órganos del gobierno cambian títulos Quizá sea confuso el hecho de que la Reserva Federal pueda comprar y vender títulos de gobierno (certificados y bonos). En realidad, dos
órganos del gobierno participan en los mercados financieros por diferentes razones y usted debe tener
presente la distinción.
En primer lugar, recuerde que el Departamento del Tesoro se encarga de recaudar los impuestos
y pagar las cuentas del gobierno federal. Los cheques del salario para los burócratas, los pagos a General Dynamics por un nuevo barco para la armada, los cheques del Seguro Social para los jubilados,
etc., se giran contra las cuentas que tiene la Tesorería. Los impuestos recolectados por el Servicio de
Recaudación Interna, que es una dependencia de la Tesorería, se depositan en esas cuentas.
Si el gasto total del gobierno excede la recaudación fiscal, la ley exige a la Tesorería que tome en
préstamo la diferencia. Recuerde que el déficit del gobierno es (G − T), es decir, las compras del gobierno menos impuestos netos. (G − T) es la cantidad que la Tesorería tiene que pedir cada año para
financiar el déficit. Esto significa que:
La Tesorería no puede imprimir dinero para financiar el déficit.
La Tesorería toma préstamos expidiendo certificados, bonos y notas que pagan intereses. Estos
títulos del gobierno, o pagarés, se venden a individuos e instituciones. A menudo, los compran
otros países, así como ciudadanos estadounidenses. Como dijimos en el capítulo 9, el monto total
de los títulos gubernamentales de propiedad privada es la deuda federal privada.
La Reserva Federal no es la Tesorería, sino un organismo descentralizado con licencia del Congreso estadounidense para comprar y vender en el mercado abierto títulos gubernamentales sin liquidar.
Los bonos y certificados que al principio vendió la Tesorería para financiar el déficit se revenden y
cambian continuamente entre ciudadanos, empresas, bancos, fondos de pensiones, etc. Como veremos, la participación de la Fed en este comercio incide en la cantidad de las reservas del sistema.
Como la Reserva Federal posee títulos gubernamentales, parte de la deuda del gobierno se la
debe a la propia Fed. Recuerde que el rubro de activos más grande del sistema de la Fed son estos títulos, que no son más que certificados y bonos expedidos por la Tesorería para financiar el déficit.
Fueron adquiridos por la Fed al paso del tiempo, mediante compras directas en mercados abiertos
hechas para ampliar el circulante a medida que crecía la economía.
Mecanismo de las operaciones de mercado abierto ¿Qué efecto tienen en el circulante las operaciones de mercado abierto? Vuelva a la tabla 10.1. Como se aprecia, la mayor parte de los
activos de la Reserva Federal consisten en los títulos gubernamentales de los que acabamos de hablar.
Supongamos que la Fed quiere aminorar el circulante. Si puede reducir el volumen de reservas
bancarias del lado de los pasivos en su balance general, obligará a los bancos a disminuir sus propios
depósitos (para cumplir con la proporción de reservas requeridas). Como estos depósitos son parte
del circulante, éste se va a contraer.
¿Qué pasará si la Fed vende una parte de sus títulos gubernamentales al público en general? Se
reducen los títulos que posee la Fed porque los títulos que venda quedarán en poder de alguien
más. ¿Cómo pagan los compradores por sus títulos? Giran cheques contra sus bancos a favor de la
Reserva Federal.
Veamos con mayor detalle cómo funciona esto, mediante la tabla 10.4. En la sección 1, la Reserva
Federal tiene primero 100,000 millones de dólares en títulos del gobierno. Sus pasivos constan de
20,000 millones en depósitos (que son las reservas de los bancos comerciales) y 80,000 millones en dinero. Con una proporción de reservas requeridas de 20 por ciento, las reservas de 20,000 millones
pueden respaldar 100,000 millones de dólares en depósitos en los bancos comerciales. El sistema
bancario comercial otorga préstamos a toda su capacidad. En la sección 1 se muestra también la posición financiera de una persona física, Jane Q. Public, quien tiene activos por 5,000 millones (una
gran cuenta de cheques depositada en el banco) y ninguna deuda, por lo que su valor neto es de
5,000 millones de dólares.
Ahora imaginemos que la Fed le vende a Jane 5,000 millones en títulos gubernamentales. Jane
los liquida con un cheque pagadero a la Reserva Federal, tomado de su banco. La Fed toma de la
cuenta de reserva del banco de Jane 5,000 millones. Los balances generales de todos los participantes
de la transacción se muestran en la sección 2. Observe que el circulante (dinero más depósitos) ha
bajado de 180,000 millones a 175,000 millones.
La historia no termina ahí. Como resultado de la venta de títulos de la Fed, el monto de las reservas bajó de 20,000 millones a 15,000 millones, mientras que los depósitos se redujeron de 100,000
millones a 95,000 millones de dólares. Con una proporción de reservas requeridas de 20 por ciento,

los bancos deben tener reservas de 0.20 × 95,000 millones, es decir, 19,000 millones. Los bancos están
abajo de la proporción obligatoria en 4,000 millones de dólares [19,000 millones (la cantidad que deberían tener) menos 15,000 millones (la cantidad que tienen)]. Para acatar las normas federales, los
bancos deben reducir sus préstamos y depósitos.9
La posición de equilibrio final se muestra en la sección 3, donde los bancos comerciales redujeron sus préstamos en 20,000 millones. Observe que el cambio en los depósitos de la sección 1 a la
3 es de 25,000 millones, que es cinco veces el tamaño del cambio de las reservas que la Fed produjo
con su venta de títulos por 5,000 millones en el mercado abierto. Esto corresponde con exactitud a
nuestro análisis anterior del multiplicador del dinero. El cambio en dinero (–25,000 millones) es
igual al multiplicador del dinero (5) por el cambio de las reservas (–5,000 millones).
Ahora consideremos lo que sucede cuando la Fed compra un título del gobierno. Supongamos
que tengo 100 dólares en certificados de la Tesorería, que la Fed me compra. Ésta gira un cheque
para mí por 100 dólares y yo entrego mis certificados de la Tesorería. A continuación, llevo mi
cheque de 100 dólares y lo deposito en el banco de mi localidad. Esto aumenta las reservas de mi
banco en 100 dólares y comienza un nuevo episodio de la historia de la expansión del dinero. Con
un requisito de reservas de 20 por ciento, mi banco ahora puede prestar 80 dólares. Si esta suma
se gasta y termina de vuelta en el banco, puede prestar 64 dólares, etc. (Repase la figura 10.3.) La
Fed puede ampliar el circulante comprando títulos del gobierno a los particulares que los tengan;
asimismo, lo reduce vendiendo los mismos títulos.
Cada día hábil, la Mesa de Mercados Abiertos del Banco de la Reserva Federal de Nueva York
compra o vende títulos por millones de dólares, casi siempre a grandes corredores que fungen como
intermediarios entre la Fed y los mercados privados. Resumamos de esta manera el efecto de las
operaciones en los mercados abiertos.
■ Si la Fed compra títulos en mercados abiertos, aumentan las reservas y se incrementa el
circulante en un monto igual al multiplicador del dinero por el cambio en las reservas.
■ Si la Fed vende títulos en mercados abiertos, disminuyen las reservas y el circulante
decrece en un monto igual al multiplicador del dinero por el cambio en las reservas.

Las operaciones de mercado abierto son el medio que prefiere la Fed para controlar el circulante,
por varias razones. En primer lugar, las operaciones de mercado abierto pueden hacerse con relativa
precisión. Si la Fed necesita cambiar el circulante en un monto pequeño, puede comprar o vender un
volumen reducido de títulos gubernamentales. Si quiere un cambio mayor del circulante, vende o
compra más. En segundo lugar, las operaciones en el mercado abierto son extremadamente flexibles.
Si la Reserva Federal decide dar marcha atrás, puede pasar con facilidad de comprar a vender títulos.
Por último, las operaciones de mercado abierto tienen un efecto bastante previsible sobre el circulante. Como los bancos están obligados a cumplir los requisitos de reservas, una venta en mercados
abiertos de 100 dólares de títulos gubernamentales reduce las reservas en 100 dólares, lo cual disminuye el circulante en 100 dólares por el multiplicador del dinero.
¿De dónde saca la Reserva Federal el dinero para comprar títulos gubernamentales cuando
quiere extender el circulante? Lo crea, así de simple. En efecto, le informa al banco al que le compró
un título de 100 dólares que su cuenta de reservas (depósitos) en la Fed tiene desde ese momento 100
dólares más. Aquí radica el poder de la Reserva Federal y de cualquier banco central. La Fed tiene la
capacidad de crear dinero a voluntad. En Estados Unidos, la Fed ejerce esta facultad cuando crea dinero para adquirir títulos gubernamentales.
LA CURVA DE LA OFERTA PARA EL DINERO
Ahora sabemos cómo la Reserva Federal manipula el circulante controlando el monto de las reservas de la economía. Si la Fed quiere en cierta fecha que la cantidad de dinero sea de un billón
350,000 millones de dólares, para alcanzar este objetivo cambia la tasa de descuento, varía la proporción de reservas requeridas o realiza operaciones en el mercado abierto (como dijimos antes,
casi siempre usa como herramienta las operaciones en el mercado abierto). Ya que la Fed puede
escoger cualquier nivel para el circulante, es útil empezar con el caso en que escoge un valor independiente de cualquier factor de la economía. En otras palabras, vamos a suponer por ahora que
el valor que elija para el circulante no depende de rubros como las tasas de interés, inflación, desempleo o producción total. En el capítulo 15 abandonamos esta suposición. Ahí veremos que en
la práctica, la Fed escoge un valor del circulante para atinar a cierto valor de las tasas de interés y
que la elección que haga de dichas tasas depende, por ejemplo, de la inflación y el desempleo. Pero
eso sería adelantarnos al curso de la historia.
Si la elección que hace la Fed del valor del circulante no depende de la tasa de interés, en la figura 10.5 podremos trazar la curva de la demanda como una línea vertical; por lo tanto, mientras
no lleguemos al capítulo 15, vamos a suponer que la curva del circulante es vertical.

### Resumen

GENERALIDADES DEL DINERO
1. El dinero tiene tres características distintivas: 1) como medio de
pago o de cambio, 2) como depósito de valor y 3) como unidad de
cuenta. La alternativa al uso del dinero es el trueque, en el que algunos bienes se intercambian directamente por otros. El trueque
es caro e ineficiente en una economía con muchos productos.
2. Los bienes de cambio son artículos que se usan como dinero y
que también tienen valor intrínseco en algún otro uso; por
ejemplo, oro y cigarrillos. De hecho, el dinero simbólico no
tiene otro valor aparte de su uso como dinero. Para garantizar
la aceptación del dinero simbólico, los gobiernos ejercen sus
facultades para declararlo dinero de curso legal y prometen al
público que la moneda no se va a degradar por la rápida expansión
del circulante.
3. Hay varias definiciones de dinero. La moneda más los depósitos
en demanda más cheques de viajeros más otros depósitos convertibles en cheques componen el M1, el dinero de transacciones:
aquél que puede usarse en forma directa para comprar cosas. La
suma de las cuentas de ahorro y las cuentas de mercado de dinero
(dinero afín) más el M1, da el M2, o dinero general.
CÓMO CREAN DINERO LOS BANCOS
4. La proporción de reservas requeridas es el porcentaje de los
depósitos de un banco que debe conservar como reservas
en el banco central del país, la Reserva Federal.
5. Los bancos crean dinero haciendo préstamos. Cuando un banco
hace un préstamo a un cliente, crea un depósito en la cuenta de
éste. Dicho depósito se convierte en parte del circulante. Los
bancos pueden crear dinero sólo cuando tienen reservas excedentes,
excedentes sobre la cantidad fijada por la proporción de reservas
requeridas.
6. El multiplicador del dinero es el múltiplo por el cual el circulante
total aumenta con todo incremento monetario de las reservas.
El multiplicador del dinero es igual a 1/proporción de reservas
requeridas.
EL SISTEMA DE LA RESERVA FEDERAL
7. La función más importante de la Reserva Federal (Fed) es controlar
el circulante de Estados Unidos. La Fed tiene otras funciones:
compensa los pagos entre bancos, es responsable de muchas
normas y criterios que rigen las prácticas bancarias y funge como
prestamista de último recurso para bancos en problemas que no
encuentran otra fuente de financiamiento. La Fed actúa también
como el banco del gobierno estadounidense.
CÓMO CONTROLA LA RESERVA FEDERAL EL CIRCULANTE
8. La clave para entender cómo la Fed controla el circulante es la
función de las reservas. Si la Fed quiere aumentar el circulante,
genera más reservas, lo cual confiere libertad a los bancos para
crear nuevos depósitos. Si desea disminuir el circulante, reduce
las reservas.
9. La Fed tiene tres herramientas para controlar el circulante: 1)
cambios en la proporción de las reservas requeridas, 2) cambios
en la tasa de descuento (la tasa de interés que pagan los bancos
miembros cuando piden prestado a la Fed) o 3) realizar operaciones
en el mercado abierto (la compraventa de títulos gubernamentales
ya existentes). Para aumentar el circulante, la Fed puede crear
más reservas disminuyendo la tasa de descuento o comprando
títulos gubernamentales, o bien, para aumentar el número de
depósitos que pueden crearse dada cierta cantidad de reservas,
se disminuye la proporción de las reservas requeridas. Para
reducir el circulante, la Fed aminora sus reservas: eleva la tasa
de descuento o vende títulos gubernamentales o aumenta la
proporción de las reservas requeridas.
10. Si el proceder de la Reserva Federal con respecto al circulante
no estuviera bajo el influjo de las tasas de interés, la curva del
circulante sería una línea vertical.

Hay muchas tasas de interés, pero vamos a suponer que sólo hay una. Esto simplifica nuestro

LA DEMANDA DE DINERO
Los factores y las fuerzas que determinan la demanda de dinero son temas centrales de la macroeconomía. Como veremos, la tasa de interés y el nivel del ingreso nacional (Y) influyen en cuánto
dinero quieren conservar los hogares y las empresas.
Antes de pasar adelante, debemos poner de relieve un punto que puede ser problemático.
Cuando hablamos de la demanda de dinero, no nos hacemos preguntas como éstas: “¿cuánto dinero
te gustaría tener?” “¿Qué ingresos quisieras ganar?” “¿Cuánta riqueza te complacería?” (La respuesta
a estas preguntas podría ser: “lo más que sea posible”.) Por el contrario, estamos interesados en
cuánto de sus activos financieros quiere conservar en la forma de dinero, que no genera intereses,
en comparación con cuánto espera tener en títulos que dan intereses, como los bonos. Damos por
hecho el total de los activos financieros; lo que nos interesa aquí es cómo se dividen esos activos
entre dinero y títulos que generan intereses.EL MOTIVO DE LAS TRANSACCIONES
La suma de efectivo que se desea conservar requiere de un equilibrio entre la liquidez del dinero y
el ingreso por intereses que ofrecen otros activos. La principal razón para tener dinero en lugar de
activos que generan intereses es que sirve para comprar cosas. Los economistas llaman a este hecho el
motivo de las transacciones. Este razonamiento para tener dinero es el meollo del análisis que sigue.1
Premisas Para mantener la claridad en nuestro análisis sobre la demanda de dinero, tenemos
que postular algunas simplificaciones. En primer lugar, suponemos que los hogares sólo tienen acceso a dos clases de activos: bonos y dinero. Por “bonos” entendemos títulos de todo tipo que generan
intereses. Por “dinero” aceptamos la moneda en circulación y los depósitos, presumiendo que en
ninguno de los dos casos se pagan intereses.2
En segundo lugar, pensamos que el ingreso del hogar común está “acumulado”: llega una vez
al mes, en los primeros días. En cambio, el gasto se prolonga todo el tiempo, pues ocurre a un ritmo uniforme a lo largo del mes; es decir, cada día se gasta la misma suma (figura 11.1). La discordancia entre la entrada de dinero al hogar y la salida para gastos se llama desincronización entre el
ingreso y el gasto.
Por último, suponemos que el gasto del mes es exactamente igual al ingreso de ese periodo. Como nos concentramos en la demanda de dinero para transacciones y no en su uso como depósito de
valor, esta suposición es por completo razonable.

mensual retiene como dinero y qué cantidad reserva como bonos que generan intereses? Imaginemos
que Jim decide depositar todo su pago en su cuenta de cheques. Digamos que gana 1,200 dólares

bancaria de Jim está en cero. Con puntualidad, recibe su siguiente pago mensual, lo deposita y el

Una estadística útil que tenemos que calcular es el saldo promedio de la cuenta de Jim. Así, él gasta su dinero a un ritmo constante de 40 dólares por día (40 por 30 días del mes = 1,200). Su saldo
promedio es su saldo inicial (1,200 dólares) más el saldo final (0) dividido entre 2; es decir, (1,200
+ 0)/2 = 600 dólares. Durante la primera mitad del mes Jim tiene depositado más que su promedio
de 600; en la segunda mitad, tiene menos que el promedio.
¿Hay algo erróneo en esta estrategia? Sí. Si Jim sigue el plan descrito, está renunciando a ganar
intereses sobre sus fondos, los cuales generaría si pusiera parte de su efectivo en bonos con rendimiento en lugar de su cuenta de cheques. ¿Cómo podría administrar su dinero para que le dieran
más intereses?
En lugar de depositar todo el pago en su cuenta de cheques al comenzar el mes, Jim podría
meter la mitad en la cuenta y comprar un bono con la otra mitad. De esta manera, se quedaría sin
dinero en su cuenta a la mitad del mes. A un ritmo de 40 dólares diarios en su cuenta, el depósito
inicial de 600 apenas duraría 15 días. Jim tendría que vender su bono a la mitad del mes y depositar los 600 dólares de la venta en su cuenta para pagar sus cuentas durante la segunda mitad.

de los fondos en bonos que pueden redituar intereses, en lugar de tener dinero “ocioso”. ¿Por qué detenerse ahí? Otra posibilidad sería que Jim depositara sólo 400 dólares en la cuenta de cheques a comienzos del mes y que comprara dos bonos de 400. Los 400 dólares de la cuenta van a durar nada más

promedio de posesión de bonos que generan intereses.
Cabe imaginar que Jim va más lejos. ¿Por qué no tener toda la riqueza en forma de bonos
(que dan intereses) y hacer transferencias de bonos a dinero cada vez que se hace una compra? Si
vender bonos, transferir fondos a cuentas de cheques y trasladarse al banco no costaran nada, Jim
nunca tendría dinero más que por un instante. Cada vez que necesitara pagar algo o emitir un cheque, iría o llamaría al banco, transferiría la suma exacta de la transacción a su cuenta de cheques y
retiraría el dinero o giraría el cheque para completar la operación. Si lo hiciera de manera continua,
obtendría la mayor cantidad posible de intereses sobre sus fondos, porque siempre tendría activos
que los generaran.
En la práctica, resulta caro este manejo del dinero. Hay comisiones de correduría y otros gastos
por comprar o vender bonos; además, hay que perder tiempo en la fila del banco o el cajero automático. Asimismo, es oneroso tener activos que no generan intereses, dado que se pierden ingresos
potenciales por este rubro.
Tenemos un problema de compensación del tipo que es común en la economía. Pasar más a
menudo de bonos a dinero aumenta los ingresos por intereses de Jim (porque cuanto más cambia,
menos, en promedio, guarda en su chequera y más conserva en bonos), pero esto aumenta los costos
de administrar el dinero. Menos traspasos significa más intereses perdidos (porque en promedio,
las posesiones de dinero son mayores), pero hay menos costos administrativos (menos compraventa
de bonos, menos tiempo perdido en las sucursales bancarias, menos traslados al banco, etcétera).
El saldo óptimo Hay un saldo promedio de dinero que le da a Jim los mayores rendimientos,
si se consideran tanto los intereses ganados por los bonos como los costos que se pagan por transferir
de bonos a dinero. Este nivel se llama saldo óptimo.
¿Qué efecto tiene la tasa de interés en los traspasos que hace Jim y, por consiguiente, en el saldo
promedio de dinero que decide conservar? Si la tasa de interés fuera de sólo 2 por ciento, no tendría
caso renunciar a tanta liquidez por tener bonos en lugar de efectivo o saldo en la chequera. En cambio,
si la tasa de interés fuera de 30 por ciento, el costo de oportunidad de conservar dinero en lugar de
bonos sería muy alto; esperaríamos que la gente tuviera la mayor parte de sus fondos como bonos y
dedicara mucho tiempo a manejar sus saldos de dinero. La tasa de interés representa el costo de oportunidad de tener dinero (y, por consiguiente, no tener bonos que pagan intereses). Cuanto más alta sea
la tasa de interés, mayor será el costo de oportunidad de tener dinero y menos dinero quiere tener la
gente. Esto nos lleva a concluir:
Cuando las tasas de interés son elevadas, la gente quiere aprovechar los rendimientos altos
de los bonos, así que prefiere tener poco dinero.

EL MOTIVO DE LA ESPECULACIÓN
Se han propuesto varias teorías para explicar por qué la cantidad de dinero que los hogares quieren tener aumenta cuando bajan las tasas de interés y se reduce cuando éstas se incrementan. Una
de las teorías considera las expectativas de los hogares y la relación entre las tasas de interés y los
valores de los bonos.
Para entender esta teoría, debe percatarse de que el valor en el mercado de la mayor parte de
los bonos que generan intereses guarda una relación inversa con la tasa de interés. Supongamos
que hace un año adquirí un bono en 1,000 dólares, a 8 por ciento. Ahora digamos que las tasas de
interés del mercado aumentan a 10 por ciento. Si pongo a la venta mi bono en 1,000 dólares, nadie lo
compraría, porque cualquiera puede obtener un bono nuevo y ganar 10 por ciento en el mercado,
en lugar de 8 por ciento que da mi bono. No obstante, en algún precio de venta inferior mi bono se
vuelve atractivo para los compradores. La razón es que un precio más bajo acrecienta el rendimiento
real para el comprador del bono. Supongamos que le vendo mi bono en 500 dólares. Como el bono
paga 8 por ciento anual sobre los 1,000 dólares originales (es decir, 80 dólares al año), de hecho le paga una suma anual que corresponde a 16 por ciento de su inversión en el bono (500 × 0.16 = 80).
Si usted me comprara el mismo bono en 800 dólares, en los hechos le pagaría 10 por ciento de interés (800 × 0.10 = 80). El punto es simple:
Cuando las tasas de interés del mercado bajan, aumenta el valor de los bonos; cuando
las tasas suben, el valor de los bonos se reduce.
Ahora consideremos mi deseo de tener saldos de dinero en lugar de bonos. Si las tasas de interés
del mercado son más altas que lo normal, prevería que bajen más adelante. Si en efecto se reducen,
aumenta el valor de los bonos que compré cuando eran altas. Cuando las tasas de interés son altas, el
costo de oportunidad de tener saldos en efectivo es elevado y hay un motivo de especulación para tener bonos en lugar de efectivo. Así, yo “especulo” que las tasas de interés menguarán en el futuro.
Del mismo modo, cuando las tasas de interés del mercado son menores a lo normal, esperaría
que crecieran más adelante. El aumento de las tasas de interés traería una mengua del valor de los
bonos. Así, cuando las tasas de interés son bajas, es buen momento para tener dinero y no bonos.
Cuando las tasas han disminuido, el costo de oportunidad de tener saldos en efectivo es bajo; además, hay un motivo especulativo para conservar más dinero. ¿Por qué debo invertir dinero en bonos
si espero que las tasas de interés aumenten más adelante? (Para más detalles sobre la relación entre
los mercados de bonos y de dinero, véase el apartado Para profundizar, “El mercado de bonos, el
mercado de dinero y el motivo de la especulación”.)
DEMANDA TOTAL DE DINERO
Hasta aquí hemos hablado sólo de la demanda de saldos en cuentas de cheques. Ahora bien, el total del dinero demandado en una economía es la suma de la demanda de saldos en cuentas y efectivo de hogares y empresas.
La compensación que hacen las compañías es igual que en el caso de Jim. Como los hogares,
las empresas tienen que administrar su dinero. Deben completar la nómina y tienen compras por
hacer; reciben efectivo y cheques de ventas y muchas firmas que atienden al público necesitan
cambio, efectivo en la caja registradora. Así, al igual que Jim, necesitan dinero para realizar las
transacciones ordinarias.
Sin embargo, las empresas, tanto como los hogares, pueden tener sus activos en formas que generen intereses. Manejan sus activos tal como los hogares: tienen una parte en efectivo, otra en su cuenta de cheques y una más en bonos. Una tasa de interés más alta encarece el costo de oportunidad del
dinero para las empresas, al igual que para los hogares, y con ello se reduce la demanda de dinero.
El mismo equilibrio ha de hacerse con el dinero. Todos cargamos dinero de bolsillo (pero no miles) para las transacciones comunes. En promedio, llevamos lo que vamos a necesitar y no más, porque hay costos: el riesgo de sufrir un robo y los intereses no devengados.
En todo momento hay una demanda de dinero, en la forma de efectivo y saldos de cuentas bancarias. Aunque hogares y empresas necesitan tener saldo para las transacciones cotidianas, esta demanda tiene un límite. Para hogares y empresas, la cantidad de dinero demandada en cualquier momento depende del costo de oportunidad de conservar el
dinero, un costo que está determinado por la tasa de interés.
VOLUMEN DE TRANSACCIONES Y NIVEL DE PRECIOS
La curva de la demanda del dinero que aparece en la figura 11.4 está en función de la tasa de interés.
Aparte de esta tasa, otros factores influyen en la cantidad total de dinero que se quiere tener. Un factor
es el valor en dinero de las transacciones hechas en determinado periodo.
Supongamos que el ingreso de Jim aumenta. En lugar de dedicar 1,200 dólares a sus compras de
cada mes, va a gastar más. Así, necesita tener más dinero. La razón es simple: para comprar más, necesita más dinero.

Lo que es verdad para Jim, también lo es para el conjunto de la economía. La demanda total
de dinero en la economía depende del volumen total en dinero de las transacciones hechas. Por su
parte, el volumen de las transacciones de la economía depende de dos cosas: el número total y el
monto promedio de las transacciones. No hay datos sobre el número real de las transacciones de
la economía, pero quizá la producción total (ingreso, Y) sea un indicador razonable. Un aumento
de la producción total (del producto interno bruto real, PIB real) significa que hay más actividad
económica. Las empresas producen y venden más, hay más trabajadores en las nóminas y el ingreso
de los hogares se incrementa. En suma, hay más transacciones y las empresas y hogares retienen más
dinero si realizan más transacciones. Así, un aumento de la producción total (ingreso) acrecienta
la demanda del dinero. En la figura 11.5 se muestra un desplazamiento de la curva de la demanda
de dinero, que es resultado del incremento en Y.
Dada una tasa de interés, una producción mayor significa un incremento del número
de transacciones y más demanda de dinero. La curva de la demanda de dinero se desplaza a la derecha cuando Y aumenta. Asimismo, una reducción de Y significa una
disminución del número de transacciones y una baja de la demanda de dinero. La
curva de la demanda del dinero se desplaza a la izquierda cuando Y baja.
La suma del dinero que necesitan las empresas y los hogares para facilitar sus transacciones
cotidianas también depende del monto promedio de cada transacción, que a su vez depende de los
precios o, mejor, del nivel de precios. Si todos los precios, incluyendo el de la mano de obra (la tasa
salarial) fueran del doble, empresas y hogares necesitarían más saldos de dinero para realizar sus
transacciones diarias: cada transacción requeriría el doble de dinero. Si el precio del desayuno
aumentara de 3.50 a 7 dólares, uno tendría que cargar más efectivo. Si las facturas del fin de mes
llegan por el doble que antes, uno meterá más dinero en la cuenta de cheques.
Los aumentos del nivel de precios desplazan la curva de la demanda a la derecha y las
disminuciones la mueven hacia la izquierda. Aunque el número de las transacciones
no cambie, sí varía el dinero necesario para realizarlas.

REPASO DE LOS DETERMINANTES DE LA DEMANDA DE DINERO
En la tabla 11.1 se resume todo lo que hemos dicho acerca de la demanda del dinero. En primer
lugar, como la tasa de interés (r) es el costo de oportunidad de tener saldos de dinero, tanto para
empresas como para hogares, es probable que los aumentos de las tasas de interés reduzcan la demanda de dinero; las reducciones de estas tasas aumentan la cantidad del dinero demandado. Así,
la cantidad de dinero demandado guarda una relación inversa con la tasa de interés.
La demanda de dinero depende también del volumen en dinero de las transacciones de determinado periodo. El volumen en dinero de las transacciones depende de la producción total
(ingreso, Y) y del nivel de precios, P. La relación entre la demanda de dinero y Y, y la correspondencia entre la demanda de dinero y P son positivas. Los aumentos en Y o P mueven la curva de
la demanda a la derecha; las reducciones de Y o P la desplazan a la izquierda.
Algunas dificultades comunes Tenemos que señalar varios escollos al reflexionar sobre la demanda de dinero. En primer lugar, cuando en capítulos anteriores nos referíamos a la
demanda de bienes y servicios, hablábamos de la demanda como variable de flujo, es decir, como
algo que se mide al paso del tiempo. Si usted dice que su demanda de café es de tres tazas, tiene
que especificar si se refiere a tres tazas por hora, por día o por semana. En macroeconomía, consumo y ahorro son variables de flujo. Consumimos y ahorramos continuamente, pero expresamos el
consumo y el ahorro en términos de periodos, como 600 dólares por mes.
La demanda de dinero no es una medida de flujo, sino de existencias, medida en un momento dado. Responde a la pregunta sobre cuánto dinero quieren tener empresas y hogares en
determinado momento, dados la tasa de interés actual, el volumen de la actividad económica y
el nivel de precios.
En segundo lugar, muchas personas piensan que la demanda de dinero y el ahorro son casi
la misma cosa; no lo son. Supongamos que en un año un hogar tiene ingresos de 50,000 dólares
y gastos de 47,000. Ahorró 3,000 ese año. A comienzos del año el hogar no tenía deudas y sus activos sumaban 100,000 dólares. Como el hogar ahorró 3,000 en el año, al finalizar éste tiene activos
por 103,000. Parte de los 103,000 dólares los tiene en acciones, parte en bonos, algo en otros títulos y algo más como dinero. El monto que escoja tener en la forma de efectivo es su demanda de
dinero. Dependiendo de la tasa de interés y las transacciones del hogar, la parte de los 103,000 dólares que quiera tener como dinero sería lo mismo de pocos cientos a muchos miles. La cantidad
de sus activos que retenga un hogar en la forma de dinero es diferente de cuánto gaste de su ingreso durante el año.
Por último, recuerde la diferencia entre un desplazamiento de la curva de la demanda y un
movimiento sobre la curva. La curva de la demanda de dinero de la figura 11.4 muestra los saldos
óptimos de dinero como función de la tasa de interés ceteris paribus, en igualdad de circunstancias.
Los cambios de las tasas de interés producen movimientos sobre la curva: cambios en la cantidad
del dinero demandado. Los cambios del PIB real (Y) o del nivel de precios (P) causan desplazamientos en la curva, como se muestra en la figura 11.5: cambios en la demanda.
TASA DE INTERÉS DE EQUILIBRIO
Ahora estamos en posición de considerar una de las preguntas macroeconómicas fundamentales:
¿cómo se determina la tasa de interés en la economía?

Los mercados financieros (que llamamos mercados de dinero) funcionan muy bien en Estados Unidos. Casi todos estos mercados son sanos, en el sentido de que alcanzan un equilibrio en
el que la cantidad de la demanda es igual a la ofrecida. En el mercado de dinero:
El punto donde la cantidad de dinero demandado es igual al monto de dinero ofrecido
determina la tasa de interés de equilibrio en la economía.
Esta explicación suena simple, pero requiere elaboración.

AUMENTOS DE Y Y DESPLAZAMIENTO DE LA CURVA DE LA
DEMANDA DE DINERO
Los cambios del circulante no son los únicos factores que influyen en la tasa de interés de equilibrio. Las variaciones de la demanda de dinero pueden hacer otro tanto.
Recuerde que la demanda de dinero depende tanto de la tasa de interés como del volumen
de las transacciones. Como medida aproximada del volumen de las transacciones tomamos Y, el
nivel de la producción total (ingreso). Recuerde también que la relación entre la demanda de dinero y Y es positiva: aumentos en Y representan un estado mayor de la actividad económica real.
Cuanto más se produce, más alto es el ingreso y hay más transacciones en la economía. Por consiguiente, la demanda de dinero por parte de empresas y hogares es mayor en total.
Un aumento de Y desplaza a la derecha la curva de la demanda de dinero.
En la figura 11.8 se ilustra este cambio. Y se acrecienta y hace que la demanda de dinero se
desplace de M d
0 a M d
1
. El resultado es un incremento en el nivel de equilibrio de la tasa de interés,
de 7 a 14 por ciento. Una reducción de Y desplazaría a Md hacia la izquierda y la tasa de interés de
equilibrio bajaría.
La curva de la demanda de dinero también se desplaza cuando cambia el nivel de precios. Si
éste aumenta, la curva de la demanda de dinero se desplaza a la derecha, porque la gente necesita
más dinero para realizar las transacciones cotidianas. En cambio, si no varía el circulante, la tasa
de interés tiene que subir para disminuir la cantidad del dinero demandado en concordancia con
el dinero ofrecido: se trata de un movimiento sobre la curva de la demanda.
Un aumento en el nivel de precios se parece a un incremento de Y en que ambos sucesos ensanchan la demanda de dinero. El resultado es que sube la tasa de interés de
equilibrio.

Si el nivel de precios baja, la curva de la demanda de dinero se desplaza a la izquierda porque
la gente necesita menos dinero para sus transacciones. Ahora bien, si no cambia el circulante, la
tasa de interés tiene que bajar para elevar la cantidad de dinero demandada al nivel de la oferta.
Una disminución del nivel de precios da lugar a una reducción de la tasa de interés de
equilibrio.

A CONTINUACIÓN: LA RESERVA FEDERAL
Y LA POLÍTICA MONETARIA
Ahora sabemos que la Reserva Federal puede cambiar la tasa de interés variando el circulante. Si
la Fed aumenta el circulante, la tasa de interés baja; si lo reduce, la tasa de interés sube.
Sin embargo, no hemos dicho para qué quiere la Fed cambiar la tasa de interés o qué pasa en
la economía cuando la tasa cambia. Ya hemos dado un indicio: una tasa de interés baja estimula el
gasto, sobre todo de inversión; una tasa elevada reduce el gasto. Al cambiar la tasa de interés, la
Fed modifica la producción total (ingreso). En el capítulo siguiente vamos a combinar nuestros
análisis de los mercados de bienes y de dinero, y veremos qué efecto tiene la tasa de interés en el
nivel de equilibrio de la producción total (ingreso, Y) en el mercado de bienes.
El centro de la política monetaria del gobierno es la aplicación que hace la Fed de sus facultades para influir en los acontecimientos del mercado de bienes, así como en el mercado del dinero. Cuando la Fed actúa para constreñir el circulante, con la intención de restringir la economía,
los economistas dicen que se trata de una política monetaria estricta. Por el contrario, cuando la
Fed amplía el circulante y estimula la economía, tiene una política monetaria laxa. La Fed actuó
con energía para ampliar el circulante y reducir las tasas de interés en 1975, 1982 y a comienzos de
1991. Estas políticas monetarias laxas contribuyeron a la recuperación económica de las recesiones de esos años. Las políticas monetarias estrictas abatieron el gasto total en 1974 y 1981, y contribuyeron a las recesiones de esos años. En el verano de 1981, las restricciones monetarias impulsaron algunas tasas de interés importantes a más de 20 por ciento.
En el capítulo 15 vamos a estudiar el efecto de la economía en el proceder de la Fed. En ese
capítulo también abordaremos las políticas más recientes de la Fed y examinaremos sus efectos
sobre la economía.

### Resumen

1. El interés es el pago que entregan los prestatarios a los prestamistas
por el uso de sus fondos. La tasa de interés es el pago anual de
intereses expresado como porcentaje del préstamo; es igual al
monto del interés recibido por año, dividido entre el monto del
préstamo. Aunque hay muchas tasas de interés en Estados Unidos,
vamos a suponer que en la economía sólo hay una. Esto simplifica
nuestro análisis, sin dejar de proporcionarnos un medio para
entender la relación recíproca que guardan los componentes de la
macroeconomía.
LA DEMANDA DE DINERO
2. La demanda de dinero depende negativamente de la tasa de
interés. Cuanto mayor es la tasa, más grande es el costo de
oportunidad de conservar el dinero y menos dinero quiere
retener la gente (se renuncia a más intereses). Un aumento en la
tasa de interés reduce la cantidad demandada de dinero y la
curva de la demanda tiende a disminuir.
3. El volumen de las transacciones de la economía incide en la
demanda del dinero. El volumen total en dinero de las transacciones depende del número total de éstas y su monto promedio.
4. Una medida razonable del número de transacciones de la
economía es la producción total (ingreso, Y). Cuando Y aumenta,
hay más actividad económica, se produce y se vende más y hay
más personas en las nóminas: se generan más transacciones en
la economía. Un aumento de Y hace que la curva de la demanda
de dinero se desplace a la derecha. Este efecto es resultado de
que hogares y empresas necesitan más dinero si realizan más
transacciones. Una reducción de Y hace que la curva de la
demanda de dinero se desplace a la izquierda.
5. Los cambios en el nivel de precios afectan el monto promedio
en dinero de cada transacción. Los incrementos en el nivel de
precios aumentan la demanda de dinero (desplazan la curva de
la demanda a la derecha) porque hogares y empresas necesitan
más dinero para sus gastos. Las reducciones del nivel de precios aminoran la demanda de dinero (desplazan la curva de la demanda de dinero a la izquierda).
TASA DE INTERÉS DE EQUILIBRIO
6. El punto donde el volumen del circulante es igual a la cantidad
de la demanda de dinero determina la tasa de interés de equilibrio
de la economía. Un exceso de circulante lleva a empresas y
hogares a comprar más bonos, lo cual hace bajar la tasa de
interés. Un exceso de la demanda de dinero hace que hogares
y empresas se aparten de los bonos, y esto hace subir las tasas
de interés.
7. La Fed influye en la tasa de interés de equilibrio si cambia el
circulante mediante alguna de sus tres herramientas: proporción
de reservas requeridas, tasa de descuento u operaciones en
mercados abiertos.
8. Un incremento en el nivel de precios es como un incremento de Y,
en el sentido de que en los dos casos se produce un aumento de la
demanda de dinero. El resultado es que sube la tasa de interés de
equilibrio. Una reducción del nivel de precios abate la demanda
de dinero y hace bajar la tasa de interés de equilibrio.
9. Una política monetaria estricta comprende las iniciativas de la
Fed para constreñir el circulante, con la intención de restringir
la economía. Una política monetaria laxa se refiere a las
políticas de la Reserva Federal para ampliar el circulante con
la intención de estimular la economía. La Fed escoge una
política por diferentes motivos y en ocasiones distintas.
